# 1. Definition

Let $\mathbf{Z}$ be the concatenated random matrix
$\mathbf{Z} = \begin{bmatrix} \mathbf{X} \\ \mathbf{Y} \end{bmatrix} \in \mathbb{R}^{n+m}$.


The random variable $\mathbf{Z}$ is said to have a **joint multivariate normal distribution** if every linear combination of its components follows a univariate normal distribution.

For $\mathbf{Z}$ to possess a **joint density function** (meaning its distribution is absolutely continuous with respect to the Lebesgue measure on $\mathbb{R}^{n+m}$), its covariance matrix must be non-singular (positive definite).


The joint density function $f_{\mathbf{X}, \mathbf{Y}}(\mathbf{z})$ for $\mathbf{z} \in \mathbb{R}^{n+m}$ is given by:

$$f_{\mathbf{X}, \mathbf{Y}}(\mathbf{z}) = \frac{1}{\sqrt{(2\pi)^{n+m} |\boldsymbol{\Sigma}|}} \exp\left( -\frac{1}{2} (\mathbf{z} - \boldsymbol{\mu})^T \boldsymbol{\Sigma}^{-1} (\mathbf{z} - \boldsymbol{\mu}) \right).$$

Here:

* **Mean ($\boldsymbol{\mu}$):** $\boldsymbol{\mu} = E[\mathbf{Z}] = \begin{bmatrix} \boldsymbol{\mu}_X \\ \boldsymbol{\mu}_Y \end{bmatrix} \in \mathbb{R}^{n+m}$, where $\boldsymbol{\mu}_X \in \mathbb{R}^n$ and $\boldsymbol{\mu}_Y \in \mathbb{R}^m$.
* **Covariance Matrix ($\boldsymbol{\Sigma}$):** A symmetric, positive definite matrix $\boldsymbol{\Sigma} \in \mathbb{R}^{(n+m) \times (n+m)}$ partitioned as:
$$\boldsymbol{\Sigma} = \begin{bmatrix} \boldsymbol{\Sigma}_{XX} & \boldsymbol{\Sigma}_{XY} \\ \boldsymbol{\Sigma}_{YX} & \boldsymbol{\Sigma}_{YY} \end{bmatrix}$$
    * $\boldsymbol{\Sigma}_{XX} = \text{Var}(\mathbf{X})$ (size $n \times n$)
    * $\boldsymbol{\Sigma}_{YY} = \text{Var}(\mathbf{Y})$ (size $m \times m$)
    * $\boldsymbol{\Sigma}_{XY} = \text{Cov}(\mathbf{X}, \mathbf{Y}) = \boldsymbol{\Sigma}_{YX}^T$ (size $n \times m$)
* **Determinant ($|\boldsymbol{\Sigma}|$):** The normalization factor ensuring the total integral is 1.

# 2. Marginal Distributions

To derive the marginal density of $\mathbf{X}$ (or $\mathbf{X}$) from the joint density of $(\mathbf{X}, \mathbf{Y})$, we must integrate the joint density over all possible values of $\mathbf{Y}$ (or $\mathbf{X}$ to find the marginal of  $\mathbf{Y}$).

The challenge lies in the quadratic form within the exponential. To solve this rigorously, we use the **block matrix inversion** (Schur complement) and the **completing the square** technique.


From the definition the joint density $f_{\mathbf{X}, \mathbf{Y}}(\mathbf{x}, \mathbf{y}) \propto \exp(-\frac{1}{2} Q)$, where the quadratic form $Q$ is:
$$Q = \begin{bmatrix} \mathbf{x} - \boldsymbol{\mu}_x \\ \mathbf{y} - \boldsymbol{\mu}_y \end{bmatrix}^T \begin{bmatrix} \boldsymbol{\Sigma}_{XX} & \boldsymbol{\Sigma}_{XY} \\ \boldsymbol{\Sigma}_{YX} & \boldsymbol{\Sigma}_{YY} \end{bmatrix}^{-1} \begin{bmatrix} \mathbf{x} - \boldsymbol{\mu}_x \\ \mathbf{y} - \boldsymbol{\mu}_y \end{bmatrix}$$

Using the **Schur Complement** of $\boldsymbol{\Sigma}_{XX}$, the inverse of the block covariance matrix $\boldsymbol{\Sigma}^{-1}$ can be written as:
$$\boldsymbol{\Sigma}^{-1} = \begin{bmatrix} \mathbf{I} & -\boldsymbol{\Sigma}_{XX}^{-1}\boldsymbol{\Sigma}_{XY} \\ \mathbf{0} & \mathbf{I} \end{bmatrix}^T \begin{bmatrix} \boldsymbol{\Sigma}_{XX}^{-1} & \mathbf{0} \\ \mathbf{0} & \mathbf{S}_{YY}^{-1} \end{bmatrix} \begin{bmatrix} \mathbf{I} & -\boldsymbol{\Sigma}_{XX}^{-1}\boldsymbol{\Sigma}_{XY} \\ \mathbf{0} & \mathbf{I} \end{bmatrix}$$
where $\mathbf{S}_{YY} = \boldsymbol{\Sigma}_{YY} - \boldsymbol{\Sigma}_{YX}\boldsymbol{\Sigma}_{XX}^{-1}\boldsymbol{\Sigma}_{XY}$.



## 2.1 Marginal Density of $X$

Letting $\widetilde{\mathbf{x}} = \mathbf{x} - \boldsymbol{\mu}_x$ and $\widetilde{\mathbf{y}} = \mathbf{y} - \boldsymbol{\mu}_y$, we can decompose the exponent $Q$ into two independent parts:
$$Q = \widetilde{\mathbf{x}}^T \boldsymbol{\Sigma}_{XX}^{-1} \widetilde{\mathbf{x}} + (\widetilde{\mathbf{y}} - \mathbf{A}\widetilde{\mathbf{x}})^T \mathbf{S}_{YY}^{-1} (\widetilde{\mathbf{y}} - \mathbf{A}\widetilde{\mathbf{x}})$$
where $\mathbf{A} = \boldsymbol{\Sigma}_{YX}\boldsymbol{\Sigma}_{XX}^{-1}$.

The joint density now factors as:
$$f_{\mathbf{X}, \mathbf{Y}}(\mathbf{x}, \mathbf{y}) = C \cdot \exp\left( -\frac{1}{2} \widetilde{\mathbf{x}}^T \boldsymbol{\Sigma}_{XX}^{-1} \widetilde{\mathbf{x}} \right) \cdot \exp\left( -\frac{1}{2} (\widetilde{\mathbf{y}} - \mathbf{m})^T \mathbf{S}_{YY}^{-1} (\widetilde{\mathbf{y}} - \mathbf{m}) \right)$$
where $\mathbf{m} = \mathbf{A}\widetilde{\mathbf{x}}$ and $C$ is the normalization constant.

---

The marginal density $f_{\mathbf{X}}(\mathbf{x})$ is defined as:
$$f_{\mathbf{X}}(\mathbf{x}) = \int_{\mathbb{R}^m} f_{\mathbf{X}, \mathbf{Y}}(\mathbf{x}, \mathbf{y}) d\mathbf{y}$$

Because the first exponential term depends only on $\mathbf{x}$, it can be pulled out of the integral:
$$f_{\mathbf{X}}(\mathbf{x}) = C \cdot e^{-\frac{1}{2} \widetilde{\mathbf{x}}^T \boldsymbol{\Sigma}_{XX}^{-1} \widetilde{\mathbf{x}}} \int_{\mathbb{R}^m} \exp\left( -\frac{1}{2} (\widetilde{\mathbf{y}} - \mathbf{m})^T \mathbf{S}_{YY}^{-1} (\widetilde{\mathbf{y}} - \mathbf{m}) \right) d\mathbf{y}$$

The integral is simply the kernel of a multivariate normal distribution for $\mathbf{Y}$ (centered at $\mathbf{m}$ with covariance $\mathbf{S}_{YY}$). By the property of probability densities, this integral evaluates to the reciprocal of the normalization constant for $\mathbf{Y}$:
$$\int_{\mathbb{R}^m} (\dots) d\mathbf{y} = \sqrt{(2\pi)^m |\mathbf{S}_{YY}|}$$

---

Using the determinant identity for block matrices, $|\boldsymbol{\Sigma}| = |\boldsymbol{\Sigma}_{XX}| \cdot |\mathbf{S}_{YY}|$. When we combine the constants:
$$f_{\mathbf{X}}(\mathbf{x}) = \frac{\sqrt{(2\pi)^m |\mathbf{S}_{YY}|}}{\sqrt{(2\pi)^{n+m} |\boldsymbol{\Sigma}_{XX}||\mathbf{S}_{YY}|}} \exp\left( -\frac{1}{2} (\mathbf{x} - \boldsymbol{\mu}_x)^T \boldsymbol{\Sigma}_{XX}^{-1} (\mathbf{x} - \boldsymbol{\mu}_x) \right)$$

Simplifying the constants yields the **Marginal Density of X**:
$$f_{\mathbf{X}}(\mathbf{x}) = \frac{1}{\sqrt{(2\pi)^n |\boldsymbol{\Sigma}_{XX}|}} \exp\left( -\frac{1}{2} (\mathbf{x} - \boldsymbol{\mu}_x)^T \boldsymbol{\Sigma}_{XX}^{-1} (\mathbf{x} - \boldsymbol{\mu}_x) \right)$$

This proves that $\mathbf{X} \sim N(\boldsymbol{\mu}_x, \boldsymbol{\Sigma}_{XX})$.

## 2.2 Marginal Density of $Y$

The derivation for $f_{\mathbf{Y}}(\mathbf{y})$ is mathematically symmetric to the derivation for $\mathbf{X}$. The goal is to integrate out $\mathbf{x}$ to isolate the distribution of $\mathbf{Y}$.


We use the **Schur complement of $\boldsymbol{\Sigma}_{YY}$** (instead of $\boldsymbol{\Sigma}_{XX}$) to factor the quadratic form $Q$. The inverse covariance matrix $\boldsymbol{\Sigma}^{-1}$ is decomposed such that the terms involving $\mathbf{x}$ are "completed" into a square:

$$Q = \widetilde{\mathbf{y}}^T \boldsymbol{\Sigma}_{YY}^{-1} \widetilde{\mathbf{y}} + (\widetilde{\mathbf{x}} - \mathbf{B}\widetilde{\mathbf{y}})^T \mathbf{S}_{XX}^{-1} (\widetilde{\mathbf{x}} - \mathbf{B}\widetilde{\mathbf{y}})$$

Where:
* $\widetilde{\mathbf{y}} = \mathbf{y} - \boldsymbol{\mu}_y$
* $\mathbf{B} = \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}$ (the regression coefficients of $\mathbf{X}$ on $\mathbf{Y}$)
* $\mathbf{S}_{XX} = \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}$ (the Schur complement of $\boldsymbol{\Sigma}_{YY}$)

---

The marginal $f_{\mathbf{Y}}(\mathbf{y})$ is calculated by integrating the joint density over $d\mathbf{x}$:

$$f_{\mathbf{Y}}(\mathbf{y}) = \int_{\mathbb{R}^n} f_{\mathbf{X}, \mathbf{Y}}(\mathbf{x}, \mathbf{y}) d\mathbf{x}$$

The term $\exp\left(-\frac{1}{2} \widetilde{\mathbf{y}}^T \boldsymbol{\Sigma}_{YY}^{-1} \widetilde{\mathbf{y}}\right)$ is constant with respect to $\mathbf{x}$ and moves outside the integral. The remaining integral over $\mathbb{R}^n$ covers the kernel of a Gaussian in $\mathbf{X}$, which evaluates to $\sqrt{(2\pi)^n |\mathbf{S}_{XX}|}$.


Using the determinant identity $|\boldsymbol{\Sigma}| = |\boldsymbol{\Sigma}_{YY}| \cdot |\mathbf{S}_{XX}|$, the constants cancel out just as before, leaving:

$$f_{\mathbf{Y}}(\mathbf{y}) = \frac{1}{\sqrt{(2\pi)^m |\boldsymbol{\Sigma}_{YY}|}} \exp\left( -\frac{1}{2} (\mathbf{y} - \boldsymbol{\mu}_y)^T \boldsymbol{\Sigma}_{YY}^{-1} (\mathbf{y} - \boldsymbol{\mu}_y) \right)$$



# 3. Conditional Density

To derive the conditional density $f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y})$, we use the definition of conditional probability for densities:

$$f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y}) = \frac{f_{\mathbf{X}, \mathbf{Y}}(\mathbf{x}, \mathbf{y})}{f_{\mathbf{Y}}(\mathbf{y})}$$

Since we have already decomposed the joint density and the marginal density of $\mathbf{Y}$ in the previous steps, this becomes a task of algebraic subtraction in the exponent.

---

From our previous derivation, we factored the joint quadratic form $Q$ using the Schur complement of $\boldsymbol{\Sigma}_{YY}$:
$$Q_{joint} = (\mathbf{y} - \boldsymbol{\mu}_y)^T \boldsymbol{\Sigma}_{YY}^{-1} (\mathbf{y} - \boldsymbol{\mu}_y) + (\mathbf{x} - \boldsymbol{\mu}^*)^T \mathbf{S}_{XX}^{-1} (\mathbf{x} - \boldsymbol{\mu}^*)$$

Where:
* $\mathbf{S}_{XX} = \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}$
* $\boldsymbol{\mu}^* = \boldsymbol{\mu}_x + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y} - \boldsymbol{\mu}_y)$

The marginal density $f_{\mathbf{Y}}(\mathbf{y})$ has the exponent:
$$Q_{marginal} = (\mathbf{y} - \boldsymbol{\mu}_y)^T \boldsymbol{\Sigma}_{YY}^{-1} (\mathbf{y} - \boldsymbol{\mu}_y)$$

When we divide the densities, we subtract the exponents ($e^{A} / e^{B} = e^{A-B}$):
$$Q_{conditional} = Q_{joint} - Q_{marginal} = (\mathbf{x} - \boldsymbol{\mu}^*)^T \mathbf{S}_{XX}^{-1} (\mathbf{x} - \boldsymbol{\mu}^*)$$

---

Now we look at the coefficients.
* **Joint constant:** $\frac{1}{\sqrt{(2\pi)^{n+m} |\boldsymbol{\Sigma}|}}$
* **Marginal constant:** $\frac{1}{\sqrt{(2\pi)^m |\boldsymbol{\Sigma}_{YY}|}}$

Dividing the joint constant by the marginal constant gives:
$$\frac{1}{\sqrt{(2\pi)^{n+m} |\boldsymbol{\Sigma}|}} \cdot \sqrt{(2\pi)^m |\boldsymbol{\Sigma}_{YY}|} = \frac{1}{\sqrt{(2\pi)^n \frac{|\boldsymbol{\Sigma}|}{|\boldsymbol{\Sigma}_{YY}|}}}$$

Using the determinant identity $|\boldsymbol{\Sigma}| = |\boldsymbol{\Sigma}_{YY}| \cdot |\mathbf{S}_{XX}|$, the term $\frac{|\boldsymbol{\Sigma}|}{|\boldsymbol{\Sigma}_{YY}|}$ simplifies exactly to $|\mathbf{S}_{XX}|$.

---

Combining the simplified constant and the remaining quadratic form, we get:

$$f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y}) = \frac{1}{\sqrt{(2\pi)^n |\mathbf{S}_{XX}|}} \exp\left( -\frac{1}{2} (\mathbf{x} - \boldsymbol{\mu}^*)^T \mathbf{S}_{XX}^{-1} (\mathbf{x} - \boldsymbol{\mu}^*) \right)$$

This is the density of a multivariate normal distribution:
$$\mathbf{X} \mid \mathbf{Y} = \mathbf{y} \sim N(\boldsymbol{\mu}_x + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y} - \boldsymbol{\mu}_y), \quad \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX})$$

---

**Summary of the Parameters**
* **Conditional Mean:** $\boldsymbol{\mu}_{X|Y} = \boldsymbol{\mu}_x + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y} - \boldsymbol{\mu}_y)$
    * *Insight:* The mean is adjusted by how much $\mathbf{y}$ deviates from its expected value, scaled by the "regression" coefficient $\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}$.
* **Conditional Covariance:** $\boldsymbol{\Sigma}_{X|Y} = \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}$
    * *Insight:* Notice this does **not** depend on the specific value of $\mathbf{y}$. The variance is always reduced (or stays the same) compared to the marginal $\boldsymbol{\Sigma}_{XX}$.

Does the fact that the conditional variance is independent of the observed value $y$ spark any ideas regarding its application in linear regression or Kalman filters?

# 4. Numerical Demonstration

```python
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# 1. Define Parameters
mu = np.array([0, 0])
# Positive correlation (Sigma_xy = 0.7)
sigma = np.array([[1.0, 0.7],
                  [0.7, 1.0]])

# Generate Samples
n_samples = 5000
data = np.random.multivariate_normal(mu, sigma, n_samples)
x_samples, y_samples = data[:, 0], data[:, 1]

# 2. Calculate Theoretical Conditional Distribution P(X | Y = y_fixed)
y_fixed = 1.2
# Formulas derived earlier:
# mu_cond = mu_x + (sigma_xy / sigma_yy) * (y_fixed - mu_y)
# var_cond = sigma_xx - (sigma_xy**2 / sigma_yy)
mu_cond = mu[0] + (sigma[0, 1] / sigma[1, 1]) * (y_fixed - mu[1])
var_cond = sigma[0, 0] - (sigma[0, 1]**2 / sigma[1, 1])

# Generate conditional samples (filtering the joint data for points where Y is near y_fixed)
condition_mask = np.abs(y_samples - y_fixed) < 0.1
x_conditional_samples = x_samples[condition_mask]

# 3. Create Subplots
fig = make_subplots(
    rows=2, cols=2,
    subplot_titles=(
        "Joint Density P(X, Y)", "Marginal P(Y)",
        "Marginal P(X)", f"Conditional P(X | Y={y_fixed})"
    ),
    vertical_spacing=0.15,
    horizontal_spacing=0.1
)

# --- Joint Plot ---
fig.add_trace(go.Histogram2dContour(x=x_samples, y=y_samples, colorscale='Viridis', name="Joint"), row=1, col=1)
fig.add_hline(y=y_fixed, line_dash="dash", line_color="red", annotation_text="Slice for P(X|Y)", row=1, col=1)

# --- Marginal Y ---
fig.add_trace(go.Histogram(y=y_samples, name="Marginal Y", marker_color='blue', opacity=0.6), row=1, col=2)

# --- Marginal X ---
fig.add_trace(go.Histogram(x=x_samples, name="Marginal X", marker_color='green', opacity=0.6), row=2, col=1)

# --- Conditional X | Y ---
fig.add_trace(go.Histogram(x=x_conditional_samples, name="Cond X|Y", marker_color='red', opacity=0.7), row=2, col=2)

# Update Layout
fig.update_layout(
    height=800, width=900,
    title_text="Multivariate Normal: Joint, Marginal, and Conditional",
    showlegend=False
)

fig.show()
```

* **The Joint Plot:** Notice the "tilt" of the ellipse. This represents the correlation $\boldsymbol{\Sigma}_{XY}$. The red dashed line represents the value we are "conditioning" on.
* **The Marginals:** These are the bell curves $N(\mu_x, \sigma_{xx})$ and $N(\mu_y, \sigma_{yy})$ that you get when you "squash" the joint cloud onto one axis.
* **The Conditional:** Look at how the red histogram is **centered differently** than the green marginal histogram. Because $X$ and $Y$ are positively correlated, observing a high $Y$ ($1.2$) shifts the expected value of $X$ to the right.
* **Precision:** Notice the conditional histogram is "thinner" (lower variance) than the marginal $X$ histogram, confirming that $\boldsymbol{\Sigma}_{X|Y} < \boldsymbol{\Sigma}_{XX}$.
