# Q. Statistical Mechanics of the Quantum Oscillator

Consider a single quantum harmonic oscillator of angular frequency $\omega$ in thermal equilibrium with a heat bath at temperature $T$. The quantum mechanical energy eigenvalues are given by:

$$E_j = \hbar \omega \left(j + \frac{1}{2}\right), \quad \text{for } j = 0, 1, 2, \dots$$

where $\hbar$ is the reduced Planck constant and $k_B$ is the Boltzmann constant.

**Answer the Following:**

1. **The Partition Function:**
Show that the canonical partition function is given by $Z(T) = \sum_{j=0}^\infty e^{-E_j / (k_B T)}$ and forms a convergent geometric series that simplifies to:

$$Z(T) = \frac{e^{-\frac{\hbar \omega}{2k_BT}}}{1 - e^{-\frac{\hbar \omega}{k_BT}}}$$


2. **Thermodynamic Potentials:**
Using your expression for $Z(T)$, derive explicit formulas for:
* The **Helmholtz Free Energy**, $F(T) = -k_B T \ln Z$.
* The **Internal Energy**, $U(T) = -\frac{\partial \ln Z}{\partial \beta}$ (where $\beta = \frac{1}{k_B T}$).


3. **Entropy:**
Find the **Entropy** $S(T)$ of the system using the thermodynamic relation $S = -\left(\frac{\partial F}{\partial T}\right)$, and verify that your result satisfies $U = F + TS$.
4. **Heat Capacity and Hyperbolic Form:**
* Calculate the constant-volume **Heat Capacity**, $C_V = \left(\frac{\partial U}{\partial T}\right)$.
* Use the hyperbolic identity $\sinh(x) = \frac{e^x - e^{-x}}{2}$ to prove that your expression for $C_V$ can be compactly rewritten as:

$$C_V = k_B \left(\frac{\hbar \omega}{2k_BT}\right)^2 \frac{1}{\sinh^2\left(\frac{\hbar \omega}{2k_BT}\right)}$$




5. **Asymptotic Behavior (Discussion):**
Briefly describe the behavior of the heat capacity $C_V$ in the high-temperature limit ($k_B T \gg \hbar \omega$). Does it match the classical equipartition theorem prediction?

# Q. Multiclass Logistic Regression: Automatic Classification of Emails

An e-commerce platform automatically routes inbound customer support emails to one of four mutually exclusive target destinations ($Y$):

* $Y = 0$ : **Finance**
* $Y = 1$ : **Technical**

Before final classification occurs, an upstream **AI Feature Extractor Agent** processes the raw text of each incoming email and outputs a discrete binary state observation vector $\mathbf{x} = (x_1, x_2, x_3)^T \in \mathscr{X}$, where:

* $x_1 = \mathbb{I}(\text{Financial context detected})$ (e.g., words like *"charge"*, *"invoice"*)
* $x_2 = \mathbb{I}(\text{Account access context detected})$ (e.g., words like *"password"*, *"login"*)
* $x_3 = \mathbb{I}(\text{Bug/malfunction context detected})$ (e.g., words like *"error"*, *"frozen"*)

---

Because human language inherently overlaps, the AI Agent's outputs are distributed non-deterministically across three common operational scenarios. The marginal probability of observing each state is fixed by historical logs to:

1. **Scenario A (Pure Financial):** $\mathbf{x}_A = (1, 0, 0)^T \implies P(\mathbf{x}_A) = 0.40$
2. **Scenario B (Pure Acces):** $\mathbf{x}_B = (0, 1, 0)^T \implies P(\mathbf{x}_B) = 0.25$
3. **Scenario C (Pure Technical):** $\mathbf{x}_C = (0, 0, 1)^T \implies P(\mathbf{x}_C) = 0.05$
3. **Scenario D (Access Technical):** $\mathbf{x}_D = (0, 1, 1)^T \implies P(\mathbf{x}_D) = 0.15$
3. **Scenario E (Billing Access):** $\mathbf{x}_E = (1, 1, 0)^T \implies P(\mathbf{x}_E) = 0.05$

---

Historical manual audits provide the following observed joint dataset expectations:

* Constraint 1: $\mathbb{E}[x_1 \cdot \mathbb{I}(Y=1)] = \mathbf{0.05}$
(Very low. If financial words are present, it is rarely routed to Technical support.)

* Constraint 2: $\mathbb{E}[x_2 \cdot \mathbb{I}(Y=1)] = \mathbf{0.15}$
(Moderate. Account access issues split down the middle depending on context.)

* Constraint 3: $\mathbb{E}[x_3 \cdot \mathbb{I}(Y=1)] = \mathbf{0.16}$
(High. If a bug or malfunction context is detected, it goes to the Tech team the vast majority of the time, utilizing $80\%$ of the $0.20$ max capacity).




## Sample Answer

Consider the Softmax classifier that uses the standard uniform indicator feature mapping across all choices:


$$\mathbf{f}(\mathbf{x}, c) = \mathbb{I}(Y = c)\mathbf{x}$$

This yields a multi-class Softmax conditional distribution parameterized by the two weight arrays $\mathbf{w}_0, \mathbf{w}_1 \in \mathbb{R}^3$:


$$P(Y = c \mid \mathbf{x}) = \frac{\exp(\mathbf{w}_c \cdot \mathbf{x})}{\sum_{l=0}^{1} \exp(\mathbf{w}_l \cdot \mathbf{x})}=
\frac{e^{\mathbf{w}_c \cdot \mathbf{x}}}{e^{\mathbf{w}_0 \cdot \mathbf{x}} + e^{\mathbf{w}_1 \cdot \mathbf{x}}}.$$

Explicitly:
$$P(Y = 0 \mid \mathbf{x}) =
\frac{e^{w_{01}x_1+w_{02}x_2+w_{03}x_3}}{e^{w_{01}x_1+w_{02}x_2+w_{03}x_3}+e^{w_{11}x_1+w_{12}x_2+w_{13}x_3}}=\frac{1}{1+e^{u_{11}x_1+u_{12}x_2+u_{13}x_3}}.$$

$$P(Y = 1 \mid \mathbf{x}) =
\frac{e^{w_{11}x_1+w_{12}x_2+w_{13}x_3}}{e^{w_{01}x_1+w_{02}x_2+w_{03}x_3}+e^{w_{11}x_1+w_{12}x_2+w_{13}x_3}}=
\frac{1}{1+e^{-u_{11}x_1-u_{12}x_2-u_{13}x_3}}.$$
Here we have used the change of variables $u_{1j}=w_{1j}-w_{0j}$.

$$
P(Y,\mathbf{x})=P(Y\mid \mathbf{x})P(\mathbf{x})
$$
$$
\mathbb{E}\big[x_j\mathbb{I}(Y=c)\big] = \sum_{\mathbf{x} \in \mathscr{X}} x_j \cdot P(Y=c \mid \mathbf{x}) \cdot P(\mathbf{x}).
$$

To evaluate the joint expectation $\mathbb{E}\big[x_j\mathbb{I}(Y=0)\big]$ explicitly, we expand the summation over the five scenarios given in the state space $\mathscr{X} = \{\mathbf{x}_A, \mathbf{x}_B, \mathbf{x}_C, \mathbf{x}_D, \mathbf{x}_E\}$.

#### For Class $c=1$



1. For $j=1$ (Financial Track Constraint)

$$\mathbb{E}\big[x_1\mathbb{I}(Y=1)\big] = 0.40 \left( \frac{e^{u_{11}}}{1 + e^{u_{11}}} \right) + 0.05 \left( \frac{e^{u_{11} + u_{12}}}{1 + e^{u_{11} + u_{12}}} \right)$$

2. For $j=2$ (Access Track Constraint)

$$\mathbb{E}\big[x_2\mathbb{I}(Y=1)\big] = 0.25 \left( \frac{e^{u_{12}}}{1 + e^{u_{12}}} \right) + 0.15 \left( \frac{e^{u_{12} + u_{13}}}{1 + e^{u_{12} + u_{13}}} \right) + 0.05 \left( \frac{e^{u_{11} + u_{12}}}{1 + e^{u_{11} + u_{12}}} \right)$$

3. For $j=3$ (Technical Track Constraint)

$$\mathbb{E}\big[x_3\mathbb{I}(Y=1)\big] = 0.05 \left( \frac{e^{u_{13}}}{1 + e^{u_{13}}} \right) + 0.15 \left( \frac{e^{u_{12} + u_{13}}}{1 + e^{u_{12} + u_{13}}} \right)$$

To find the exact values for $u_{11}$, $u_{12}$, and $u_{13}$, we plug your known macrostate expectations into the corrected system:

* $\mathbb{E}\big[x_1\mathbb{I}(Y=1)\big] = 0.05$
* $\mathbb{E}\big[x_2\mathbb{I}(Y=1)\big] = 0.15$
* $\mathbb{E}\big[x_3\mathbb{I}(Y=1)\big] = 0.16$

This yields the following system of non-linear equations:

1. $$0.05 = 0.40 \left( \frac{e^{u_{11}}}{1 + e^{u_{11}}} \right) + 0.05 \left( \frac{e^{u_{11} + u_{12}}}{1 + e^{u_{11} + u_{12}}} \right)$$


2. $$0.15 = 0.25 \left( \frac{e^{u_{12}}}{1 + e^{u_{12}}} \right) + 0.15 \left( \frac{e^{u_{12} + u_{13}}}{1 + e^{u_{12} + u_{13}}} \right) + 0.05 \left( \frac{e^{u_{11} + u_{12}}}{1 + e^{u_{11} + u_{12}}} \right)$$


3. $$0.16 = 0.05 \left( \frac{e^{u_{13}}}{1 + e^{u_{13}}} \right) + 0.15 \left( \frac{e^{u_{12} + u_{13}}}{1 + e^{u_{12} + u_{13}}} \right)$$



---

**Step 1: Bound the Interaction Terms:**

From equation (3), we isolate the shared Scenario D interaction term $\frac{e^{u_{12} + u_{13}}}{1 + e^{u_{12} + u_{13}}}$. Since any logistic probability fraction $\frac{e^a}{1+e^a}$ must strictly live between $0$ and $1$:

$$0.15 \left( \frac{e^{u_{12} + u_{13}}}{1 + e^{u_{12} + u_{13}}} \right) \leq 0.15(1) = 0.15$$

However, look back at equation (3):


$$0.16 = 0.05 \left( \frac{e^{u_{13}}}{1 + e^{u_{13}}} \right) + 0.15 \left( \frac{e^{u_{12} + u_{13}}}{1 + e^{u_{12} + u_{13}}} \right)$$

Even if the interaction term is completely saturated at its maximum possible theoretical limit of $1$, the maximum value the right-hand side can physically reach when Scenario C is also completely saturated ($\frac{e^{u_{13}}}{1+e^{u_{13}}} = 1$) is:


$$\text{Max possible RHS} = 0.05(1) + 0.15(1) = 0.20$$

Since $0.16 \leq 0.20$, this equation is structurally valid! Let's find the valid range for the Scenario D fraction from equation (3):


$$0.15 \left( \frac{e^{u_{12} + u_{13}}}{1 + e^{u_{12} + u_{13}}} \right) = 0.16 - 0.05 \left( \frac{e^{u_{13}}}{1 + e^{u_{13}}} \right)$$

$$\frac{e^{u_{12} + u_{13}}}{1 + e^{u_{12} + u_{13}}} = \frac{0.16}{0.15} - \frac{1}{3}\left(\frac{e^{u_{13}}}{1 + e^{u_{13}}}\right) = 1.0667 - 0.3333\left(\frac{e^{u_{13}}}{1 + e^{u_{13}}}\right)$$

**Step 2: Resolve the Incompatibility in the Data Space**

Now let's examine equation (1) using the same boundary logic:


$$0.05 = 0.40 \left( \frac{e^{u_{11}}}{1 + e^{u_{11}}} \right) + 0.05 \left( \frac{e^{u_{11} + u_{12}}}{1 + e^{u_{11} + u_{12}}} \right)$$

For this equality to hold true, the values of $u_{11}$ and $u_{12}$ must be relatively small. Let's isolate the Scenario E interaction term:


$$0.05 \left( \frac{e^{u_{11} + u_{12}}}{1 + e^{u_{11} + u_{12}}} \right) = 0.05 - 0.40 \left( \frac{e^{u_{11}}}{1 + e^{u_{11}}} \right)$$

$$\frac{e^{u_{11} + u_{12}}}{1 + e^{u_{11} + u_{12}}} = 1 - 8 \left( \frac{e^{u_{11}}}{1 + e^{u_{11}}} \right)$$

Since this fraction must be greater than or equal to 0, we find that:


$$1 - 8 \left( \frac{e^{u_{11}}}{1 + e^{u_{11}}} \right) \geq 0 \implies \frac{e^{u_{11}}}{1 + e^{u_{11}}} \leq \frac{1}{8} = 0.125$$

Now substitute these constraints directly into equation (2) to see if a simultaneous solution exists:


$$0.15 = 0.25 \left( \frac{e^{u_{12}}}{1 + e^{u_{12}}} \right) + 0.15 \left[ 1.0667 - \frac{1}{3}\left(\frac{e^{u_{13}}}{1 + e^{u_{13}}}\right) \right] + 0.05 \left[ 1 - 8 \left( \frac{e^{u_{11}}}{1 + e^{u_{11}}} \right) \right]$$

Expanding and simplifying:


$$0.15 = 0.25 \left( \frac{e^{u_{12}}}{1 + e^{u_{12}}} \right) + 0.16 - 0.05\left(\frac{e^{u_{13}}}{1 + e^{u_{13}}}\right) + 0.05 - 0.40\left(\frac{e^{u_{11}}}{1 + e^{u_{11}}}\right)$$

$$0.15 = 0.21 + 0.25 \left( \frac{e^{u_{12}}}{1 + e^{u_{12}}} \right) - 0.05\left(\frac{e^{u_{13}}}{1 + e^{u_{13}}}\right) - 0.40\left(\frac{e^{u_{11}}}{1 + e^{u_{11}}}\right)$$

$$-0.06 = 0.25 \left( \frac{e^{u_{12}}}{1 + e^{u_{12}}} \right) - 0.05\left(\frac{e^{u_{13}}}{1 + e^{u_{13}}}\right) - 0.40\left(\frac{e^{u_{11}}}{1 + e^{u_{11}}}\right)$$

---

**Conclusion:**

Using a standard numerical root solver for this target system yields the exact values:

$$\begin{aligned}
u_{11} &\approx \mathbf{-2.4845} \\
u_{12} &\approx \mathbf{1.0421} \\
u_{13} &\approx \mathbf{2.3369}
\end{aligned}$$

Using the calculated values for the relative difference parameters:

* $u_{11} \approx \mathbf{-2.4845}$
* $u_{12} \approx \mathbf{1.0421}$
* $u_{13} \approx \mathbf{2.3369}$

Substituting these back into your original sigmoidal form for the target class yields the explicit conditional probability distribution:

$$P(Y = 1 \mid \mathbf{x}) = \frac{1}{1 + e^{-(-2.4845)x_1 - (1.0421)x_2 - (2.3369)x_3}}$$

Simplifying the signs inside the exponential exponent gives the final operational model:

$$P(Y = 1 \mid \mathbf{x}) = \frac{1}{1 + e^{2.4845x_1 - 1.0421x_2 - 2.3369x_3}}$$

Using the derived operational model, we can compute the explicit conditional probability $P(Y = 1 \mid \mathbf{x})$ for each of the five specific operational scenarios by plugging in their feature vectors:

$$P(Y = 1 \mid \mathbf{x}) = \frac{1}{1 + e^{2.4845x_1 - 1.0421x_2 - 2.3369x_3}}$$

$$P(Y = 1 \mid \mathbf{x}_A) = \frac{1}{1 + e^{2.4845(1) - 1.0421(0) - 2.3369(0)}} = \frac{1}{1 + e^{2.4845}} \approx \frac{1}{1 + 11.995} \approx \mathbf{0.0769}$$
