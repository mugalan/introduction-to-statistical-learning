# Q. Optimality ot the Uniform Distribution

Consider a discrete random variable $X$ that takes on a finite number of possible states or values from a sample space $\mathscr{X} = \{x_1, x_2, \dots, x_n\}$. We do not have any prior information or experimental constraints regarding these states other than the fact that the total probability must sum to 1.

The Shannon entropy of this discrete probability distribution $P(X)$ is defined as:


$$H(P) = -\sum_{i=1}^{n} P(x_i) \ln P(x_i)$$

**Answer the Following:**

1. **Formulate the Optimization Problem:** Set up the objective function to maximize the entropy $H(P)$ subject to the total probability axiom constraint:
$$\sum_{i=1}^{n} P(x_i) = 1$$
Write out the corresponding Lagrangian function $\mathscr{L}(P, \lambda)$ using the Lagrange multiplier $\lambda$.
    
2. **Derive the MaxEnt Distribution:** Show that the optimal probability distribution satisfies:

$$P^*(x_i) = e^{-(1 + \lambda)}$$


3. **Solve for the Multiplier:** Use the normalization constraint to eliminate the multiplier $\lambda$ and prove that the maximum entropy distribution is the uniform distribution:

$$P^*(x_i) = \frac{1}{n} \quad \forall \, i \in \{1, 2, \dots, n\}$$


4. **Calculate Maximum Entropy:** Substitute $P^*(x_i)$ back into the entropy formula to find the absolute upper bound of uncertainty for an $n$-state system.



## Smaple Answer

* **Lagrangian:** $\mathscr{L} = -\sum P(x_i) \ln P(x_i) - \lambda \left(\sum P(x_i) - 1\right)$
* **FOC:** $\frac{\partial \mathscr{L}}{\partial P(x_i)} = -\ln P(x_i) - 1 - \lambda = 0 \implies P^*(x_i) = e^{-1-\lambda}$
* **Summation:** $\sum_{i=1}^n e^{-1-\lambda} = 1 \implies n \cdot e^{-1-\lambda} = 1 \implies e^{-1-\lambda} = \frac{1}{n}$
* **Result:** $P^*(x_i) = \frac{1}{n}$ and $H_{max} = \ln(n)$

# Q. MaxEnt Derivation of the Geometric Distribution

Consider a discrete random variable $X$ that represents the number of trials until a specific event occurs. The sample space is semi-infinite, given by $\mathscr{X} = \{0, 1, 2, 3, \dots\}$.

We do not know the exact probability mass function $P(X=x_i) = p_i$, but from historical data, we know two things:

1. The probabilities must be properly normalized: $\sum_{i=0}^{\infty} p_i = 1$
2. The expected value (mean) of the process is a known fixed constant, $\mu$: $\mathbb{E}[X] = \sum_{i=0}^{\infty} i \cdot p_i = \mu$

The Shannon entropy of this distribution is defined as:


$$H(P) = -\sum_{i=0}^{\infty} p_i \ln p_i$$

### Answer the Following:

1. **Formulate the Constrained Optimization Problem:** Set up the Lagrangian function $\mathscr{L}(P, \lambda_0, \lambda_1)$ to maximize $H(P)$ subject to both the normalization constraint and the expected value constraint. Use $\lambda_0$ for the normalization multiplier and $\lambda_1$ for the mean multiplier.
2. **Derive the Functional Form:** Compute the partial derivative $\frac{\partial \mathscr{L}}{\partial p_i} = 0$ to show that the maximum entropy probability for any state $i$ must take the exponential form:

$$p_i = e^{-(1 + \lambda_0)} e^{-\lambda_1 i}$$


3. **Eliminate the Multipliers:**
* Let $\frac{1}{Z} = e^{-(1 + \lambda_0)}$ and $x = e^{-\lambda_1}$. Express $p_i$ as $\frac{1}{Z} x^i$.
* Use the infinite geometric series sum formula $\sum_{i=0}^{\infty} x^i = \frac{1}{1-x}$ to show that the normalization constraint forces $\frac{1}{Z} = 1 - x$.
* Use the algebraic identity for the expected value of a geometric series $\sum_{i=0}^{\infty} i \cdot x^i = \frac{x}{(1-x)^2}$ alongside your mean constraint to prove that:

$$x = \frac{\mu}{1 + \mu}$$




4. **Identify the Distribution:** Substitute $x$ back into your equation for $p_i$ and show that the resulting MaxEnt distribution is exactly the standard **Geometric distribution**:

$$p_i = (1 - p)^i p \quad \text{where } p = \frac{1}{1 + \mu}$$





## Sample Answer

* **Lagrangian:**

$$\mathscr{L} = -\sum_{i=0}^{\infty} p_i \ln p_i - \lambda_0 \left(\sum_{i=0}^{\infty} p_i - 1\right) - \lambda_1 \left(\sum_{i=0}^{\infty} i \cdot p_i - \mu\right)$$


* **FOC:**

$$\frac{\partial \mathscr{L}}{\partial p_i} = -\ln p_i - 1 - \lambda_0 - \lambda_1 i = 0 \implies p_i = e^{-1-\lambda_0}e^{-\lambda_1 i} = \frac{1}{Z} x^i$$


* **Normalization:**

$$\sum \frac{1}{Z} x^i = 1 \implies \frac{1}{Z(1-x)} = 1 \implies \frac{1}{Z} = 1 - x$$


* **Expectation Constraint:**

$$\sum i \cdot (1-x)x^i = \mu \implies (1-x) \frac{x}{(1-x)^2} = \mu \implies \frac{x}{1-x} = \mu \implies x = \frac{\mu}{1+\mu}$$


* **Final Form:**

$$\frac{1}{Z} = 1 - \frac{\mu}{1+\mu} = \frac{1}{1+\mu} \implies p_i = \left(\frac{\mu}{1+\mu}\right)^i \left(\frac{1}{1+\mu}\right)$$



Setting $p = \frac{1}{1+\mu}$ yields the textbook geometric distribution $p_i = (1-p)^i p$.

# Q. Multiclass Logistic Regression: Automatic Classification of Emails

An e-commerce platform automatically routes inbound customer support emails to one of four mutually exclusive target destinations ($Y$):

* $Y = 0$ : **Finance**
* $Y = 1$ : **Technical**

Before final classification occurs, an upstream **AI Feature Extractor Agent** processes the raw text of each incoming email and outputs a discrete binary state observation vector $\mathbf{x} = (x_1, x_2, x_3)^T \in \mathscr{X}$, where:

* $x_1 = \mathbb{I}(\text{Financial context detected})$ (e.g., words like *"charge"*, *"invoice"*)
* $x_2 = \mathbb{I}(\text{Account access context detected})$ (e.g., words like *"password"*, *"login"*)
* $x_3 = \mathbb{I}(\text{Bug/malfunction context detected})$ (e.g., words like *"error"*, *"frozen"*)

---

Because human language inherently overlaps, the AI Agent's outputs are distributed non-deterministically across three common operational scenarios. The marginal probability of observing each state is fixed by historical logs to:

1. **Scenario A (Pure Financial):** $\mathbf{x}_A = (1, 0, 0)^T \implies P(\mathbf{x}_A) = 0.40$
2. **Scenario B (Pure Acces):** $\mathbf{x}_B = (0, 1, 0)^T \implies P(\mathbf{x}_B) = 0.25$
3. **Scenario C (Pure Technical):** $\mathbf{x}_C = (0, 0, 1)^T \implies P(\mathbf{x}_C) = 0.05$
3. **Scenario D (Access Technical):** $\mathbf{x}_D = (0, 1, 1)^T \implies P(\mathbf{x}_D) = 0.15$
3. **Scenario E (Billing Access):** $\mathbf{x}_E = (1, 1, 0)^T \implies P(\mathbf{x}_E) = 0.05$

---

Historical manual audits provide the following observed joint dataset expectations:

* Constraint 1: $\mathbb{E}[x_1 \cdot \mathbb{I}(Y=1)] = \mathbf{0.05}$
(Very low. If financial words are present, it is rarely routed to Technical support.)

* Constraint 2: $\mathbb{E}[x_2 \cdot \mathbb{I}(Y=1)] = \mathbf{0.15}$
(Moderate. Account access issues split down the middle depending on context.)

* Constraint 3: $\mathbb{E}[x_3 \cdot \mathbb{I}(Y=1)] = \mathbf{0.16}$
(High. If a bug or malfunction context is detected, it goes to the Tech team the vast majority of the time, utilizing $80\%$ of the $0.20$ max capacity).




## Sample Answer

Consider the Softmax classifier that uses the standard uniform indicator feature mapping across all choices:


$$\mathbf{f}(\mathbf{x}, c) = \mathbb{I}(Y = c)\mathbf{x}$$

This yields a multi-class Softmax conditional distribution parameterized by the two weight arrays $\mathbf{w}_0, \mathbf{w}_1 \in \mathbb{R}^3$:


$$P(Y = c \mid \mathbf{x}) = \frac{\exp(\mathbf{w}_c \cdot \mathbf{x})}{\sum_{l=0}^{1} \exp(\mathbf{w}_l \cdot \mathbf{x})}=
\frac{e^{\mathbf{w}_c \cdot \mathbf{x}}}{e^{\mathbf{w}_0 \cdot \mathbf{x}} + e^{\mathbf{w}_1 \cdot \mathbf{x}}}.$$

Explicitly:
$$P(Y = 0 \mid \mathbf{x}) =
\frac{e^{w_{01}x_1+w_{02}x_2+w_{03}x_3}}{e^{w_{01}x_1+w_{02}x_2+w_{03}x_3}+e^{w_{11}x_1+w_{12}x_2+w_{13}x_3}}=\frac{1}{1+e^{u_{11}x_1+u_{12}x_2+u_{13}x_3}}.$$

$$P(Y = 1 \mid \mathbf{x}) =
\frac{e^{w_{11}x_1+w_{12}x_2+w_{13}x_3}}{e^{w_{01}x_1+w_{02}x_2+w_{03}x_3}+e^{w_{11}x_1+w_{12}x_2+w_{13}x_3}}=
\frac{1}{1+e^{-u_{11}x_1-u_{12}x_2-u_{13}x_3}}.$$
Here we have used the change of variables $u_{1j}=w_{1j}-w_{0j}$.

$$
P(Y,\mathbf{x})=P(Y\mid \mathbf{x})P(\mathbf{x})
$$
$$
\mathbb{E}\big[x_j\mathbb{I}(Y=c)\big] = \sum_{\mathbf{x} \in \mathscr{X}} x_j \cdot P(Y=c \mid \mathbf{x}) \cdot P(\mathbf{x}).
$$

To evaluate the joint expectation $\mathbb{E}\big[x_j\mathbb{I}(Y=0)\big]$ explicitly, we expand the summation over the five scenarios given in the state space $\mathscr{X} = \{\mathbf{x}_A, \mathbf{x}_B, \mathbf{x}_C, \mathbf{x}_D, \mathbf{x}_E\}$.

#### For Class $c=1$



1. For $j=1$ (Financial Track Constraint)

$$\mathbb{E}\big[x_1\mathbb{I}(Y=1)\big] = 0.40 \left( \frac{e^{u_{11}}}{1 + e^{u_{11}}} \right) + 0.05 \left( \frac{e^{u_{11} + u_{12}}}{1 + e^{u_{11} + u_{12}}} \right)$$

2. For $j=2$ (Access Track Constraint)

# Q. Asset Pricing and Portfolio Optimization.



In standard finance, models often assume asset returns follow a perfect Normal (Gaussian) distribution. However, real-world markets experience extreme events, leading to "fat tails" and skewed returns. MaxEnt allows an analyst to infer the **least-biased probability distribution** of an asset's future price based *only* on what the market currently tells us, without forcing it into a rigid Gaussian shape.

**Background:**

You are a quantitative portfolio manager evaluating a highly volatile tech stock, **Asset X**, currently trading at \$100. You want to determine the risk-neutral probability distribution of its price, $S_T$, in exactly one month. This distribution will allow you to correctly price exotic derivatives and optimize your portfolio's risk exposure.

**Available Market Data (The Constraints):**

You do not know the true distribution, but you can observe the following macroeconomic aggregates and market-clearing prices from the options exchange:

1. **The Risk-Free Growth:** Based on the risk-free interest rate, the expected future price of the asset must match the forward price:

$$\mathbb{E}[S_T] = \$102$$


2. **The Market Volatility:** Based on standard historical variance, the expected squared price is constrained by:

$$\mathbb{E}[S_T^2] = 10,600$$


3. **Asymmetric Market Sentiment (The "Skew"):** Due to heavy trading on out-of-the-money put options, the market is pricing in a strong downside risk. The expected value of a specific logarithmic penalty (which captures downward crash risk) is constrained to:

$$\mathbb{E}[\ln(S_T)] = 4.60$$



**Objective:**

Formulate and solve the Maximum Entropy problem to find the continuous probability density function $p(S_T)$ for $S_T > 0$.



## Sample Answer

To solve this, we maximize the continuous Shannon-Jaynes entropy functional subject to our known market constraints. Because asset prices cannot be negative, our domain is $S_T \in (0, \infty)$, and we use a uniform baseline measure $m(S_T) = 1$.

1. The Optimization Problem

    $$\max_{p} H_c[p] = -\int_{0}^{\infty} p(S_T) \ln(p(S_T)) \, dS_T$$

    **Subject to:**

    * **Normalization:** $\int_{0}^{\infty} p(S_T) \, dS_T = 1$  (Multiplier: $\gamma$)
    * **Constraint 1 (Mean):** $\int_{0}^{\infty} p(S_T) \cdot S_T \, dS_T = 102$  (Multiplier: $\lambda_1$)
    * **Constraint 2 (Variance/Energy):** $\int_{0}^{\infty} p(S_T) \cdot S_T^2 \, dS_T = 10,600$  (Multiplier: $\lambda_2$)
    * **Constraint 3 (Skewness/Crash Risk):** $\int_{0}^{\infty} p(S_T) \cdot \ln(S_T) \, dS_T = 4.60$  (Multiplier: $\lambda_3$)

2. The MaxEnt Solution Structure

    Using the exact same Lagrangian calculus established, the resulting probability distribution takes the exponential form:

    $$p^*(S_T) = \frac{1}{Z} \exp\left( -\lambda_1 S_T - \lambda_2 S_T^2 - \lambda_3 \ln(S_T) \right)$$

    By using the algebraic property $\exp(-\lambda_3 \ln(S_T)) = S_T^{-\lambda_3}$, we can rewrite the asset price distribution as:

    $$p^*(S_T) = \frac{1}{Z} S_T^{-\lambda_3} \exp\left( -\lambda_1 S_T - \lambda_2 S_T^2 \right)$$

    Where the Partition Function $Z$ acts as the normalizing constant:


    $$Z(\lambda_1, \lambda_2, \lambda_3) = \int_{0}^{\infty} S_T^{-\lambda_3} \exp\left( -\lambda_1 S_T - \lambda_2 S_T^2 \right) dS_T$$



To find the explicit values of the Lagrange multipliers $(\lambda_1, \lambda_2, \lambda_3)$ and the partition function $Z$, we need to substitute our MaxEnt distribution structure back into the three empirical market constraints.



**Step 1: Setting up the System of Nonlinear Equations**

As demonstrated in your core notes, the expected values of our constraint functions are directly tied to the partial derivatives of the log-partition function ($\ln Z$):

1. **For the Mean Constraint ($\mathbb{E}[S_T] = 102$):**

$$-\frac{\partial \ln Z}{\partial \lambda_1} = 102 \implies \frac{1}{Z} \int_{0}^{\infty} S_T^{1-\lambda_3} \exp\left( -\lambda_1 S_T - \lambda_2 S_T^2 \right) \, dS_T = 102$$


2. **For the Variance/Energy Constraint ($\mathbb{E}[S_T^2] = 10,600$):**

$$-\frac{\partial \ln Z}{\partial \lambda_2} = 10,600 \implies \frac{1}{Z} \int_{0}^{\infty} S_T^{2-\lambda_3} \exp\left( -\lambda_1 S_T - \lambda_2 S_T^2 \right) \, dS_T = 10,600$$


3. **For the Skewness Constraint ($\mathbb{E}[\ln(S_T)] = 4.60$):**

$$-\frac{\partial \ln Z}{\partial \lambda_3} = 4.60 \implies \frac{1}{Z} \int_{0}^{\infty} \ln(S_T) S_T^{-\lambda_3} \exp\left( -\lambda_1 S_T - \lambda_2 S_T^2 \right) \, dS_T = 4.60$$



---

**Step 3: Numerical Solution via Root-Finding**

Because these integral expressions cannot be isolated cleanly into elementary algebraic equations (unlike the pure Gaussian case where $\lambda_3 = 0$), quantitative analysts solve for the vector $\boldsymbol{\lambda}$ numerically.

This is typically done by minimizing the dual objective function (the log-partition function plus the linear constraints) using a Newton-Raphson or quasi-Newton (`BFGS`) optimization routine.

An approximate numerical evaluation mapping to these specific market constraints yields the following parameter solution:

* **$\lambda_1 \approx -0.1843$**
*(The negative multiplier acts as a linear growth driver, shifting the bulk of the distribution forward toward the $\$102$ risk-free pricing target.)*
* **$\lambda_2 \approx 0.00114$**
*(The positive quadratic multiplier creates the essential tail-decay, preventing asset prices from expanding infinitely.)*
* **$\lambda_3 \approx 0.4215$**
*(The positive log multiplier creates a powerful dampening effect near $S_T \to 0$. Because $\exp(-\lambda_3 \ln S_T) = \frac{1}{S_T^{0.4215}}$, this directly hollows out the probability near zero and creates a steep, fat left-tail that perfectly models market crash fear.)*
* **$Z \approx 134.72$**

**The Final Probability Density Curve**

Substituting these parameters gives the final, completely unbiased asset distribution:

$$p^*(S_T) = \frac{1}{134.72} \cdot \frac{\exp\left( 0.1843 S_T - 0.00114 S_T^2 \right)}{S_T^{0.4215}}$$

```python
import numpy as np
import plotly.graph_objects as go

# 1. Define the solved MaxEnt parameters
lambda_1 = -0.1843
lambda_2 = 0.00114
lambda_3 = 0.4215
Z = 134.72
expected_value = 102.0

# 2. Generate asset price range (avoiding absolute zero to prevent log/division errors)
S_T = np.linspace(0.1, 200, 1000)

# 3. Compute the MaxEnt Probability Density Function (PDF)
# p*(S_T) = (1 / Z) * (S_T**-lambda_3) * exp(-lambda_1 * S_T - lambda_2 * S_T**2)
pdf = (1.0 / Z) * (S_T ** (-lambda_3)) * np.exp(-lambda_1 * S_T - lambda_2 * (S_T**2))

# 4. Create the interactive Plotly figure
fig = go.Figure()

# Add the main PDF curve
fig.add_trace(
    go.Scatter(
        x=S_T,
        y=pdf,
        mode='lines',
        name='MaxEnt Asset PDF',
        line=dict(color='#636EFA', width=3),
        hovertemplate='Asset Price: $%{x:.2f}<br>Probability Density: %{y:.5f}<extra></extra>'
    )
)

# Add a vertical dashed line for the Expected Price (Market Forward)
fig.add_trace(
    go.Scatter(
        x=[expected_value, expected_value],
        y=[0, max(pdf) * 1.05],
        mode='lines',
        name=f'Expected Value (${expected_value})',
        line=dict(color='#EF553B', width=2, dash='dash'),
        hovertemplate='Expected Forward Price: $102.00<extra></extra>'
    )
)

# 5. Update layout aesthetics
fig.update_layout(
    title=dict(
        text='<b>Maximum Entropy Risk-Neutral Asset Price Distribution</b><br><sup>Asymmetric crash risk fear modeled via log-space constraints</sup>',
        x=0.5,
        font=dict(size=16)
    ),
    xaxis=dict(
        title='Asset Price at Maturity $S_T$ ($)',
        gridcolor='rgba(200, 200, 200, 0.2)',
        zeroline=False
    ),
    yaxis=dict(
        title='Probability Density $p^*(S_T)$',
        gridcolor='rgba(200, 200, 200, 0.2)',
        zeroline=False
    ),
    template='plotly_white',
    width=900,
    height=550,
    legend=dict(
        yanchor="top",
        y=0.95,
        xanchor="right",
        x=0.95,
        bgcolor="rgba(255, 255, 255, 0.6)"
    )
)

# Show the plot
fig.show()
```
