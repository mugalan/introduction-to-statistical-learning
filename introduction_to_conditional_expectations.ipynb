# 1. Definition of the Conditional Expectation $(\mathbb{E}(X|Y))$



Let $(\Omega, \mathfrak{A}, P)$ be a probability space.

* Let $X: \Omega \to \mathbb{R}$ be a random variable that is integrable ($X \in L^1$).
* Let $Y: \Omega \to \mathbb{R}$ be another random variable.
* Let $\sigma(Y)$ be the **sub-$\sigma$-algebra** generated by $Y$. This represents "the information contained in $Y$."

---

**The Existence: Conditional Expectation**

The conditional expectation of $X$ given $Y$, denoted $\mathbb E[X|Y]$ (or more formally $\mathbb E[X|\sigma(Y)]$), is defined as the unique (almost surely) random variable $Z$ that satisfies two conditions:

1. **Measurability:** $Z$ is measurable with respect to $\sigma(Y)$.
That is $Z$ must be a function of $Y$:
$$Z = g(Y).$$
2. **The Integral Property (Partial Averaging):** For every set $G \in \sigma(Y)$:

$$\int_G Z \, dP = \int_G X \, dP$$



*(Translation: On any set of outcomes that $Y$ can distinguish, the "average" of our estimate $Z$ must match the "average" of the actual $X$.)*

The existence and uniqueness of $Z$ are guaranteed by the **Radon-Nikodym Theorem** and is beyond the scope of this course.



# 2 The Conditional Density ($f_{X\mid Y}$)

##2.1 The Joint Probability ($P(X\in B\mid Y\in A)$)

Let
$$
Z_B\triangleq \mathbb E[\mathbb 1_{\{X\in B\}}\mid Y]\triangleq P(X\in B\mid Y).
$$

Since the event

$$
Y^{-1}(A)
$$

belongs to $\sigma(Y)$, the defining property of conditional expectation gives

$$
\int_{Y^{-1}(A)} Z_B\,dP=\int_{Y^{-1}(A)} \mathbb 1_{\{X\in B\}}\,dP.
$$

Substituting $P(X\in B\mid Y)$ for $Z_B$,

$$
\int_{Y^{-1}(A)} P(X\in B\mid Y)\,dP=\int_{Y^{-1}(A)} \mathbb 1_{\{X\in B\}}\,dP.
$$

Now note that

$$
\mathbb 1_{\{Y\in A\}}\mathbb 1_{\{X\in B\}}=\mathbb 1_{\{X\in B,\;Y\in A\}},
$$

so

$$
\int_{Y^{-1}(A)} \mathbb 1_{{X\in B}}\,dP
=\int_{\Omega} \mathbb 1_{\{X\in B\}}\mathbb 1_{\{Y\in A\}}\,dP=P(X\in B,\;Y\in A).
$$

Therefore

$$
P(X\in B,\;Y\in A)=\int_{Y^{-1}(A)} P(X\in B\mid Y)\,dP.
$$

---

$Z_B$ is measurable with respect to $\sigma(Y)$.
Hence from the Doob-Dynkin Lemma $Z_B$ must be a function of $Y$. We can write $Z_B = h_B(Y)$ and then the above expression can be written as:

$$
P(X\in B,\;Y\in A)=\int_{Y^{-1}(A)} h_B(Y)\,dP.
$$

Now applying the pushforward rule:
$$
\int_{Y^{-1}(A)} h_B(Y)\,dP=\int_A h_B(y)\,dP_Y(y),
$$

we have

$$
P(X\in B,\;Y\in A)=\int_A P(X\in B\mid Y=y)\,dP_Y(y).
$$

If $Y$ has density $f_Y$, so that

$$
dP_Y(y)=f_Y(y)\,dy,
$$

then this becomes

$$
P(X\in B,\;Y\in A)=\int_A P(X\in B\mid Y=y)\,f_Y(y)\,dy.
$$


A concise interpretation is:

> The joint probability $P(X\in B,\;Y\in A)$ is obtained by averaging the conditional probability $P(X\in B\mid Y=y)$ over all values $y\in A$, weighted by the distribution of $Y$.


Denote by $f_{X,Y}(x,y)$ the density of the joint distribution of $(X,Y)$ then
$$
P(X\in B,\;Y\in A)=\iint_{A\times B} f_{X,Y}(x,)\,dx\,\,dy.
$$



## 2.2 The Conditional Random Variable ($X \mid Y=y$)

The conditional measure $P_{X \mid Y=y}$ is defined by
$$P_{X \mid Y=y}(B)\triangleq P(X \in B \mid Y=y).$$
If this measure is absolutely continuous with respect to the Lebesgue measure, then there exists a density $f_{X|Y}(x \mid y)$ such that:
$$
P(X \in B \mid Y=y)=P_{X \mid Y=y}(B)
=\int_B dP_{X \mid Y=y}= \int_B f_{X|Y}(x \mid y) \, dx.
$$
Specifically, if this measure is absolutely continuous with respect to the Lebesgue measure, the density $f_{X|Y}(x \mid y)$ is the Radon-Nikodym derivative of $P_{X \mid Y=y}$ with respect to the Lebesgue measure:
$$f_{X|Y}(x \mid y) = \frac{dP_{X \mid Y=y}}{dx}(x).$$

We define
$$X \mid Y=y$$
to be the unique random variable whose law is the conditional probability measure $P_{X \mid Y=y}$. Uniqueness follows from the uniqueness of the Radon-Nikodym derivative.

The expectation of this random variable is defined by

$$
\mathbb{E}(X\mid Y=y)\triangleq \int_{\Omega}X(\omega) dP(\omega \mid Y=y)=\int_{\mathscr{X}}x\,dP_{X \mid Y=y}(x)=\int_{\mathscr{X}}xf_{X|Y}(x \mid y)\,dx
$$

**Note that we have used parenthesis to distinguish from the conditional random varable $\mathbb{E}[X\mid Y]=g(Y)$.**



## 2.3 The Product Rule of the Conditional Density

Substituting the above expression, $
P(X \in B \mid Y=y)=\int_B f_{X|Y}(x \mid y) \, dx
$, into the integral $
P(X\in B,\;Y\in A)=\int_A P(X\in B\mid Y=y)\,f_Y(y)\,dy
$ we have:
$$P(X \in B, Y \in A)
= \int_A \int_B f_{X|Y}(x \mid y) \, f_Y(y)\,dx \, dy.
$$

Since $P(X \in B, Y \in A) = \iint_{A \times B} f_{X,Y}(x,y) \, dx \, dy$ where $f_{X,Y}(x,y)$ is the joint probability density,
we obtain the product rule

$$
f_{X,Y}(x,y)=f_{X|Y}(x\mid y)f_Y(y)
$$
for almost every $(x,y)$.
The equality holds for the joint density $f_{X,Y}$ by the uniqueness of the Radon-Nikodym derivative. **Compare this with the Radon-Nikodym definition of Baye's theorem presented in section 3.3.2.**


---

**Discrete Product Rule**

The relationship $f_{X,Y}(x,y) = f_{X|Y}(x|y)f_Y(y)$ represents the "Product Rule" of probability, which in the discrete case is expressed as
$$p_{X,Y}(x,y) = p_{X|Y}(x|y)p_Y(y).$$ Rigorously, this formula is the realization of the measure-theoretic definition: $P(X \in B | Y) = \mathbb E[\mathbb{1}_B(X) | Y]$. In this context, $p_Y(y)$ is the marginal probability (the "weight" of a specific outcome of $Y$), and $p_{X|Y}(x|y)$ is the conditional probability (the "rescaled slice" of the joint distribution where $Y$ is fixed). By multiplying the localized probability of $X$ within a specific group by the total probability of that group occurring, we recover the joint probability of both events happening simultaneously.


## 2.4 The Disintegration Theorem

For any $A\in \mathscr{Y}$ the **Disintegration Theorem** states:

$$
\int_{Y^{-1}(A)} X \, dP =\int_A \mathbb{E}(X\mid Y=y)\, dP_Y(y)= \int_A \left( \int_{\mathscr{X}} x \, dP_{X \mid Y=y}(x) \right) dP_Y(y).
$$
Here **$\mathbb{E}(X \mid Y=y)$ is the expectation of the random variable $X\mid Y=y$ with respect to the conditional measure $P_{X \mid Y=y}$**.


The formal proof of this theorem is beyond the scope of this course. Intuitively, what it says is that the global average of a random variable $X$ over a specific region of the sample space can be calculated by "slicing" the sample space into fibers where $Y$ is fixed at a constant value $y$.

Within each fiber, we calculate the local average of $X$—this is the inner integral, which yields the conditional expectation $\mathbb{E}(X \mid Y=y)$. The theorem then tells us that the total average over the region $Y \in A$ is simply the weighted sum (or integral) of these local averages, where the weights are determined by the probability distribution of $Y$ itself.

In short, it justifies the "divide and conquer" strategy for computing expectations: you can break any complex distribution into a family of simpler conditional distributions, compute the averages locally, and then re-aggregate them to recover the global expectation.

**Note that $\mathbb{E}(X \mid Y=y)$ is the expectation of the random variable $(X\mid Y=y)$ with respect to the conditional measure $P_{X \mid Y=y}$. In contrast $\mathbb{E}[X \mid Y=y]=g(y)$ is the value of the conditional expectation random variable $\mathbb{E}[X|Y]$ evaluated at $Y=y$. However we will show below that the both values are the same and hence $\mathbb{E}(X \mid Y=y)=\mathbb{E}[X \mid Y=y]$.**

## 2.5 Expectation of $(X\mid Y=y)$ is the Same as the Conditional Expectation Evaluated at $Y=y$

We apply the definition of conditional expectation for a general random variable $X$ and using the Doob-Dynkin lemma we have for any $A\in \mathscr{Y}$:

$$
\int_{Y^{-1}(A)} X \, dP =
\int_{Y^{-1}(A)} \mathbb E[X \mid Y] \, dP =\int_A \mathbb E[X \mid Y=y] \, dP_Y(y).
$$

Note that in the previous section we have shown that the Disintegration Theorem gives
$$
\int_{Y^{-1}(A)} X \, dP =\int_A \mathbb{E}(X\mid Y=y)\, dP_Y(y)= \int_A \left( \int_{\mathscr{X}} x \, dP_{X \mid Y=y}(x) \right) dP_Y(y),
$$
where $\mathbb{E}(X\mid Y=y)$ was the expectation of the random variable $X\mid Y=y$ with respect to the conditional measure $P_{X \mid Y=y}$.

Thus we see that
$$\mathbb{E}[X\mid Y=y]=\mathbb{E}(X\mid Y=y)= \int_{-\infty}^{\infty} x \, dP_{X \mid Y=y}(x).
$$
The first term, $\mathbb{E}[X|Y=y]$, represents the evaluation of the $\sigma(Y)$-measurable random variable, while the second, $\mathbb{E}(X|Y=y)$, represents the integral against the conditional measure. The Disintegration Theorem proves these two objects are, in fact, the same $P_Y$-almost everywhere.


Given the assumption that the measure $P_{X \mid Y=y}$ is absolutely continuous with respect to the Lebesgue measure, we have the expression:

$$
\mathbb E[X \mid Y=y] = \int_{-\infty}^{\infty} x \, f_{X|Y}(x \mid y) \, dx
$$

Defining $g(y) \triangleq \mathbb E[X \mid Y=y]$, we arrive at:

$$
g(y) = \int_{-\infty}^{\infty} x \, f_{X|Y}(x \mid y) \, dx.
$$


---

**Summary: The Relationship Between $\mathbb E[X\mid Y]$ and $f_{X\mid Y}$**

> The conditional expectation $\mathbb E[X\mid Y]$ is the $\sigma(Y)$-measurable random variable:
$$\mathbb E[X\mid Y]=g(Y).$$ It'sevaluation at the random value $Y=y$ is given by the expression
> $$g(y)=\mathbb E[X\mid Y=y]=\int_{\mathscr{X}} x\,f_{X\mid Y}(x\mid y)\,dx.$$



## 2.6 Examples

### 2.6.1 Two Coin Toss Example

Let’s take a two-coin toss where **$X$** is the result of the first toss and **$Y$** is the total number of "Heads" across both tosses ($H=1, T=0$).

In this scenario, the joint probability **$p_{X,Y}(x,y)$** is the probability that the first coin is $x$ **and** the total count is $y$. To find the conditional probability **$p_{X|Y}(x|y)$**, we "slice" the joint table at a specific value of $y$ and divide by the marginal probability $p_Y(y)$. For instance, if we observe that there is exactly **one head total ($Y=1$)**, our "strategy" $P(X \in B | Y)$ updates: since the total could have come from $(H,T)$ or $(T,H)$, the probability that the first coin was a Head ($X=1$) becomes $0.5$. This perfectly demonstrates the product rule: the joint probability of the first coin being Heads while the total is one ($0.25$) is equal to the conditional probability ($0.5$) multiplied by the marginal probability of getting one head total ($0.5$).

| Outcome $(\omega)$ | $X$ (1st Coin) | $Y$ (Total Heads) | $P(\omega)$ |
| --- | --- | --- | --- |
| **(H, H)** | 1 | 2 | 0.25 |
| **(H, T)** | 1 | 1 | 0.25 |
| **(T, H)** | 0 | 1 | 0.25 |
| **(T, T)** | 0 | 0 | 0.25 |

---

**Verification of the Product Rule for $X=1, Y=1$:**

* **Marginal $p_Y(1)$**: $P(Y=1) = 0.25 + 0.25 = 0.5$ (The "weight" of the group).
* **Conditional $p_{X|Y}(1|1)$**: $0.25 / 0.5 = 0.5$ (The "rescaled slice").
* **Product**: $0.5 \times 0.5 = 0.25$, which is exactly the joint probability $p_{X,Y}(1,1)$.

### 2.6.2 Foundations of the Kalman Filter

##### The Joint and Conditional Densities

Let $(X, Y)$ be a bivariate normal vector with mean vector $\boldsymbol{\mu} = (\mu_X, \mu_Y)^\top$ and covariance matrix:
$$\Sigma = \begin{pmatrix} \sigma_X^2 & \rho\sigma_X\sigma_Y \\ \rho\sigma_X\sigma_Y & \sigma_Y^2 \end{pmatrix}$$
where $\rho$ is the correlation coefficient, so $\text{Cov}(X,Y) = \rho\sigma_X\sigma_Y$.

The conditional density $f_{X|Y}(x|y)$ is given by the ratio:
$$f_{X|Y}(x|y) = \frac{f_{X,Y}(x,y)}{f_Y(y)}$$

To compute the conditional density $f_{X|Y}(x|y)$, we use the definition of conditional probability for continuous variables. The process involves algebraic manipulation of the exponents to separate the terms involving $x$ from those involving only $y$.

$$f_{X,Y}(x,y) = \frac{1}{2\pi\sigma_X\sigma_Y\sqrt{1-\rho^2}} \exp\left( -\frac{1}{2(1-\rho^2)} Q(x,y) \right)$$
where $Q(x,y)$ is the quadratic form:
$$Q(x,y) = \frac{(x-\mu_X)^2}{\sigma_X^2} - \frac{2\rho(x-\mu_X)(y-\mu_Y)}{\sigma_X\sigma_Y} + \frac{(y-\mu_Y)^2}{\sigma_Y^2}$$


To find the conditional density, we focus on the terms in the exponent containing $x$. We treat $y$ as a constant and "complete the square" for the variable $x$:

$$Q(x,y) = \left[ \frac{(x-\mu_X)}{\sigma_X} - \rho\frac{(y-\mu_Y)}{\sigma_Y} \right]^2 + \left(1-\rho^2\right)\frac{(y-\mu_Y)^2}{\sigma_Y^2}$$

Substituting this back into the joint PDF allows us to factor the expression:

$$f_{X,Y}(x,y) = \underbrace{\frac{1}{\sqrt{2\pi\sigma_X^2(1-\rho^2)}} \exp\left( -\frac{(x - \mu_{X|Y})^2}{2\sigma_X^2(1-\rho^2)} \right)}_{f_{X|Y}(x|y)} \times \underbrace{\frac{1}{\sqrt{2\pi\sigma_Y^2}} \exp\left( -\frac{(y-\mu_Y)^2}{2\sigma_Y^2} \right)}_{f_Y(y)}$$

By dividing the joint density by the marginal $f_Y(y)$, we are left with $f_{X|Y}(x|y)$. We can now read off the mean and variance of this Gaussian distribution:


$$\mu_{X|Y} = \mu_X + \rho\frac{\sigma_X}{\sigma_Y}(y - \mu_Y) = \mu_X + \frac{\text{Cov}(X,Y)}{\text{Var}(Y)}(y - \mu_Y)$$

$$\sigma_{X|Y}^2 = \sigma_X^2(1-\rho^2)$$

>**Summary of the Result:**
The conditional density is itself a univariate Gaussian:
$$f_{X|Y}(x|y) = \frac{1}{\sqrt{2\pi\sigma_{X|Y}^2}} \exp\left( -\frac{(x - \mu_{X|Y})^2}{2\sigma_{X|Y}^2} \right)$$
This proves that for jointly Gaussian variables, the conditional distribution is always Gaussian, with a mean that is a linear function of $y$ and a variance that is constant (independent of the specific value of $y$).



#####Conditional Expectation in the Bivariate Normal Distribution

We define $g(y) = \mathbb{E}[X \mid Y=y]$. By the Disintegration Theorem, this is computed as the integral against the conditional PDF:
$$g(y) = \int_{-\infty}^{\infty} x f_{X|Y}(x \mid y) \, dx$$

Substitute the Gaussian form of $f_{X|Y}(x \mid y)$ derived above. Let $\mu(y) = \mu_X + \frac{\text{Cov}(X,Y)}{\text{Var}(Y)}(y - \mu_Y)$ and $\sigma_{X|Y}^2 = \sigma_X^2(1-\rho^2)$. Then:
$$g(y) = \int_{-\infty}^{\infty} x \left[ \frac{1}{\sqrt{2\pi\sigma_{X|Y}^2}} \exp\left( -\frac{(x - \mu(y))^2}{2\sigma_{X|Y}^2} \right) \right] dx$$

This integral is simply the first moment (mean) of a Gaussian random variable with mean $\mu(y)$. Thus:
$$g(y) = \mu(y) = \mu_X + \frac{\text{Cov}(X,Y)}{\text{Var}(Y)}(y - \mu_Y)$$

Therefore:
$$\mathbb{E}[X \mid Y=y] = \mu_X + \frac{\text{Cov}(X,Y)}{\text{Var}(Y)}(y - \mu_Y)$$

and hence
$$\mathbb{E}[X \mid Y] = \mu_X + \frac{\text{Cov}(X,Y)}{\text{Var}(Y)}(Y - \mu_Y)$$


>In physical terms, this result demonstrates that when two variables are statistically coupled, observing one ($Y$) provides a window into the hidden state of the other ($X$). The expression represents a weighted balance between **prior knowledge** and **new evidence**: the term $(Y - \mu_Y)$ captures the "surprise" or deviation in our measurement, while the ratio $\frac{\text{Cov}(X,Y)}{\text{Var}(Y)}$—the gain—scales that surprise based on how strongly the two variables move together relative to the volatility of our observation.
>
>Instead of taking the measurement at face value, the conditional expectation "shrinks" the estimate toward the historical mean $\mu_X$ whenever the sensor is noisy or the correlation is weak. It represents the mathematically optimal way to extract signal from noise, ensuring that our estimate is not just a guess, but the most probable physical state of $X$ given the specific information "leaked" through $Y$.
>This is in fact the **Kalman-Filter**.

##### Optimality of the estimate

The conditional expectation random variable $\widehat{X}\triangleq \mathbb{E}[X \mid Y]$ was shown in the previoys section to be:
$$\widehat{X} = \mu_X + k(Y - \mu_Y), \quad \text{where } k = \frac{\text{Cov}(X,Y)}{\text{Var}(Y)}$$

---

Lets find the correlation between the estimation error $Z = X - \widehat{X}$ and our data $Y$.

$$\text{Cov}(X - \widehat{X}, Y) = \text{Cov}(X, Y) - \text{Cov}(\widehat{X}, Y)$$

Substitute $\widehat{X} = \mu_X + k(Y - \mu_Y)$:
$$\text{Cov}(\widehat{X}, Y) = \text{Cov}(\mu_X + kY - k\mu_Y, Y)$$
Since constants do not affect covariance:
$$\text{Cov}(\widehat{X}, Y) = k \cdot \text{Cov}(Y, Y) = k \cdot \text{Var}(Y)$$

Now, substitute the definition of $k = \frac{\text{Cov}(X,Y)}{\text{Var}(Y)}$:
$$\text{Cov}(\widehat{X}, Y) = \left( \frac{\text{Cov}(X,Y)}{\text{Var}(Y)} \right) \cdot \text{Var}(Y) = \text{Cov}(X,Y)$$

Returning to our original equation:
$$\text{Cov}(X - \widehat{X}, Y) = \text{Cov}(X, Y) - \text{Cov}(X, Y) = \mathbf{0}$$

**Physical Significance:** This proves that $\widehat{X}$ has extracted **all** possible information from $Y$. If there were any covariance left between the error and $Y$, you could use that relationship to improve your estimate further.

---

In summary orthogonality means optimality because it ensures that

1. **No Signal is Left:**
That is any predictable patterns in the error do not exists.

2. **Shortest Path:**
We have found the point in the "Information Space" closest to the true "Hidden State.

3. **Minimum Variance:**
Any deviation from the orthogonal projection will strictly increase the variance of your error.

In Fault Diagnostics: If your error $Z$ were still correlated with $Y$, it would mean you could use the power meter $Y$ to predict your own mistakes—which means you weren't using the power meter correctly in the first place! Orthogonality is the proof that you have maximized your Information Gain.

##### The "Slice" Visual



Think of the joint distribution $f_{X,Y}$ as a 3D mountain range.

1. **Fixing $Y=y$** is like taking a vertical "chainsaw cut" through that mountain at a specific $y$-coordinate.
2. The resulting **silhouette** of the cut is the conditional distribution.
3. However, the area under that silhouette isn't 1 yet. We must **rescale** it (divide by the marginal $f_Y(y)$) to make it a valid probability distribution.


````python
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots
from scipy.stats import multivariate_normal, norm

# --- Parameters ---
mu_x, mu_y = 0, 0
sigma_x, sigma_y = 1, 1
rho = 0.8
y_obs = 1.5

# Math for Conditional Distribution
mu_cond = mu_x + rho * (sigma_x / sigma_y) * (y_obs - mu_y)
var_cond = sigma_x**2 * (1 - rho**2)
std_cond = np.sqrt(var_cond)

# --- Data Generation ---
x_grid = np.linspace(-3.5, 3.5, 200)
y_grid = np.linspace(-3.5, 3.5, 200)
X, Y = np.meshgrid(x_grid, y_grid)
pos = np.dstack((X, Y))
Z = multivariate_normal([mu_x, mu_y], [[sigma_x**2, rho*sigma_x*sigma_y], [rho*sigma_x*sigma_y, sigma_y**2]]).pdf(pos)

# --- Create Plot ---
fig = make_subplots(rows=1, cols=2,
                    subplot_titles=("1. Joint Density (The 'Hill')", "2. Conditional Density (The 'Slice')"))

# Left: Joint Contour
fig.add_trace(go.Contour(x=x_grid, y=y_grid, z=Z, colorscale='Viridis', showscale=False), row=1, col=1)
fig.add_shape(type="line", x0=-3.5, y0=y_obs, x1=3.5, y1=y_obs, line=dict(color="Red", width=3, dash="dash"), row=1, col=1)

# Right: Conditional PDF
x_pdf = np.linspace(-3.5, 3.5, 500)
y_pdf = norm.pdf(x_pdf, mu_cond, std_cond)
prior_pdf = norm.pdf(x_pdf, mu_x, sigma_x)

fig.add_trace(go.Scatter(x=x_pdf, y=prior_pdf, name='Prior P(X)', line=dict(color='gray', dash='dot')), row=1, col=2)
fig.add_trace(go.Scatter(x=x_pdf, y=y_pdf, name='Posterior P(X|Y)', fill='tozeroy', line=dict(color='blue', width=3)), row=1, col=2)
fig.add_vline(x=mu_cond, line=dict(color="Red", dash="dot"), row=1, col=2)

fig.update_layout(height=500, title_text="Information Gain: From Joint to Conditional Distribution", template="plotly_white")
fig.show()
````

**What this visualization reveals**

Dimensionality Collapse: The left plot exists in the joint $\mathbb{R}^2$ space. By observing $Y=1.5$ (the dashed red line), we "slice" the joint density. The right plot represents the profile of that slice, effectively collapsing the $Y$ dimension into a single, updated 1D distribution for $X$.

The "Gain" in Action (Mean Shift): Because $X$ and $Y$ are positively correlated ($\rho=0.8$), the high observed value of $Y$ "drags" our estimate of $X$ with it. Our best estimate $E[X \mid Y]$ shifts from the prior mean of $0$ to the posterior mean of $1.2$.

Information Gain (Variance Shrinkage): Notice that the blue curve on the right is "taller and skinnier" than the original spread of $X$. Mathematically, the variance has shrunk from $1.0$ to $0.36$. This narrowing represents the Information Gain: the physical reduction of uncertainty through observation.

The Math: How Observation "Shrinks" UncertaintyIn Gaussian systems, the most striking property is that the Conditional Variance is always less than (or equal to) the Prior Variance.$$\text{Var}(X \mid Y) = \sigma_X^2 (1 - \rho^2)$$

*Why this is critical for Engineering:*

If $\rho = 0$
(Independence): The variance remains $\sigma_X^2$. Observing $Y$ tells you nothing; uncertainty is unchanged.

If $\rho = 1$.
(Perfect Correlation): The variance becomes $0$. Observing $Y$ tells you $X$ exactly; uncertainty is eliminated.

The "Information" is in the Correlation: The term $(1 - \rho^2)$ is the factor by which the "noise" is suppressed. In our plot, $\rho=0.8$, so the variance shrinks by $1 - 0.64 = 0.36$. We have "removed" $64\%$ of the original uncertainty.

## 3.1 Law of Total Expectation

This is arguably one of the most elegant result in probability theory.

If you know the expected value of $X$ for every possible "slice" of $Y$, then the overall average of $X$ is simply the weighted average of those slices.

---

**The Measure-Theoretic Statement**

The law states:


$$\mathbb E[X] = \mathbb E[\mathbb E[X | \sigma(Y)]]$$

**The Intuition:**

* The inner expectation, $\mathbb E[X | \sigma(Y)]$, is your **Strategy** (the "Best Guess" function).
* The outer expectation, $\mathbb E[\cdot]$, averages that strategy over all possible realizations of $Y$.

---

**The Analytical Derivation (Using Densities)**

To see why this works in the lab, let's write it out using the "slice and divide" logic we derived:

1. **Start with the Conditional Mean for a point $y$:**

$$\mathbb E[X | Y=y] = \int_{\mathscr{X}} x \cdot f_{X|Y}(x|y) \, dx$$


2. **Take the expectation of that function over all $y$:**

$$\mathbb E[\mathbb E[X|Y]] = \int_{\mathscr{Y}} \left( \int_{\mathscr{X}} x \cdot f_{X|Y}(x|y) \, dx \right) f_Y(y) \, dy$$


3. **Use the Product Rule ($f_{X|Y} \cdot f_Y = f_{X,Y}$):**

$$\mathbb E[\mathbb E[X|Y]] = \int_{\mathscr{Y}} \int_{\mathscr{X}} x \cdot f_{X,Y}(x,y) \, dx \, dy$$


4. **Change the Order of Integration (Fubini's Theorem):**

$$\mathbb E[\mathbb E[X|Y]] = \int_{\mathscr{X}} x \left( \int_{\mathscr{Y}} f_{X,Y}(x,y) \, dy \right) dx$$


5. **Recognize the Marginal ($ \int f_{X,Y} dy = f_X(x)$):**

$$\mathbb E[\mathbb E[X|Y]] = \int_{\mathscr{X}} x \cdot f_X(x) \, dx = \mathbb E[X]$$



---

**The "Tower Property" Analogy**

Think of this as a **Tower**:

* **Ground Floor:** $\mathbb E[X]$. The raw average of the whole population.
* **First Floor:** $\mathbb E[X|Y]$. You’ve broken the population into groups (by $Y$). Each group has its own average.
* **The Result:** If you take the average of all the group averages (weighted by group size), you land back on the Ground Floor.

---





**Why this is Important:**

It allows us to solve complex problems by **Conditioning**. If you don't know how to calculate $\mathbb E[X]$ directly, you can:

1. Pick a variable $Y$ that "explains" $X$ well.
2. Calculate the mean of $X$ for each "slice" of $Y$ (usually easier).
3. Average those results.

#### Python Verification: "The Path of Means"

This code proves that the average of our "Red Regression Line" (from earlier) is exactly the global mean of $X$.

````python
import numpy as np

# Parameters
rho = 0.8
n_samples = 1_000_000

# 1. Generate Joint Data
y_samples = np.random.normal(5, 2, n_samples) # Mean 5, Std 2
x_samples = 0.8 * (y_samples - 5) + np.random.normal(10, 1, n_samples) # Mean 10

# 2. Calculate the Global Mean (Ground Floor)
global_mean_x = np.mean(x_samples)

# 3. Calculate Conditional Means (First Floor)
# For each y, our "Best Guess" strategy is h(y)
def h(y):
    # \mathbb E[X|Y=y] = mu_x + rho * (sigma_x/sigma_y) * (y - mu_y)
    # In this specific simulated case, the logic is:
    return 10 + 0.8 * (y - 5)

conditional_means = h(y_samples)

# 4. Law of Iterated Expectations: \mathbb E[h(Y)] should equal \mathbb E[X]
average_of_averages = np.mean(conditional_means)

print(f"Global Mean \\mathbb E[X]:              {global_mean_x:.4f}")
print(f"Average of Conditional Means:  {average_of_averages:.4f}")
print(f"Difference:                    {abs(global_mean_x - average_of_averages):.6f}")
````

## 3.2 Conditional Variance ($\mathrm{Var}(X|Y)$)

In probability and statistical learning, we often seek to quantify how much of the uncertainty in a random variable $X$ can be explained by another variable $Y$.  

The Conditional Variance: $\mathrm{Var}(X|Y)$
The conditional variance is the variance of $X$ given a specific outcome of $Y$. Formally, it is a $\sigma(Y)$-measurable random variable defined as:
$$\mathrm{Var}(X|Y) = \mathbb E[X^2 | Y] - (\mathbb E[X|Y])^2$$
It measures the **residual uncertainty** of $X$ that remains after $Y$ is known. If you were to observe $Y=y$, the conditional variance tells you the spread of the remaining probability distribution of $X$.



## 3.3 Law of Total Variance

For any two random variables $X$ and $Y$ with finite variance, the total variance of $X$ can be split into two distinct parts:

$$\mathrm{Var}(X) = \mathbb E[\mathrm{Var}(X|Y)] + \mathrm{Var}(\mathbb E[X|Y])$$

* **$\mathrm{Var}(X)$**: Total Uncertainty (The "Total Spread").
* **$\mathbb E[\mathrm{Var}(X|Y)]$**: Unexplained Variance (The "Average Spread of the Slices").
* **$\mathrm{Var}(\mathbb E[X|Y])$**: Explained Variance (The "Spread of the Means").

---

**The Analytical Derivation**

We use the definition of variance $\mathrm{Var}(Z) = \mathbb E[Z^2] - (\mathbb E[Z])^2$ and the **Law of Iterated Expectations** to prove this.

**Step 1: Expand the Right Hand Side.**
Recall $\mathrm{Var}(X|Y) = \mathbb E[X^2|Y] - (\mathbb E[X|Y])^2$.
Taking the expectation of this "Slice Variance":


$$\mathbb E[\mathrm{Var}(X|Y)] = \mathbb E[\mathbb E[X^2|Y]] - \mathbb E[(\mathbb E[X|Y])^2]$$

By the Law of Iterated Expectations, $\mathbb E[\mathbb E[X^2|Y]] = \mathbb E[X^2]$. So:


$$\mathbb E[\mathrm{Var}(X|Y)] = \mathbb E[X^2] - \mathbb E[(\mathbb E[X|Y])^2] \quad \text{--- (Equation A)}$$

**Step 2: Expand the Variance of the Mean.**


$$\mathrm{Var}(\mathbb E[X|Y]) = \mathbb E[(\mathbb E[X|Y])^2] - (\mathbb E[\mathbb E[X|Y]])^2$$

Again, by Iterated Expectations, $\mathbb E[\mathbb E[X|Y]] = \mathbb E[X]$. So:


$$\mathrm{Var}(\mathbb E[X|Y]) = \mathbb E[(\mathbb E[X|Y])^2] - (\mathbb E[X])^2 \quad \text{--- (Equation B)}$$

**Step 3: Add (A) and (B).**
Notice that the $\mathbb E[(\mathbb E[X|Y])^2]$ terms cancel out:


$$\mathbb E[\mathrm{Var}(X|Y)] + \mathrm{Var}(\mathbb E[X|Y]) = \mathbb E[X^2] - (\mathbb E[X])^2$$

This is exactly the definition of **$\mathrm{Var}(X)$**.

---

**The Physical Intuition**

Imagine you are measuring the height ($X$) of students across different ages ($Y$).

* **$\mathrm{Var}(X)$**: The total variation in height if you ignore age.
* **$\mathbb E[\mathrm{Var}(X|Y)]$**: The variation in height *within* each age group (e.g., the spread of 10-year-olds). This is what your model **cannot** explain.
* **$\mathrm{Var}(\mathbb E[X|Y])$**: How much the *average* height changes as you move from age 5 to age 20. This is the information that Age ($Y$) **successfully explains**.

---

> **The Shrinkage Principle:** Since $\mathrm{Var}(\mathbb E[X|Y]) \geq 0$, it is mathematically guaranteed that $\mathbb E[\mathrm{Var}(X|Y)] \leq \mathrm{Var}(X)$. Observing $Y$ can never make you *more* confused on average; it either helps or stays the same.



````python
import numpy as np

# Generate correlated data
n = 10**6
y = np.random.normal(0, 10, n)
# X is dependent on Y + some noise
x = 0.5 * y + np.random.normal(0, 5, n)

# 1. Total Variance
total_var = np.var(x)

# 2. Explained Variance (Variance of the "Prediction" \mathbb E[X|Y])
# For this linear case, \mathbb E[X|Y] is simply 0.5 * Y
prediction = 0.5 * y
explained_var = np.var(prediction)

# 3. Unexplained Variance (Average variance of the slices/residuals)
residuals = x - prediction
unexplained_var = np.mean(residuals**2) # Since mean of residuals is 0

print(f"Total Variance:      {total_var:.4f}")
print(f"Explained + Unexpl.: {explained_var + unexplained_var:.4f}")
print(f"R-squared (Explained/Total): {explained_var/total_var:.4f}")
````

### 3.3.1 The Two Coin Toss Example

**Global Variance $\mathrm{Var}(X)$**

First, we look at $X$ alone. Since $X$ is a Bernoulli trial (fair coin), its mean is $\mathbb E[X] = 0.5$.


$$\mathrm{Var}(X) = \mathbb E[X^2] - (\mathbb E[X])^2 = (1^2 \cdot 0.5 + 0^2 \cdot 0.5) - (0.5)^2 = 0.5 - 0.25 = \mathbf{0.25}$$


This represents our **Total Uncertainty** about the first coin before we know anything about the total count.

---

**The Partition (Law of Total Variance)**

Now we apply: $\mathrm{Var}(X) = \mathbb E[\mathrm{Var}(X|Y)] + \mathrm{Var}(\mathbb E[X|Y])$.

**Part A: The "Strategy" $\mathbb E[X|Y]$ (Explained Variance)**

We calculate the "Best Guess" of $X$ for every possible value of $Y$:

* If $Y=0$, then $X$ must be $0$. So $\mathbb E[X|Y=0] = \mathbf{0}$.
* If $Y=1$, then $X$ could be $0$ or $1$ (equally likely). So $\mathbb E[X|Y=1] = \mathbf{0.5}$.
* If $Y=2$, then $X$ must be $1$. So $\mathbb E[X|Y=2] = \mathbf{1}$.

Now, we find the variance of these guesses (weighted by $p_Y$):


$$Var(\mathbb E[X|Y]) = \underbrace{(0^2 \cdot 0.25 + 0.5^2 \cdot 0.5 + 1^2 \cdot 0.25)}_{\mathbb E[(\mathbb E[X|Y])^2]} - \underbrace{(0.5)^2}_{(\mathbb E[X])^2}$$

$$\mathrm{Var}(\mathbb E[X|Y]) = (0 + 0.125 + 0.25) - 0.25 = \mathbf{0.125}$$


**This is the "Explained Variance":** Knowing $Y$ explains $50\%$ of the uncertainty in $X$.

**Part B: The "Residuals" $\mathbb E[Var(X|Y)]$ (Unexplained Variance)**

We calculate the "Spread" of $X$ within each slice of $Y$:

* Slice $Y=0$: $X$ is always $0 \implies Var(X|Y=0) = \mathbf{0}$.
* Slice $Y=1$: $X$ is a fair coin $\implies Var(X|Y=1) = \mathbf{0.25}$.
* Slice $Y=2$: $X$ is always $1 \implies Var(X|Y=2) = \mathbf{0}$.

Now, average these slices:


$$\mathbb E[Var(X|Y)] = (0 \cdot 0.25) + (0.25 \cdot 0.5) + (0 \cdot 0.25) = \mathbf{0.125}$$


**This is the "Unexplained Variance":** The uncertainty left over when the total heads is 1.

---

**Summary of the Results**

| Component | Value | Interpretation |
| --- | --- | --- |
| **$Var(X)$** | **0.25** | Total uncertainty (Raw data). |
| **$Var(\mathbb E[X | Y])$** | **0.125** |
| **$\mathbb E[Var(X | Y)]$** | **0.125** |

**Check:** $0.125 (\text{Explained}) + 0.125 (\text{Unexplained}) = 0.25 (\text{Total})$.

---

**Conclusion:**

By observing $Y$, we have cut our uncertainty about $X$ exactly in half. In "Information Theory" terms, we would say $Y$ provides information about $X$. If $Y$ and $X$ were independent, the explained variance would be $0$, and we would be just as confused as when we started.

#5. Entropy

In the most rigorous sense—the one that bridges Information Theory and Measure Theory—**Entropy is a property of the probability measure.**

However, in practice, we usually describe it as a property of the **distribution** (the Law) of a random variable. To be precise: the random variable $X$ is simply the "delivery vehicle" for the measure.

## 5.1 The Measure-Theoretic View (Relative Entropy)

**Relative entropy** (also called **Kullback-Leibler divergence**) is a measure of how one probability measure differs from another.

If $P$ and $Q$ are two probability measures on the same measurable space $(\Omega,\mathfrak F)$, and $P$ is absolutely continuous with respect to $Q$ ($P \ll Q$), then the relative entropy of $P$ with respect to $Q$ is defined by

$$
D_{KL}(P\|Q)=\int_{\Omega}\ln\left(\frac{dP}{dQ}\right)\,dP
$$

equivalently,

$$
D_{KL}(P\|Q)=\int_{\Omega}\frac{dP}{dQ}\ln\left(\frac{dP}{dQ}\right)\,dQ.
$$

Here $\frac{dP}{dQ}$ is the **Radon-Nikodym derivative** of $P$ with respect to $Q$.

So relative entropy is fundamentally a comparison between two probability measures. It does not require introducing a random variable explicitly; it only requires that the two measures are defined on the same measurable space.

---

For two discrete probability distributions $P$ and $Q$ on the same countable alphabet $\mathfrak X$, the relative entropy is

$$
D_{KL}(P\|Q)=\sum_{x\in\mathfrak X} P(x)\ln\left(\frac{P(x)}{Q(x)}\right).
$$

In the discrete case, the Radon-Nikodym derivative reduces to the ratio

$$
\frac{dP}{dQ}(x)=\frac{P(x)}{Q(x)}
\quad\text{whenever } Q(x)>0.
$$

**Condition for finiteness:**

For $D_{KL}(P\|Q)$ to be well defined as an extended value, it is enough to require $P \ll Q$. Under this condition, the KL divergence may still be finite or $+\infty$.

For the discrete case, absolute continuity means:

> whenever $Q(x)=0$, we must also have $P(x)=0$.

If there exists some $x$ such that $P(x)>0$ but $Q(x)=0$, then

$$
D_{KL}(P\|Q)=+\infty.
$$



##5.1.1 Non-negativity of the Kullback-Leibler divergence

For the continuous case:
$$D_{KL}(p \parallel q) = \int p(x) \ln \left( \frac{p(x)}{q(x)} \right) dx = \int p(x) \left( -\ln \frac{q(x)}{p(x)} \right) dx$$

Because $-\ln(x)$ is a convex function, Jensen’s Inequality tells us that the average of the function is greater than or equal to the function of the average:
$$\mathbb{E}_p \left[ -\ln \frac{q(x)}{p(x)} \right] \geq -\ln \left( \mathbb{E}_p \left[ \frac{q(x)}{p(x)} \right] \right)$$

When you evaluate the right side:
$$\mathbb{E}_p \left[ \frac{q(x)}{p(x)} \right] = \int p(x) \frac{q(x)}{p(x)} dx = \int q(x) dx = 1$$
Since $-\ln(1) = 0$, we are left with:
$$D_{KL}(P \parallel Q) \geq 0$$

It is exactly **$0$** if and only if $P = Q$ almost everywhere.

## 5.2 The Distribution View (Differential Entropy)

Suppose we have the distribution of our random variable $X$, denoted by the probability measure $P_X$, and a reference measure $\mu$ on $\mathbb{R}^n$ (such as the standard Lebesgue measure $\lambda^n$).

If $P_X$ is absolutely continuous with respect to $\mu$ ($P_X \ll \mu$), the **Radon-Nikodym theorem** guarantees that there exists a non-negative, measurable function $f(x)$—the density—such that:

$$dP_X(x) = f(x) \, d\mu(x) \quad \text{or, equivalently,} \quad \frac{dP_X}{d\mu}(x) = f(x)$$

Using the original definition of Kullback-Leibler divergence the **relative entropy** of the probability measure $P_X$ with respect to the reference measure $\mu$ is:

$$D_{KL}(P_X \| \mu) = \int_{\mathbb{R}^n} \ln\left(\frac{dP_X}{d\mu}\right) \, dP_X$$


Now, we replace the abstract Radon-Nikodym derivative $\frac{dP_X}{d\mu}$ with our explicit density function $f(x)$, and change the integrator from $dP_X$ to $f(x) \, d\mu(x)$:

$$D_{KL}(P_X \| \mu) = \int_{\mathbb{R}^n} \ln(f(x)) \cdot f(x) \, d\mu(x)$$

Rearranging the terms inside the integral gives:

$$D_{KL}(P_X \| \mu) = \int_{\mathbb{R}^n} f(x) \ln f(x) \, d\mu(x)$$


In information theory, **entropy** is intuitively understood as the uncertainty or the information content of a distribution. Mathematically, it is the negative of the relative entropy with respect to a uniform or reference background.

Thus, we define the way **differential entropy** $h_\mu(X)$ as the negative of the KL divergence between $P_X$ and $\mu$:

$$h_\mu(X) = -D_{KL}(P_X \| \mu)$$

By substituting our result from above into this definition, we get:

$$h_\mu(X) = -\int_{\mathbb{R}^n} f(x) \ln f(x) \, d\mu(x)=-\mathbb{E}\left(\ln(f(X))\right).$$


### 5.2.1 Example: The Differential Entropy of a Normally Distributed Random Variable

To find the differential entropy of a normally distributed random variable $X \sim \mathscr{N}(\mu, \sigma^2)$, we evaluate its entropy with respect to the Lebesgue measure on $\mathbb{R}$.

The probability density function $f(x)$ of a normal distribution with mean $\mu$ and variance $\sigma^2$ is:

$$f(x) = \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left( -\frac{(x - \mu)^2}{2\sigma^2} \right)$$

By the definition of differential entropy $h(X)$, we have:

$$h(X) = -\int_{-\infty}^{\infty} f(x) \ln f(x) \, dx = -\mathbb{E}[\ln f(X)]$$

---

Taking the natural logarithm of the PDF gives:

$$\ln f(x) = \ln\left( \frac{1}{\sqrt{2\pi\sigma^2}} \right) - \frac{(x - \mu)^2}{2\sigma^2} = -\frac{1}{2}\ln(2\pi\sigma^2) - \frac{(x - \mu)^2}{2\sigma^2}$$

Now we compute the expected value $-\mathbb{E}[\ln f(X)]$:

$$h(X) = -\mathbb{E}\left[ -\frac{1}{2}\ln(2\pi\sigma^2) - \frac{(X - \mu)^2}{2\sigma^2} \right]$$

Using the linearity of expectation, this separates into two terms:

$$h(X) = \frac{1}{2}\ln(2\pi\sigma^2) + \frac{1}{2\sigma^2}\mathbb{E}[(X - \mu)^2]$$


By definition, the expected squared deviation from the mean is the variance of the distribution:

$$\mathbb{E}[(X - \mu)^2] = \sigma^2$$

Substituting this back into our expression:

$$h(X) = \frac{1}{2}\ln(2\pi\sigma^2) + \frac{1}{2\sigma^2} \cdot \sigma^2 = \frac{1}{2}\ln(2\pi\sigma^2) + \frac{1}{2}=\frac{1}{2}\ln(2\pi e \sigma^2)$$

---

**Key Insights:**

* **Independence from the mean:** Notice that the mean $\mu$ does not appear in the final result. Differential entropy is translation-invariant; shifting the distribution doesn't change its uncertainty.
* **Dependence on variance:** The entropy scales directly with $\sigma^2$. As the variance increases, the distribution spreads out, leading to more uncertainty and higher entropy.
* **Maximum Entropy:** Under the constraint of a fixed variance $\sigma^2$, the Gaussian distribution has the **highest possible differential entropy** of any real-valued continuous random variable.

#### Visualizing Entropy as "Spread" of Measure

The higher the entropy, the more "spread out" the probability measure is.

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import entropy

# Define two probability measures (distributions)
# P1: Concentrated (Low Entropy)
# P2: Spread out (High Entropy)
x = np.linspace(0, 10, 100)
p1 = np.exp(-(x-5)**2 / (2 * 0.5**2))
p1 /= p1.sum()

p2 = np.exp(-(x-5)**2 / (2 * 2.0**2))
p2 /= p2.sum()

plt.figure(figsize=(10, 4))
plt.fill_between(x, p1, alpha=0.5, label=f'Low Entropy: {entropy(p1):.2f}')
plt.fill_between(x, p2, alpha=0.5, label=f'High Entropy: {entropy(p2):.2f}')
plt.title("Entropy as a Property of the Measure's Concentration")
plt.legend()
plt.show()
````

## 5.3 Entropy of a Discrete Random Variable (Shannon Entropy)

Just as differential entropy in the continuous setting is the negative relative entropy of $P_X$ with respect to the Lebesgue measure, **discrete entropy** is the negative relative entropy of $P_X$ with respect to the **counting measure**.

Let $X$ be a discrete random variable taking values in a countable set $S$. We consider the measurable space $(S, \mathscr{P}(S))$, where $\mathscr{P}(S)$ is the power set of $S$.


On $(S, \mathscr{P}(S))$, we establish our reference measure as the **counting measure** $\mu$, defined for any subset $A \subset S$ as the number of elements in $A$.

The distribution (law) of $X$ is a probability measure $P_X$ on $(S, \mathscr{P}(S))$. Since the only set $A \subset S$ for which $\mu(A) = 0$ is the empty set $\emptyset$, and $P_X(\emptyset) = 0$, the law $P_X$ is trivially **absolutely continuous** with respect to the counting measure $\mu$ ($P_X \ll \mu$).

By the Radon-Nikodym theorem, there exists a unique density function $p_X: S \to [0, 1]$, which is the Radon-Nikodym derivative of $P_X$ with respect to $\mu$:

$$\frac{dP_X}{d\mu}(x) = p_X(x)$$

For any point $x \in S$, this derivative is exactly the probability mass function:

$$p_X(x) = P_X(\{x\})$$

Thus, we can express the measure $P_X$ in terms of the counting measure as:

$$dP_X(x) = p_X(x) \, d\mu(x)$$

Using the original definition of the Kullback-Leibler divergence, we evaluate the relative entropy of the probability measure $P_X$ with respect to the non-probabilistic counting measure $\mu$ on $S$:

$$D_{KL}(P_X \| \mu) = \int_{S} \ln\left(\frac{dP_X}{d\mu}\right) \, dP_X$$

Substituting the density $p_X = \frac{dP_X}{d\mu}$ into the integral, we obtain:

$$D_{KL}(P_X \| \mu) = \int_{S} \ln(p_X(x)) \, dP_X(x)$$

Since $P_X$ is a discrete measure, this integral over the state space $S$ reduces exactly to a sum over the countable points:

$$D_{KL}(P_X \| \mu) = \sum_{x \in S} p_X(x) \ln p_X(x)$$

The **Shannon entropy** $H(X)$ is defined as the negative of this divergence. This reflects the uncertainty of the distribution relative to the uniform baseline represented by the counting measure:

$$H(X) = -D_{KL}(P_X \| \mu)$$

By expanding this equality, we recover the standard expressions for discrete entropy:

$$H(X) = -\int_{S} \ln\left(\frac{dP_X}{d\mu}\right) \, dP_X = -\sum_{x \in S} p_X(x) \ln p_X(x) = -\mathbb{E}[\ln p_X(X)]$$




## 5.4 The Entropy of a 'Spiked' Distribution

### 5.4.1 The Discrete Setting

Let our state space be the single-point set $S = \{x_0\}$.
* The law of our random variable $X$ is the delta measure $P_X = \delta_{x_0}$, which assigns a probability of $1$ to the point $x_0$:
  $$p_X(x_0) = P_X(\{x_0\}) = 1$$
* Our reference measure is the **counting measure** $\mu$ on $S$. Since the set has exactly one element, $\mu(\{x_0\}) = 1$.

Now, let's plug this into our entropy formula:

$$H_{\mu}(X) = -\sum_{x \in S} p_X(x) \ln p_X(x)$$

Since $S$ only contains the point $x_0$, this sum has exactly one term:

$$H_{\mu}(X) = - (1 \cdot \ln(1)) = - (1 \cdot 0) = 0$$

> **Conclusion:** When the delta measure is correctly treated as a discrete distribution on its support, its entropy is exactly **$0$ nats**. This matches our intuition perfectly: there is exactly zero uncertainty about the outcome of $X$.



### 5.4.2 The Continuous Setting

What happens if we view the delta measure as a limit of continuous random variables on $\mathbb{R}$, evaluated with respect to the **Lebesgue measure**?

Consider a sequence of normal distributions $X_\sigma \sim \mathscr{N}(x_0, \sigma^2)$. As the variance $\sigma^2 \to 0$, the probability density function $f_\sigma(x)$ collapses into the Dirac delta function $\delta(x - x_0)$.

Using the formula we derived for the differential entropy of a normal distribution:

$$h(X_\sigma) = \frac{1}{2} \ln(2 \pi e \sigma^2)$$

To find the differential entropy of the delta distribution, we take the limit as $\sigma \to 0$:

$$\lim_{\sigma \to 0} h(X_\sigma) = \lim_{\sigma \to 0} \frac{1}{2} \ln(2 \pi e \sigma^2) = -\infty$$



````python
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# 1. Setup data
x_range = np.linspace(-2, 2, 1000)
sigmas = [1.0, 0.5, 0.2, 0.1, 0.05]

def gaussian_pdf(x, sigma):
    return (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * (x / sigma)**2)

def diff_entropy(sigma):
    # h(X) = 0.5 * ln(2 * pi * e * sigma^2)
    return 0.5 * np.log(2 * np.pi * np.e * (sigma**2))

entropy_sigmas = np.logspace(-3, 0.5, 300)
entropy_vals = diff_entropy(entropy_sigmas)

# 2. Create Plotly Figure with Subplots
fig = make_subplots(
    rows=1, cols=2,
    subplot_titles=(
        "PDF Collapse: $\\sigma \\to 0$",
        "Entropy Divergence: $h(X_\\sigma) \\to -\\infty$"
    ),
    horizontal_spacing=0.15
)

# Left Plot: PDF traces
for s in sigmas:
    fig.add_trace(
        go.Scatter(x=x_range, y=gaussian_pdf(x_range, s), name=f"σ={s}"),
        row=1, col=1
    )

# Right Plot: Entropy curve
fig.add_trace(
    go.Scatter(
        x=entropy_sigmas,
        y=entropy_vals,
        name="Differential Entropy",
        line=dict(color='red', width=3)
    ),
    row=1, col=2
)

# Layout adjustments
fig.update_xaxes(title_text="x", row=1, col=1)
fig.update_yaxes(title_text="Density", row=1, col=1)

fig.update_xaxes(title_text="Standard Deviation (σ)", type="log", row=1, col=2)
fig.update_yaxes(title_text="Entropy (nats)", row=1, col=2)

fig.update_layout(
    title="The Delta Measure as a Limit of Gaussian Distributions",
    template="plotly_white",
    height=500,
    showlegend=True
)

fig.show()
````

### 5.4.3 Resolving the Paradox ?

You might ask: *Why does the continuous approach say $-\infty$ while the discrete approach says $0$?*

This occurs because the **Lebesgue measure** and the **counting measure** treat points differently:
* To the counting measure, a point has a reference size of **$1$**.
* To the Lebesgue measure, a point has a reference size of **$0$**.

When you squeeze a continuous distribution down to a single point, you are concentrating all of its probability mass into a region of Lebesgue measure zero. The probability density $f(x)$ shoots up to infinity, so $\ln f(x) \to \infty$.

Thus, the $-\infty$ differential entropy simply tells us that a point has **infinitely less uncertainty** than a continuous interval of Lebesgue measure.

**Summary:**

Depending on your reference measure, the delta measure has:
* **$0$ entropy** relative to the counting measure (there is no uncertainty).
* **$-\infty$ entropy** relative to the Lebesgue measure (it is infinitely more certain than any continuous spread).

## 5.5 Comparison of Differential Entropy and Shannon Entropy


1. Shannon Entropy (Discrete)
For a discrete random variable $X$, Shannon entropy is defined as:
$$H(X) = -\sum_{i} P(x_i) \log P(x_i)$$

* **Range:** $[0, \infty)$
* **Why it can't be negative:** Since $0 \le P(x_i) \le 1$, the term $\log P(x_i)$ is always $\le 0$. Multiplying by the negative sign in the formula ensures $H(X)$ is always non-negative.
* **Physical Meaning:** It represents the average number of bits (or nats) needed to identify an outcome. You can't need "negative bits" to describe something.
* **Minimum Value:** $0$ (occurs when one outcome has probability $1$; there is no uncertainty).

---

2. Differential Entropy (Continuous)
For a continuous random variable $X$ with PDF $f(x)$, differential entropy is defined as:
$$h(X) = -\int_{-\infty}^{\infty} f(x) \log f(x) dx$$

* **Range:** $(-\infty, \infty)$
* **Why it can be negative:** Unlike probabilities, a probability **density** $f(x)$ can be much greater than $1$.
    > **Example:** If $X$ is uniform on $[0, 0.1]$, then $f(x) = 10$. The $\log(10)$ term is positive, so $-\int 10 \log(10) dx$ becomes **negative**.
* **Physical Meaning:** It does **not** represent the absolute number of bits to describe the value (which would be infinite for any continuous variable). Instead, it describes the volume of the support relative to the Lebesgue measure.
* **The Delta Limit:** As you saw in the previous plot, as a distribution narrows (like the Dirac Delta), $f(x) \to \infty$, which forces $h(X) \to -\infty$.

---


| Feature | Shannon Entropy $H(X)$ | Differential Entropy $h(X)$ |
| :--- | :--- | :--- |
| **Domain** | Discrete | Continuous |
| **Range** | $[0, \infty)$ | $(-\infty, \infty)$ |
| **$P(x)$ vs $f(x)$** | Probabilities ($\le 1$) | Densities (can be $> 1$) |
| **Coordinate Change** | Invariant | Changes with scaling ($h(aX) = h(X) + \ln|a|$) |
| **Interpretation** | Absolute uncertainty | Relative uncertainty |


## 5.6 Conditional Entropy

###5.6.1 Conditional Differential Entropy, $(h(X\mid Y))$


The conditional differential entropy $h(X \mid Y)$ measures the amount of uncertainty remaining in the random variable $X$, **on average**, after the random variable $Y$ has been observed.

Mathematically, it is the expected value (over all possible outcomes of $Y$) of the differential entropy of the conditional random variable $X \mid Y = y$.

Recall that:
1. For a fixed $y$, the random variable $X \mid Y=y$ has the law $P_{X \mid Y=y}$ and the PDF $f_{X \mid Y}(x \mid y)$.
2. The differential entropy of this specific conditional random variable is:
   $$h(X \mid Y = y) = -\int_{\mathscr{X}} f_{X \mid Y}(x \mid y) \ln f_{X \mid Y}(x \mid y) \, dx$$
3. The conditional entropy $h(X \mid Y)$ is defined as the expected value of $h(X \mid Y = y)$ across all possible values of the conditioning random variable $Y$:
   $$h(X \mid Y) \triangleq \mathbb{E}_Y [ h(X \mid Y = Y) ] = \int_{\mathscr{Y}} h(X \mid Y = y) f_Y(y) \, dy$$


Substituting the expression for $h(X \mid Y = y)$ into the expectation gives the complete formula for the conditional differential entropy:

$$h(X \mid Y) = -\int_{\mathscr{Y}} \left( \int_{\mathscr{X}} f_{X \mid Y}(x \mid y) \ln f_{X \mid Y}(x \mid y) \, dx \right) f_Y(y) \, dy$$

Since $f_{X,Y}(x, y) = f_{X \mid Y}(x \mid y) f_Y(y)$, we can merge the terms into a single joint expectation:

$$h(X \mid Y) = -\int_{\mathscr{Y}} \int_{\mathscr{X}} f_{X,Y}(x, y) \ln f_{X \mid Y}(x \mid y) \, dx \, dy$$



### 5.6.2 Conditional Shannon Entropy, $(H(X\mid Y))$

The conditional entropy $H(X \mid Y)$ measures the amount of uncertainty remaining in the discrete random variable $X$, **on average**, after the discrete random variable $Y$ has been observed.

Mathematically, it is the expected value (over all possible outcomes of $Y$) of the entropy of the conditional random variable $X \mid Y = y$.


1.  For a fixed $y$, the random variable $X \mid Y=y$ follows the conditional probability mass function (PMF) $P(x \mid y)$.
2.  The entropy of this specific conditional random variable is:
$$H(X \mid Y = y) = -\sum_{x \in \mathscr{X}} P(x \mid y) \log P(x \mid y)$$
3.  The conditional entropy $H(X \mid Y)$ is defined as the expected value of $H(X \mid Y = y)$ across all possible values of the conditioning random variable $Y$:
$$H(X \mid Y) \triangleq \mathbb{E}_Y [ H(X \mid Y = Y) ] = \sum_{y \in \mathscr{Y}} P(y) H(X \mid Y = y)$$

Substituting the expression for $H(X \mid Y = y)$ into the expectation gives the full formula for discrete conditional entropy:

$$H(X \mid Y) = -\sum_{y \in \mathscr{Y}} P(y) \sum_{x \in \mathscr{X}} P(x \mid y) \log P(x \mid y)$$

Since the joint probability is defined as $P(x, y) = P(x \mid y) P(y)$, we can simplify the terms into a single joint expectation:

$$H(X \mid Y) = -\sum_{x \in \mathscr{X}, y \in \mathscr{Y}} P(x, y) \log P(x \mid y)$$

> **Note:** Unlike differential entropy, which can be negative, discrete conditional entropy $H(X \mid Y)$ is always non-negative ($H(X \mid Y) \geq 0$).

### 5.6.3 The case of independent $X$ and $Y$

If $X$ and $Y$ are independent, the following two properties hold:
1.  **Joint PDF:** The joint density is the product of the marginal densities:

$$f_{X,Y}(x, y) = f_X(x) f_Y(y)$$

2.  **Conditional PDF:** The conditional density of $X$ given $Y$ is simply the marginal density of $X$:
    
$$f_{X|Y}(x|y) = \frac{f_{X,Y}(x, y)}{f_Y(y)} = \frac{f_X(x)f_Y(y)}{f_Y(y)} = f_X(x)$$


Starting with the definition of conditional entropy provided:
$$h(X \mid Y) = -\int_{\mathscr{Y}} \int_{\mathscr{X}} f_{X,Y}(x, y) \ln f_{X \mid Y}(x \mid y) \, dx \, dy$$

We substitute the independence identities ($f_{X,Y} = f_X f_Y$ and $f_{X|Y} = f_X$) into the integral:
$$h(X \mid Y) = -\int_{\mathscr{Y}} \int_{\mathscr{X}} f_X(x) f_Y(y) \ln f_X(x) \, dx \, dy$$


Since the term $\ln f_X(x)$ does not depend on $y$, we can rearrange the double integral as a product of two separate integrals:

$$h(X \mid Y) = \left( \int_{\mathscr{Y}} f_Y(y) \, dy \right) \left( -\int_{\mathscr{X}} f_X(x) \ln f_X(x) \, dx \right)=h(X).$$

Following this same reasoning one can show that if $X$ and $Y$ are independent
$$H(X\mid Y)= H(X).$$


## 5.7  Entropy of the Conditional Expectation, $(h(\mathbb{E}(X \mid Y)))$

Recall that $\mathbb{E}[X \mid Y] = g(Y)$ is a *random variable* that is a deterministic function of $Y$.

When we evaluate $h(\mathbb{E}(X \mid Y))$, we are calculating the differential entropy of this **new random variable**, $Z = g(Y) = \mathbb{E}[X \mid Y]$.

Because $Z = g(Y)$ is a function of $Y$, it lives on its own state space $\mathscr{Z} \subseteq \mathbb{R}^n$. To compute its entropy, we must determine its own probability density function, which we will call $f_Z(z)$.


The density $f_Z(z)$ is obtained by mapping the marginal distribution of $Y$ through the function $g(y) = \int_{\mathscr{X}} x f_{X \mid Y}(x \mid y) \, dx$.

If $g(y)$ is a smooth, invertible function, the density of $Z$ can be found via the standard transformation of variables using the Jacobian determinant:

$$f_Z(z) = f_Y(g^{-1}(z)) \left| \det \frac{\partial g^{-1}(z)}{\partial z} \right|$$

Once this density is established, the differential entropy $h(\mathbb{E}(X \mid Y))$ is evaluated as:

$$h(\mathbb{E}(X \mid Y)) = h(Z) = -\int_{\mathscr{Z}} f_Z(z) \ln f_Z(z) \, dz.$$

## 5.8 Mutual Information

### 5.8.1 For Continuous Random Variables

Mutual Information measures **how far the joint distribution $P_{X,Y}$ is from the case where $X$ and $Y$ are independent** ($P_X \otimes P_Y$).

$$I(X;Y) \triangleq D_{KL}(P_{X,Y} \, \| \, P_X P_Y)$$

Using the Radon-Nikodym derivative (the ratio of densities), this is defined as:


$$I(X;Y) = \int_{\mathfrak{X} \times \mathfrak{Y}} f_{X,Y}(x,y) \ln \left( \frac{f_{X,Y}(x,y)}{f_X(x) f_Y(y)} \right) dx \, dy$$

---

**Relationship Between Mutual Information and Conditional Entropy**

Starting from mutual information,

$$
I(X;Y)=\iint f_{X,Y}(x,y)\ln\frac{f_{X,Y}(x,y)}{f_X(x)f_Y(y)}\,dx\,dy,
$$

use

$$
f_{X,Y}(x,y)=f_{X|Y}(x|y)f_Y(y)
$$

to get

$$
I(X;Y)=\iint f_{X,Y}(x,y)\ln\frac{f_{X|Y}(x|y)}{f_X(x)}\,dx\,dy.
$$

Then

$$
I(X;Y)
=\iint f_{X,Y}(x,y)\ln f_{X|Y}(x|y)\,dx\,dy
=\iint f_{X,Y}(x,y)\ln f_X(x)\,dx\,dy
-h(X|Y)+h(X).
$$

Hence

$$
I(X;Y)=h(X)-h(X|Y).
$$

Interestingly, even though $h(X)$ can be negative, Mutual Information $I(X;Y)$ is always non-negative ($I(X;Y) \ge 0$), even in the continuous case. This is because $I(X;Y)$ measures the reduction in uncertainty. Even if the starting uncertainty is negative, knowing $Y$ can only make you "more certain" (more negative) or leave you where you are. The difference—the information gained—remains a positive quantity.Is the idea of "negative uncertainty" starting to click, or does it still feel a bit like mathematical sorcery?

---

**Summary**

> Mutual Information is the **"Universal Information Gain."** It is the reduction in the entropy of $X$ achieved by observing $Y$. In other words it measures the percentage of **total uncertainty** (entropy) removed.




### 5.8.2 For Discrete Random Variables

**Mutual Information:**

$I(X;Y)$ measures the statistical dependence between two discrete variables by quantifying how much the joint distribution $P(x,y)$ deviates from independence $P(x)P(y)$.

$$I(X;Y) = D_{KL}(P(x,y) \| P(x)P(y)) = \sum_{x,y} P(x,y) \log \left( \frac{P(x,y)}{P(x)P(y)} \right)$$

---

**Relationship with Entropy:**

By substituting $P(x,y) = P(x|y)P(y)$, we can decompose the formula:

$$I(X;Y) = \sum_{x,y} P(x,y) \log P(x|y) - \sum_{x,y} P(x,y) \log P(x)$$
$$I(X;Y) = -H(X|Y) + H(X)$$

Resulting in the fundamental identity:
$$I(X;Y) = H(X) - H(X|Y)$$

---

**Summary:**

* **Information Gain:** $I(X;Y)$ is the reduction in uncertainty of $X$ due to the knowledge of $Y$.
* **Non-negativity:** Unlike differential entropy, both $H(X)$ and $I(X;Y)$ are always $\geq 0$ in the discrete case.
* **Symmetry:** It captures the information shared between variables, such that $I(X;Y) = I(Y;X)$.

### 5.8.3 Examples

#### 5.8.3.1 Example: The Two coin flip problem

Suppose we flip two independent, fair coins.
* Let **$X \in \{0, 1\}$** be the outcome of the first coin ($1 = \text{Heads}$, $0 = \text{Tails}$).
* Let **$Y \in \{0, 1, 2\}$** be the total number of heads observed across both coins.

We want to quantify the amount of information that $Y$ reveals about $X$ using three metrics from information theory: **Entropy $H(X)$**, **Conditional Entropy $H(X|Y)$**, and **Mutual Information $I(X;Y)$**. Because we want our answers in **bits**, we will use base-2 logarithms ($\log_2$).

---

There are $4$ equally likely outcomes for the two coin flips: $(\text{T,T})$, $(\text{T,H})$, $(\text{H,T})$, and $(\text{H,H})$. From these, we construct the joint probability distribution $P(X=x, Y=y)$:

| $X$ (First Coin) \ $Y$ (Total Heads) | $Y = 0$ | $Y = 1$ | $Y = 2$ | Marginal $P(X=x)$ |
| :--- | :--- | :--- | :--- | :--- |
| **$X = 0$** (Tails) | $0.25$ | $0.25$ | $0.00$ | **$0.50$** |
| **$X = 1$** (Heads) | $0.00$ | $0.25$ | $0.25$ | **$0.50$** |
| **Marginal $P(Y=y)$** | **$0.25$** | **$0.50$** | **$0.25$** | **$1.00$** |

---

The marginal entropy $H(X)$ measures the total a priori uncertainty about the first coin before observing $Y$.

$$H(X) = -\sum_{x \in \{0,1\}} P(X=x) \log_2 P(X=x)$$

Substituting our marginal probabilities for $X$:

$$H(X) = -\left(0.5 \log_2(0.5) + 0.5 \log_2(0.5)\right) = -\left(0.5(-1) + 0.5(-1)\right) = \mathbf{1 \text{ bit}}$$

> **Interpretation:** A fair coin flip has exactly $1$ bit of uncertainty.

---

The conditional entropy $H(X|Y)$ measures the uncertainty remaining in $X$ *after* we learn the value of $Y$. It is the weighted average of the uncertainties $H(X|Y=y)$ across all possible values of $Y$.

$$H(X|Y) = \sum_{y \in \{0,1,2\}} P(Y=y) H(X|Y=y)$$


Let us calculate the uncertainty $H(X \mid Y=y)$ for each specific outcome of $Y$:

**Case A: If $Y = 0$ (Zero total heads):**

We know with certainty that both coins were Tails. Therefore, $P(X=0 \mid Y=0) = 1$ and $P(X=1 \mid Y=0) = 0$.
$$H(X|Y=0) = -(1 \log_2 1 + 0 \log_2 0) = \mathbf{0 \text{ bits}}$$

**Case B: If $Y = 2$ (Two total heads):**

We know with certainty that both coins were Heads. Therefore, $P(X=0 \mid Y=2) = 0$ and $P(X=1 \mid Y=2) = 1$.
$$H(X|Y=2) = -(0 \log_2 0 + 1 \log_2 1) = \mathbf{0 \text{ bits}}$$

**Case C: If $Y = 1$ (Exactly one total head):**

The single head could be on either the first coin or the second coin with equal probability. Therefore, $P(X=0 \mid Y=1) = 0.5$ and $P(X=1 \mid Y=1) = 0.5$.
$$H(X|Y=1) = -(0.5 \log_2 0.5 + 0.5 \log_2 0.5) = \mathbf{1 \text{ bit}}$$

**Bringing it all together:**

We average these specific entropies, weighting them by the marginal probabilities $P(Y=y)$:

$$H(X|Y) = (0.25 \times 0 \text{ bits}) + (0.50 \times 1 \text{ bit}) + (0.25 \times 0 \text{ bits}) = \mathbf{0.5 \text{ bits}}$$

---


The mutual information $I(X;Y)$ measures the reduction in uncertainty about $X$ that results from learning $Y$:

$$I(X;Y) = H(X) - H(X|Y)$$

Plugging in our derived values:

$$I(X;Y) = 1 \text{ bit} - 0.5 \text{ bits} = \mathbf{0.5 \text{ bits}}$$

---

**Physical Interpretation:**

* **Before observing $Y$:** You have $1$ bit of uncertainty about the outcome of the first coin ($X$).
* **After observing $Y$:** On average, half of that uncertainty is eliminated.
    * If $Y=0$ or $Y=2$ ($50\%$ of the time), you know exactly what $X$ did; your residual uncertainty drops to $0$ bits.
    * If $Y=1$ ($50\%$ of the time), you gain no information about $X$; your uncertainty remains at $1$ bit.

Thus, knowing the total number of heads across two coin flips provides exactly **$0.5$ bits** of information about the outcome of the first coin.

#### 5.8.3.2 The problem of noisy measurements

Let $X$ represent the true value of a physical quantity (such as temperature or voltage), and let $Y$ be a noisy reading from a sensor. We model the noise $N$ as additive, meaning:

$$Y = X + N$$

We assume that:
* The true signal $X \sim \mathscr{N}(0, \sigma_X^2)$ is a Gaussian random variable.
* The sensor noise $N \sim \mathscr{N}(0, \sigma_N^2)$ is a Gaussian random variable.
* $X$ and $N$ are completely independent.

**Goal:** Derive the expressions for the original uncertainty $h(X)$, the information gained $I(X;Y)$, and the remaining uncertainty $h(X \mid Y)$ given by:

$$h(X \mid Y) = h(X) - I(X;Y)$$

---


**Step 1: Calculate the Original Uncertainty $h(X)$:**

Since $X$ is a normally distributed random variable with variance $\sigma_X^2$, its differential entropy (measured in nats) follows the standard Gaussian entropy formula derived earlier:

$$h(X) = \frac{1}{2} \ln(2\pi e \sigma_X^2)$$

---

**Step 2: Calculate the Mutual Information $I(X;Y)$:**

To calculate the information gained, we use the alternative formula for mutual information:

$$I(X;Y) = h(Y) - h(Y \mid X)$$

We must evaluate both $h(Y)$ and $h(Y \mid X)$:


Because $X$ and $N$ are independent Gaussian random variables, their sum $Y = X + N$ is also a Gaussian random variable. Its variance is the sum of their individual variances:

$$\text{Var}(Y) = \sigma_X^2 + \sigma_N^2$$

Thus, the differential entropy of the noisy sensor reading $Y$ is:

$$h(Y) = \frac{1}{2} \ln(2\pi e (\sigma_X^2 + \sigma_N^2))$$


The conditional entropy $h(Y \mid X)$ represents the uncertainty remaining in the measurement $Y$ once we know the exact true value $X$.

Substituting $Y = X + N$ gives:

$$h(Y \mid X) = h(X + N \mid X) = h(N \mid X)$$

Because $N$ and $X$ are independent, conditioning on $X$ does not change the uncertainty of $N$:

$$h(N \mid X) = h(N) = \frac{1}{2} \ln(2\pi e \sigma_N^2)$$


Now we substitute $h(Y)$ and $h(Y \mid X)$ into our mutual information equation:

$$I(X;Y) = \frac{1}{2} \ln(2\pi e (\sigma_X^2 + \sigma_N^2)) - \frac{1}{2} \ln(2\pi e \sigma_N^2)$$

Using the properties of logarithms to combine the terms:

$$I(X;Y) = \frac{1}{2} \ln\left( \frac{2\pi e (\sigma_X^2 + \sigma_N^2)}{2\pi e \sigma_N^2} \right) = \mathbf{\frac{1}{2} \ln\left(1 + \frac{\sigma_X^2}{\sigma_N^2}\right)}$$

> **Interpretation:** The term $\frac{\sigma_X^2}{\sigma_N^2}$ is the **Signal-to-Noise Ratio (SNR)**. The information gained by reading the sensor is directly proportional to the SNR.

---

**Step 3: Calculate the Remaining Uncertainty $h(X \mid Y)$:**

Using the identity $h(X \mid Y) = h(X) - I(X;Y)$, we can now compute the residual uncertainty about the true value $X$ after reading the sensor $Y$:

$$h(X \mid Y) = \frac{1}{2} \ln(2\pi e \sigma_X^2) - \frac{1}{2} \ln\left(1 + \frac{\sigma_X^2}{\sigma_N^2}\right)$$

Combining the logarithms:

$$h(X \mid Y) = \frac{1}{2} \ln\left( \frac{2\pi e \sigma_X^2}{1 + \frac{\sigma_X^2}{\sigma_N^2}} \right)$$

We can simplify the denominator by finding a common denominator:

$$1 + \frac{\sigma_X^2}{\sigma_N^2} = \frac{\sigma_N^2 + \sigma_X^2}{\sigma_N^2}$$

Plugging this back in:

$$h(X \mid Y) = \frac{1}{2} \ln\left( \frac{2\pi e \sigma_X^2}{\frac{\sigma_N^2 + \sigma_X^2}{\sigma_N^2}} \right) = \mathbf{\frac{1}{2} \ln\left(2\pi e \frac{\sigma_X^2 \sigma_N^2}{\sigma_X^2 + \sigma_N^2}\right)}$$

---

**Physical Interpretation:**

The results reveal a clear relationship between sensor noise ($\sigma_N^2$) and information flow:

```
                  ┌──────────────────────┐
                  │ Original Signal h(X) │
                  └──────────┬───────────┘
                             ▼
              ┌──────────────┴──────────────┐
              │ Sensor Measurement Noise    │
              └──────────────┬──────────────┘
                             ▼
             ┌───────────────┴───────────────┐
             │       Signal to Noise         │
             │     SNR = σ_X^2 / σ_N^2       │
             └───────────────┬───────────────┘
                             │
            ┌────────────────┴────────────────┐
            ▼                                 ▼
   High Noise (σ_N^2 → ∞)            Low Noise (σ_N^2 → 0)
   • SNR → 0                         • SNR → ∞
   • I(X;Y) ≈ 0                      • I(X;Y) is large
   • h(X|Y) ≈ h(X)                   • h(X|Y) → -∞ (absolute certainty)
```

---

**Scenario A: The Very Noisy Sensor ($\sigma_N^2 \to \infty$):**

If the sensor has high noise, then $\sigma_N^2$ is much larger than the signal variance $\sigma_X^2$:
* The SNR $\frac{\sigma_X^2}{\sigma_N^2} \approx 0$.
* The information gain is $I(X;Y) = \frac{1}{2} \ln(1 + 0) = \mathbf{0 \text{ nats}}$.
* The remaining uncertainty is $h(X|Y) \approx h(X)$.

> **Conclusion:** A bad sensor reduces almost no uncertainty about the true physical quantity.

**Scenario B: The Highly Accurate Sensor ($\sigma_N^2 \to 0$):**

If the sensor is highly precise, then $\sigma_N^2$ is tiny compared to $\sigma_X^2$:
* The SNR $\frac{\sigma_X^2}{\sigma_N^2} \to \infty$.
* The information gain is $I(X;Y) \to \mathbf{\infty \text{ nats}}$.
* The remaining uncertainty is $h(X \mid Y) = \frac{1}{2} \ln(2 \pi e \cdot 0) = \mathbf{-\infty \text{ nats}}$.

---

> **Conclusion:** An ultra-precise sensor collapses the uncertainty of the true quantity down to zero (which corresponds to $-\infty$ differential entropy).

#### 5.8.3.3 The conditional expectation $E[X \mid Y]$ for the linear Gaussian model

In this section, we derive the distribution of the conditional expectation $E[X \mid Y]$ for the linear Gaussian model. This is a fundamental result in estimation theory, showing how the "best estimate" itself behaves as a random variable.

---

1. The Model Setup

Consider the signal-plus-noise model:
* **Signal:** $X \sim \mathscr{N}(0, \sigma_X^2)$
* **Noise:** $N \sim \mathscr{N}(0, \sigma_N^2)$, where $X \perp N$.
* **Observation:** $Y = X + N$.

Since $Y$ is a linear combination of independent Gaussians, the joint distribution of $(X, Y)$ is bivariate normal:
$$\begin{pmatrix} X \\ Y \end{pmatrix} \sim \mathscr{N} \left( \begin{pmatrix} 0 \\ 0 \end{pmatrix}, \begin{pmatrix} \sigma_X^2 & \sigma_X^2 \\ \sigma_X^2 & \sigma_X^2 + \sigma_N^2 \end{pmatrix} \right)$$
*(Note: $\text{Cov}(X, Y) = E[X(X+N)] = E[X^2] = \sigma_X^2$.)*

---

2. Deriving the Conditional Expectation $E[X \mid Y]$

For a bivariate normal distribution, the conditional expectation is given by the linear projection:
$$E(X \mid Y = y) = \mu_X + \frac{\text{Cov}(X, Y)}{\text{Var}(Y)}(y - \mu_Y)$$

Substituting our values:
$$E[X \mid Y = y] = 0 + \frac{\sigma_X^2}{\sigma_X^2 + \sigma_N^2}(y - 0)$$

Let $k = \frac{\sigma_X^2}{\sigma_X^2 + \sigma_N^2}$. This constant $k$ is often called the **Kalman Gain** or the **Shrinkage Factor**. It represents the "reliability" of the sensor.
Thus, the random variable is:
$$\widehat{X} = E[X \mid Y] = kY$$

---

3. Distribution of the Random Variable $E[X \mid Y]$

To find the distribution of $\widehat{X} = kY$, we recall that a linear transformation of a Gaussian is still Gaussian.

**A. The Mean:**
$$E[\widehat{X}] = E[kY] = k E[Y] = 0$$

**B. The Variance:**
$$\text{Var}(\widehat{X}) = \text{Var}(kY) = k^2 \text{Var}(Y)$$
Substitute $k$ and $\text{Var}(Y)$:
$$\text{Var}(\widehat{X}) = \left( \frac{\sigma_X^2}{\sigma_X^2 + \sigma_N^2} \right)^2 (\sigma_X^2 + \sigma_N^2)$$
$$\text{Var}(\widehat{X}) = \frac{(\sigma_X^2)^2}{\sigma_X^2 + \sigma_N^2}$$

**Conclusion:**
$$E[X \mid Y] \sim \mathscr{N}\left( 0, \frac{\sigma_X^4}{\sigma_X^2 + \sigma_N^2} \right)$$

---

4. Decomposition of Variance

A profound result in probability is that the total variance of $X$ can be decomposed into the "explained" variance and the "unexplained" (residual) variance:
$$\text{Var}(X) = \text{Var}(E[X \mid Y]) + E[\text{Var}(X \mid Y)]$$

From our derivation:
1.  **Total Variance:** $\sigma_X^2$
2.  **Explained Variance (Variance of the Estimate):** $\frac{\sigma_X^4}{\sigma_X^2 + \sigma_N^2}$
3.  **Unexplained Variance (Error Variance):** $\sigma_X^2 - \frac{\sigma_X^4}{\sigma_X^2 + \sigma_N^2} = \frac{\sigma_X^2 \sigma_N^2}{\sigma_X^2 + \sigma_N^2}$

---

5. Practical Interpretation
* **Perfect Sensor ($\sigma_N^2 \to 0$):** $k \to 1$. The estimate $E[X \mid Y] \to Y$. The variance of the estimate matches the variance of $X$ perfectly.
* **Useless Sensor ($\sigma_N^2 \to \infty$):** $k \to 0$. The estimate $E[X \mid Y] \to 0$. The variance of the estimate is $0$ because we always guess the mean.

> **Key Takeaway:** The variance of the conditional expectation $E[X \mid Y]$ is always **less than or equal to** the variance of $X$. This is because the estimate is a "smoothed" or "shrunken" version of the reality, filtered through the lens of a noisy observation.

## 5.9 Joint Entropy and Conditional Entropy

1. **Differential Entropy (Continuous Case):**
For continuous random variables with a joint density $f(x,y)$, the relationship is mathematically identical because the joint density follows the same product rule:
$$f(x,y) = f(x)f(y|x).$$
* The differential entropy is defined via the Lebesgue integral:
$$h(X,Y) = -\int \int f(x,y) \log f(x,y) \, dx dy.$$
* Just as in the discrete case, the linearity of the integral and the property of the logarithm allow the joint entropy to be split into the marginal and conditional components:
$$h(X,Y) = h(X) + h(Y|X).$$

2. **Shannon Entropy (Discrete Case):**
For discrete random variables, the joint probability can be decomposed using the product rule:
$$P(X, Y) = P(X)P(Y|X).$$
* The joint entropy is defined as:
$$H(X,Y) = -\sum_{x,y} p(x,y) \log p(x,y).$$
* By substituting the product rule into the logarithm:
$$\log p(x,y) = \log p(x) + \log p(y|x)$$.
* Distributing the summation and expectation leads to the identity:
$$H(X,Y) = H(X) + H(Y|X).$$
