# 1. Definition of the Conditional Expectation $(\mathbb{E}(X|Y))$

Let $(\Omega, \mathfrak{A}, P)$ be a probability space.

* Let $X: \Omega \to \mathbb{R}$ be a random variable that is integrable ($X \in L^1$).
* Let $Y: \Omega \to \mathbb{R}$ be another random variable.
* Let $\sigma(Y)$ be the **sub-$\sigma$-algebra** generated by $Y$. This represents "the information contained in $Y$."

---

**The Existence: Conditional Expectation**

The conditional expectation of $X$ given $Y$, denoted $\mathbb E[X|Y]$ (or more formally $\mathbb E[X|\sigma(Y)]$), is defined as the unique (almost surely) random variable $Z$ that satisfies two conditions:

1. **Measurability:** $Z$ is measurable with respect to $\sigma(Y)$.
That is byt the Doob-Dynkin lemma $Z$ must be a function of $Y$:
$$Z = g(Y).$$
2. **The Integral Property (Partial Averaging):** For every set $G \in \sigma(Y)$:

$$\int_G Z \, dP = \int_G X \, dP$$



*(Translation: On any set of outcomes that $Y$ can distinguish, the "average" of our estimate $Z$ must match the "average" of the actual $X$.)*

The existence and uniqueness of $Z$ are guaranteed by the **Radon-Nikodym Theorem** and is beyond the scope of this course.

# 2 The Conditional Density ($f_{X\mid Y}$)

##2.1 The Joint Probability ($P(X\in B\mid Y\in A)$)

Let
$$
Z_B\triangleq \mathbb E[\mathbb 1_{\{X\in B\}}\mid Y].
$$

Since the event

$$
Y^{-1}(A)
$$

belongs to $\sigma(Y)$, the defining property of conditional expectation gives

$$
\int_{Y^{-1}(A)} Z_B\,dP=\int_{Y^{-1}(A)} \mathbb 1_{\{X\in B\}}\,dP
=\int_{\Omega} \mathbb 1_{\{Y\in A\}}\mathbb 1_{\{X\in B\}}\,dP.
$$

Now note that

$$
\mathbb 1_{\{Y\in A\}}\mathbb 1_{\{X\in B\}}=\mathbb 1_{\{X\in B,\;Y\in A\}},
$$

so

$$
\int_{Y^{-1}(A)} \mathbb 1_{{X\in B}}\,dP
=\int_{\Omega} \mathbb 1_{\{X\in B\}}\mathbb 1_{\{Y\in A\}}\,dP=P(X\in B,\;Y\in A).
$$

Therefore

$$
P(X\in B,\;Y\in A)=\int_{Y^{-1}(A)} Z_B\,dP.
$$

---

$Z_B$ is measurable with respect to $\sigma(Y)$.
Hence from the Doob-Dynkin Lemma $Z_B$ must be a function of $Y$. We can write $Z_B = g_B(Y)$ and then the above expression can be written as:

$$
P(X\in B,\;Y\in A)=\int_{Y^{-1}(A)} g_B(Y)\,dP.
$$

Now applying the pushforward rule:
$$
\int_{Y^{-1}(A)} g_B(Y)\,dP=\int_A g_B(y)\,dP_Y(y),
$$

we have

$$
P(X\in B,\;Y\in A)=\int_A g_B(y)\,dP_Y(y).
$$

If $Y$ has density $f_Y$, so that

$$
dP_Y(y)=f_Y(y)\,dy,
$$

then this becomes

$$
P(X\in B,\;Y\in A)=\int_A g_B(y)\,f_Y(y)\,dy.
$$


A concise interpretation is:

> The joint probability $P(X\in B,\;Y\in A)$ is obtained by averaging the conditional probability $P(X\in B\mid Y=y)$ over all values $y\in A$, weighted by the distribution of $Y$.


Denote by $f_{X,Y}(x,y)$ the density of the joint distribution of $(X,Y)$ then
$$
P(X\in B,\;Y\in A)=\iint_{A\times B}dP_{X,Y}=\int_{A}\int_{B} f_{X,Y}(x,y)\,dx\,\,dy.
$$



## 2.2 The Conditional Probability Measure ($P_{X \mid Y=y}$)

If for $P_Y$-almost every $y$, the mapping $B \mapsto g_B(y)$ satisfies the axioms of a probability measure, then we define the conditional measure $P_{X \mid Y=y}$ by
$$P_{X \mid Y=y}(B)\triangleq g_B(y).$$
If this measure is absolutely continuous with respect to the Lebesgue measure, then there exists a density $f_{X|Y}(x \mid y)$ such that:
$$
g_B(y)
=\int_B dP_{X \mid Y=y}= \int_B f_{X|Y}(x \mid y) \, dx.
$$
Specifically, if this measure is absolutely continuous with respect to the Lebesgue measure, the density $f_{X|Y}(x \mid y)$ is the Radon-Nikodym derivative of $P_{X \mid Y=y}$ with respect to the Lebesgue measure:
$$f_{X|Y}(x \mid y) = \frac{dP_{X \mid Y=y}}{dx}(x).$$

## 2.2.1 The conditional Probability, $P(X\in B\mid Y=y)$

In measure theory, the "probability" of an event is simply the **measure of the set** representing that event. The expression
$$
g_B(y)
=\int_B dP_{X \mid Y=y}=P_{X \mid Y=y}(B).
$$
indicates that $g_B(y)$ gives the conditional probality of $X\in B$ given that $Y=y$. That is
$$
g_B(y)\triangleq P(X\in B\mid Y=y).
$$

---

**Summary:**

* **Step A:** Define $Z_B = \mathbb{E}[\mathbb{1}_{\{X\in B\}} \mid Y]$.
* **Step B:** Use Doob-Dynkin to show $Z_B = g_B(Y)$.
* **Step C:** Establish $P_{X \mid Y=y}(B) = g_B(y)$ as a regular conditional measure.
* **Step D:** Conclude that since $g_B(y)$ calculates the mass of $B$ given $y$, it **is** the conditional probability $P(X \in B \mid Y=y)$.


## 2.3 The Product Rule of the Conditional Density

### 2.3.2 Continuous Case

Substituting the above expression, $
g_B(y)=\int_B f_{X|Y}(x \mid y) \, dx
$, into the integral $
P(X\in B,\;Y\in A)=\int_A g_B(y)\,f_Y(y)\,dy
$ we have:
$$P(X \in B, Y \in A)
= \int_A \int_B f_{X|Y}(x \mid y) \, f_Y(y)\,dx \, dy.
$$

Since $P(X \in B, Y \in A) = \iint_{A \times B} f_{X,Y}(x,y) \, dx \, dy$ where $f_{X,Y}(x,y)$ is the joint probability density,
we obtain the product rule

$$
f_{X,Y}(x,y)=f_{X|Y}(x\mid y)f_Y(y)
$$
for almost every $(x,y)$.
The equality holds for the joint density $f_{X,Y}$ by the uniqueness of the Radon-Nikodym derivative. **Compare this with the Radon-Nikodym definition of Baye's theorem.**




### 2.3.2 Discrete Case

Here is a rigorous and complete paragraph that integrates your machinery with the discrete and continuous interpretations of the Product Rule:

The relationship $f_{X,Y}(x,y) = f_{X|Y}(x|y)f_Y(y)$ represents the **Product Rule** of probability, which in the discrete case is expressed as
$$p_{X,Y}(x,y) = p_{X|Y}(x|y)p_Y(y),$$
where
$$p_{X,Y}(x,y) = P(X=x, Y=y)$$
is the joint probability and
$$p_Y(y) = P(Y=y)$$
is the marginal probability. Rigorously, this formula is the realization of the measure-theoretic definition $h_B(y) = \mathbb{E}[\mathbb{1}_{\{X \in B\}} \mid Y=y]$ developed earlier. In this context,
$p_{X|Y}(x|y)$ is the **conditional probability** $P(X=x \mid Y=y)$, which corresponds to the value of the function $h_{\{x\}}(y)$. That is
$$h_{\{x\}}(y)=p_{X|Y}(x|y)=P(X=x \mid Y=y)$$

By multiplying the localized probability of $X$ within a specific "slice" where $Y=y$ by the total probability mass of that slice occurring ($p_Y(y)$), we recover the joint probability of both events happening simultaneously. This framework ensures that the discrete product rule is not merely a heuristic, but a specific case of the Radon-Nikodym identity $dP_{X,Y} = dP_{X|Y} \, dP_Y$.

The relationship $f_{X,Y}(x,y) = f_{X|Y}(x|y)f_Y(y)$ represents the Product Rule of probability for the continuous case. In the discrete case this is expressed as
$$p_{X,Y}(x,y) = p_{X|Y}(x|y)p_Y(y),$$
where
$$p_{X|Y}(x|y)=P(X=x \mid Y=y).$$

$$P(X \in B, Y \in A) = \sum_{y \in A} P(X \in B \mid Y=y) P(Y=y)$$

## 2.4 The Disintegration Theorem

Let $(X, Y): \Omega \to \mathscr{X} \times \mathscr{Y}$ be the joint map. By the **Pushforward Rule**, the integral of $X$ over the preimage $Y^{-1}(A)$, for any $A\in \sigma(Y)$, can be rewritten as an integral over the product space $\mathscr{X} \times \mathscr{Y}$ with respect to the joint distribution $P_{X,Y}$ with density $f_{X,Y}(x,y)$:

$$\int_{Y^{-1}(A)} X(\omega) \, dP(\omega) = \int_{\mathscr{X} \times \mathscr{Y}} x \cdot \mathbb{1}_A(y) \, dP_{X,Y}(x,y)= \int_{\mathscr{X}}\int_{\mathscr{Y}} x \cdot \mathbb{1}_A(y) \, f_{X,Y}(x,y)\,dxdy.$$

Since $f_{X,Y}(x,y)=f_{X|Y}(x\mid y)f_Y(y)$ we have
$$\int_{\mathscr{X}}\int_{\mathscr{Y}} x \cdot \mathbb{1}_A(y) \, f_{X,Y}(x,y)\,dxdy = \int_{\mathscr{Y}} \left( \int_{\mathscr{X}} x \cdot \mathbb{1}_A(y) \, f_{X|Y}(x\mid y)dx\right) dy$$

Since $\mathbb{1}_A(y)$ does not depend on $x$, it can be pulled out of the inner integral (with respect to $x$):

$$\int_{\mathscr{X}}\int_{\mathscr{Y}} x \cdot \mathbb{1}_A(y) \, f_{X,Y}(x,y)\,dxdy = \int_{\mathscr{Y}} \mathbb{1}_A(y) \, \left( \int_{\mathscr{X}} x f_{X|Y}(x\mid y)dx\right) dy$$

The indicator function $\mathbb{1}_A(y)$ restricts the integration over $\mathscr{Y}$ to the set $A$. Therefore, we have:

$$\int_{\mathscr{X}}\int_{\mathscr{Y}} x \cdot \mathbb{1}_A(y) \, f_{X,Y}(x,y)\,dxdy = \int_A \left( \int_{\mathscr{X}} x \, dP_{X \mid Y=y}(x) \right) dP_Y(y).$$

This tells us that

$$\int_{Y^{-1}(A)} X \, dP = \int_A \left( \int_{\mathscr{X}} x \, dP_{X \mid Y=y}(x) \right) dP_Y(y)$$

---

>**Note on the result:**
The Disintegration Theorem essentially "slices" the joint measure $P_{X,Y}$ into local measures $P_{X|Y=y}$ sitting on the fibers of the map $Y$.

## 2.5 The Conditional Expectation Evaluated at $Y=y$

We apply the definition of conditional expectation for a general random variable $X$ and using the Doob-Dynkin lemma we have for any $A\in \mathscr{Y}$:

$$
\int_{Y^{-1}(A)} X \, dP =
\int_{Y^{-1}(A)} \mathbb E[X \mid Y] \, dP =\int_A \mathbb E[X \mid Y=y] \, dP_Y(y).
$$

Note that in the previous section we have shown that the Disintegration Theorem gives
$$
\int_{Y^{-1}(A)} X \, dP = \int_A \left( \int_{\mathscr{X}} x \, dP_{X \mid Y=y}(x) \right) dP_Y(y).
$$

Thus we see that
$$\mathbb{E}[X\mid Y=y]= \int_{-\infty}^{\infty} x \, dP_{X \mid Y=y}(x).
$$
The first term, $\mathbb{E}[X|Y=y]$, represents the evaluation of the $\sigma(Y)$-measurable random variable $g(Y)=\mathbb{E}[X\mid Y]$ at $Y=y$.


---

**Summary:**

> The conditional expectation $\mathbb E[X\mid Y]$ is the $\sigma(Y)$-measurable random variable:
$$\mathbb E[X\mid Y]=g(Y).$$ It'sevaluation at the random value $Y=y$ is given by the expression
> $$g(y)=\mathbb E[X\mid Y=y]=\int_{\mathscr{X}} x\,f_{X\mid Y}(x\mid y)\,dx.$$
> In the discrete case we have $$g(y) = \mathbb{E}[X \mid Y=y] = \sum_{x \in \mathscr{X}} x \, p_{X|Y}(x \mid y),$$where $p_{X|Y}(x \mid y) = P(X=x \mid Y=y)$ is the discrete conditional probability mass function.


## 2.6 Conditional Random Variable, $X \mid Y=y$

In the standard framework of probability, we often talk about the distribution of $X$ across the entire sample space. However, once we observe $Y=y$, the "global" distribution of $X$ becomes obsolete; we need a mathematical object that represents our updated knowledge of $X$ within this specific "slice" of reality.


While $X$ is a mapping from the sample space $\Omega$ to $\mathscr{Y}$, the **conditional random variable** $X \mid Y=y$ (read as "$X$ given $Y=y$") is a conceptualization of $X$ acting on a "localized" probability space. It is defined as a random variable whose law is the **Regular Conditional Distribution** $P_{X \mid Y=y}$. Formally, for a fixed $y \in \mathscr{Y}$ such that $f_Y(y) > 0$, $X \mid Y=y$ is a random variable where:
$$P(X \in B \mid Y=y) = \int_B f_{X|Y}(x \mid y) \, dx.$$

---

Unlike the original $X$, which is a single map defined on $(\Omega, \mathfrak{A}, P)$, the conditional variable $X \mid Y=y$ can be viewed as a **family of random variables** indexed by the possible outcomes of $Y$.

* **Realization:** If $Y$ is fixed at $y$, $X \mid Y=y$ behaves like a standard random variable with its own mean, variance, and moments, all of which are now functions of $y$.
* **The Density:** The function $f_{X|Y}(x \mid y)$ serves as the probability density function (PDF) for this variable. Just as we write $X \sim f_X(x)$, we now write:
$$(X \mid Y=y) \sim f_{X|Y}(x \mid y).$$
* **The Expectation:** The "best guess" for the value of this conditional random variable is the conditional mean you derived earlier:
$$\mathbb{E}(X \mid Y=y) \triangleq \int_{\mathscr{X}} x f_{X|Y}(x \mid y) \, dx=\mathbb{E}[X \mid Y=y].$$

---

The relationship between the global random variable $X$ and the conditional family $X \mid Y=y$ is bridged by the **Law of Total Probability**. The global distribution is the mixture of these conditional variables, weighted by the marginal distribution of $Y$:
$$f_X(x) = \int_{\mathscr{Y}} f_{X|Y}(x \mid y) f_Y(y) \, dy.$$

In essence, $X \mid Y=y$ is the version of $X$ that lives exclusively on the "slice" of the universe where $Y$ has already happened and its value is known.

## 3.1 The Factorization Property

Let $X$, $Y$, and $W$ be random variables such that $Y$ is $\sigma(X)$ measurable and $W$ is $\sigma(Y)$ measurable.

Because $W$ is constant on the smallest "building blocks" of $\sigma(Y)$, the averaging process (the expectation) doesn't affect it; it only averages $X$. We can state this rigorously as:
$$\mathbb{E}[XW \mid Y] = W \mathbb{E}[X \mid Y] \quad \text{a.s.}$$


The proof outlined below is beyond the scope of this course. However it is provided for completeness.

---

*Step 1:*
Suppose $W$ is a simple function, $W = \sum_{i=1}^n c_i \mathbb{1}_{B_i}$, where $B_i \in \sigma(Y)$.
For any $A \in \sigma(Y)$, the set $A \cap B_i$ is also in $\sigma(Y)$. We use the defining property of $\mathbb{E}[X \mid Y]$:
$$\int_A XW \, dP = \int_A X \left( \sum c_i \mathbb{1}_{B_i} \right) dP = \sum c_i \int_{A \cap B_i} X \, dP$$

Applying the definition $\int_{A \cap B_i} X \, dP = \int_{A \cap B_i} \mathbb{E}[X \mid Y] \, dP$:
$$= \sum c_i \int_{A \cap B_i} \mathbb{E}[X \mid Y] \, dP = \int_A \left( \sum c_i \mathbb{1}_{B_i} \right) \mathbb{E}[X \mid Y] \, dP = \int_A W \mathbb{E}[X \mid Y] \, dP$$

*Step 2:*
If $W \geq 0$ is $\sigma(Y)$-measurable, there exists a sequence of simple $\sigma(Y)$-measurable functions $W_n$ such that $W_n \uparrow W$ (monotone convergence).
By the **Monotone Convergence Theorem for Conditional Expectations**:
$$\mathbb{E}[X W_n \mid Y] \uparrow \mathbb{E}[XW \mid Y]$$
Since $W_n$ are simple, we have $W_n \mathbb{E}[X \mid Y] \uparrow W \mathbb{E}[X \mid Y]$. The identity holds.

*Step 3:*
For a general $W$, we decompose it into its positive and negative parts: $W = W^+ - W^-$. By the linearity of the integral and the results from Step 2:
$$\mathbb{E}[X(W^+ - W^-) \mid Y] = \mathbb{E}[XW^+ \mid Y] - \mathbb{E}[XW^- \mid Y]$$
$$= W^+ \mathbb{E}[X \mid Y] - W^- \mathbb{E}[X \mid Y] = (W^+ - W^-) \mathbb{E}[X \mid Y] = W \mathbb{E}[X \mid Y]$$

---

### Conclusion
We have shown that $W \mathbb{E}[X \mid Y]$ satisfies the defining integral identity for $\mathbb{E}[XW \mid Y]$ for all $A \in \sigma(Y)$. Because it is also $\sigma(Y)$-measurable, by the **uniqueness** of the conditional expectation:
$$\mathbb{E}[XW \mid Y] = W \mathbb{E}[X \mid Y] \quad \text{a.s.}$$

## 3.2 Law of Total Expectation (or equivalently the Law of Iterated Expectations)

This is arguably one of the most elegant result in probability theory.

If you know the expected value of $X$ for every possible "slice" of $Y$, then the overall average of $X$ is simply the weighted average of those slices.

---

**The Measure-Theoretic Statement**

The law states:


$$\mathbb E[X] = \mathbb E[\mathbb E[X | Y]]$$

**The Intuition:**

* The inner expectation, $\mathbb E[X | Y]$, is your **Strategy** (the "Best Guess" function).
* The outer expectation, $\mathbb E[\cdot]$, averages that strategy over all possible realizations of $Y$.

---

**The Analytical Derivation (Using Densities)**

To see why this works in the lab, let's write it out using the "slice and divide" logic we derived:

1. **Start with the Conditional Mean for a point $y$:**

$$\mathbb E[X | Y=y] = \int_{\mathscr{X}} x \cdot f_{X|Y}(x|y) \, dx$$


2. **Take the expectation of that function over all $y$:**

$$\mathbb E[\mathbb E[X|Y]] = \int_{\mathscr{Y}} \left( \int_{\mathscr{X}} x \cdot f_{X|Y}(x|y) \, dx \right) f_Y(y) \, dy$$


3. **Use the Product Rule ($f_{X|Y} \cdot f_Y = f_{X,Y}$):**

$$\mathbb E[\mathbb E[X|Y]] = \int_{\mathscr{Y}} \int_{\mathscr{X}} x \cdot f_{X,Y}(x,y) \, dx \, dy$$


4. **Change the Order of Integration (Fubini's Theorem):**

$$\mathbb E[\mathbb E[X|Y]] = \int_{\mathscr{X}} x \left( \int_{\mathscr{Y}} f_{X,Y}(x,y) \, dy \right) dx$$


5. **Recognize the Marginal ($ \int f_{X,Y} dy = f_X(x)$):**

$$\mathbb E[\mathbb E[X|Y]] = \int_{\mathscr{X}} x \cdot f_X(x) \, dx = \mathbb E[X]$$



---

**The "Tower Property" Analogy**

Think of this as a **Tower**:

* **Ground Floor:** $\mathbb E[X]$. The raw average of the whole population.
* **First Floor:** $\mathbb E[X|Y]$. You’ve broken the population into groups (by $Y$). Each group has its own average.
* **The Result:** If you take the average of all the group averages (weighted by group size), you land back on the Ground Floor.

---





**Why this is Important:**

It allows us to solve complex problems by **Conditioning**. If you don't know how to calculate $\mathbb E[X]$ directly, you can:

1. Pick a variable $Y$ that "explains" $X$ well.
2. Calculate the mean of $X$ for each "slice" of $Y$ (usually easier).
3. Average those results.

#### Python Verification: "The Path of Means"

This code proves that the average of our "Red Regression Line" (from earlier) is exactly the global mean of $X$.

```python
import numpy as np

# Parameters
rho = 0.8
n_samples = 1_000_000

# 1. Generate Joint Data
y_samples = np.random.normal(5, 2, n_samples) # Mean 5, Std 2
x_samples = 0.8 * (y_samples - 5) + np.random.normal(10, 1, n_samples) # Mean 10

# 2. Calculate the Global Mean (Ground Floor)
global_mean_x = np.mean(x_samples)

# 3. Calculate Conditional Means (First Floor)
# For each y, our "Best Guess" strategy is h(y)
def h(y):
    # \mathbb E[X|Y=y] = mu_x + rho * (sigma_x/sigma_y) * (y - mu_y)
    # In this specific simulated case, the logic is:
    return 10 + 0.8 * (y - 5)

conditional_means = h(y_samples)

# 4. Law of Iterated Expectations: \mathbb E[h(Y)] should equal \mathbb E[X]
average_of_averages = np.mean(conditional_means)

print(f"Global Mean \\mathbb E[X]:              {global_mean_x:.4f}")
print(f"Average of Conditional Means:  {average_of_averages:.4f}")
print(f"Difference:                    {abs(global_mean_x - average_of_averages):.6f}")
```

## 3.3 Conditional Variance ($\mathrm{Var}(X|Y)$)

In probability and statistical learning, we often seek to quantify how much of the uncertainty in a random variable $X$ can be explained by another variable $Y$.  

The conditional variance, $\mathrm{Var}(X|Y)$, is the variance of $X$ given a specific outcome of $Y$. Formally, it is a $\sigma(Y)$-measurable random variable defined as:
$$\mathrm{Var}(X|Y) \triangleq \mathbb E[X^2 | Y] - (\mathbb E[X|Y])^2$$
It measures the **residual uncertainty** of $X$ that remains after $Y$ is known. If you were to observe $Y=y$, the conditional variance tells you the spread of the remaining probability distribution of $X$.

Below we will show that
$$
\mathbb{E}\left(\mathrm{Var}(X|Y)\right)
$$
will give the **Mean Square Error (MSE)** of the estmate $\mathbb{E}[X\mid Y]$.

## 3.4 Law of Total Variance

For any two random variables $X$ and $Y$ with finite variance, the total variance of $X$ can be split into two distinct parts:

$$\mathrm{Var}(X) = \mathbb E[\mathrm{Var}(X|Y)] + \mathrm{Var}(\mathbb E[X|Y])$$

* **$\mathrm{Var}(X)$**: Total Uncertainty (The "Total Spread").
* **$\mathbb E[\mathrm{Var}(X|Y)]$**: **Unexplained Variance** (The "Average Spread of the Slices").
* **$\mathrm{Var}(\mathbb E[X|Y])$**: **Explained Variance** (The "Spread of the Means").

The **Explained Variance** represents the part of $X$'s variance that is "accounted for" or explained by the linear relationship with $Y$. **Unexplained Variance** represents the remaining uncertainty in $X$ after $Y$ is known, often referred to as the estimation error or residual variance.

---

**The Analytical Derivation**

We use the definition of variance $\mathrm{Var}(Z) = \mathbb E[Z^2] - (\mathbb E[Z])^2$ and the **Law of Iterated Expectations** to prove this.

**Step 1: Expand the Right Hand Side.**
Recall $\mathrm{Var}(X|Y) = \mathbb E[X^2|Y] - (\mathbb E[X|Y])^2$.
Taking the expectation of this "Slice Variance":


$$\mathbb E[\mathrm{Var}(X|Y)] = \mathbb E[\mathbb E[X^2|Y]] - \mathbb E[(\mathbb E[X|Y])^2]$$

By the Law of Iterated Expectations, $\mathbb E[\mathbb E[X^2|Y]] = \mathbb E[X^2]$. So:


$$\mathbb E[\mathrm{Var}(X|Y)] = \mathbb E[X^2] - \mathbb E[(\mathbb E[X|Y])^2] \quad \text{--- (Equation A)}$$

**Step 2: Expand the Variance of the Mean.**


$$\mathrm{Var}(\mathbb E[X|Y]) = \mathbb E[(\mathbb E[X|Y])^2] - (\mathbb E[\mathbb E[X|Y]])^2$$

Again, by Iterated Expectations, $\mathbb E[\mathbb E[X|Y]] = \mathbb E[X]$. So:


$$\mathrm{Var}(\mathbb E[X|Y]) = \mathbb E[(\mathbb E[X|Y])^2] - (\mathbb E[X])^2 \quad \text{--- (Equation B)}$$

**Step 3: Add (A) and (B).**
Notice that the $\mathbb E[(\mathbb E[X|Y])^2]$ terms cancel out:


$$\mathbb E[\mathrm{Var}(X|Y)] + \mathrm{Var}(\mathbb E[X|Y]) = \mathbb E[X^2] - (\mathbb E[X])^2$$

This is exactly the definition of **$\mathrm{Var}(X)$**.

---

**The Physical Intuition**

Imagine you are measuring the height ($X$) of students across different ages ($Y$).

* **$\mathrm{Var}(X)$**: The total variation in height if you ignore age.
* **$\mathbb E[\mathrm{Var}(X|Y)]$**: The variation in height *within* each age group (e.g., the spread of 10-year-olds). This is what your model **cannot** explain.
* **$\mathrm{Var}(\mathbb E[X|Y])$**: How much the *average* height changes as you move from age 5 to age 20. This is the information that Age ($Y$) **successfully explains**.

---

> **The Shrinkage Principle:** Since $\mathrm{Var}(\mathbb E[X|Y]) \geq 0$, it is mathematically guaranteed that $\mathbb E[\mathrm{Var}(X|Y)] \leq \mathrm{Var}(X)$. Observing $Y$ can never make you *more* confused on average; it either helps or stays the same.



## 3.5 Optimality of the Conditional Estimate

Let $X$ be a random variable with $\mathbb{E}[X^2] < \infty$. Let $Y$ is a $\sigma(X)$-measurable random variable and $Z$ be a $\sigma(Y)$-measurable random variable with $\mathbb{E}[Z^2] < \infty$.

Consider
$$W=\mathbb{E}[X\mid Y]-Z.$$

Lets find $\mathrm{Var}(W)=\mathrm{Var}\left(\mathbb{E}[X\mid Y]-Z\right)$.

Note that
\begin{align*}
\mathbb{E}[(X - Z)^2] &= \mathbb{E}\left[\left((X - \mathbb{E}[X \mid Y]) + (\mathbb{E}[X \mid Y] - Z)\right)^2\right]\\
&= \mathbb{E}[(X - \mathbb{E}[X \mid Y])^2] + 2\mathbb{E}[(X-\mathbb{E}[X \mid Y])W]+ \mathbb{E}[(\mathbb{E}[X \mid Y] - Z)^2]\\
&=\mathbb{E}[(X - \mathbb{E}[X \mid Y])^2] + \mathbb{E}[(\mathbb{E}[X \mid Y] - Z)^2]
\end{align*}

Since $W$ is $\sigma(Y)$-measurable the factorization property implies that
$$\mathbb{E}[(X-\mathbb{E}[X \mid Y])W]=\mathbb{E}\left[\mathbb{E}[(X-\mathbb{E}[X \mid Y])W\mid Y]|Y\right]=\mathbb{E}\left[W\mathbb{E}[(X-\mathbb{E}[X \mid Y])\mid Y]\right]$$.

Since
$$\mathbb{E}[(X-\mathbb{E}[X \mid Y])\mid Y]=\mathbb{E}[X\mid Y]-\mathbb{E}[\mathbb{E}[X \mid Y]\mid Y]=\mathbb{E}[X\mid Y]-\mathbb{E}[X\mid Y]=0,$$
we have $\mathbb{E}[(X-\mathbb{E}[X \mid Y])W]=0$.

Thus
\begin{align*}
\mathbb{E}[(X - Z)^2] &= \mathbb{E}[(X - \mathbb{E}[X \mid Y])^2] + \mathbb{E}[(\mathbb{E}[X \mid Y] - Z)^2]
\end{align*}

Since $\mathbb{E}[(\mathbb{E}[X \mid Y] - Z)^2]\geq 0$ we have
\begin{align*}
\mathbb{E}[(X - \mathbb{E}[X \mid Y])^2] &\leq \mathbb{E}[(X - Z)^2].
\end{align*}

>This result proves that $\mathbb{E}[X \mid Y]$ is the **optimal estimate** of $X$ given the information in $Y$, in the sense that it is the unique minimizer of the **Mean Square Error (MSE)**. Any other $\sigma(Y)$-measurable estimate $Z$ will necessarily result in a larger (or equal) expected squared error.

---

Using the Law of Iterated Expectations we have:

\begin{align*}
\mathbb{E}[(X - \mathbb{E}[X \mid Y])^2] &= \mathbb{E}\big[ \mathbb{E}[(X - \mathbb{E}[X \mid Y])^2 \mid Y] \big] \\
&=\mathbb{E}\big[ \mathbb{E}[X^2 \mid Y]-2\mathbb{E}[X \mathbb{E}[X \mid Y]\mid Y]+\mathbb{E}[(\mathbb{E}[X \mid Y])^2 \mid Y] \big] \\
&=\mathbb{E}\big[ \mathbb{E}[X^2 \mid Y]-2\mathbb{E}[X \mid Y]\mathbb{E}[X \mid Y]+(\mathbb{E}[X \mid Y])^2 \big] \\
&=\mathbb{E}\big[ \mathbb{E}[X^2 \mid Y] - (\mathbb{E}[X \mid Y])^2 \big] \\
&= \mathbb{E}[\text{Var}(X \mid Y)]
\end{align*}

>This identity serves as the bridge between **estimation theory** and **uncertainty quantification**. It proves that the mean squared error of the optimal estimator—the average "distance" between the true value $X$ and our best guess $\mathbb{E}[X \mid Y]$—is exactly equal to the average residual uncertainty left in $X$ after $Y$ is revealed.

Essentially, it formalizes the idea that even an "ideal" model cannot eliminate all errors; rather, it "shrinks" the total error down to the unavoidable noise inherent in the conditional distribution. In a practical sense, it tells us that the **minimum possible error** we can ever achieve when predicting $X$ using $Y$ is precisely the **Expected Conditional Variance**, marking the mathematical limit of our predictive power.

### 2.6.1 Two Coin Toss Example

Let’s take a two-coin toss where **$X$** is the result of the first toss and **$Y$** is the total number of "Heads" across both tosses ($H=1, T=0$).

| Outcome $(\omega)$ | $X$ (1st Coin) | $Y$ (Total Heads) | $P(\omega)$ |
| --- | --- | --- | --- |
| **(H, H)** | 1 | 2 | 0.25 |
| **(H, T)** | 1 | 1 | 0.25 |
| **(T, H)** | 0 | 1 | 0.25 |
| **(T, T)** | 0 | 0 | 0.25 |

Let $\Omega = \{HH, HT, TH, TT\}$ with $P(\omega) = 1/4$ for each outcome.

Difine the random variables
* **$X$ (First coin is Heads):** $X(HH)=1, X(HT)=1, X(TH)=0, X(TT)=0$.
* **$Y$ (Total number of Heads):** $Y(HH)=2, Y(HT)=1, Y(TH)=1, Y(TT)=0$.


The random variable $Y$ induces a partition of $\Omega$ based on its values $\{0, 1, 2\}$:
* $A_0 = Y^{-1}(0)=\{TT\}$ where $P(Y=0) = 1/4$.
* $A_1 = Y^{-1}(1)=\{HT, TH\}$ where $P(Y=1) = 2/4 = 1/2$.
* $A_2 = Y^{-1}(2)=\{HH\}$ where $P(Y=2) = 1/4$.

Using the product rule $p_{X|Y}(x \mid y) = \frac{P(X=x, Y=y)}{P(Y=y)}$ for $x=1$ (Heads):

* **For $y=0$:**
    $$p_{X|Y}(1 \mid 0) = \frac{P(X=1, Y=0)}{P(Y=0)} = \frac{P(\emptyset)}{1/4} = 0$$
* **For $y=1$:**
    $$p_{X|Y}(1 \mid 1) = \frac{P(X=1, Y=1)}{P(Y=1)} = \frac{P(\{HT\})}{1/2} = \frac{1/4}{1/2} = 1/2$$
* **For $y=2$:**
    $$p_{X|Y}(1 \mid 2) = \frac{P(X=1, Y=2)}{P(Y=2)} = \frac{P(\{HH\})}{1/4} = \frac{1/4}{1/4} = 1$$

Now using your discrete summary formula
$$g(y) = \mathbb{E}[X\mid Y=y]=\sum_{x\in \{0,1\}} x \cdot p_{X|Y}(x|y)$$
we have
* **$g(0) = \mathbb{E}[X \mid Y=0] = (1 \cdot 0) = 0$**
* **$g(1) = \mathbb{E}[X \mid Y=1] = (1 \cdot 1/2) = 1/2$**
* **$g(2) = \mathbb{E}[X \mid Y=2] = (1 \cdot 1) = 1$**


By the Doob-Dynkin Lemma, $\mathbb{E}[X \mid Y] = g(Y)$. We can write this explicitly as:
$$\mathbb{E}[X \mid Y] = 0 \cdot \mathbb{1}_{\{Y=0\}} + \frac{1}{2} \mathbb{1}_{\{Y=1\}} + 1 \cdot \mathbb{1}_{\{Y=2\}}$$

Equivalently, in this specific case, you might notice that $h(y) = y/2$. Thus:
$$\mathbb{E}[X \mid Y] = \frac{Y}{2}$$


We check the defining property $\int_A \mathbb{E}[X|Y] \, dP = \int_A X \, dP$ for the most "interesting" set $A = \{HT, TH\}$ (where $Y=1$):
* **LHS:** $h(1) \cdot P(Y=1) = \frac{1}{2} \cdot \frac{1}{2} = \mathbf{1/4}$.
* **RHS:** $X(HT)P(HT) + X(TH)P(TH) = (1 \cdot 1/4) + (0 \cdot 1/4) = \mathbf{1/4}$.

**Interpretation:**
If you know there is exactly **one** head ($Y=1$), but you don't know which coin it was, your "best estimate" that the first coin was heads is **$50\%$**. If you know there are **two** heads, you are **$100\%$** sure the first was heads. If there are **zero** heads, you are **$0\%$** sure. The machinery perfectly captures the intuition.

---

To calculate the **conditional variance** $\mathrm{Var}(X|Y)$ for the two-coin toss, we use the definition provided in your machinery:
$$\mathrm{Var}(X|Y) \triangleq \mathbb E[X^2 | Y] - (\mathbb E[X|Y])^2$$

Since $X$ is an indicator variable ($X \in \{0,1\}$), a useful property is that $X^2 = X$. Therefore, $\mathbb E[X^2|Y] = \mathbb E[X|Y]$. This simplifies the variance formula for this specific case to:
$$\mathrm{Var}(X|Y) = \mathbb E[X|Y] - (\mathbb E[X|Y])^2$$

---

Let us calculate $\mathrm{Var}(X|Y=y)$ for each realization
We evaluate the variance "slice by slice" using the formula $g(y) - g(y)^2$:

* **For $y=0$ (Zero Heads):**
    $$\mathrm{Var}(X|Y=0) = 0 - (0)^2 = \mathbf{0}$$
    *Logic: If you know there are zero heads, you are certain $X=0$. There is no uncertainty (variance).*

* **For $y=1$ (One Head):**
    $$\mathrm{Var}(X|Y=1) = \frac{1}{2} - \left(\frac{1}{2}\right)^2 = \frac{1}{2} - \frac{1}{4} = \mathbf{1/4}$$
    *Logic: If you know there is one head, $X$ could be 0 or 1 with equal probability. This is the state of maximum uncertainty.*

* **For $y=2$ (Two Heads):**
    $$\mathrm{Var}(X|Y=2) = 1 - (1)^2 = \mathbf{0}$$
    *Logic: If you know there are two heads, you are certain $X=1$. Again, zero uncertainty.*


Using the indicator functions for the sets generated by $Y$:
$$\mathrm{Var}(X|Y) = 0 \cdot \mathbb{1}_{\{Y=0\}} + \frac{1}{4} \mathbb{1}_{\{Y=1\}} + 0 \cdot \mathbb{1}_{\{Y=2\}}$$
Or more simply:
$$\mathrm{Var}(X|Y) = \frac{1}{4} \mathbb{1}_{\{Y=1\}}$$

---

Recall $\mathbb E[X|Y]$ takes values $\{0, 1/2, 1\}$ with probabilities $\{1/4, 1/2, 1/4\}$.

**Total Variance $\mathrm{Var}(X)$:**
Since $P(X=1) = 0.5$ and $P(X=0) = 0.5$, $\mathrm{Var}(X) = 0.5 \times 0.5 = \mathbf{0.25}$.

**Unexplained Variance $\mathbb E[\mathrm{Var}(X|Y)]$:**
Using the values we just found:
$$\mathbb E[\mathrm{Var}(X|Y)] = (0 \cdot \tfrac{1}{4}) + (\tfrac{1}{4} \cdot \tfrac{1}{2}) + (0 \cdot \tfrac{1}{4}) = \mathbf{0.125}$$

Since $\mathbb E[\mathrm{Var}(X|Y)]=\mathbb{E}[(X - \mathbb{E}[X \mid Y])^2]$ we see that the unexplained variance corresponds to the **error in the optimal estimate**.

**Explained Variance $\mathrm{Var}(\mathbb E[X|Y])$:**

$\mathbb E[(\mathbb E[X|Y])^2] = (0^2 \cdot \tfrac{1}{4}) + ((\tfrac{1}{2})^2 \cdot \tfrac{1}{2}) + (1^2 \cdot \tfrac{1}{4}) = 0 + \tfrac{1}{8} + \tfrac{1}{4} = 0.375$.
$\mathrm{Var}(\mathbb E[X|Y]) = 0.375 - (0.5)^2 = 0.375 - 0.25 = \mathbf{0.125}$.

**Final Sum:**
$$0.125 (\text{Unexplained}) + 0.125 (\text{Explained}) = \mathbf{0.25}$$
The Law of Total Variance is satisfied perfectly. In this case, exactly **50%** of the uncertainty about the first coin is explained by knowing the total number of heads.

---

**Conclusion:**

By observing $Y$, we have cut our uncertainty about $X$ exactly in half. In "Information Theory" terms, we would say $Y$ provides information about $X$. If $Y$ and $X$ were independent, the explained variance would be $0$, and we would be just as confused as when we started.

### 2.6.2 Foundations of the Kalman Filter

#### 2.6.2.1 The Conditional Density of a Bivariate Gaussian

Consider the [bivariate normal distribution discussed here](https://github.com/mugalan/introduction-to-statistical-learning/blob/main/Multivariate_Gaussian_Distributions.ipynb).
For the case $n=m=1$, the general matrix results simplify through the scalar correlation $\rho$. We start with the joint density $f_{X,Y}(x,y) = f_{X|Y}(x|y)f_Y(y)$.

1. Conditional Mean ($\mu_{X|Y}$):
The general vector formula was $\boldsymbol{\mu}_{X|Y} = \boldsymbol{\mu}_X + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y} - \boldsymbol{\mu}_Y)$.  
Substituting the scalar values $\boldsymbol{\Sigma}_{XY} = \rho\sigma_X\sigma_Y$ and $\boldsymbol{\Sigma}_{YY} = \sigma_Y^2$:
$$\mu_{X|Y} = \mu_X + \frac{\rho\sigma_X\sigma_Y}{\sigma_Y^2}(y - \mu_Y) = \mu_X + \rho\frac{\sigma_X}{\sigma_Y}(y - \mu_Y)$$

2. Conditional Variance ($\sigma_{X|Y}^2$):
The general covariance was the Schur complement $\boldsymbol{\Sigma}_{X|Y} = \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}$.  
Substituting the scalar values:
$$\sigma_{X|Y}^2 = \sigma_X^2 - \frac{(\rho\sigma_X\sigma_Y)^2}{\sigma_Y^2} = \sigma_X^2 - \rho^2\sigma_X^2 = \sigma_X^2(1-\rho^2)$$

3. Factorization Result
By completing the square in the joint quadratic form $Q(x,y)$, we partition the joint density into:
$$f_{X,Y}(x,y) = \mathscr{N}(x; \mu_{X|Y}, \sigma_{X|Y}^2) \cdot \mathscr{N}(y; \mu_Y, \sigma_Y^2).$$
Dividing by the marginal $f_Y(y)$ yields the conditional density:
$$f_{X|Y}(x|y) = \frac{1}{\sqrt{2\pi\sigma_X^2(1-\rho^2)}} \exp\left( -\frac{(x - \mu_{X|Y})^2}{2\sigma_X^2(1-\rho^2)} \right)$$

**Core Insight:** The "linear regression" term $\rho\frac{\sigma_X}{\sigma_Y}$ and the "variance reduction" factor $(1-\rho^2)$ are the specific scalar realizations of the Schur complement and block-matrix inversion used in the general $n$-dimensional proof.

>**Summary of the Result:**
The conditional density is itself a univariate Gaussian:
$$f_{X|Y}(x|y) = \frac{1}{\sqrt{2\pi\sigma_{X|Y}^2}} \exp\left( -\frac{(x - \mu_{X|Y})^2}{2\sigma_{X|Y}^2} \right).$$ The random variable that corresponds to this density is the conditional random variable that is denoted by $(X \mid Y=y)$.


#### 2.6.2.2 Conditional Expectation in the Bivariate Normal Distribution

We define $g(y) = \mathbb{E}[X \mid Y=y]$. We have seen that:
$$g(y) = \int_{-\infty}^{\infty} x f_{X|Y}(x \mid y) \, dx$$

Substitute the Gaussian form of $f_{X|Y}(x \mid y)$ derived above. Let $\mu(y) = \mu_X + \rho\frac{\sigma_X}{\sigma_Y}(y - \mu_Y)$ and $\sigma_{X|Y}^2 = \sigma_X^2(1-\rho^2)$. Then:
$$g(y) = \int_{-\infty}^{\infty} x \left[ \frac{1}{\sqrt{2\pi\sigma_{X|Y}^2}} \exp\left( -\frac{(x - \mu(y))^2}{2\sigma_{X|Y}^2} \right) \right] dx$$

This integral is simply the first moment (mean) of a Gaussian random variable with mean $\mu(y)$. Thus:
$$g(y) = \mu(y) = \mu_X + \rho\frac{\sigma_X}{\sigma_Y}(y - \mu_Y)$$

Therefore:
$$\mathbb{E}[X \mid Y=y] = \mu_X + \rho\frac{\sigma_X}{\sigma_Y}(y - \mu_Y)$$

and hence
$$\mathbb{E}[X \mid Y] = \mu_X + \rho\frac{\sigma_X}{\sigma_Y}(Y - \mu_Y)$$


>In physical terms, this result demonstrates that when two variables are statistically coupled, observing one ($Y$) provides a window into the hidden state of the other ($X$). The expression represents a weighted balance between **prior knowledge** and **new evidence**: the term $(Y - \mu_Y)$ captures the "surprise" or deviation in our measurement, while the ratio $\rho\frac{\sigma_X}{\sigma_Y}$—the gain—scales that surprise based on how strongly the two variables move together relative to the volatility of our observation.
>
>Instead of taking the measurement at face value, the conditional expectation "shrinks" the estimate toward the historical mean $\mu_X$ whenever the sensor is noisy or the correlation is weak. It represents the mathematically optimal way to extract signal from noise, ensuring that our estimate is not just a guess, but the most probable physical state of $X$ given the specific information "leaked" through $Y$.
>This is in fact the **Kalman-Filter**.

##### **MSE**

Given the bivariate normal distribution where $X, Y \in \mathbb{R}$, we use the scalar parameters $\sigma_X^2$, $\sigma_Y^2$, and $\rho$.

From the conditional density, we know:
$$(X \mid Y=y) \sim \mathscr{N}\left(\mu_X + \rho\frac{\sigma_X}{\sigma_Y}(y - \mu_Y), \quad \sigma_X^2(1 - \rho^2)\right)$$

1. Conditional Variance as a Constant:
The conditional variance for a fixed $y$ is:
$$\mathrm{Var}(X \mid Y=y) = \sigma_X^2(1 - \rho^2)$$
Since this expression is **independent of $y$**, the conditional variance as a random variable is the constant:
$$\mathrm{Var}(X \mid Y) = \sigma_X^2(1 - \rho^2)$$

2. Expected Conditional Variance (MSE):
Because the random variable is a constant, its expectation is simply that constant value:
$$\mathbb{E}[\mathrm{Var}(X \mid Y)] = \mathbb{E}[\sigma_X^2(1 - \rho^2)] = \sigma_X^2(1 - \rho^2)$$

3. The Final Estimation Error:
Applying the identity that the Mean Square Error (MSE) of the conditional expectation is equal to the expected conditional variance, we have:

$$\text{MSE} = \mathbb{E}[(X - \mathbb{E}[X \mid Y])^2] = \sigma_X^2(1 - \rho^2)$$

---

**Core Comparison:**

| Component | Multivariate Result         | Scalar Result ($n=m=1$) |
| :------- | :--------------------------- | :----------------------- |
| **Best Predictor** | $\boldsymbol{\mu}_x + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{Y} - \boldsymbol{\mu}_y)$ | $\mu_X + \rho\frac{\sigma_X}{\sigma_Y}(Y - \mu_Y)$ |
| **Estimation Error (MSE)** | $\boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}$ | $\sigma_X^2(1 - \rho^2)$ |

**Conclusion:** In the scalar case, the "unexplained variance" is simply the total variance of $X$ scaled down by the factor $(1-\rho^2)$. This represents the minimum possible mean square error achievable by any predictor of $X$ based on $Y$.

##### **The "Slice" Visual**



Think of the joint distribution $f_{X,Y}$ as a 3D mountain range.

1. **Fixing $Y=y$** is like taking a vertical "chainsaw cut" through that mountain at a specific $y$-coordinate.
2. The resulting **silhouette** of the cut is the conditional distribution.
3. However, the area under that silhouette isn't 1 yet. We must **rescale** it (divide by the marginal $f_Y(y)$) to make it a valid probability distribution.


```python
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots
from scipy.stats import multivariate_normal, norm

# --- Parameters ---
mu_x, mu_y = 0, 0
sigma_x, sigma_y = 1, 1
rho = 0.8
y_obs = 1.5

# Math for Conditional Distribution
mu_cond = mu_x + rho * (sigma_x / sigma_y) * (y_obs - mu_y)
var_cond = sigma_x**2 * (1 - rho**2)
std_cond = np.sqrt(var_cond)

# --- Data Generation ---
x_grid = np.linspace(-3.5, 3.5, 200)
y_grid = np.linspace(-3.5, 3.5, 200)
X, Y = np.meshgrid(x_grid, y_grid)
pos = np.dstack((X, Y))
Z = multivariate_normal([mu_x, mu_y], [[sigma_x**2, rho*sigma_x*sigma_y], [rho*sigma_x*sigma_y, sigma_y**2]]).pdf(pos)

# --- Create Plot ---
fig = make_subplots(rows=1, cols=2,
                    subplot_titles=("1. Joint Density (The 'Hill')", "2. Conditional Density (The 'Slice')"))

# Left: Joint Contour
fig.add_trace(go.Contour(x=x_grid, y=y_grid, z=Z, colorscale='Viridis', showscale=False), row=1, col=1)
fig.add_shape(type="line", x0=-3.5, y0=y_obs, x1=3.5, y1=y_obs, line=dict(color="Red", width=3, dash="dash"), row=1, col=1)

# Right: Conditional PDF
x_pdf = np.linspace(-3.5, 3.5, 500)
y_pdf = norm.pdf(x_pdf, mu_cond, std_cond)
prior_pdf = norm.pdf(x_pdf, mu_x, sigma_x)

fig.add_trace(go.Scatter(x=x_pdf, y=prior_pdf, name='Prior P(X)', line=dict(color='gray', dash='dot')), row=1, col=2)
fig.add_trace(go.Scatter(x=x_pdf, y=y_pdf, name='Posterior P(X|Y)', fill='tozeroy', line=dict(color='blue', width=3)), row=1, col=2)
fig.add_vline(x=mu_cond, line=dict(color="Red", dash="dot"), row=1, col=2)

fig.update_layout(height=500, title_text="Information Gain: From Joint to Conditional Distribution", template="plotly_white")
fig.show()
```

**What this visualization reveals**

Dimensionality Collapse: The left plot exists in the joint $\mathbb{R}^2$ space. By observing $Y=1.5$ (the dashed red line), we "slice" the joint density. The right plot represents the profile of that slice, effectively collapsing the $Y$ dimension into a single, updated 1D distribution for $X$.

The "Gain" in Action (Mean Shift): Because $X$ and $Y$ are positively correlated ($\rho=0.8$), the high observed value of $Y$ "drags" our estimate of $X$ with it. Our best estimate $E[X \mid Y]$ shifts from the prior mean of $0$ to the posterior mean of $1.2$.

Information Gain (Variance Shrinkage): Notice that the blue curve on the right is "taller and skinnier" than the original spread of $X$. Mathematically, the variance has shrunk from $1.0$ to $0.36$. This narrowing represents the Information Gain: the physical reduction of uncertainty through observation.

The Math: How Observation "Shrinks" UncertaintyIn Gaussian systems, the most striking property is that the Conditional Variance is always less than (or equal to) the Prior Variance.$$\text{Var}(X \mid Y) = \sigma_X^2 (1 - \rho^2)$$

*Why this is critical for Engineering:*

If $\rho = 0$
(Independence): The variance remains $\sigma_X^2$. Observing $Y$ tells you nothing; uncertainty is unchanged.

If $\rho = 1$.
(Perfect Correlation): The variance becomes $0$. Observing $Y$ tells you $X$ exactly; uncertainty is eliminated.

The "Information" is in the Correlation: The term $(1 - \rho^2)$ is the factor by which the "noise" is suppressed. In our plot, $\rho=0.8$, so the variance shrinks by $1 - 0.64 = 0.36$. We have "removed" $64\%$ of the original uncertainty.
