# 0. The Evolution of Entropy: From Communication to Statistical Inference

The concept of entropy has undergone one of the most fascinating conceptual evolutions in science. Originating in 19th-century thermodynamics as a macroscopic measure of wasted heat energy, it was completely reinvented in the 20th century. Today, entropy is understood not merely as a property of physical matter, but as the foundational mathematical measure of **information, uncertainty, and inference**.

---

1. Shannon's Mathematical Milestone (1948)

In 1948, an American mathematician and electronic engineer named **Claude Shannon** published a groundbreaking paper titled *"A Mathematical Theory of Communication."* Shannon set out to solve a highly practical engineering problem: how to efficiently and reliably transmit messages across a noisy communication channel.

To achieve this, Shannon needed to quantify a concept that had never been mathematically defined before: **information**.

Shannon realized that the "information content" of a message is directly tied to how surprising it is. If an event is certain to happen ($p=1$), being told it occurred provides zero new information. Conversely, learning that a highly improbable event occurred provides a large amount of information.

**The Axiomatic Derivation of Uncertainty:**

Shannon sought a mathematical function $H(p_1, p_2, \dots, p_k)$ that could measure the total amount of uncertainty choice inherent in a discrete probability distribution. He demanded that this function satisfy three intuitive axioms:

1. **Continuity:** The measure $H$ should be continuous in the probabilities $p_j$. Small changes in probabilities should yield small changes in uncertainty.
2. **Monotonicity:** If all $k$ outcomes are equally likely ($p_j = 1/k$), then $H(1/k, \dots, 1/k)$ should be a monotonically increasing function of $k$. More choices mean more uncertainty.
3. **Additivity (Composition):** If a choice is broken down into successive choices, the total $H$ should be the weighted sum of the individual values of $H$.

In a brilliant mathematical proof, Shannon demonstrated that the **only** function satisfying these criteria takes the form:

$$H(X) = -K \sum_{j=1}^k p_j \ln p_j$$

Where $K$ is a positive scaling constant. Setting $K=1$ and using the natural logarithm gives the uncertainty unit in **nats**, while using log base 2 measures uncertainty in **bits**.

> **Shannon's Conclusion:** Information is mathematically equivalent to the removal of uncertainty. The function $H(X)$, which he named **Information Entropy**, represents the absolute minimum number of bits required to encode and transmit messages from a source without losing information.

---

2. The Great Convergence: Why "Entropy"?

A famous historical anecdote highlights the cross-disciplinary bridge built by Shannon's work. When deciding what to call his new measure of uncertainty, Shannon consulted the legendary mathematician John von Neumann. Shannon recalled the conversation:

> *"Von Neumann told me, 'You should call it entropy, for two reasons. In the first place your uncertainty function has been used in statistical mechanics under that name, so it already has a name. In the second place, and more important, nobody knows what entropy really is, so in a debate you will always have the advantage.'"*

Von Neumann was entirely right. The algebraic structure of Shannon’s information entropy was a mirror image of the statistical entropy developed by Ludwig Boltzmann and J. Willard Gibbs in the late 1800s to describe the disordered microstates of gas molecules:

$$S = -k_B \sum_{j} p_j \ln p_j$$

For nearly a decade, scientists debated whether this structural similarity was a mere mathematical coincidence or a sign of a deeper physical law. The answer came in 1957.

---

3. Jaynes' Principle of Maximum Entropy (1957)

The true philosophical unification of information theory and the physical world was achieved by **Edwin T. Jaynes** in his seminal 1957 paper, *"Information Theory and Statistical Mechanics."* Jaynes made a radical conceptual leap. He asserted that statistical mechanics should not be viewed as a physical theory wrapped around atoms, but rather as an exercise in **statistical inference** wrapped around missing data.

**Shifting from Frequencies to States of Knowledge:**

Before Jaynes, probabilities in physics were strictly interpreted as *frequencies*—the fractional number of times a system visits a state over infinite time. Jaynes discarded this restriction, adopting a subjective (or Bayesian) view: a probability distribution represents our **state of knowledge** about a system based on our available data.

Jaynes formalized this into an objective procedure for model building:

$$\text{Maximize } H = -\sum_{j=1}^k p_j \ln p_j \quad \text{subject to what you know.}$$

```
    [ Macroscopic Observations ]  ---> (e.g., Average Energy / Data Features)
                 │
                 ▼
    [ Principle of Maximum Entropy ]
                 │
                 ▼
    [ Most Unbiased Distribution ] ---> (Uniformly spreads remaining uncertainty)

```

Jaynes proved that if you maximize Shannon's entropy subject to the constraint of a fixed average internal energy, the mathematical machinery *automatically* outputs the laws of thermodynamics and the canonical Boltzmann distribution.

**The Legacy of the Conceptual Leap:**

Jaynes' contribution completely liberated entropy from the confines of physics. By demonstrating that maximizing entropy is the only mathematically defensible way to assign probabilities without injecting human bias, he turned MaxEnt into a universal tool.

Today, whether we are deriving the state of an engineered system, calculating the distribution of text in a natural language processing model, or training a multiclass logistic classifier, we are executing Jaynes' foundational philosophy: **trust the facts you have observed, and maximize your entropy about the rest.**

# 1. The Measure-Theoretic View (Relative Entropy)

**Relative entropy** (also called **Kullback-Leibler divergence**) is a measure of how one probability measure differs from another.

If $P$ and $Q$ are two probability measures on the same measurable space $(\Omega,\mathfrak F)$, and $P$ is absolutely continuous with respect to $Q$ ($P \ll Q$), then the relative entropy of $P$ with respect to $Q$ is defined by

$$
D_{KL}(P\|Q)=\int_{\Omega}\ln\left(\frac{dP}{dQ}\right)\,dP
$$

equivalently,

$$
D_{KL}(P\|Q)=\int_{\Omega}\frac{dP}{dQ}\ln\left(\frac{dP}{dQ}\right)\,dQ.
$$

Here $\frac{dP}{dQ}$ is the **Radon-Nikodym derivative** of $P$ with respect to $Q$.

So relative entropy is fundamentally a comparison between two probability measures. It does not require introducing a random variable explicitly; it only requires that the two measures are defined on the same measurable space.

---

For two discrete probability distributions $P$ and $Q$ on the same countable alphabet $\mathfrak X$, the relative entropy is

$$
D_{KL}(P\|Q)=\sum_{x\in\mathfrak X} P(x)\ln\left(\frac{P(x)}{Q(x)}\right).
$$

In the discrete case, the Radon-Nikodym derivative reduces to the ratio

$$
\frac{dP}{dQ}(x)=\frac{P(x)}{Q(x)}
\quad\text{whenever } Q(x)>0.
$$

**Condition for finiteness:**

For $D_{KL}(P\|Q)$ to be well defined as an extended value, it is enough to require $P \ll Q$. Under this condition, the KL divergence may still be finite or $+\infty$.

For the discrete case, absolute continuity means:

> whenever $Q(x)=0$, we must also have $P(x)=0$.

If there exists some $x$ such that $P(x)>0$ but $Q(x)=0$, then

$$
D_{KL}(P\|Q)=+\infty.
$$



## 1.1 Non-negativity of the Kullback-Leibler divergence

For the continuous case:
$$D_{KL}(p \parallel q) = \int p(x) \ln \left( \frac{p(x)}{q(x)} \right) dx = \int p(x) \left( -\ln \frac{q(x)}{p(x)} \right) dx$$

Because $-\ln(x)$ is a convex function, Jensen’s Inequality tells us that the average of the function is greater than or equal to the function of the average:
$$\mathbb{E}_p \left[ -\ln \frac{q(x)}{p(x)} \right] \geq -\ln \left( \mathbb{E}_p \left[ \frac{q(x)}{p(x)} \right] \right)$$

When you evaluate the right side:
$$\mathbb{E}_p \left[ \frac{q(x)}{p(x)} \right] = \int p(x) \frac{q(x)}{p(x)} dx = \int q(x) dx = 1$$
Since $-\ln(1) = 0$, we are left with:
$$D_{KL}(P \parallel Q) \geq 0$$

It is exactly **$0$** if and only if $P = Q$ almost everywhere.

# 2. The Distribution View (Differential Entropy)

Suppose we have the distribution of our random variable $X$, denoted by the probability measure $P_X$, and a reference measure $\mu$ on $\mathbb{R}^n$ (such as the standard Lebesgue measure $\lambda^n$).

If $P_X$ is absolutely continuous with respect to $\mu$ ($P_X \ll \mu$), the **Radon-Nikodym theorem** guarantees that there exists a non-negative, measurable function $f(x)$—the density—such that:

$$dP_X(x) = f(x) \, d\mu(x) \quad \text{or, equivalently,} \quad \frac{dP_X}{d\mu}(x) = f(x)$$

Using the original definition of Kullback-Leibler divergence the **relative entropy** of the probability measure $P_X$ with respect to the reference measure $\mu$ is:

$$D_{KL}(P_X \| \mu) = \int_{\mathbb{R}^n} \ln\left(\frac{dP_X}{d\mu}\right) \, dP_X$$


Now, we replace the abstract Radon-Nikodym derivative $\frac{dP_X}{d\mu}$ with our explicit density function $f(x)$, and change the integrator from $dP_X$ to $f(x) \, d\mu(x)$:

$$D_{KL}(P_X \| \mu) = \int_{\mathbb{R}^n} \ln(f(x)) \cdot f(x) \, d\mu(x)$$

Rearranging the terms inside the integral gives:

$$D_{KL}(P_X \| \mu) = \int_{\mathbb{R}^n} f(x) \ln f(x) \, d\mu(x)$$


In information theory, **entropy** is intuitively understood as the uncertainty or the information content of a distribution. Mathematically, it is the negative of the relative entropy with respect to a uniform or reference background.

Thus, we define the way **differential entropy** $h_\mu(X)$ as the negative of the KL divergence between $P_X$ and $\mu$:

$$h_\mu(X) = -D_{KL}(P_X \| \mu)$$

By substituting our result from above into this definition, we get:

$$h_\mu(X) = -\int_{\mathbb{R}^n} f(x) \ln f(x) \, d\mu(x)=-\mathbb{E}\left(\ln(f(X))\right).$$


## 2.1 Example: The Differential Entropy of a Normally Distributed Random Variable

To find the differential entropy of a normally distributed random variable $X \sim \mathscr{N}(\mu, \sigma^2)$, we evaluate its entropy with respect to the Lebesgue measure on $\mathbb{R}$.

The probability density function $f(x)$ of a normal distribution with mean $\mu$ and variance $\sigma^2$ is:

$$f(x) = \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left( -\frac{(x - \mu)^2}{2\sigma^2} \right)$$

By the definition of differential entropy $h(X)$, we have:

$$h(X) = -\int_{-\infty}^{\infty} f(x) \ln f(x) \, dx = -\mathbb{E}[\ln f(X)]$$

---

Taking the natural logarithm of the PDF gives:

$$\ln f(x) = \ln\left( \frac{1}{\sqrt{2\pi\sigma^2}} \right) - \frac{(x - \mu)^2}{2\sigma^2} = -\frac{1}{2}\ln(2\pi\sigma^2) - \frac{(x - \mu)^2}{2\sigma^2}$$

Now we compute the expected value $-\mathbb{E}[\ln f(X)]$:

$$h(X) = -\mathbb{E}\left[ -\frac{1}{2}\ln(2\pi\sigma^2) - \frac{(X - \mu)^2}{2\sigma^2} \right]$$

Using the linearity of expectation, this separates into two terms:

$$h(X) = \frac{1}{2}\ln(2\pi\sigma^2) + \frac{1}{2\sigma^2}\mathbb{E}[(X - \mu)^2]$$


By definition, the expected squared deviation from the mean is the variance of the distribution:

$$\mathbb{E}[(X - \mu)^2] = \sigma^2$$

Substituting this back into our expression:

$$h(X) = \frac{1}{2}\ln(2\pi\sigma^2) + \frac{1}{2\sigma^2} \cdot \sigma^2 = \frac{1}{2}\ln(2\pi\sigma^2) + \frac{1}{2}=\frac{1}{2}\ln(2\pi e \sigma^2)$$

---

**Key Insights:**

* **Independence from the mean:** Notice that the mean $\mu$ does not appear in the final result. Differential entropy is translation-invariant; shifting the distribution doesn't change its uncertainty.
* **Dependence on variance:** The entropy scales directly with $\sigma^2$. As the variance increases, the distribution spreads out, leading to more uncertainty and higher entropy.
* **Maximum Entropy:** Under the constraint of a fixed variance $\sigma^2$, the Gaussian distribution has the **highest possible differential entropy** of any real-valued continuous random variable.

#### Visualizing Entropy as "Spread" of Measure

The higher the entropy, the more "spread out" the probability measure is.

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import entropy

# Define two probability measures (distributions)
# P1: Concentrated (Low Entropy)
# P2: Spread out (High Entropy)
x = np.linspace(0, 10, 100)
p1 = np.exp(-(x-5)**2 / (2 * 0.5**2))
p1 /= p1.sum()

p2 = np.exp(-(x-5)**2 / (2 * 2.0**2))
p2 /= p2.sum()

plt.figure(figsize=(10, 4))
plt.fill_between(x, p1, alpha=0.5, label=f'Low Entropy: {entropy(p1):.2f}')
plt.fill_between(x, p2, alpha=0.5, label=f'High Entropy: {entropy(p2):.2f}')
plt.title("Entropy as a Property of the Measure's Concentration")
plt.legend()
plt.show()
````

# 3. Entropy of a Discrete Random Variable (Shannon Entropy)

Just as differential entropy in the continuous setting is the negative relative entropy of $P_X$ with respect to the Lebesgue measure, **discrete entropy** is the negative relative entropy of $P_X$ with respect to the **counting measure**.

Let $X$ be a discrete random variable taking values in a countable set $S$. We consider the measurable space $(S, \mathscr{P}(S))$, where $\mathscr{P}(S)$ is the power set of $S$.


On $(S, \mathscr{P}(S))$, we establish our reference measure as the **counting measure** $\mu$, defined for any subset $A \subset S$ as the number of elements in $A$.

The distribution (law) of $X$ is a probability measure $P_X$ on $(S, \mathscr{P}(S))$. Since the only set $A \subset S$ for which $\mu(A) = 0$ is the empty set $\emptyset$, and $P_X(\emptyset) = 0$, the law $P_X$ is trivially **absolutely continuous** with respect to the counting measure $\mu$ ($P_X \ll \mu$).

By the Radon-Nikodym theorem, there exists a unique density function $p_X: S \to [0, 1]$, which is the Radon-Nikodym derivative of $P_X$ with respect to $\mu$:

$$\frac{dP_X}{d\mu}(x) = p_X(x)$$

For any point $x \in S$, this derivative is exactly the probability mass function:

$$p_X(x) = P_X(\{x\})$$

Thus, we can express the measure $P_X$ in terms of the counting measure as:

$$dP_X(x) = p_X(x) \, d\mu(x)$$

Using the original definition of the Kullback-Leibler divergence, we evaluate the relative entropy of the probability measure $P_X$ with respect to the non-probabilistic counting measure $\mu$ on $S$:

$$D_{KL}(P_X \| \mu) = \int_{S} \ln\left(\frac{dP_X}{d\mu}\right) \, dP_X$$

Substituting the density $p_X = \frac{dP_X}{d\mu}$ into the integral, we obtain:

$$D_{KL}(P_X \| \mu) = \int_{S} \ln(p_X(x)) \, dP_X(x)$$

Since $P_X$ is a discrete measure, this integral over the state space $S$ reduces exactly to a sum over the countable points:

$$D_{KL}(P_X \| \mu) = \sum_{x \in S} p_X(x) \ln p_X(x)$$

The **Shannon entropy** $H(X)$ is defined as the negative of this divergence. This reflects the uncertainty of the distribution relative to the uniform baseline represented by the counting measure:

$$H(X) = -D_{KL}(P_X \| \mu)$$

By expanding this equality, we recover the standard expressions for discrete entropy:

$$H(X) = -\int_{S} \ln\left(\frac{dP_X}{d\mu}\right) \, dP_X = -\sum_{x \in S} p_X(x) \ln p_X(x) = -\mathbb{E}[\ln p_X(X)]$$




# 4. The Entropy of a 'Spiked' Distribution

## 4.1 The Discrete Setting

Let our state space be the single-point set $S = \{x_0\}$.
* The law of our random variable $X$ is the delta measure $P_X = \delta_{x_0}$, which assigns a probability of $1$ to the point $x_0$:
  $$p_X(x_0) = P_X(\{x_0\}) = 1$$
* Our reference measure is the **counting measure** $\mu$ on $S$. Since the set has exactly one element, $\mu(\{x_0\}) = 1$.

Now, let's plug this into our entropy formula:

$$H_{\mu}(X) = -\sum_{x \in S} p_X(x) \ln p_X(x)$$

Since $S$ only contains the point $x_0$, this sum has exactly one term:

$$H_{\mu}(X) = - (1 \cdot \ln(1)) = - (1 \cdot 0) = 0$$

> **Conclusion:** When the delta measure is correctly treated as a discrete distribution on its support, its entropy is exactly **$0$ nats**. This matches our intuition perfectly: there is exactly zero uncertainty about the outcome of $X$.



## 4.2 The Continuous Setting

What happens if we view the delta measure as a limit of continuous random variables on $\mathbb{R}$, evaluated with respect to the **Lebesgue measure**?

Consider a sequence of normal distributions $X_\sigma \sim \mathscr{N}(x_0, \sigma^2)$. As the variance $\sigma^2 \to 0$, the probability density function $f_\sigma(x)$ collapses into the Dirac delta function $\delta(x - x_0)$.

Using the formula we derived for the differential entropy of a normal distribution:

$$h(X_\sigma) = \frac{1}{2} \ln(2 \pi e \sigma^2)$$

To find the differential entropy of the delta distribution, we take the limit as $\sigma \to 0$:

$$\lim_{\sigma \to 0} h(X_\sigma) = \lim_{\sigma \to 0} \frac{1}{2} \ln(2 \pi e \sigma^2) = -\infty$$



````python
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# 1. Setup data
x_range = np.linspace(-2, 2, 1000)
sigmas = [1.0, 0.5, 0.2, 0.1, 0.05]

def gaussian_pdf(x, sigma):
    return (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * (x / sigma)**2)

def diff_entropy(sigma):
    # h(X) = 0.5 * ln(2 * pi * e * sigma^2)
    return 0.5 * np.log(2 * np.pi * np.e * (sigma**2))

entropy_sigmas = np.logspace(-3, 0.5, 300)
entropy_vals = diff_entropy(entropy_sigmas)

# 2. Create Plotly Figure with Subplots
fig = make_subplots(
    rows=1, cols=2,
    subplot_titles=(
        "PDF Collapse: $\\sigma \\to 0$",
        "Entropy Divergence: $h(X_\\sigma) \\to -\\infty$"
    ),
    horizontal_spacing=0.15
)

# Left Plot: PDF traces
for s in sigmas:
    fig.add_trace(
        go.Scatter(x=x_range, y=gaussian_pdf(x_range, s), name=f"σ={s}"),
        row=1, col=1
    )

# Right Plot: Entropy curve
fig.add_trace(
    go.Scatter(
        x=entropy_sigmas,
        y=entropy_vals,
        name="Differential Entropy",
        line=dict(color='red', width=3)
    ),
    row=1, col=2
)

# Layout adjustments
fig.update_xaxes(title_text="x", row=1, col=1)
fig.update_yaxes(title_text="Density", row=1, col=1)

fig.update_xaxes(title_text="Standard Deviation (σ)", type="log", row=1, col=2)
fig.update_yaxes(title_text="Entropy (nats)", row=1, col=2)

fig.update_layout(
    title="The Delta Measure as a Limit of Gaussian Distributions",
    template="plotly_white",
    height=500,
    showlegend=True
)

fig.show()
````

## 4.3 Resolving the Paradox: Why does the continuous approach say $-\infty$ while the discrete approach says $0$?

You might ask: *Why does the continuous approach say $-\infty$ while the discrete approach says $0$?*

This occurs because the **Lebesgue measure** and the **counting measure** treat points differently:
* To the counting measure, a point has a reference size of **$1$**.
* To the Lebesgue measure, a point has a reference size of **$0$**.

When you squeeze a continuous distribution down to a single point, you are concentrating all of its probability mass into a region of Lebesgue measure zero. The probability density $f(x)$ shoots up to infinity, so $\ln f(x) \to \infty$.

Thus, the $-\infty$ differential entropy simply tells us that a point has **infinitely less uncertainty** than a continuous interval of Lebesgue measure.

**Summary:**

Depending on your reference measure, the delta measure has:
* **$0$ entropy** relative to the counting measure (there is no uncertainty).
* **$-\infty$ entropy** relative to the Lebesgue measure (it is infinitely more certain than any continuous spread).

# 5. Comparison of Differential Entropy and Shannon Entropy


1. Shannon Entropy (Discrete)
For a discrete random variable $X$, Shannon entropy is defined as:
$$H(X) = -\sum_{i} P(x_i) \ln P(x_i)$$

* **Range:** $[0, \infty)$
* **Why it can't be negative:** Since $0 \le P(x_i) \le 1$, the term $\ln P(x_i)$ is always $\le 0$. Multiplying by the negative sign in the formula ensures $H(X)$ is always non-negative.
* **Physical Meaning:** It represents the average number of bits (or nats) needed to identify an outcome. You can't need "negative bits" to describe something.
* **Minimum Value:** $0$ (occurs when one outcome has probability $1$; there is no uncertainty).

---

2. Differential Entropy (Continuous)
For a continuous random variable $X$ with PDF $f(x)$, differential entropy is defined as:
$$h(X) = -\int_{-\infty}^{\infty} f(x) \ln f(x) dx$$

* **Range:** $(-\infty, \infty)$
* **Why it can be negative:** Unlike probabilities, a probability **density** $f(x)$ can be much greater than $1$.
    > **Example:** If $X$ is uniform on $[0, 0.1]$, then $f(x) = 10$. The $\ln(10)$ term is positive, so $-\int 10 \ln(10) dx$ becomes **negative**.
* **Physical Meaning:** It does **not** represent the absolute number of bits to describe the value (which would be infinite for any continuous variable). Instead, it describes the volume of the support relative to the Lebesgue measure.
* **The Delta Limit:** As you saw in the previous plot, as a distribution narrows (like the Dirac Delta), $f(x) \to \infty$, which forces $h(X) \to -\infty$.

---


| Feature | Shannon Entropy $H(X)$ | Differential Entropy $h(X)$ |
| :--- | :--- | :--- |
| **Domain** | Discrete | Continuous |
| **Range** | $[0, \infty)$ | $(-\infty, \infty)$ |
| **$P(x)$ vs $f(x)$** | Probabilities ($\le 1$) | Densities (can be $> 1$) |
| **Coordinate Change** | Invariant | Changes with scaling ($h(aX) = h(X) + \ln|a|$) |
| **Interpretation** | Absolute uncertainty | Relative uncertainty |


# 6. Conditional Entropy

## 6.1 Conditional Differential Entropy, $(h(X\mid Y))$


The conditional differential entropy $h(X \mid Y)$ measures the amount of uncertainty remaining in the random variable $X$, **on average**, after the random variable $Y$ has been observed.

Mathematically, it is the expected value (over all possible outcomes of $Y$) of the differential entropy of the conditional random variable $X \mid Y = y$.

Recall that:
1. For a fixed $y$, the random variable $X \mid Y=y$ has the law $P_{X \mid Y=y}$ and the PDF $f_{X \mid Y}(x \mid y)$.
2. The differential entropy of this specific conditional random variable is:
   $$h(X \mid Y = y) = -\int_{\mathscr{X}} f_{X \mid Y}(x \mid y) \ln f_{X \mid Y}(x \mid y) \, dx$$
3. The conditional entropy $h(X \mid Y)$ is defined as the expected value of $h(X \mid Y = y)$ across all possible values of the conditioning random variable $Y$:
   $$h(X \mid Y) \triangleq \mathbb{E}_Y [ h(X \mid Y = Y) ] = \int_{\mathscr{Y}} h(X \mid Y = y) f_Y(y) \, dy$$


Substituting the expression for $h(X \mid Y = y)$ into the expectation gives the complete formula for the conditional differential entropy:

$$h(X \mid Y) = -\int_{\mathscr{Y}} \left( \int_{\mathscr{X}} f_{X \mid Y}(x \mid y) \ln f_{X \mid Y}(x \mid y) \, dx \right) f_Y(y) \, dy$$

Since $f_{X,Y}(x, y) = f_{X \mid Y}(x \mid y) f_Y(y)$, we can merge the terms into a single joint expectation:

$$h(X \mid Y) = -\int_{\mathscr{Y}} \int_{\mathscr{X}} f_{X,Y}(x, y) \ln f_{X \mid Y}(x \mid y) \, dx \, dy$$



## 6.2 Conditional Shannon Entropy, $(H(X\mid Y))$

The conditional entropy $H(X \mid Y)$ measures the amount of uncertainty remaining in the discrete random variable $X$, **on average**, after the discrete random variable $Y$ has been observed.

Mathematically, it is the expected value (over all possible outcomes of $Y$) of the entropy of the conditional random variable $X \mid Y = y$.


1.  For a fixed $y$, the random variable $X \mid Y=y$ follows the conditional probability mass function (PMF) $P(x \mid y)$.
2.  The entropy of this specific conditional random variable is:
$$H(X \mid Y = y) = -\sum_{x \in \mathscr{X}} P(x \mid y) \ln P(x \mid y)$$
3.  The conditional entropy $H(X \mid Y)$ is defined as the expected value of $H(X \mid Y = y)$ across all possible values of the conditioning random variable $Y$:
$$H(X \mid Y) \triangleq \mathbb{E}_Y [ H(X \mid Y = Y) ] = \sum_{y \in \mathscr{Y}} P(y) H(X \mid Y = y)$$

Substituting the expression for $H(X \mid Y = y)$ into the expectation gives the full formula for discrete conditional entropy:

$$H(X \mid Y) = -\sum_{y \in \mathscr{Y}} P(y) \sum_{x \in \mathscr{X}} P(x \mid y) \ln P(x \mid y)$$

Since the joint probability is defined as $P(x, y) = P(x \mid y) P(y)$, we can simplify the terms into a single joint expectation:

$$H(X \mid Y) = -\sum_{x \in \mathscr{X}, y \in \mathscr{Y}} P(x, y) \ln P(x \mid y)$$

> **Note:** Unlike differential entropy, which can be negative, discrete conditional entropy $H(X \mid Y)$ is always non-negative ($H(X \mid Y) \geq 0$).

## 6.3 The case of independent $X$ and $Y$

If $X$ and $Y$ are independent, the following two properties hold:
1.  **Joint PDF:** The joint density is the product of the marginal densities:

$$f_{X,Y}(x, y) = f_X(x) f_Y(y)$$

2.  **Conditional PDF:** The conditional density of $X$ given $Y$ is simply the marginal density of $X$:
    
$$f_{X|Y}(x|y) = \frac{f_{X,Y}(x, y)}{f_Y(y)} = \frac{f_X(x)f_Y(y)}{f_Y(y)} = f_X(x)$$


Starting with the definition of conditional entropy provided:
$$h(X \mid Y) = -\int_{\mathscr{Y}} \int_{\mathscr{X}} f_{X,Y}(x, y) \ln f_{X \mid Y}(x \mid y) \, dx \, dy$$

We substitute the independence identities ($f_{X,Y} = f_X f_Y$ and $f_{X|Y} = f_X$) into the integral:
$$h(X \mid Y) = -\int_{\mathscr{Y}} \int_{\mathscr{X}} f_X(x) f_Y(y) \ln f_X(x) \, dx \, dy$$


Since the term $\ln f_X(x)$ does not depend on $y$, we can rearrange the double integral as a product of two separate integrals:

$$h(X \mid Y) = \left( \int_{\mathscr{Y}} f_Y(y) \, dy \right) \left( -\int_{\mathscr{X}} f_X(x) \ln f_X(x) \, dx \right)=h(X).$$

Following this same reasoning one can show that if $X$ and $Y$ are independent
$$H(X\mid Y)= H(X).$$


# 7.  Entropy of the Conditional Expectation, $(h(\mathbb{E}(X \mid Y)))$

Recall that $\mathbb{E}[X \mid Y] = g(Y)$ is a *random variable* that is a deterministic function of $Y$.

When we evaluate $h(\mathbb{E}(X \mid Y))$, we are calculating the differential entropy of this **new random variable**, $Z = g(Y) = \mathbb{E}[X \mid Y]$.

Because $Z = g(Y)$ is a function of $Y$, it lives on its own state space $\mathscr{Z} \subseteq \mathbb{R}^n$. To compute its entropy, we must determine its own probability density function, which we will call $f_Z(z)$.


The density $f_Z(z)$ is obtained by mapping the marginal distribution of $Y$ through the function $g(y) = \int_{\mathscr{X}} x f_{X \mid Y}(x \mid y) \, dx$.

If $g(y)$ is a smooth, invertible function, the density of $Z$ can be found via the standard transformation of variables using the Jacobian determinant:

$$f_Z(z) = f_Y(g^{-1}(z)) \left| \det \frac{\partial g^{-1}(z)}{\partial z} \right|$$

Once this density is established, the differential entropy $h(\mathbb{E}(X \mid Y))$ is evaluated as:

$$h(\mathbb{E}(X \mid Y)) = h(Z) = -\int_{\mathscr{Z}} f_Z(z) \ln f_Z(z) \, dz.$$

# 8. Mutual Information

## 8.1 For Continuous Random Variables

Mutual Information measures **how far the joint distribution $P_{X,Y}$ is from the case where $X$ and $Y$ are independent** ($P_X \otimes P_Y$).

$$I(X;Y) \triangleq D_{KL}(P_{X,Y} \, \| \, P_X P_Y)$$

Using the Radon-Nikodym derivative (the ratio of densities), this is defined as:


$$I(X;Y) = \int_{\mathfrak{X} \times \mathfrak{Y}} f_{X,Y}(x,y) \ln \left( \frac{f_{X,Y}(x,y)}{f_X(x) f_Y(y)} \right) dx \, dy$$

---

**Relationship Between Mutual Information and Conditional Entropy**

Starting from mutual information,

$$
I(X;Y)=\iint f_{X,Y}(x,y)\ln\frac{f_{X,Y}(x,y)}{f_X(x)f_Y(y)}\,dx\,dy,
$$

use

$$
f_{X,Y}(x,y)=f_{X|Y}(x|y)f_Y(y)
$$

to get

$$
I(X;Y)=\iint f_{X,Y}(x,y)\ln\frac{f_{X|Y}(x|y)}{f_X(x)}\,dx\,dy.
$$

Then

$$
I(X;Y)
=\iint f_{X,Y}(x,y)\ln f_{X|Y}(x|y)\,dx\,dy
=\iint f_{X,Y}(x,y)\ln f_X(x)\,dx\,dy
-h(X|Y)+h(X).
$$

Hence

$$
I(X;Y)=h(X)-h(X|Y).
$$

Interestingly, even though $h(X)$ can be negative, Mutual Information $I(X;Y)$ is always non-negative ($I(X;Y) \ge 0$), even in the continuous case. This is because $I(X;Y)$ measures the reduction in uncertainty. Even if the starting uncertainty is negative, knowing $Y$ can only make you "more certain" (more negative) or leave you where you are. The difference—the information gained—remains a positive quantity.Is the idea of "negative uncertainty" starting to click, or does it still feel a bit like mathematical sorcery?

---

**Summary**

> Mutual Information is the **"Universal Information Gain."** It is the reduction in the entropy of $X$ achieved by observing $Y$. In other words it measures the percentage of **total uncertainty** (entropy) removed.




## 8.2 For Discrete Random Variables

**Mutual Information:**

$I(X;Y)$ measures the statistical dependence between two discrete variables by quantifying how much the joint distribution $P(x,y)$ deviates from independence $P(x)P(y)$.

$$I(X;Y) = D_{KL}(P(x,y) \| P(x)P(y)) = \sum_{x,y} P(x,y) \ln \left( \frac{P(x,y)}{P(x)P(y)} \right)$$

---

**Relationship with Entropy:**

By substituting $P(x,y) = P(x|y)P(y)$, we can decompose the formula:

$$I(X;Y) = \sum_{x,y} P(x,y) \ln P(x|y) - \sum_{x,y} P(x,y) \ln P(x)$$
$$I(X;Y) = -H(X|Y) + H(X)$$

Resulting in the fundamental identity:
$$I(X;Y) = H(X) - H(X|Y)$$

---

**Summary:**

* **Information Gain:** $I(X;Y)$ is the reduction in uncertainty of $X$ due to the knowledge of $Y$.
* **Non-negativity:** Unlike differential entropy, both $H(X)$ and $I(X;Y)$ are always $\geq 0$ in the discrete case.
* **Symmetry:** It captures the information shared between variables, such that $I(X;Y) = I(Y;X)$.

## 8.3 Examples

### 8.3.1 Example: The Two coin flip problem

Suppose we flip two independent, fair coins.
* Let **$X \in \{0, 1\}$** be the outcome of the first coin ($1 = \text{Heads}$, $0 = \text{Tails}$).
* Let **$Y \in \{0, 1, 2\}$** be the total number of heads observed across both coins.

We want to quantify the amount of information that $Y$ reveals about $X$ using three metrics from information theory: **Entropy $H(X)$**, **Conditional Entropy $H(X|Y)$**, and **Mutual Information $I(X;Y)$**. Because we want our answers in **bits**, we will use base-2 logarithms ($\ln_2$).

---

There are $4$ equally likely outcomes for the two coin flips: $(\text{T,T})$, $(\text{T,H})$, $(\text{H,T})$, and $(\text{H,H})$. From these, we construct the joint probability distribution $P(X=x, Y=y)$:

| $X$ (First Coin) \ $Y$ (Total Heads) | $Y = 0$ | $Y = 1$ | $Y = 2$ | Marginal $P(X=x)$ |
| :--- | :--- | :--- | :--- | :--- |
| **$X = 0$** (Tails) | $0.25$ | $0.25$ | $0.00$ | **$0.50$** |
| **$X = 1$** (Heads) | $0.00$ | $0.25$ | $0.25$ | **$0.50$** |
| **Marginal $P(Y=y)$** | **$0.25$** | **$0.50$** | **$0.25$** | **$1.00$** |

---

The marginal entropy $H(X)$ measures the total a priori uncertainty about the first coin before observing $Y$.

$$H(X) = -\sum_{x \in \{0,1\}} P(X=x) \ln_2 P(X=x)$$

Substituting our marginal probabilities for $X$:

$$H(X) = -\left(0.5 \ln_2(0.5) + 0.5 \ln_2(0.5)\right) = -\left(0.5(-1) + 0.5(-1)\right) = \mathbf{1 \text{ bit}}$$

> **Interpretation:** A fair coin flip has exactly $1$ bit of uncertainty.

---

The conditional entropy $H(X|Y)$ measures the uncertainty remaining in $X$ *after* we learn the value of $Y$. It is the weighted average of the uncertainties $H(X|Y=y)$ across all possible values of $Y$.

$$H(X|Y) = \sum_{y \in \{0,1,2\}} P(Y=y) H(X|Y=y)$$


Let us calculate the uncertainty $H(X \mid Y=y)$ for each specific outcome of $Y$:

**Case A: If $Y = 0$ (Zero total heads):**

We know with certainty that both coins were Tails. Therefore, $P(X=0 \mid Y=0) = 1$ and $P(X=1 \mid Y=0) = 0$.
$$H(X|Y=0) = -(1 \ln_2 1 + 0 \ln_2 0) = \mathbf{0 \text{ bits}}$$

**Case B: If $Y = 2$ (Two total heads):**

We know with certainty that both coins were Heads. Therefore, $P(X=0 \mid Y=2) = 0$ and $P(X=1 \mid Y=2) = 1$.
$$H(X|Y=2) = -(0 \ln_2 0 + 1 \ln_2 1) = \mathbf{0 \text{ bits}}$$

**Case C: If $Y = 1$ (Exactly one total head):**

The single head could be on either the first coin or the second coin with equal probability. Therefore, $P(X=0 \mid Y=1) = 0.5$ and $P(X=1 \mid Y=1) = 0.5$.
$$H(X|Y=1) = -(0.5 \ln_2 0.5 + 0.5 \ln_2 0.5) = \mathbf{1 \text{ bit}}$$

**Bringing it all together:**

We average these specific entropies, weighting them by the marginal probabilities $P(Y=y)$:

$$H(X|Y) = (0.25 \times 0 \text{ bits}) + (0.50 \times 1 \text{ bit}) + (0.25 \times 0 \text{ bits}) = \mathbf{0.5 \text{ bits}}$$

---


The mutual information $I(X;Y)$ measures the reduction in uncertainty about $X$ that results from learning $Y$:

$$I(X;Y) = H(X) - H(X|Y)$$

Plugging in our derived values:

$$I(X;Y) = 1 \text{ bit} - 0.5 \text{ bits} = \mathbf{0.5 \text{ bits}}$$

---

**Physical Interpretation:**

* **Before observing $Y$:** You have $1$ bit of uncertainty about the outcome of the first coin ($X$).
* **After observing $Y$:** On average, half of that uncertainty is eliminated.
    * If $Y=0$ or $Y=2$ ($50\%$ of the time), you know exactly what $X$ did; your residual uncertainty drops to $0$ bits.
    * If $Y=1$ ($50\%$ of the time), you gain no information about $X$; your uncertainty remains at $1$ bit.

Thus, knowing the total number of heads across two coin flips provides exactly **$0.5$ bits** of information about the outcome of the first coin.

### 8.3.2 The problem of noisy measurements

Let $X$ represent the true value of a physical quantity (such as temperature or voltage), and let $Y$ be a noisy reading from a sensor. We model the noise $N$ as additive, meaning:

$$Y = X + N$$

We assume that:
* The true signal $X \sim \mathscr{N}(0, \sigma_X^2)$ is a Gaussian random variable.
* The sensor noise $N \sim \mathscr{N}(0, \sigma_N^2)$ is a Gaussian random variable.
* $X$ and $N$ are completely independent.

**Goal:** Derive the expressions for the original uncertainty $h(X)$, the information gained $I(X;Y)$, and the remaining uncertainty $h(X \mid Y)$ given by:

$$h(X \mid Y) = h(X) - I(X;Y)$$

---


**Step 1: Calculate the Original Uncertainty $h(X)$:**

Since $X$ is a normally distributed random variable with variance $\sigma_X^2$, its differential entropy (measured in nats) follows the standard Gaussian entropy formula derived earlier:

$$h(X) = \frac{1}{2} \ln(2\pi e \sigma_X^2)$$

---

**Step 2: Calculate the Mutual Information $I(X;Y)$:**

To calculate the information gained, we use the alternative formula for mutual information:

$$I(X;Y) = h(Y) - h(Y \mid X)$$

We must evaluate both $h(Y)$ and $h(Y \mid X)$:


Because $X$ and $N$ are independent Gaussian random variables, their sum $Y = X + N$ is also a Gaussian random variable. Its variance is the sum of their individual variances:

$$\text{Var}(Y) = \sigma_X^2 + \sigma_N^2$$

Thus, the differential entropy of the noisy sensor reading $Y$ is:

$$h(Y) = \frac{1}{2} \ln(2\pi e (\sigma_X^2 + \sigma_N^2))$$


The conditional entropy $h(Y \mid X)$ represents the uncertainty remaining in the measurement $Y$ once we know the exact true value $X$.

Substituting $Y = X + N$ gives:

$$h(Y \mid X) = h(X + N \mid X) = h(N \mid X)$$

Because $N$ and $X$ are independent, conditioning on $X$ does not change the uncertainty of $N$:

$$h(N \mid X) = h(N) = \frac{1}{2} \ln(2\pi e \sigma_N^2)$$


Now we substitute $h(Y)$ and $h(Y \mid X)$ into our mutual information equation:

$$I(X;Y) = \frac{1}{2} \ln(2\pi e (\sigma_X^2 + \sigma_N^2)) - \frac{1}{2} \ln(2\pi e \sigma_N^2)$$

Using the properties of logarithms to combine the terms:

$$I(X;Y) = \frac{1}{2} \ln\left( \frac{2\pi e (\sigma_X^2 + \sigma_N^2)}{2\pi e \sigma_N^2} \right) = \mathbf{\frac{1}{2} \ln\left(1 + \frac{\sigma_X^2}{\sigma_N^2}\right)}$$

> **Interpretation:** The term $\frac{\sigma_X^2}{\sigma_N^2}$ is the **Signal-to-Noise Ratio (SNR)**. The information gained by reading the sensor is directly proportional to the SNR.

---

**Step 3: Calculate the Remaining Uncertainty $h(X \mid Y)$:**

Using the identity $h(X \mid Y) = h(X) - I(X;Y)$, we can now compute the residual uncertainty about the true value $X$ after reading the sensor $Y$:

$$h(X \mid Y) = \frac{1}{2} \ln(2\pi e \sigma_X^2) - \frac{1}{2} \ln\left(1 + \frac{\sigma_X^2}{\sigma_N^2}\right)$$

Combining the logarithms:

$$h(X \mid Y) = \frac{1}{2} \ln\left( \frac{2\pi e \sigma_X^2}{1 + \frac{\sigma_X^2}{\sigma_N^2}} \right)$$

We can simplify the denominator by finding a common denominator:

$$1 + \frac{\sigma_X^2}{\sigma_N^2} = \frac{\sigma_N^2 + \sigma_X^2}{\sigma_N^2}$$

Plugging this back in:

$$h(X \mid Y) = \frac{1}{2} \ln\left( \frac{2\pi e \sigma_X^2}{\frac{\sigma_N^2 + \sigma_X^2}{\sigma_N^2}} \right) = \mathbf{\frac{1}{2} \ln\left(2\pi e \frac{\sigma_X^2 \sigma_N^2}{\sigma_X^2 + \sigma_N^2}\right)}$$

---

**Physical Interpretation:**

The results reveal a clear relationship between sensor noise ($\sigma_N^2$) and information flow:

```
                  ┌──────────────────────┐
                  │ Original Signal h(X) │
                  └──────────┬───────────┘
                             ▼
              ┌──────────────┴──────────────┐
              │ Sensor Measurement Noise    │
              └──────────────┬──────────────┘
                             ▼
             ┌───────────────┴───────────────┐
             │       Signal to Noise         │
             │     SNR = σ_X^2 / σ_N^2       │
             └───────────────┬───────────────┘
                             │
            ┌────────────────┴────────────────┐
            ▼                                 ▼
   High Noise (σ_N^2 → ∞)            Low Noise (σ_N^2 → 0)
   • SNR → 0                         • SNR → ∞
   • I(X;Y) ≈ 0                      • I(X;Y) is large
   • h(X|Y) ≈ h(X)                   • h(X|Y) → -∞ (absolute certainty)
```

---

**Scenario A: The Very Noisy Sensor ($\sigma_N^2 \to \infty$):**

If the sensor has high noise, then $\sigma_N^2$ is much larger than the signal variance $\sigma_X^2$:
* The SNR $\frac{\sigma_X^2}{\sigma_N^2} \approx 0$.
* The information gain is $I(X;Y) = \frac{1}{2} \ln(1 + 0) = \mathbf{0 \text{ nats}}$.
* The remaining uncertainty is $h(X|Y) \approx h(X)$.

> **Conclusion:** A bad sensor reduces almost no uncertainty about the true physical quantity.

**Scenario B: The Highly Accurate Sensor ($\sigma_N^2 \to 0$):**

If the sensor is highly precise, then $\sigma_N^2$ is tiny compared to $\sigma_X^2$:
* The SNR $\frac{\sigma_X^2}{\sigma_N^2} \to \infty$.
* The information gain is $I(X;Y) \to \mathbf{\infty \text{ nats}}$.
* The remaining uncertainty is $h(X \mid Y) = \frac{1}{2} \ln(2 \pi e \cdot 0) = \mathbf{-\infty \text{ nats}}$.

---

> **Conclusion:** An ultra-precise sensor collapses the uncertainty of the true quantity down to zero (which corresponds to $-\infty$ differential entropy).

### 8.3.3 The conditional expectation $E[X \mid Y]$ for the linear Gaussian model

In this section, we derive the distribution of the conditional expectation $E[X \mid Y]$ for the linear Gaussian model. This is a fundamental result in estimation theory, showing how the "best estimate" itself behaves as a random variable.

---

1. The Model Setup

Consider the signal-plus-noise model:
* **Signal:** $X \sim \mathscr{N}(0, \sigma_X^2)$
* **Noise:** $N \sim \mathscr{N}(0, \sigma_N^2)$, where $X \perp N$.
* **Observation:** $Y = X + N$.

Since $Y$ is a linear combination of independent Gaussians, the joint distribution of $(X, Y)$ is bivariate normal:
$$\begin{pmatrix} X \\ Y \end{pmatrix} \sim \mathscr{N} \left( \begin{pmatrix} 0 \\ 0 \end{pmatrix}, \begin{pmatrix} \sigma_X^2 & \sigma_X^2 \\ \sigma_X^2 & \sigma_X^2 + \sigma_N^2 \end{pmatrix} \right)$$
*(Note: $\text{Cov}(X, Y) = E[X(X+N)] = E[X^2] = \sigma_X^2$.)*

---

2. Deriving the Conditional Expectation $E[X \mid Y]$

For a bivariate normal distribution, the conditional expectation is given by the linear projection:
$$E(X \mid Y = y) = \mu_X + \frac{\text{Cov}(X, Y)}{\text{Var}(Y)}(y - \mu_Y)$$

Substituting our values:
$$E[X \mid Y = y] = 0 + \frac{\sigma_X^2}{\sigma_X^2 + \sigma_N^2}(y - 0)$$

Let $k = \frac{\sigma_X^2}{\sigma_X^2 + \sigma_N^2}$. This constant $k$ is often called the **Kalman Gain** or the **Shrinkage Factor**. It represents the "reliability" of the sensor.
Thus, the random variable is:
$$\widehat{X} = E[X \mid Y] = kY$$

---

3. Distribution of the Random Variable $E[X \mid Y]$

To find the distribution of $\widehat{X} = kY$, we recall that a linear transformation of a Gaussian is still Gaussian.

**A. The Mean:**
$$E[\widehat{X}] = E[kY] = k E[Y] = 0$$

**B. The Variance:**
$$\text{Var}(\widehat{X}) = \text{Var}(kY) = k^2 \text{Var}(Y)$$
Substitute $k$ and $\text{Var}(Y)$:
$$\text{Var}(\widehat{X}) = \left( \frac{\sigma_X^2}{\sigma_X^2 + \sigma_N^2} \right)^2 (\sigma_X^2 + \sigma_N^2)$$
$$\text{Var}(\widehat{X}) = \frac{(\sigma_X^2)^2}{\sigma_X^2 + \sigma_N^2}$$

**Conclusion:**
$$E[X \mid Y] \sim \mathscr{N}\left( 0, \frac{\sigma_X^4}{\sigma_X^2 + \sigma_N^2} \right)$$

---

4. Decomposition of Variance

A profound result in probability is that the total variance of $X$ can be decomposed into the "explained" variance and the "unexplained" (residual) variance:
$$\text{Var}(X) = \text{Var}(E[X \mid Y]) + E[\text{Var}(X \mid Y)]$$

From our derivation:
1.  **Total Variance:** $\sigma_X^2$
2.  **Explained Variance (Variance of the Estimate):** $\frac{\sigma_X^4}{\sigma_X^2 + \sigma_N^2}$
3.  **Unexplained Variance (Error Variance):** $\sigma_X^2 - \frac{\sigma_X^4}{\sigma_X^2 + \sigma_N^2} = \frac{\sigma_X^2 \sigma_N^2}{\sigma_X^2 + \sigma_N^2}$

---

5. Practical Interpretation
* **Perfect Sensor ($\sigma_N^2 \to 0$):** $k \to 1$. The estimate $E[X \mid Y] \to Y$. The variance of the estimate matches the variance of $X$ perfectly.
* **Useless Sensor ($\sigma_N^2 \to \infty$):** $k \to 0$. The estimate $E[X \mid Y] \to 0$. The variance of the estimate is $0$ because we always guess the mean.

> **Key Takeaway:** The variance of the conditional expectation $E[X \mid Y]$ is always **less than or equal to** the variance of $X$. This is because the estimate is a "smoothed" or "shrunken" version of the reality, filtered through the lens of a noisy observation.

# 9. Joint Entropy and Conditional Entropy

1. **Differential Entropy (Continuous Case):**
For continuous random variables with a joint density $f(x,y)$, the relationship is mathematically identical because the joint density follows the same product rule:
$$f(x,y) = f(x)f(y|x).$$
* The differential entropy is defined via the Lebesgue integral:
$$h(X,Y) = -\int \int f(x,y) \ln f(x,y) \, dx dy.$$
* Just as in the discrete case, the linearity of the integral and the property of the logarithm allow the joint entropy to be split into the marginal and conditional components:
$$h(X,Y) = h(X) + h(Y|X).$$

2. **Shannon Entropy (Discrete Case):**
For discrete random variables, the joint probability can be decomposed using the product rule:
$$P(X, Y) = P(X)P(Y|X).$$
* The joint entropy is defined as:
$$H(X,Y) = -\sum_{x,y} p(x,y) \ln p(x,y).$$
* By substituting the product rule into the logarithm:
$$\ln p(x,y) = \ln p(x) + \ln p(y|x)$$.
* Distributing the summation and expectation leads to the identity:
$$H(X,Y) = H(X) + H(Y|X).$$
