# Rigorous Formulation of Bayesian Inference via Conditional Expectation

[cite_start]This section provides a mathematically rigorous overview of Bayesian inference using measure theory and regular conditional distributions, based on the foundational probability concepts outlined in our statistical learning toolkit[cite: 32, 53, 154].

---

## 1. The Measure-Theoretic Setup

[cite_start]Let $(\Omega, \mathfrak{A}, P)$ be our underlying probability space[cite: 33, 90, 156]. [cite_start]In Bayesian inference, the latent parameters and observable data are modeled as random variables defined on this space[cite: 53, 90]:

* [cite_start]**The Parameter / Latent State ($X$):** A measurable map $X: \Omega \rightarrow \mathbb{X}$ representing the hidden quantity of interest[cite: 55, 56, 90]. [cite_start]Its global distribution before seeing data is governed by the marginal law $P_X$[cite: 63, 67, 248].
* [cite_start]**The Observed Data ($Y$):** A measurable map $Y: \Omega \rightarrow \mathbb{Y}$ representing the collected evidence[cite: 55, 56, 90].

[cite_start]Together, they form a single joint random ordered pair $Z = (X, Y)$ mapping into the product space $(\mathbb{X} \times \mathbb{Y}, \mathfrak{U}_X \otimes \mathfrak{U}_Y)$ governed by the joint law $P_{X,Y}$[cite: 91, 92, 93].

---

## 2. Prior, Likelihood, and Posterior via Regular Conditional Laws

When conditioning on continuous data, classical calculus definitions break down if an event has a probability of zero ($P(Y=y) = 0$). [cite_start]To remain rigorous, we define these relationships via **Radon-Nikodym derivatives** and **Regular Conditional Laws**[cite: 43, 86, 251].

### The Prior Density
[cite_start]If the marginal law $P_X$ is absolutely continuous with respect to the Lebesgue measure $\lambda$, the **Prior PDF** is defined by[cite: 85, 86, 88]:
$$f_X(x) \triangleq \frac{dP_X}{d\lambda}(x)$$

### The Likelihood
[cite_start]For a fixed localized slice of reality where $X = x$, the data behaves according to the regular conditional law $P_{Y|X=x}$[cite: 250, 251, 252]. [cite_start]This gives us our **Likelihood Density**[cite: 201, 202, 257]:
$$f_{Y|X}(y|x) \triangleq \frac{dP_{Y|X=x}}{dy}(y)$$

### The Posterior Density
[cite_start]By utilizing the **Product Rule of Conditional Density**, the joint density factorizes as $f_{X,Y}(x,y) = f_{X|Y}(x|y)f_Y(y)$[cite: 207, 212]. [cite_start]Applying Bayes' theorem, the updated **Posterior PDF** for the conditional random variable $(X|Y=y)$ is formulated as[cite: 251, 257, 259]:

\begin{align}
f_{X|Y}(x|y) &= \frac{f_{X,Y}(x,y)}{f_Y(y)} \\
&= \frac{f_{Y|X}(y|x)f_X(x)}{\int_{\mathbb{X}} f_{Y|X}(y|x')f_X(x')dx'}
\end{align}

[cite_start]Where the denominator represents the marginal density of $Y$ derived by integrating out the parameter (The Law of Total Probability)[cite: 115, 262, 265]:
$$f_Y(y) = \int_{\mathbb{X}} f_{Y|X}(y|x)f_X(x)dx$$

## 3. The Conditional Expectation Formulation (Optimal Estimation)

[cite_start]If our goal is to produce an optimal point estimate of the parameter $X$ given our data $Y$, we construct the conditional expectation operator $\mathbb{E}[X|Y]$[cite: 154, 155, 317].

### Definition and Partial Averaging
[cite_start]The Bayesian point estimate is a $\sigma(Y)$-measurable random variable $Z = g(Y) = \mathbb{E}[X|Y]$ uniquely defined via the partial averaging property[cite: 161, 163, 165, 167]:
$$\int_{G} \mathbb{E}[X|Y] dP = \int_{G} X dP \quad \forall G \in \sigma(Y)$$

[cite_start]When evaluated at a specific observation slice $Y=y$, the **Posterior Mean** computes the functional average over the entire domain with respect to the localized posterior measure[cite: 234, 240, 241, 280]:
$$g(y) = \mathbb{E}[X|Y=y] = \int_{\mathbb{X}} x f_{X|Y}(x|y) dx$$

### The Optimality of the Posterior Mean
[cite_start]Let $Z$ be *any* other arbitrary estimator that is a function of our data (meaning $Z$ is $\sigma(Y)$-measurable)[cite: 304]. [cite_start]The conditional expectation formulation is the **unique minimizer of the Mean Square Error (MSE)**[cite: 306, 307]:
$$\mathbb{E}[(X - \mathbb{E}[X|Y])^2] \le \mathbb{E}[(X - Z)^2]$$

[cite_start]The residual, unexplained uncertainty that remains after calculating this optimal estimate is trapped in the **Conditional Variance**[cite: 291, 302]:
$$Var(X|Y) \triangleq \mathbb{E}[X^2|Y] - (\mathbb{E}[X|Y])^2$$

## 4. Canonical Application: The Multivariate Gaussian Case

[cite_start]When the parameters and data are jointly Gaussian, this measure-theoretic machinery collapses into a beautiful, closed-form analytical solution[cite: 314, 315]. [cite_start]Given[cite: 316]:

$$Z = \begin{bmatrix} X \\ Y \end{bmatrix} \sim \mathcal{N}\left(\begin{bmatrix} \mu_X \\ \mu_Y \end{bmatrix}, \begin{bmatrix} \Sigma_{XX} & \Sigma_{XY} \\ \Sigma_{YX} & \Sigma_{YY} \end{bmatrix}\right)$$

[cite_start]The optimal Bayesian update map is exactly[cite: 317, 318]:
$$\mathbb{E}[X|Y] = \mu_X + \Sigma_{XY}\Sigma_{YY}^{-1}(Y - \mu_Y)$$

### Component Decomposition
* [cite_start]**$\mu_X$ (The Prior Mean):** Our baseline baseline "best guess" for $X$ assuming zero data information[cite: 320, 321, 322].
* **$(Y - \mu_Y)$ (The Innovation / Residual):** The error vector tracking the difference between what we observed and what we expected to observe[cite: 319].
* **$\Sigma_{XY}\Sigma_{YY}^{-1}$ (The Gain Matrix / Information Bridge):** Scales the innovation based on the cross-covariance $\Sigma_{XY}$ normalized by total data uncertainty $\Sigma_{YY}$[cite: 326, 327].

By the **Law of Total Variance**, the posterior uncertainty strips away the explained variance from our total prior variance $\Sigma_{XX}$[cite: 295, 297, 301, 325]:
$$Var(X|Y) = \Sigma_{XX} - \Sigma_{XY}\Sigma_{YY}^{-1}\Sigma_{YX}$$
