# Overview

Bayesian inference can be thought of as a theory of **learning under uncertainty**.

In many situations, we do not know the true state of the world. We may not know the value of a parameter, the correct model, the hidden cause of an observation, or the future outcome of a process. However, we are rarely in a state of complete ignorance. We usually begin with some prior information: previous experience, physical constraints, expert knowledge, symmetry assumptions, or even a deliberately vague state of belief.

Bayesian inference formalizes this situation by saying:

> Knowledge is not all-or-nothing; it comes in degrees of belief.

These degrees of belief are represented by probability measures.

The unknown quantity, say $\theta$, is treated as uncertain. This does **not necessarily mean** that $\theta$ is physically random. It may be fixed but unknown. The probability distribution over $\theta$ represents our state of knowledge about it.

Before observing data, we have a **prior belief** represented by a probability measure
$\Pi$
on a measurable parameter space $(\Theta,\mathfrak T)$.


After an observation, the data does not usually reveal $\theta$ exactly. Rather, it changes which values of $\theta$ are more or less plausible. Bayesian inference updates the prior belief $\Pi$ into a **posterior belief.**


So the philosophical idea of  Bayesian inference is:

> Bayesian inference is rational belief revision in the presence of evidence.



This is one of the key differences between Bayesian and purely frequentist thinking. In a frequentist view, the parameter $\theta$ is fixed and probability describes the long-run behavior of repeated data. In a Bayesian view, probability may also describe uncertainty about $\theta$ itself, conditional on the information currently available.

Thus, Bayesian inference is not merely a computational rule. It is a coherent framework for answering the question:

> Given what I believed before, and given what I have now observed, what should I believe next?

In measure-theoretic language, this becomes simply:

> Bayesian inference is conditioning a joint probability measure on the observed data.

In the following we rigorously define this notion.

# Mathematical Formulation

Let $(\Omega,\mathfrak A,P)$ be an underlying probability space. Let

$$
X:\Omega\to \mathscr X
$$

be a random variable representing the unknown parameter, latent state, or hidden quantity of interest (traditionaly it is common to use the variable $\theta$ to denote paramters. However in order to stick to the general conditioning notation adopted in this course we will use $X$ instead). Here $(\mathscr X,\mathfrak U_X)$ is a measurable space. Let

$$
Y:\Omega\to \mathscr Y
$$

be a random variable representing the observed data where $
(\mathscr Y,\mathfrak U_Y)
$$
is another measurable space.

Thus, $X$ and $Y$ are measurable maps satisfying
$X^{-1}(A)\in\mathfrak A$ for all $A\in\mathfrak U_X$,
and $Y^{-1}(B)\in\mathfrak A$ for all $B\in\mathfrak U_Y$.
Together, they define a joint random variable

$$
Z=(X,Y):\Omega\to \mathscr X\times\mathscr Y,
$$

where the product space is equipped with the product $\sigma$-algebra $\mathfrak U_X\otimes\mathfrak U_Y$.

The joint law of $(X,Y)$ is the push-forward measure

$$
P_{X,Y}\triangleq (X,Y)_{\#} P.
$$

That is, for every measurable set $C\in \mathfrak U_X\otimes\mathfrak U_Y$,

$$
P_{X,Y}(C)=P\big((X,Y)^{-1}(C)\big).
$$

The marginal law of $X$ is

$$
P_X\triangleq X_{\#}P,
$$

and the marginal law of $Y$ is

$$
P_Y\triangleq Y_{\#}P.
$$

Explicitly,

$$
P_X(A)=P(X\in A),
\qquad A\in\mathfrak U_X,
$$

and

$$
P_Y(B)=P(Y\in B),
\qquad B\in\mathfrak U_Y.
$$

In the Bayesian interpretation, $P_X$ is the **prior law** of the unknown quantity $X$ before observing the data $Y$.



# Prior, Likelihood, and Posterior via Regular Conditional Laws



In Bayesian inference, the prior, likelihood, and posterior are not fundamentally densities. They are probability measures or conditional probability kernels. Densities appear only when suitable dominating measures exist.

Assume that regular conditional distributions exist. Then there exists, for each $x\in\mathscr X$,
a probability measure on $(\mathscr Y,\mathfrak U_Y)$, such that for each $B\in\mathfrak U_Y$, the probability that the observed data lies in $B$, conditional on the parameter value $X=x$ is given by
$$
P_{Y\mid X=x}(B).
$$


Similarly, the posterior is described by a regular probability measure on $(\mathscr X,\mathfrak U_X)$
such that for each $A\in\mathfrak U_X$, the probability that the parameter $X$ lies in $A$, conditional on the observation $Y=y$ is given by
$$
P_{X\mid Y=y}(A)
$$

This is the **posterior law of $X$ after observing $Y=y$**.



>Thus, measure-theoretically, Bayesian inference is the passage from the prior law
$$
P_X
$$
>to the posterior law
$$
P_{X\mid Y=y}
$$
after observing the data value $Y=y$.

For every $A\in\mathfrak U_X$ and every $B\in\mathfrak U_Y$,

$$
P(X\in A,Y\in B)=\int_B P_{X\mid Y=y}(A)\,dP_Y(y).
$$

Equivalently,

$$
P_{X,Y}(A\times B)=\int_B P_{X\mid Y=y}(A)\,dP_Y(y).
$$

This identity says that the joint law $P_{X,Y}$ can be disintegrated with respect to the marginal law $P_Y$ of the observed data:

$$
dP_{X,Y}(x,y)=
dP_{X\mid Y=y}(x)\,dP_Y(y).
$$

On the other hand, before observing $Y$, the Bayesian model may be described by the prior law of $X$ together with the conditional law of $Y$ given $X$. That is,
$$
dP_{X,Y}(x,y)=dP_{Y\mid X=x}(y)\,dP_X(x).
$$

Thus, the same joint probability measure admits two factorizations.

Therefore, the Bayesian update is the passage

$$
dP_X(x)
\quad\longmapsto\quad
dP_{X\mid Y=y}(x),
$$
after observing the value $Y=y$.

The measure-theoretic form of Bayes' rule is the identity
$$
dP_{Y\mid X=x}(y)\,dP_X(x)=
dP_{X\mid Y=y}(x)\,dP_Y(y).
$$
It can be written in the compact symbolic form:

$$
dP_{X\mid Y=y}(x)=
\frac{
dP_{Y\mid X=x}(y)\,dP_X(x)
}{
dP_Y(y)
}.
$$
>This expression should be interpreted carefully: it is not ordinary division of measures in two independent variables. It is just short hand for the identity preceding it.

The marginal probability of finding $Y\in B$ is given by
$$
P_{Y}(B)=\int_{\mathscr{X}} P_{Y\mid X=x}(B)\,dP_X(x).
$$

# Density Form of Bayes' Rule

Suppose now that $\mathscr X$ and $\mathscr Y$ are Euclidean spaces, and suppose that the relevant measures admit densities with respect to suitable reference measures.

Let $\lambda_X$ be a reference measure on $\mathscr X$, and let $\lambda_Y$ be a reference measure on $\mathscr Y$.

Assume that the prior law $P_X$ is absolutely continuous with respect to $\lambda_X$. Then the prior density is defined by

$$
dP_X(x)=f_X(x){d\lambda_X}(x).
$$

Assume also that, for each $x$, the conditional law $P_{Y\mid X=x}$ is absolutely continuous with respect to $\lambda_Y$. Then the likelihood density is defined by

$$
dP_{Y\mid X=x}=f_{Y\mid X}(y\mid x)d\lambda_Y(y).
$$

The **word likelihood** refers to the same function viewed as a function of $x$ after the data value $y$ has been fixed. Thus, for fixed $y$,

$$
x\mapsto f_{Y\mid X}(y\mid x)
$$

is the likelihood function.

If the joint law $P_{X,Y}$ admits a density $f_{X,Y}$ with respect to $\lambda_X\otimes\lambda_Y$, then

$$
f_{X,Y}(x,y)=f_{Y\mid X}(y\mid x)f_X(x).
$$

The marginal density of $Y$ is obtained by integrating out $x$:

$$
f_Y(y)=\int_{\mathscr X} f_{Y\mid X}(y\mid x)f_X(x)\,d\lambda_X(x).
$$

Whenever $f_Y(y)>0$, the posterior density is

$$
f_{X\mid Y}(x\mid y)=\frac{f_{X,Y}(x,y)}{f_Y(y)}.
$$

Therefore,

$$
f_{X\mid Y}(x\mid y)=\frac{
f_{Y\mid X}(y\mid x)f_X(x)
}{
\int_{\mathscr X} f_{Y\mid X}(y\mid x')f_X(x')\,d\lambda_X(x')
}.
$$

This is the usual density form of Bayes' rule.

In symbolic Bayesian notation, this is often summarized as

$$
\text{posterior}
\propto
\text{likelihood}\times \text{prior}.
$$

More explicitly,

$$
f_{X\mid Y}(x\mid y)
\propto
f_{Y\mid X}(y\mid x)f_X(x),
$$

where the constant of proportionality is determined by the condition

$$
\int_{\mathscr X} f_{X\mid Y}(x\mid y)\,d\lambda_X(x)=1.
$$



# Conditional Expectation as Bayesian Point Estimation

The posterior distribution gives the full updated uncertainty about $X$ after observing $Y=y$. If we want a single point estimate of $X$, a natural Bayesian choice is the posterior mean.

Measure-theoretically, this is described by the conditional expectation

$$
\mathbb E[X\mid Y].
$$

Assume that $X$ is integrable, i.e.

$$
\mathbb E[|X|]<\infty.
$$

Then $\mathbb E[X\mid Y]$ is the unique $\sigma(Y)$-measurable random variable satisfying

$$
\int_G \mathbb E[X\mid Y]\,dP=\int_G X,dP
\qquad
\forall G\in\sigma(Y).
$$

Since $\mathbb E[X\mid Y]$ is $\sigma(Y)$-measurable, there exists a measurable function $g:\mathscr Y\to\mathscr X$ such that

$$
\mathbb E[X\mid Y]=g(Y).
$$

The function $g$ is the Bayesian estimator associated with squared-error loss.

For a particular observation $Y=y$, the posterior mean is

$$
g(y)=\mathbb E[X\mid Y=y].
$$

In terms of the posterior measure, this is

$$
g(y)=\int_{\mathscr X} x\,dP_{X\mid Y=y}(x).
$$

If the posterior admits a density, then

$$
g(y)=\int_{\mathscr X} x\,f_{X\mid Y}(x\mid y)\,d\lambda_X(x).
$$

Thus, the Bayesian point estimate is obtained by averaging over the posterior distribution.



## Optimality of the Posterior Mean

The posterior mean is not merely a convenient estimate. It is optimal under mean-square error.

Let $Z$ be any other estimator of $X$ based on the observed data $Y$. This means that $Z$ is $\sigma(Y)$-measurable.

Then

$$
\mathbb E\left[|X-\mathbb E[X\mid Y]|^2\right]
\le
\mathbb E\left[|X-Z|^2\right].
$$

Therefore, $\mathbb E[X\mid Y]$ is the optimal estimator of $X$ among all $\sigma(Y)$-measurable estimators under squared-error loss.

Equivalently,

$$
\mathbb E[X\mid Y]=\operatorname*{arg,min}_{Z\in L^2(\sigma(Y))}
\mathbb E\left[|X-Z|^2\right].
$$

The remaining uncertainty after observing $Y$ is described by the conditional variance

$$
\operatorname{Var}(X\mid Y)=\mathbb E[X^2\mid Y]
\left(\mathbb E[X\mid Y]\right)^2
$$

in the scalar case.

For vector-valued $X$, the conditional covariance is

$$
\operatorname{Cov}(X\mid Y)=\mathbb E\left[
\left(X-\mathbb E[X\mid Y]\right)
\left(X-\mathbb E[X\mid Y]\right)^T
\mid Y
\right].
$$



# Examples

## The Multivariate Gaussian Case

A central example occurs when $X$ and $Y$ are jointly Gaussian.

Suppose

$$
\begin{bmatrix}
X\\
Y
\end{bmatrix}
\sim
\mathscr N
\left(
\begin{bmatrix}
\mu_X\\
\mu_Y
\end{bmatrix},
\begin{bmatrix}
\Sigma_{XX} & \Sigma_{XY}\\
\Sigma_{YX} & \Sigma_{YY}
\end{bmatrix}
\right).
$$

Assume that $\Sigma_{YY}$ is invertible.

Then the conditional distribution of $X$ given $Y=y$ is Gaussian:

$$
X\mid Y=y
\sim
\mathscr N
\left(
\mu_{X\mid y},
\Sigma_{X\mid y}
\right),
$$

where

$$
\mu_{X\mid y}=\mu_X
+
\Sigma_{XY}\Sigma_{YY}^{-1}(y-\mu_Y),
$$

and

$$
\Sigma_{X\mid y}=\Sigma_{XX}-
\Sigma_{XY}\Sigma_{YY}^{-1}\Sigma_{YX}.
$$

Therefore, as a random variable,

$$
\mathbb E[X\mid Y]=\mu_X
+
\Sigma_{XY}\Sigma_{YY}^{-1}(Y-\mu_Y).
$$

The matrix

$$
\Sigma_{XY}\Sigma_{YY}^{-1}
$$

acts as a gain matrix. It determines how strongly the innovation

$$
Y-\mu_Y
$$

should change the prior mean $\mu_X$.

Thus, in the Gaussian case, Bayesian inference has a closed-form update:

$$
\text{posterior mean}=\text{prior mean}
+
\text{gain}\times\text{innovation}.
$$

This is the same mathematical structure that appears in linear Gaussian estimation and the Kalman filter.



## Example: Estimating the Bias of a Coin

Suppose we are given a coin, but we do not know whether it is fair. The coin may be biased toward heads, biased toward tails, or approximately fair. Let

$$
\Theta\in[0,1]
$$

denote the unknown probability that the coin lands heads.

The value of $\Theta$ is fixed for the coin, but it is unknown to us. Therefore, in the Bayesian formulation, we represent our uncertainty about $\Theta$ using a probability distribution on the interval $[0,1]$.

Before tossing the coin, we begin with a prior belief about possible values of $\Theta$. For example, if we have no strong reason to prefer one value over another, we may start with a uniform prior on $[0,1]$.

We then toss the coin several times and observe the number of heads and tails. These observations provide evidence about the unknown value of $\Theta$. If we observe many heads, then larger values of $\Theta$ become more plausible. If we observe many tails, then smaller values of $\Theta$ become more plausible.

Bayesian inference updates the prior distribution of $\Theta$ into a posterior distribution after seeing the data. The posterior distribution describes our revised uncertainty about the probability of heads.


Let $\Theta \sim \operatorname{Beta}(\alpha,\beta)$.

The prior density is

$$
f_\Theta(\theta)=\frac{1}{B(\alpha,\beta)}
\theta^{\alpha-1}(1-\theta)^{\beta-1},
\qquad 0<\theta<1.
$$


Now suppose we toss the coin $n$ times and observe $h$ heads and $t=n-h$ tails. The likelihood is

$$P(Y=(h,t)\mid \Theta=\theta)=\binom{n}{h}\theta^h(1-\theta)^t.$$


The posterior density is proportional to

$$
f_{\Theta\mid Y}(\theta\mid Y)
\propto
P(Y=(h,t)\mid \Theta=\theta)f_\Theta(\theta).
$$

Therefore,

$$
f_{\Theta\mid Y}(\theta\mid h,t)
\propto
\theta^h(1-\theta)^t
\theta^{\alpha-1}(1-\theta)^{\beta-1}.
$$

So

$$
f_{\Theta\mid Y}(\theta\mid h,t)
\propto
\theta^{\alpha+h-1}(1-\theta)^{\beta+t-1}.
$$

Hence,

$$
\Theta\mid Y=(h,t)
\sim
\operatorname{Beta}(\alpha+h,\beta+t).
$$

```python
import numpy as np
import plotly.graph_objects as go
from scipy.stats import beta

def plot_beta_prior_posterior(n, h, alpha=1, beta_param=1, num_points=1000):
    """
    Plot the prior and posterior distributions for the unknown coin bias theta.

    Parameters
    ----------
    n : int
        Total number of coin tosses.
    h : int
        Number of heads observed.
    alpha : float
        Prior Beta distribution first parameter.
    beta_param : float
        Prior Beta distribution second parameter.
    num_points : int
        Number of grid points for plotting.
    """

    if not isinstance(n, int) or not isinstance(h, int):
        raise ValueError("n and h must be integers.")

    if n < 0:
        raise ValueError("n must be non-negative.")

    if h < 0 or h > n:
        raise ValueError("h must satisfy 0 <= h <= n.")

    if alpha <= 0 or beta_param <= 0:
        raise ValueError("alpha and beta_param must be positive.")

    t = n - h

    posterior_alpha = alpha + h
    posterior_beta = beta_param + t

    theta = np.linspace(0, 1, num_points)

    prior_density = beta.pdf(theta, alpha, beta_param)
    posterior_density = beta.pdf(theta, posterior_alpha, posterior_beta)

    prior_mean = alpha / (alpha + beta_param)
    posterior_mean = posterior_alpha / (posterior_alpha + posterior_beta)

    fig = go.Figure()

    fig.add_trace(
        go.Scatter(
            x=theta,
            y=prior_density,
            mode="lines",
            name=f"Prior: Beta({alpha}, {beta_param})"
        )
    )

    fig.add_trace(
        go.Scatter(
            x=theta,
            y=posterior_density,
            mode="lines",
            name=f"Posterior: Beta({posterior_alpha}, {posterior_beta})"
        )
    )

    fig.add_vline(
        x=prior_mean,
        line_dash="dash",
        annotation_text=f"Prior mean = {prior_mean:.3f}",
        annotation_position="top left"
    )

    fig.add_vline(
        x=posterior_mean,
        line_dash="dash",
        annotation_text=f"Posterior mean = {posterior_mean:.3f}",
        annotation_position="top right"
    )

    fig.update_layout(
        title=f"Bayesian Update for Coin Bias: n={n}, heads={h}, tails={t}",
        xaxis_title="Coin bias θ = P(Heads)",
        yaxis_title="Density",
        template="plotly_white",
        hovermode="x unified"
    )

    fig.show()

    print(f"Prior:      Beta({alpha}, {beta_param})")
    print(f"Posterior: Beta({posterior_alpha}, {posterior_beta})")
    print(f"Prior mean:      {prior_mean:.4f}")
    print(f"Posterior mean: {posterior_mean:.4f}")
```

```python
plot_beta_prior_posterior(n=10, h=8)
```

## Example: Estimating Loan Default Probability


Suppose a bank issues many similar personal loans. Each borrower either defaults or does not default within one year.

Let $\Theta\in[0,1]$
denote the unknown probability that a randomly selected borrower defaults within one year.
The bank does not know the true value of $\Theta$, but it may have prior information from past years, industry knowledge, or expert judgment.

Before observing the current portfolio, suppose the bank believes that the default probability is usually around $5\%$. We can encode this using a Beta prior, for example

$$
\Theta\sim \operatorname{Beta}(5,95).
$$

This prior has mean

$$
\mathbb E[\Theta]=\frac{5}{5+95}=0.05.
$$

So the prior belief says:

$$
\text{default probability is probably around }5\%.
$$

Now suppose the bank observes $n=200$ loans from the current year, and $d=16$ of them default. Conditional on $\Theta=\theta$, the number of defaults is modeled as

$$
D\mid \Theta=\theta \sim \operatorname{Binomial}(n,\theta).
$$

The likelihood is

$$
P(D=d\mid \Theta=\theta)=\binom{n}{d}\theta^d(1-\theta)^{n-d}.
$$

The Bayesian update is

$$
\text{posterior}
\propto
\text{likelihood}\times \text{prior}.
$$

Since the prior is Beta and the likelihood is Binomial, the posterior is again Beta:

$$
\Theta\mid D=d
\sim
\operatorname{Beta}(\alpha+d,\beta+n-d).
$$

For our example,

$$
\Theta\sim \operatorname{Beta}(5,95),
$$

and we observe

$$
n=200,
\qquad
d=16.
$$

Therefore,

$$
\Theta\mid D=16
\sim
\operatorname{Beta}(5+16,;95+184).
$$

Hence,

$$
\Theta\mid D=16
\sim
\operatorname{Beta}(21,279).
$$

The posterior mean is

$$
\mathbb E[\Theta\mid D=16]=\frac{21}{21+279}=0.07.
$$

So the bank's estimate of the default probability moves from $5\%$
to $7\%$.

The interpretation is:

$$
\boxed{
\text{The observed data suggests default risk is higher than initially believed.}
}
$$



We can describe the update as

$$
\operatorname{Beta}(5,95)
\quad\longrightarrow\quad
\operatorname{Beta}(21,279).
$$
This example is useful because it mirrors a real financial risk problem:

$$
\text{How risky is this loan portfolio?}
$$

The prior comes from historical knowledge. The likelihood comes from the observed default count. The posterior gives the updated belief about the true default probability.

In words:

$$
\boxed{
\text{Bayesian inference updates the bank's belief about default risk after observing actual loan outcomes.}
}
$$

Graphically, this is very nice. The prior density is centered near $5\%$. After observing $16$ defaults out of $200$, the posterior density shifts to the right and becomes narrower, centered near $7\%$.

```python
import numpy as np
import plotly.graph_objects as go
from scipy.stats import beta

def plot_default_probability_update(
    n=200,
    d=16,
    alpha=5,
    beta_param=95,
    num_points=1000
):
    """
    Visualize Bayesian updating for a loan default probability.

    Model
    -----
    Theta = unknown default probability

    Prior:
        Theta ~ Beta(alpha, beta_param)

    Data:
        D | Theta = theta ~ Binomial(n, theta)

    Posterior:
        Theta | D=d ~ Beta(alpha + d, beta_param + n - d)

    Parameters
    ----------
    n : int
        Number of loans observed.
    d : int
        Number of defaults observed.
    alpha : float
        First parameter of the Beta prior.
    beta_param : float
        Second parameter of the Beta prior.
    num_points : int
        Number of grid points for plotting.
    """

    if not isinstance(n, int) or not isinstance(d, int):
        raise ValueError("n and d must be integers.")

    if n <= 0:
        raise ValueError("n must be positive.")

    if d < 0 or d > n:
        raise ValueError("d must satisfy 0 <= d <= n.")

    if alpha <= 0 or beta_param <= 0:
        raise ValueError("alpha and beta_param must be positive.")

    non_defaults = n - d

    posterior_alpha = alpha + d
    posterior_beta = beta_param + non_defaults

    theta = np.linspace(0, 0.20, num_points)

    prior_density = beta.pdf(theta, alpha, beta_param)
    posterior_density = beta.pdf(theta, posterior_alpha, posterior_beta)

    prior_mean = alpha / (alpha + beta_param)
    posterior_mean = posterior_alpha / (posterior_alpha + posterior_beta)

    observed_default_rate = d / n

    fig = go.Figure()

    fig.add_trace(
        go.Scatter(
            x=theta,
            y=prior_density,
            mode="lines",
            name=f"Prior: Beta({alpha}, {beta_param})"
        )
    )

    fig.add_trace(
        go.Scatter(
            x=theta,
            y=posterior_density,
            mode="lines",
            name=f"Posterior: Beta({posterior_alpha}, {posterior_beta})"
        )
    )

    fig.add_vline(
        x=prior_mean,
        line_dash="dash",
        annotation_text=f"Prior mean = {prior_mean:.2%}",
        annotation_position="top left"
    )

    fig.add_vline(
        x=posterior_mean,
        line_dash="dash",
        annotation_text=f"Posterior mean = {posterior_mean:.2%}",
        annotation_position="top right"
    )

    fig.add_vline(
        x=observed_default_rate,
        line_dash="dot",
        annotation_text=f"Observed rate = {observed_default_rate:.2%}",
        annotation_position="bottom right"
    )

    fig.update_layout(
        title=f"Bayesian Update of Loan Default Probability: {d} Defaults out of {n} Loans",
        xaxis_title="Default probability θ",
        yaxis_title="Density",
        template="plotly_white",
        hovermode="x unified",
        legend=dict(
            x=0.70,
            y=0.95,
            bgcolor="rgba(255,255,255,0.7)"
        )
    )

    fig.update_xaxes(tickformat=".0%")

    fig.show()

    print("Bayesian default probability update")
    print("-----------------------------------")
    print(f"Observed loans:       {n}")
    print(f"Defaults observed:    {d}")
    print(f"Non-defaults:         {non_defaults}")
    print()
    print(f"Prior:                Beta({alpha}, {beta_param})")
    print(f"Posterior:            Beta({posterior_alpha}, {posterior_beta})")
    print()
    print(f"Prior mean:           {prior_mean:.4f}  = {prior_mean:.2%}")
    print(f"Observed default rate:{observed_default_rate:.4f}  = {observed_default_rate:.2%}")
    print(f"Posterior mean:       {posterior_mean:.4f}  = {posterior_mean:.2%}")
```

```python
plot_default_probability_update(
    n=200,
    d=16,
    alpha=5,
    beta_param=95
)
```
