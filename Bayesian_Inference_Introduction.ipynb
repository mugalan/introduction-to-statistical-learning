# Overview

Bayesian inference can be thought of as a theory of **learning under uncertainty**.

In many situations, we do not know the true state of the world. We may not know the value of a parameter, the correct model, the hidden cause of an observation, or the future outcome of a process. However, we are rarely in a state of complete ignorance. We usually begin with some prior information: previous experience, physical constraints, expert knowledge, symmetry assumptions, or even a deliberately vague state of belief.

Bayesian inference formalizes this situation by saying:

> Knowledge is not all-or-nothing; it comes in degrees of belief.

These degrees of belief are represented by probability measures.

The unknown quantity, say $\theta$, is treated as uncertain. This does **not necessarily mean** that $\theta$ is physically random. It may be fixed but unknown. The probability distribution over $\theta$ represents our state of knowledge about it.

Before observing data, we have a **prior belief** represented by a probability measure
$\Pi$
on a measurable parameter space $(\Theta,\mathfrak T)$.


After an observation, the data does not usually reveal $\theta$ exactly. Rather, it changes which values of $\theta$ are more or less plausible. Bayesian inference updates the prior belief $\Pi$ into a **posterior belief.**


So the philosophical idea of  Bayesian inference is:

> Bayesian inference is rational belief revision in the presence of evidence.



This is one of the key differences between Bayesian and purely frequentist thinking. In a frequentist view, the parameter $\theta$ is fixed and probability describes the long-run behavior of repeated data. In a Bayesian view, probability may also describe uncertainty about $\theta$ itself, conditional on the information currently available.

Thus, Bayesian inference is not merely a computational rule. It is a coherent framework for answering the question:

> Given what I believed before, and given what I have now observed, what should I believe next?

In measure-theoretic language, this becomes simply:

> Bayesian inference is conditioning a joint probability measure on the observed data.

In the following we rigorously define this notion.

# Rigorous Formulation of Bayesian Inference via Conditional Expectation

[cite_start]This section provides a mathematically rigorous overview of Bayesian inference using measure theory and regular conditional distributions, based on the foundational probability concepts outlined in our statistical learning toolkit[cite: 32, 53, 154].

---

## 1. The Measure-Theoretic Setup

[cite_start]Let $(\Omega, \mathfrak{A}, P)$ be our underlying probability space[cite: 33, 90, 156]. [cite_start]In Bayesian inference, the latent parameters and observable data are modeled as random variables defined on this space[cite: 53, 90]:

* [cite_start]**The Parameter / Latent State ($X$):** A measurable map $X: \Omega \rightarrow \mathbb{X}$ representing the hidden quantity of interest[cite: 55, 56, 90]. [cite_start]Its global distribution before seeing data is governed by the marginal law $P_X$[cite: 63, 67, 248].
* [cite_start]**The Observed Data ($Y$):** A measurable map $Y: \Omega \rightarrow \mathbb{Y}$ representing the collected evidence[cite: 55, 56, 90].

[cite_start]Together, they form a single joint random ordered pair $Z = (X, Y)$ mapping into the product space $(\mathbb{X} \times \mathbb{Y}, \mathfrak{U}_X \otimes \mathfrak{U}_Y)$ governed by the joint law $P_{X,Y}$[cite: 91, 92, 93].

---

## 2. Prior, Likelihood, and Posterior via Regular Conditional Laws

When conditioning on continuous data, classical calculus definitions break down if an event has a probability of zero ($P(Y=y) = 0$). [cite_start]To remain rigorous, we define these relationships via **Radon-Nikodym derivatives** and **Regular Conditional Laws**[cite: 43, 86, 251].

### The Prior Density
[cite_start]If the marginal law $P_X$ is absolutely continuous with respect to the Lebesgue measure $\lambda$, the **Prior PDF** is defined by[cite: 85, 86, 88]:
$$f_X(x) \triangleq \frac{dP_X}{d\lambda}(x)$$

### The Likelihood
[cite_start]For a fixed localized slice of reality where $X = x$, the data behaves according to the regular conditional law $P_{Y|X=x}$[cite: 250, 251, 252]. [cite_start]This gives us our **Likelihood Density**[cite: 201, 202, 257]:
$$f_{Y|X}(y|x) \triangleq \frac{dP_{Y|X=x}}{dy}(y)$$

### The Posterior Density
[cite_start]By utilizing the **Product Rule of Conditional Density**, the joint density factorizes as $f_{X,Y}(x,y) = f_{X|Y}(x|y)f_Y(y)$[cite: 207, 212]. [cite_start]Applying Bayes' theorem, the updated **Posterior PDF** for the conditional random variable $(X|Y=y)$ is formulated as[cite: 251, 257, 259]:

\begin{align}
f_{X|Y}(x|y) &= \frac{f_{X,Y}(x,y)}{f_Y(y)} \\
&= \frac{f_{Y|X}(y|x)f_X(x)}{\int_{\mathbb{X}} f_{Y|X}(y|x')f_X(x')dx'}
\end{align}

[cite_start]Where the denominator represents the marginal density of $Y$ derived by integrating out the parameter (The Law of Total Probability)[cite: 115, 262, 265]:
$$f_Y(y) = \int_{\mathbb{X}} f_{Y|X}(y|x)f_X(x)dx$$

---

## 3. The Conditional Expectation Formulation (Optimal Estimation)

[cite_start]If our goal is to produce an optimal point estimate of the parameter $X$ given our data $Y$, we construct the conditional expectation operator $\mathbb{E}[X|Y]$[cite: 154, 155, 317].

### Definition and Partial Averaging
[cite_start]The Bayesian point estimate is a $\sigma(Y)$-measurable random variable $Z = g(Y) = \mathbb{E}[X|Y]$ uniquely defined via the partial averaging property[cite: 161, 163, 165, 167]:
$$\int_{G} \mathbb{E}[X|Y] dP = \int_{G} X dP \quad \forall G \in \sigma(Y)$$

[cite_start]When evaluated at a specific observation slice $Y=y$, the **Posterior Mean** computes the functional average over the entire domain with respect to the localized posterior measure[cite: 234, 240, 241, 280]:
$$g(y) = \mathbb{E}[X|Y=y] = \int_{\mathbb{X}} x f_{X|Y}(x|y) dx$$

### The Optimality of the Posterior Mean
[cite_start]Let $Z$ be *any* other arbitrary estimator that is a function of our data (meaning $Z$ is $\sigma(Y)$-measurable)[cite: 304]. [cite_start]The conditional expectation formulation is the **unique minimizer of the Mean Square Error (MSE)**[cite: 306, 307]:
$$\mathbb{E}[(X - \mathbb{E}[X|Y])^2] \le \mathbb{E}[(X - Z)^2]$$

[cite_start]The residual, unexplained uncertainty that remains after calculating this optimal estimate is trapped in the **Conditional Variance**[cite: 291, 302]:
$$Var(X|Y) \triangleq \mathbb{E}[X^2|Y] - (\mathbb{E}[X|Y])^2$$

---

## 4. Canonical Application: The Multivariate Gaussian Case

[cite_start]When the parameters and data are jointly Gaussian, this measure-theoretic machinery collapses into a beautiful, closed-form analytical solution[cite: 314, 315]. [cite_start]Given[cite: 316]:

$$Z = \begin{bmatrix} X \\ Y \end{bmatrix} \sim \mathcal{N}\left(\begin{bmatrix} \mu_X \\ \mu_Y \end{bmatrix}, \begin{bmatrix} \Sigma_{XX} & \Sigma_{XY} \\ \Sigma_{YX} & \Sigma_{YY} \end{bmatrix}\right)$$

[cite_start]The optimal Bayesian update map is exactly[cite: 317, 318]:
$$\mathbb{E}[X|Y] = \mu_X + \Sigma_{XY}\Sigma_{YY}^{-1}(Y - \mu_Y)$$

### Component Decomposition
* [cite_start]**$\mu_X$ (The Prior Mean):** Our baseline baseline "best guess" for $X$ assuming zero data information[cite: 320, 321, 322].
* **$(Y - \mu_Y)$ (The Innovation / Residual):** The error vector tracking the difference between what we observed and what we expected to observe[cite: 319].
* **$\Sigma_{XY}\Sigma_{YY}^{-1}$ (The Gain Matrix / Information Bridge):** Scales the innovation based on the cross-covariance $\Sigma_{XY}$ normalized by total data uncertainty $\Sigma_{YY}$[cite: 326, 327].

By the **Law of Total Variance**, the posterior uncertainty strips away the explained variance from our total prior variance $\Sigma_{XX}$[cite: 295, 297, 301, 325]:
$$Var(X|Y) = \Sigma_{XX} - \Sigma_{XY}\Sigma_{YY}^{-1}\Sigma_{YX}$$

# Mathematical Formulation

Let $(\Omega,\mathfrak A,P)$ be an underlying probability space.

Let

$$
X:\Omega\to \mathscr X
$$

be a random variable representing the unknown parameter, latent state, or hidden quantity of interest (traditionaly it is common to use the variable $\theta$ denote paramters. However in order to stick to the general conditioning notation adopted in this course we will use $X$ instead). Here

$$
(\mathscr X,\mathfrak U_X)
$$

is a measurable space.

Let

$$
Y:\Omega\to \mathbb Y
$$

be a random variable representing the observed data. Here

$$
(\mathbb Y,\mathfrak U_Y)
$$

is another measurable space.

Thus, $X$ and $Y$ are measurable maps satisfying

$$
X^{-1}(A)\in\mathfrak A
\qquad
\forall A\in\mathfrak U_X,
$$

and

$$
Y^{-1}(B)\in\mathfrak A
\qquad
\forall B\in\mathfrak U_Y.
$$

Together, they define a joint random variable

$$
Z=(X,Y):\Omega\to \mathscr X\times\mathbb Y,
$$

where the product space is equipped with the product $\sigma$-algebra

$$
\mathfrak U_X\otimes\mathfrak U_Y.
$$

The joint law of $(X,Y)$ is the push-forward measure

$$
P_{X,Y}\triangleq (X,Y)_{\#} P.
$$

That is, for every measurable set $C\in \mathfrak U_X\otimes\mathfrak U_Y$,

$$
P_{X,Y}(C)=P\big((X,Y)^{-1}(C)\big).
$$

The marginal law of $X$ is

$$
P_X\triangleq X_{\#}P,
$$

and the marginal law of $Y$ is

$$
P_Y\triangleq Y_{\#}P.
$$

Explicitly,

$$
P_X(A)=P(X\in A),
\qquad A\in\mathfrak U_X,
$$

and

$$
P_Y(B)=P(Y\in B),
\qquad B\in\mathfrak U_Y.
$$

In the Bayesian interpretation, $P_X$ is the **prior law** of the unknown quantity $X$ before observing the data $Y$.



# Prior, Likelihood, and Posterior via Regular Conditional Laws



In Bayesian inference, the prior, likelihood, and posterior are not fundamentally densities. They are probability measures or conditional probability kernels. Densities appear only when suitable dominating measures exist.

Assume that regular conditional distributions exist. Then there is a Markov kernel

$$
P_{Y\mid X}:\mathscr X\times\mathfrak U_Y\to [0,1],
$$

such that, for each $x\in\mathscr X$,

$$
P_{Y\mid X=x}(\cdot)
$$

is a probability measure on $(\mathbb Y,\mathfrak U_Y)$, and for each $B\in\mathfrak U_Y$,

$$
x\mapsto P_{Y\mid X=x}(B)
$$

is $\mathfrak U_X$-measurable.

This kernel describes the data-generating mechanism. In Bayesian terminology, it is the sampling model. For $B\in\mathfrak U_Y$,

$$
P_{Y\mid X=x}(B)
$$

is the probability that the observed data lies in $B$, conditional on the parameter value $X=x$.

Similarly, the posterior is described by a regular conditional distribution

$$
P_{X\mid Y}:\mathbb Y\times\mathfrak U_X\to [0,1].
$$

For each observed value $y\in\mathbb Y$,

$$
P_{X\mid Y=y}(\cdot)
$$

is a probability measure on $(\mathscr X,\mathfrak U_X)$.

This is the posterior law of $X$ after observing $Y=y$.

The defining property of the posterior kernel is that, for every $A\in\mathfrak U_X$ and every $B\in\mathfrak U_Y$,

$$
P(X\in A,;Y\in B)=\int_B P_{X\mid Y=y}(A),P_Y(dy).
$$

Equivalently,

$$
P_{X,Y}(A\times B)=\int_B P_{X\mid Y=y}(A),P_Y(dy).
$$

Thus, measure-theoretically, Bayesian inference is the passage from the prior law

$$
P_X
$$

to the posterior law

$$
P_{X\mid Y=y}
$$

after observing the data value $Y=y$.



# Density Form of Bayes' Rule

Suppose now that $\mathscr X$ and $\mathbb Y$ are Euclidean spaces, and suppose that the relevant measures admit densities with respect to suitable reference measures.

Let $\lambda_X$ be a reference measure on $\mathscr X$, and let $\lambda_Y$ be a reference measure on $\mathbb Y$.

Assume that the prior law $P_X$ is absolutely continuous with respect to $\lambda_X$. Then the prior density is

$$
f_X(x)\frac{dP_X}{d\lambda_X}(x).
$$

Assume also that, for each $x$, the conditional law $P_{Y\mid X=x}$ is absolutely continuous with respect to $\lambda_Y$. Then the likelihood density is

$$
f_{Y\mid X}(y\mid x)\frac{dP_{Y\mid X=x}}{d\lambda_Y}(y).
$$

The word likelihood refers to the same function viewed as a function of $x$ after the data value $y$ has been fixed. Thus, for fixed $y$,

$$
x\mapsto f_{Y\mid X}(y\mid x)
$$

is the likelihood function.

If the joint law $P_{X,Y}$ admits a density $f_{X,Y}$ with respect to $\lambda_X\otimes\lambda_Y$, then

$$
f_{X,Y}(x,y)=f_{Y\mid X}(y\mid x)f_X(x).
$$

The marginal density of $Y$ is obtained by integrating out $x$:

$$
f_Y(y)=\int_{\mathscr X} f_{Y\mid X}(y\mid x)f_X(x),\lambda_X(dx).
$$

Whenever $f_Y(y)>0$, the posterior density is

$$
f_{X\mid Y}(x\mid y)=\frac{f_{X,Y}(x,y)}{f_Y(y)}.
$$

Therefore,

$$
f_{X\mid Y}(x\mid y)=\frac{
f_{Y\mid X}(y\mid x)f_X(x)
}{
\int_{\mathscr X} f_{Y\mid X}(y\mid x')f_X(x'),\lambda_X(dx')
}.
$$

This is the usual density form of Bayes' rule.

In symbolic Bayesian notation, this is often summarized as

$$
\text{posterior}
\propto
\text{likelihood}\times \text{prior}.
$$

More explicitly,

$$
f_{X\mid Y}(x\mid y)
\propto
f_{Y\mid X}(y\mid x)f_X(x),
$$

where the constant of proportionality is determined by the condition

$$
\int_{\mathscr X} f_{X\mid Y}(x\mid y),\lambda_X(dx)=1.
$$



# Conditional Expectation as Bayesian Point Estimation

The posterior distribution gives the full updated uncertainty about $X$ after observing $Y=y$. If we want a single point estimate of $X$, a natural Bayesian choice is the posterior mean.

Measure-theoretically, this is described by the conditional expectation

$$
\mathbb E[X\mid Y].
$$

Assume that $X$ is integrable, i.e.

$$
\mathbb E[|X|]<\infty.
$$

Then $\mathbb E[X\mid Y]$ is the unique $\sigma(Y)$-measurable random variable satisfying

$$
\int_G \mathbb E[X\mid Y]\,dP=\int_G X,dP
\qquad
\forall G\in\sigma(Y).
$$

Since $\mathbb E[X\mid Y]$ is $\sigma(Y)$-measurable, there exists a measurable function $g:\mathbb Y\to\mathscr X$ such that

$$
\mathbb E[X\mid Y]=g(Y).
$$

The function $g$ is the Bayesian estimator associated with squared-error loss.

For a particular observation $Y=y$, the posterior mean is

$$
g(y)\mathbb E[X\mid Y=y].
$$

In terms of the posterior measure, this is

$$
g(y)\int_{\mathscr X} x,P_{X\mid Y=y}(dx).
$$

If the posterior admits a density, then

$$
g(y)=\int_{\mathscr X} x,f_{X\mid Y}(x\mid y),\lambda_X(dx).
$$

Thus, the Bayesian point estimate is obtained by averaging over the posterior distribution.



# Optimality of the Posterior Mean

The posterior mean is not merely a convenient estimate. It is optimal under mean-square error.

Let $Z$ be any other estimator of $X$ based on the observed data $Y$. This means that $Z$ is $\sigma(Y)$-measurable.

Then

$$
\mathbb E\left[|X-\mathbb E[X\mid Y]|^2\right]
\le
\mathbb E\left[|X-Z|^2\right].
$$

Therefore, $\mathbb E[X\mid Y]$ is the optimal estimator of $X$ among all $\sigma(Y)$-measurable estimators under squared-error loss.

Equivalently,

$$
\mathbb E[X\mid Y]=\operatorname*{arg,min}_{Z\in L^2(\sigma(Y))}
\mathbb E\left[|X-Z|^2\right].
$$

The remaining uncertainty after observing $Y$ is described by the conditional variance

$$
\operatorname{Var}(X\mid Y)=\mathbb E[X^2\mid Y]
\left(\mathbb E[X\mid Y]\right)^2
$$

in the scalar case.

For vector-valued $X$, the conditional covariance is

$$
\operatorname{Cov}(X\mid Y)=\mathbb E\left[
\left(X-\mathbb E[X\mid Y]\right)
\left(X-\mathbb E[X\mid Y]\right)^T
\mid Y
\right].
$$



# The Multivariate Gaussian Case

A central example occurs when $X$ and $Y$ are jointly Gaussian.

Suppose

$$
\begin{bmatrix}
X\
Y
\end{bmatrix}
\sim
\mathcal N
\left(
\begin{bmatrix}
\mu_X\
\mu_Y
\end{bmatrix},
\begin{bmatrix}
\Sigma_{XX} & \Sigma_{XY}\
\Sigma_{YX} & \Sigma_{YY}
\end{bmatrix}
\right).
$$

Assume that $\Sigma_{YY}$ is invertible.

Then the conditional distribution of $X$ given $Y=y$ is Gaussian:

$$
X\mid Y=y
\sim
\mathcal N
\left(
\mu_{X\mid y},
\Sigma_{X\mid y}
\right),
$$

where

$$
\mu_{X\mid y}=\mu_X
+
\Sigma_{XY}\Sigma_{YY}^{-1}(y-\mu_Y),
$$

and

$$
\Sigma_{X\mid y}=\Sigma_{XX}
\Sigma_{XY}\Sigma_{YY}^{-1}\Sigma_{YX}.
$$

Therefore, as a random variable,

$$
\mathbb E[X\mid Y]=\mu_X
+
\Sigma_{XY}\Sigma_{YY}^{-1}(Y-\mu_Y).
$$

The matrix

$$
\Sigma_{XY}\Sigma_{YY}^{-1}
$$

acts as a gain matrix. It determines how strongly the innovation

$$
Y-\mu_Y
$$

should change the prior mean $\mu_X$.

Thus, in the Gaussian case, Bayesian inference has a closed-form update:

$$
\text{posterior mean}=\text{prior mean}
+
\text{gain}\times\text{innovation}.
$$

This is the same mathematical structure that appears in linear Gaussian estimation and the Kalman filter.

## 7. Summary

The measure-theoretic formulation of Bayesian inference can be summarized as follows.

We begin with an unknown quantity $X$ and observed data $Y$, both defined on a common probability space

$$
(\Omega,\mathfrak A,P).
$$

The prior belief about $X$ is its marginal law

$$
P_X.
$$

The data-generating model is described by the conditional law

$$
P_{Y\mid X=x}.
$$

After observing $Y=y$, Bayesian inference updates the prior law $P_X$ to the posterior law

$$
P_{X\mid Y=y}.
$$

The posterior law is a regular conditional probability measure satisfying

$$
P(X\in A,;Y\in B)=\int_B P_{X\mid Y=y}(A),P_Y(dy).
$$

When densities exist, this becomes Bayes' rule:

$$
f_{X\mid Y}(x\mid y)=\frac{
f_{Y\mid X}(y\mid x)f_X(x)
}{
\int_{\mathscr X} f_{Y\mid X}(y\mid x')f_X(x'),\lambda_X(dx')
}.
$$

Finally, the posterior mean

$$
\mathbb E[X\mid Y=y]
$$

is the optimal Bayesian point estimate under squared-error loss.

Thus, measure-theoretically,

$$
\boxed{
\text{Bayesian inference is conditioning the joint law of }(X,Y)\text{ on the observed value }Y=y.
}
$$
