# Hypothesis Testing


Hypothesis testing operates on a philosophy similar to "innocent until proven guilty." We start by assuming a baseline state of no effect and see if our data provides enough evidence to reject that assumption.

* **The Null Hypothesis ($H_0$):** The status quo. It states that there is no effect, no difference, or no relationship. For example, $\mu_1 = \mu_2$ (the means of two groups are equal).
* **The Alternative Hypothesis ($H_a$ or $H_1$):** The claim you are actually trying to prove. It states that there *is* an effect or difference (e.g., $\mu_1 \neq \mu_2$).



## Mathematical Machinary

To evaluate the hypotheses, we compress our sample data into a single metric and compare it to what we would theoretically expect under the null hypothesis.

* **Test Statistic:** A standardized value calculated from your sample data (like a $z$-score, $t$-statistic, or $F$-statistic). The general formula often looks like:

$$\text{Test Statistic} = \frac{\text{Observed Value} - \text{Hypothesized Value}}{\text{Standard Error}}$$


* **The $p$-value:** The probability of obtaining a test statistic at least as extreme as the one observed, *assuming the null hypothesis is completely true*.
$$p\text{-value} = P(\text{Test Statistic} \geq \text{Observed} \mid H_0)$$
Here $P$ is the probability model (or probability distribution) of the specific test statistic, conditioned on the assumption that the null hypothesis ($H_0$) is true




## The Decision Rule & Errors

Before looking at the data, we set a threshold for risk called the **significance level ($\alpha$)**, typically set at **5%** (0.05).

* If $p\text{-value} \leq \alpha$: We **reject $H_0$**. The data is too unusual to have happened by chance alone. We say the result is "statistically significant."
* If $p\text{-value} > \alpha$: We **fail to reject $H_0$**. We do *not* "accept" the null; we simply state that we lack sufficient evidence to disprove it.

## The Standard Roadmap

Every hypothesis test follows these exact five steps:

1. **State the hypotheses** ($H_0$ and $H_a$).
2. **Formulate an analysis plan** (choose the right test—like a t-test, Chi-Square, or ANOVA—based on your data type and assumptions).
3. **Analyze sample data** (calculate the test statistic and $p$-value).
4. **Apply the decision rule** (compare the $p$-value to $\alpha$).
5. **Interpret the results** in the context of your original real-world problem.

# Initial Steps of Numerical Data Analyis

## The Multivariate i.i.d. Assumption

Let each row of an industrial dataset be modeled as an $m$-dimensional **random matrix** $\mathbf{X}_i = [X_{i,1}, X_{i,2}, \dots, X_{i,m}]^T$ governed by an underlying, unknown multivariate probability distribution. Each individual random variable is characterized by its first two theoretical moments:

* **The Population Mean:** $\boldsymbol{\mu}_i = \mathbb{E}[\mathbf{X}_i] \in \mathbb{R}^{m}$
* **The Population Covariance Matrix:** $\boldsymbol{\Sigma}_i = \mathbb{E}[(\mathbf{X}_i - \boldsymbol{\mu}_i)(\mathbf{X}_i - \boldsymbol{\mu}_i)^T] \in \mathbb{R}^{m \times m}$

When collecting a dataset of size $n$, we observe a finite sequence of these random matrices, denoted as $\{\mathbf{X}_i\}_{i=1}^n$.
The actual dataset processed by a computer is the matrix of empirical **realizations**—the concrete numerical values extracted from the system.

Formally, we define the column matrix $\mathbf{x}_i = [x_{i,1}, x_{i,2}, \dots, x_{i,m}]^T \in \mathbb{R}^{m}$ as the observed realization of the random variable $\mathbf{X}_i$. Mapping the $j$-th feature of the $i$-th sample as the scalar $x_{i,j}$ yields the structured $n \times m$ observed data matrix $\mathbf{X}_{\text{obs}}$:

$$\mathbf{X}_{\text{obs}} = \begin{bmatrix}
x_{1,1} & x_{1,2} & \dots & x_{1,m} \\
x_{2,1} & x_{2,2} & \dots & x_{2,m} \\
\vdots & \vdots & \ddots & \vdots \\
x_{n,1} & x_{n,2} & \dots & x_{n,m}
\end{bmatrix} = \begin{bmatrix} \mathbf{x}_1^T \\ \mathbf{x}_2^T \\ \vdots \\ \mathbf{x}_n^T \end{bmatrix}.$$



To justify the application of classical multivariate statistical inference, we impose the foundational assumption that the sequence of random varaibles $\{X_i\}_{i=1}^n$ is **independent and identically distributed (i.i.d.)**.

1. **Identical Distribution:** Every random variable $\mathbf{X}_i$ possesses the exact same probability law. Consequently, the distinct population moments collapse into a single, universal system profile:

$$\boldsymbol{\mu}_i = \boldsymbol{\mu} \quad \text{and} \quad \boldsymbol{\Sigma}_i = \boldsymbol{\Sigma} \quad \forall \ i \in \{1, 2, \dots, n\}$$

2. **Mutual Independence:** The joint probability distribution of the sequence collapses into the product of their individual distributions, written as $\mathbf{X}_1 \perp \mathbf{X}_2 \perp \dots \perp \mathbf{X}_n$. This guarantees that no combination of rows carries hidden, collective information about any other row, meaning the physical state captured by any subset of observations exerts zero statistical influence over the rest.


Mutual Independence implies that the joint Cumulative Distribution Function (CDF) of the entire sample factors perfectly into the product of identical marginal multivariate CDFs, expressed as:

$$F_{\mathbf{X}_1, \mathbf{X}_2, \dots, \mathbf{X}_n}(\mathbf{x}_1, \mathbf{x}_2, \dots, \mathbf{x}_n) = \prod_{i=1}^{n} F_{\mathbf{X}}(\mathbf{x}_i) \quad \forall \mathbf{x}_i \in \mathbb{R}^m$$

Furthermore, assuming the multivariate random variables are continuous and the joint distribution is absolutely continuous, this system possesses a Radon-Nikodym derivative with respect to the Lebesgue measure. This allows us to equivalently express the i.i.d. condition via the joint Probability Density Function (PDF), denoted by $f_{\mathbf{X}_1, \mathbf{X}_2, \dots, \mathbf{X}_n}$:

$$f_{\mathbf{X}_1, \mathbf{X}_2, \dots, \mathbf{X}_n}(\mathbf{x}_1, \mathbf{x}_2, \dots, \mathbf{x}_n) = \prod_{i=1}^{n} f_{\mathbf{X}}(\mathbf{x}_i) \quad \forall \mathbf{x}_i \in \mathbb{R}^m$$

where $f_{\mathbf{X}}(\mathbf{x}_i)$ represents the identical marginal PDF evaluated at the $i$-th realization.

Under this joint-density framework, the **independence** condition dictates that the conditional density of any realization $\mathbf{X}_i$ given $\mathbf{X}_j$ ($i \neq j$) collapses to its marginal density ($f_{\mathbf{X}_i \mid \{\mathbf{X}_j\}_{j\neq}} = f_{\mathbf{X}_i}$), ensuring that the row sequence possesses no serial or spatial memory. Concurrently, the **identically distributed** condition asserts that every realization originates from the exact same data-generating process governed by this time-invariant probability law, thereby sharing an identical, constant population mean $\boldsymbol{\mu} \in \mathbb{R}^m$ and a static variance-covariance matrix $\boldsymbol{\Sigma} \in \mathbb{R}^{m \times m}$. When the sample size is massive relative to the feature space ($n \gg m$), these theoretical properties ensure that our sample estimators asymptotically converge to their true population parameters via the Law of Large Numbers, establishing the baseline validity required for subsequent statistical modeling.

## Empirical Mean, Covariance and the Pearson Correlation

Below we evaluate the first and second empirical moments calculated directly from the observed realizations $\mathbf{x}_i \in \mathbb{R}^{m}$:

* **The Empirical Mean ($\widehat{\boldsymbol{\mu}}_n \in \mathbb{R}^{m \times 1}$):** The sample average mapping the spatial center of gravity of the observed feature space.

$$\widehat{\boldsymbol{\mu}}_n = \frac{1}{n} \sum_{i=1}^n \mathbf{x}_i = \begin{bmatrix} \frac{1}{n}\sum_{i=1}^n x_{i,1} \\ \frac{1}{n}\sum_{i=1}^n x_{i,2} \\ \vdots \\ \frac{1}{n}\sum_{i=1}^n x_{i,m} \end{bmatrix}=
\begin{bmatrix}\widehat{\mu}_1\\\widehat{\mu}_2\\\vdots\\\widehat{\mu}_m\\\end{bmatrix}$$


* **The Empirical Covariance Matrix ($\mathbf{S} \in \mathbb{R}^{m \times m}$):** The sample variance-covariance matrix mapping the structural variations and linear couplings between features, utilizing Bessel's correction ($n-1$) to ensure an unbiased estimator ($\mathbb{E}[\mathbf{S}] = \boldsymbol{\Sigma}$).
$$\mathbf{S} = \frac{1}{n-1} \sum_{i=1}^n (\mathbf{x}_i - \widehat{\boldsymbol{\mu}}_n)(\mathbf{x}_i - \widehat{\boldsymbol{\mu}}_n)^T = \begin{bmatrix} s_1^2 & s_{12} & \dots & s_{1m} \\ s_{21} & s_2^2 & \dots & s_{2m} \\ \vdots & \vdots & \ddots & \vdots \\ s_{m1} & s_{m2} & \dots & s_m^2 \end{bmatrix}$$
where any specific element tracking the cross-covariance between feature $j$ and feature $k$ is explicitly defined as:
$$s_{jk} = \frac{1}{n-1} \sum_{i=1}^n (x_{i,j} - \widehat{\mu}_{j})(x_{i,k} - \widehat{\mu}_{k}).$$

* **Pearson Correlation Matrix ($\mathbf{R}$):** Standardizes the sample covariance matrix, $\mathbf{S}$, between $[-1, 1]$. Because of the iid assumption, these correlations represent pure, unpolluted linear structural relationships between your industrial features (e.g., how Motor Speed and Temperature naturally scale together).
$$\mathbf{R} = \mathbf{D}^{-1/2} \mathbf{S} \mathbf{D}^{-1/2}$$
where
$$\mathbf{D} = \text{diag}(\mathbf{S}) = \begin{bmatrix} s_1^2 & 0 & \dots & 0 \\ 0 & s_2^2 & \dots & 0 \\ \vdots & \vdots & \ddots & \vdots \\ 0 & 0 & \dots & s_m^2 \end{bmatrix}.$$
Explicitly writtent down the $jk$ matrix element of $\mathbf{R}$ is
$$r_{jk} = \frac{s_{jk}}{s_j \cdot s_k} = \frac{\sum_{i=1}^n (x_{i,j} - \widehat{\mu}_j)(x_{i,k} - \widehat{\mu}_k)}{\sqrt{\sum_{i=1}^n (x_{i,j} - \widehat{\mu}_j)^2} \sqrt{\sum_{i=1}^n (x_{i,k} - \widehat{\mu}_k)^2}}.$$



## The Asymptotic Consequence: The Multivariate Central Limit Theorem

Establishing that the sequence $\{\mathbf{X}_i\}_{i=1}^n$ is i.i.d. with a well-defined population mean vector $\boldsymbol{\mu}$ and a finite population covariance matrix $\boldsymbol{\Sigma}$ satisfies the exact prerequisites for the **Lindeberg-Lévy Multivariate Central Limit Theorem (CLT)**.

The Multivariate CLT dictates that as the sample size approaches infinity ($n \to \infty$), the standardized difference between the empirical sample mean and the true population mean converges in distribution ($\xrightarrow{d}$) to a standard Multivariate Gaussian distribution:

$$\sqrt{n}\left(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu}\right) \xrightarrow{d} \mathscr{N}\left(\mathbf{0}, \boldsymbol{\Sigma}\right)$$

By tracking this distributional convergence back to the scale of the unstandardized empirical mean, the theorem establishes that for a fixed, sufficiently large sample size $n$, the estimator $\widehat{\boldsymbol{\mu}}_n$ behaves as a multivariate normal random variable:

$$\widehat{\boldsymbol{\mu}}_n \sim \mathscr{N}\left(\boldsymbol{\mu}, \frac{1}{n}\boldsymbol{\Sigma}\right)$$

This result yields a profound structural advantage: while the underlying raw industrial data points $\mathbf{x}_i$ may originate from a highly non-Gaussian, skewed, or multimodal joint density $f_{\mathbf{X}}$, the *sample mean* $\widehat{\boldsymbol{\mu}}_n$ is guaranteed to be asymptotically Multivariate Normal.

---

In practical applications, the true population parameters $\boldsymbol{\mu}$ and $\boldsymbol{\Sigma}$ remain latent and unobservable. To operationalize the CLT result, we invoke **Slutsky's Theorem**.

Because the empirical sample covariance matrix $\mathbf{S}$ is a consistent estimator of the population covariance matrix ($\mathbf{S} \xrightarrow{p} \boldsymbol{\Sigma}$ via the Weak Law of Large Numbers), we can substitute our sample statistics into the CLT formulation without altering its asymptotic convergence. This yields the operational parametric model evaluating the stability of the baseline system center by:

$${\widehat{\boldsymbol{\mu}}_n} \sim \mathscr{N}\left(\widehat{\boldsymbol{\mu}}_n, \frac{1}{n}\mathbf{S}\right)$$

This distribution mathematically proves that the uncertainty envelope surrounding the estimated baseline system center shrinks at a deterministic rate of $\frac{1}{n}$ in variance-covariance space and a rate of $\frac{1}{\sqrt{n}}$ along any single feature's standard error axis, yielding the classic Standard Error of the Mean:

$$\text{SEM}_j = \frac{s_j}{\sqrt{n}}$$

This establishes the definitive geometric ruler required to draw highly precise, statistically sound confidence ellipsoids around healthy industrial operations.

---

The integration of these layers reveals a cohesive, dual-scale parametric framework for industrial diagnostics. At the micro-scale, the Principle of Maximum Entropy dictates that a Multivariate Gaussian distribution $\mathscr{N}(\boldsymbol{\mu}, \boldsymbol{\Sigma})$ is the most mathematically unbiased and conservative model choice given only first- and second-moment constraints. We operationalize this space by using our empirical sample estimators $(\widehat{\boldsymbol{\mu}}_n, \mathbf{S})$ to approximate this distribution by:

$$\mathbf{X}_i\sim \mathscr{N}(\widehat{\boldsymbol{\mu}}_n, \mathbf{S}),$$

establishing a localized continuous boundary for real-time anomaly detection.

At the macro-scale, we leverage the Multivariate Central Limit Theorem to model the sample mean vector:

$$\widehat{\boldsymbol{\mu}}_n \sim \mathscr{N}\left(\widehat{\boldsymbol{\mu}}_n, \frac{1}{n}\mathbf{S}\right),$$

providing an unshakeable statistical foundation for tracking structural system drift and conducting baseline validation tests.

----

We can also approximate the latent population distribution $\mathscr{N}(\boldsymbol{\mu}, \boldsymbol{\Sigma})$ by finding the parameters that maximize the joint probability of observing our collected dataset. Assuming the realizations $\mathbf{x}_i$ are mutually independent, the joint log-likelihood function is defined as:

$$\ell(\boldsymbol{\mu}, \boldsymbol{\Sigma} \;|\; \mathbf{X}_{\text{obs}}) = -\frac{nm}{2}\ln(2\pi) - \frac{n}{2}\ln|\boldsymbol{\Sigma}| - \frac{1}{2}\sum_{i=1}^n (\mathbf{x}_i - \boldsymbol{\mu})^T \boldsymbol{\Sigma}^{-1} (\mathbf{x}_i - \boldsymbol{\mu})$$

Maximizing this function by taking its vector and matrix derivatives with respect to $\boldsymbol{\mu}$ and $\boldsymbol{\Sigma}^{-1}$ yields the unique, mathematically optimal Maximum Likelihood Estimators (MLE):

$$\widehat{\boldsymbol{\mu}}_{\text{MLE}} = \frac{1}{n} \sum_{i=1}^n \mathbf{x}_i = \widehat{\boldsymbol{\mu}}_n$$

$$\widehat{\boldsymbol{\Sigma}}_{\text{MLE}} = \frac{1}{n} \sum_{i=1}^n (\mathbf{x}_i - \widehat{\boldsymbol{\mu}}_n)(\mathbf{x}_i - \widehat{\boldsymbol{\mu}}_n)^T$$

While $\widehat{\boldsymbol{\mu}}_{\text{MLE}}$ is perfectly identical to our empirical mean vector $\widehat{\boldsymbol{\mu}}_n$, the pure maximum likelihood matrix $\widehat{\boldsymbol{\Sigma}}_{\text{MLE}}$ is inherently biased in finite samples ($\mathbb{E}[\widehat{\boldsymbol{\Sigma}}_{\text{MLE}}] = \frac{n-1}{n}\boldsymbol{\Sigma}$) because it measures variations against the calculated sample mean rather than the true population mean.

To resolve this systematic underestimation, we apply Bessel's correction by scaling the MLE result by $\frac{n}{n-1}$. This substitutes the biased MLE matrix with the unbiased empirical sample covariance matrix $\mathbf{S}$:

$$\mathbf{S} = \frac{n}{n-1}\widehat{\boldsymbol{\Sigma}}_{\text{MLE}} = \frac{1}{n-1} \sum_{i=1}^n (\mathbf{x}_i - \widehat{\boldsymbol{\mu}}_n)(\mathbf{x}_i - \widehat{\boldsymbol{\mu}}_n)^T$$

Consequently, the operational micro-scale model $\mathbf{X}_i \sim \mathscr{N}(\widehat{\boldsymbol{\mu}}_n, \mathbf{S})$ is fundamentally an **unbiased Maximum Likelihood framework**. This connection guarantees that the geometric boundary established for real-time anomaly detection layer is both asymptotically optimal under likelihood theory and mathematically guarded against small-sample estimation bias.

## Empirical i.i.d. Weak Testing

While strict independence and identical distribution require all higher-order probability moments to be invariant and uncoupled across realizations, testing these infinite-dimensional criteria empirically is practically impossible. Therefore, in classical data analysis, we relax our requirements to **second-order (weak) conditions**.

To validate that our data satisfies the weak i.i.d. assumption, we partition our $n$ rows into $g$ distinct, consecutive, non-overlapping blocks (or chunks) of size $n_j$ (where $\sum_{j=1}^g n_j = n$). Because $n \gg m$, each individual chunk remains statistically large enough to yield highly precise sample estimators. We then verify three necessary conditions concerning the first two joint moments:



### Condition 1: Constant Joint Mean (First Moment Homogeneity)



The expected value of the multivariate random variable must remain invariant across the sequence of realizations. The center of the distribution cannot drift or exhibit structural shifts over the row index.

* **Statistical Formulation:** Let $\widehat{\mu}_j \in \mathbb{R}^m$ be the empirical sample mean calculated for chunk $j$. We test the null hypothesis that all chunk population means are equivalent to a single global mean $\boldsymbol{\mu}$:

$$H_0: \mathbb{E}[\mathbf{X}_{i \in \text{chunk } 1}] = \mathbb{E}[\mathbf{X}_{i \in \text{chunk } 2}] = \dots = \mathbb{E}[\mathbf{X}_{i \in \text{chunk } g}] = \boldsymbol{\mu}$$


* **Verification Methodology:** This condition is evaluated using **Multivariate Analysis of Variance (MANOVA)** via Wilks' Lambda ($\Lambda$). MANOVA determines if the variance *between* the chunk mean is significantly larger than the variance *within* the chunks.



#### MANOVA framework and Wilks' Lambda ($\Lambda$) Verification Methodology

To evaluate the homogeneity of the chunk mean, we decompose the total multivariate dispersion into two matrices: the **Within-Chunk Sum of Squares and Cross-Products matrix ($\mathbf{W}$)** and the **Between-Chunk Sum of Squares and Cross-Products matrix ($\mathbf{B}$)**.

Let $g$ be the number of chunks, $n_j$ be the number of observations in chunk $j$, $\widehat{\boldsymbol{\mu}}_j \in \mathbb{R}^m$ be the sample mean of chunk $j$, and $\widehat{\boldsymbol{\mu}}_n \in \mathbb{R}^m$ be the global sample mean of the entire dataset.

---

1. Matrix Decompositions

The **Within-Chunk (Residual) matrix $\mathbf{W}$** captures the random variation of individual realizations around their respective chunk centers:


$$\mathbf{W} = \sum_{j=1}^{g} \sum_{i \in \text{chunk } j} (x_{i,j} - \widehat{\boldsymbol{\mu}}_j)(x_{i,j} - \widehat{\boldsymbol{\mu}}_j)^T$$

The **Between-Chunk (Effect) matrix $\mathbf{B}$** captures the displacement of the chunk centers from the global center (the drift):


$$\mathbf{B} = \sum_{j=1}^{g} n_j (\widehat{\boldsymbol{\mu}}_j - \widehat{\boldsymbol{\mu}}_n)(\widehat{\boldsymbol{\mu}}_j - \widehat{\boldsymbol{\mu}}_n)^T$$

Both $\mathbf{W}$ and $\mathbf{B}$ are $m \times m$ symmetric matrices. The total variation matrix is defined as $\mathbf{T} = \mathbf{B} + \mathbf{W}$.

---

2. Wilks' Lambda Statistic ($\Lambda$)

Wilks' Lambda is calculated as the ratio of the determinant of the within-chunk variation to the determinant of the total variation:

$$\Lambda = \frac{|\mathbf{W}|}{|\mathbf{B} + \mathbf{W}|} = \frac{|\mathbf{W}|}{|\mathbf{T}|}$$

Alternatively, $\Lambda$ can be expressed via the eigenvalues ($\lambda_k$) of the matrix $\mathbf{W}^{-1}\mathbf{B}$:


$$\Lambda = \prod_{k=1}^{m} \frac{1}{1 + \lambda_k}$$

---

3. Decision Rule

* **Value Interpretation:** $\Lambda$ ranges strictly between $0$ and $1$. A value near $1$ indicates that the between-chunk variation ($\mathbf{B}$) is exceptionally small compared to the within-chunk variation ($\mathbf{W}$), meaning the centers are not drifting.
* **Asymptotic Testing:** If the sample size is very large ($n \gg m$), we transform $\Lambda$ using **Bartlett's $\chi^2$ approximation** to calculate a $p$-value:

$$\chi^2 = -\left( n - 1 - \frac{m + g}{2} \right) \ln \Lambda$$



Under the null hypothesis of constant means, this statistic asymptotically follows a Chi-square distribution with $m(g - 1)$ degrees of freedom:

$$\chi^2 \sim \chi^2_{m(g - 1)}$$


* **Conclusion:** If the resulting $p$-value is greater than your significance level (e.g., $p > 0.05$), we fail to reject $H_0$ and conclude that the first joint moment remains constant across the chunks.

### Condition 2: Constant Covariance Matrix (Second Moment Homogeneity)

The dispersion profile of the multivariate random variable and the linear relationships between its $m$ dimensions must be stable across the sample sequence. The data must not exhibit heteroscedasticity or structural breaks in its correlation fabric.

* **Statistical Formulation:** Let $\mathbf{S}_j \in \mathbb{R}^{m \times m}$ be the sample variance-covariance matrix evaluated within chunk $j$. We test the null hypothesis that the true population covariance structure is identical across all chunks:

$$H_0: \boldsymbol{\Sigma}_1 = \boldsymbol{\Sigma}_2 = \dots = \boldsymbol{\Sigma}_g = \boldsymbol{\Sigma}$$


* **Verification Methodology:** This structural equivalence is evaluated using **Box's M-test**, which relies on a likelihood-ratio test statistic comparing the determinants of the individual chunk covariance matrices ($|\mathbf{S}_j|$) against the determinant of the pooled sample covariance matrix ($|\mathbf{S}_p|$).


#### Box’s M-test Verification Methodology

To evaluate whether the covariance matrices are uniform across all chunks, Box's M-test utilizes a generalized likelihood-ratio criterion. It structurally contrasts the volume of the individual chunk covariance matrices against the volume of the pooled covariance matrix.

Let $g$ be the number of chunks, $n_j$ be the number of observations in chunk $j$, and $\mathbf{S}_j \in \mathbb{R}^{m \times m}$ be the unbiased sample covariance matrix of chunk $j$:
$$\mathbf{S}_j = \frac{1}{n_j - 1} \sum_{i \in \text{chunk } j} (x_{i,j} - \widehat{\boldsymbol{\mu}}_j)(x_{i,j} - \widehat{\boldsymbol{\mu}}_j)^T.$$
The denominator utilizes Bessel's correction ($n-1$) instead of $n$ to yield an unbiased estimator of the population covariance matrix, correcting for the mathematical bias introduced by measuring variations against the calculated sample mean $\widehat{\boldsymbol{\mu}}_j$ rather than the true, hidden population mean $\boldsymbol{\mu}$.

---

1. The Pooled Covariance Matrix ($\mathbf{S}_p$)

The pooled covariance matrix represents the single, optimal estimate of the global population covariance structure, calculated by weighting each chunk covariance by its corresponding degrees of freedom:


$$\mathbf{S}_p = \frac{\sum_{j=1}^{g} (n_j - 1)\mathbf{S}_j}{n - g}$$

Where $n = \sum_{j=1}^g n_j$ is the total sample size.

---

2. The Box's $M$ Test Statistic

The core test statistic, $M$, is a function of the log-determinants of the matrices. It measures the discrepancy between the pooled matrix and the individual chunk matrices:


$$M = (n - g) \ln |\mathbf{S}_p| - \sum_{j=1}^{g} (n_j - 1) \ln |\mathbf{S}_j|$$

* **Value Interpretation:** If all chunk covariance matrices are perfectly identical ($\mathbf{S}_j = \mathbf{S}_p$), then $\ln |\mathbf{S}_p| = \ln |\mathbf{S}_j|$, causing $M$ to collapse to exactly $0$. As the chunk covariances diverge from one another, $M$ increases positively.

---

3. Asymptotic Transformation and Decision Rule

Because your sample size is massive ($n \gg m$), we use Box's asymptotic scale correction factor ($C$) to map the $M$ statistic onto a standard Chi-square ($\chi^2$) distribution. The correction factor $C$ is defined as:


$$C = \left[ \sum_{j=1}^{g} \frac{1}{n_j - 1} - \frac{1}{n - g} \right] \cdot \left[ \frac{2m^2 + 3m - 1}{6(m + 1)(g - 1)} \right]$$

Using this factor, the test statistic is transformed into:


$$\chi^2_{\text{calc}} = M(1 - C)$$

Under the null hypothesis that all population covariance matrices are equal ($H_0: \boldsymbol{\Sigma}_1 = \dots = \boldsymbol{\Sigma}_g$), this statistic asymptotically follows a Chi-square distribution:


$$\chi^2_{\text{calc}} \sim \chi^2_{\nu}$$

Where the degrees of freedom $\nu$ represent the number of unique mathematical constraints being tested:


$$\nu = \frac{m(m + 1)(g - 1)}{2}$$

---

* **Conclusion:** If the resulting $p$-value is greater than your strict alpha threshold (e.g., $p > 0.001$), we fail to reject $H_0$, mathematically establishing that the variance-covariance structure remains homoscedastic and stable across all realizations.

> **Note:** Because $n$ is very large, Box's M-test possesses extreme statistical power. It will flag even minor, practically negligible variations in variance with a tiny $p$-value. In practice, look at the relative changes in the diagonal elements (variances) of $\mathbf{S}_j$ alongside the test result.

### Condition 3: Zero Cross-Covariance Between Realizations (Linear Independence)

Distinct rows must not carry memory or linearly predict subsequent rows. Any pair of realizations separated by a row distance (lag) $k$ must yield an expected cross-covariance of zero, effectively ruling out serial autocorrelation.

* **Statistical Formulation:** For any two non-identical row indices $i$ and $j$, the cross-covariance matrix must vanish:

$$H_0: \text{Cov}(\mathbf{X}_i, \mathbf{X}_j) = \mathbb{E}\left[ (\mathbf{X}_i - \boldsymbol{\mu})(\mathbf{X}_j - \boldsymbol{\mu})^T \right] = \mathbf{0}_{m \times m} \quad \forall i \neq j$$



In a sequential framework, this implies that at any lag $k \neq 0$, the multivariate autocovariance matrix $\boldsymbol{\Gamma}(k) = \mathbf{0}_{m \times m}$.
* **Verification Methodology:** This condition is evaluated dynamically across the entire dataset using the **Multivariate Ljung-Box Test** (or Hosking's Portmanteau Test). It calculates the sample cross-correlation matrices $\mathbf{R}(k)$ up to a designated maximum lag $H$, testing if the collective serial correlation is statistically indistinguishable from a white-noise process.

#### Multivariate Ljung-Box test Verification Methodology

To dynamically evaluate whether your observations carry memory across rows, the Multivariate Ljung-Box test (Hosking's Portmanteau Test) assesses cross-covariance up to a user-defined maximum row separation lag, $H$. It calculates whether the historical dependency structure is statistically indistinguishable from a white-noise process.

Let $\mathbf{x}_i \in \mathbb{R}^m$ be the $i$-th row realization, $\widehat{\boldsymbol{\mu}}_n \in \mathbb{R}^m$ be the global sample mean, and $n$ be the total number of observations.

---

1. Sample Auto-Cross Covariance Matrices ($\widehat{\boldsymbol{\Gamma}}_k$)

First, we establish the $m \times m$ sample auto-cross covariance matrix at lag $k$ (where $k = 0, 1, 2, \dots, H$). This matrix maps the linear associations of variables against themselves and each other across a row distance of $k$:

$$\widehat{\boldsymbol{\Gamma}}_k = \frac{1}{n} \sum_{i=k+1}^{n} (\mathbf{x}_i - \widehat{\boldsymbol{\mu}}_n)(\mathbf{x}_{i-k} - \widehat{\boldsymbol{\mu}}_n)^T$$

* When $k=0$, $\widehat{\boldsymbol{\Gamma}}_0$ represents the standard static sample variance-covariance matrix of the entire dataset.

---

2. The Multivariate Ljung-Box Test Statistic ($Q_m$)

Hosking's modified multivariate portmanteau statistic combines the cross-correlations across all lags up to $H$. It weights each lag to enhance small-to-moderate sample performance, though asymptotically it aligns with a standard joint maximum likelihood framework:

$$Q_m(H) = n^2 \sum_{k=1}^{H} \frac{1}{n - k} \text{tr}\left( \widehat{\boldsymbol{\Gamma}}_k^T \widehat{\boldsymbol{\Gamma}}_0^{-1} \widehat{\boldsymbol{\Gamma}}_k \widehat{\boldsymbol{\Gamma}}_0^{-1} \right)$$

Where $\text{tr}(\cdot)$ is the **matrix trace operator** (the sum of the diagonal elements), which effectively collapses the $m \times m$ matrix result at each lag into a single representative scalar.

---

3. Decision Rule

* **Value Interpretation:** If there is zero row-to-row dependency in the data, the elements of $\widehat{\boldsymbol{\Gamma}}_k$ for all $k > 0$ will hover close to zero, minimizing the trace and causing $Q_m(H)$ to stay small. Significant serial dependency will cause $Q_m(H)$ to inflate.
* **Asymptotic Testing:** Under the null hypothesis of joint row independence ($H_0: \boldsymbol{\Gamma}_k = \mathbf{0} \;\; \forall k \neq 0$), the statistic $Q_m(H)$ asymptotically follows a Chi-square distribution:

$$Q_m(H) \sim \chi^2_{\nu}$$



Where the degrees of freedom $\nu$ account for the total number of unique cross-correlation parameters checked over the $H$ lags:

$$\nu = m^2 H$$


* **Conclusion:** If the resulting $p$-value is greater than your significance level (e.g., $p > 0.05$), we fail to reject $H_0$. This provides empirical proof that the rows are linearly independent and do not exhibit memory effects.

> **Note:** Choosing the maximum lag $H$ requires balance. If $H$ is too small, you might miss long-range row dependencies; if $H$ is too large, the test loses statistical power by adding empty degrees of freedom. A standard empirical rule of thumb for baseline exploration is setting $H = \min(20, \, n/5)$ or using $H = \ln(n)$.

# Geometric De-correlation via Principal Component Analysis (PCA)



To exploit the structure of the empirical covariance matrix for structural health monitoring, we map the raw random variables into an orthogonal, uncorrelated coordinate space. This is achieved via Principal Component Analysis (PCA) derived directly from the population moments.

Consider a zero-mean random varaible $\widetilde{\mathbf{X}}_i = (\mathbf{X}_i - \boldsymbol{\mu}) \in \mathbb{R}^{m}$, representing the centered baseline features. Its population covariance matrix $\boldsymbol{\Sigma} \in \mathbb{R}^{m \times m}$ is symmetric and positive-semidefinite:

$$\boldsymbol{\Sigma} = \mathbb{E}[\widetilde{\mathbf{X}}_i \widetilde{\mathbf{X}}_i^T]$$

By the Spectral Theorem (or the Singular Value Decomposition of a symmetric matrix), $\boldsymbol{\Sigma}$ can be factored into its orthogonal eigenvectors and real, non-negative eigenvalues:

$$\boldsymbol{\Sigma} = \mathbf{P} \mathbf{\Lambda} \mathbf{P}^T$$

Where:

* $\mathbf{P} = [\mathbf{p}_1, \mathbf{p}_2, \dots, \mathbf{p}_m] \in \mathbb{R}^{m \times m}$ is an orthogonal matrix ($\mathbf{P}^T\mathbf{P} = \mathbf{P}\mathbf{P}^T = \mathbf{I}$) containing the principal component loading vectors.
* $\mathbf{\Lambda} = \text{diag}(\lambda_1, \lambda_2, \dots, \lambda_m) \in \mathbb{R}^{m \times m}$ is a diagonal matrix containing sorted eigenvalues ($\lambda_1 \ge \lambda_2 \ge \dots \ge \lambda_m \ge 0$), representing the population variance along each principal axis.

---

We define the transformed random varable $\mathbf{Z}_i \in \mathbb{R}^{m \times 1}$ as the projection of the centered observations onto the orthogonal eigenvector basis:

$$\mathbf{Z}_i \triangleq \mathbf{P}^T \widetilde{\mathbf{X}}_i$$

To prove that this transformation completely de-correlates the feature space, we evaluate the theoretical covariance of $\mathbf{Z}_i$:

$$\mathbb{E}[\mathbf{Z}_i \mathbf{Z}_i^T] = \mathbb{E}[(\mathbf{P}^T \widetilde{\mathbf{X}}_i)(\mathbf{P}^T \widetilde{\mathbf{X}}_i)^T]$$

Utilizing the linear properties of expectation and the matrix transpose identity $(\mathbf{A}\mathbf{B})^T = \mathbf{B}^T\mathbf{A}^T$:

$$\mathbb{E}[\mathbf{Z}_i \mathbf{Z}_i^T] = \mathbf{P}^T \mathbb{E}[\widetilde{\mathbf{X}}_i \widetilde{\mathbf{X}}_i^T] \mathbf{P} = \mathbf{P}^T \boldsymbol{\Sigma} \mathbf{P}$$

Substituting the spectral decomposition $\boldsymbol{\Sigma} = \mathbf{P}\mathbf{\Lambda}\mathbf{P}^T$ into the equation yields the definitive diagonalization proof:

$$\mathbb{E}[\mathbf{Z}_i \mathbf{Z}_i^T] = \mathbf{P}^T (\mathbf{P} \mathbf{\Lambda} \mathbf{P}^T) \mathbf{P} = (\mathbf{P}^T \mathbf{P}) \mathbf{\Lambda} (\mathbf{P}^T \mathbf{P}) = \mathbf{I} \mathbf{\Lambda} \mathbf{I} = \mathbf{\Lambda}$$

Because $\mathbf{\Lambda}$ is strictly diagonal, the cross-covariances between distinct principal components are exactly zero ($\mathbb{E}[Z_{i,j}Z_{i,k}] = 0$ for all $j \neq k$).

---

In practical applications, the latent matrix $\boldsymbol{\Sigma}$ is unknown. We operationalize PCA by substituting the consistent, unbiased empirical sample covariance matrix $\mathbf{S}$ derived via our Maximum Likelihood framework:

$$\mathbf{S} = \widehat{\mathbf{P}} \widehat{\mathbf{\Lambda}} \widehat{\mathbf{P}}^T$$

The observed realizations are centered using the empirical mean vector $\widehat{\boldsymbol{\mu}}_n$ and projected into the empirical principal space to obtain sample scores $\mathbf{z}_i$:

$$\mathbf{z}_i = \widehat{\mathbf{P}}^T (\mathbf{x}_i - \widehat{\boldsymbol{\mu}}_n)$$

This operational mapping guarantees that the sample covariance of the transformed data matrix collapses to the diagonal matrix $\widehat{\mathbf{\Lambda}}$, establishing an optimal coordinate system for isolating independent physical fault modes in your diagnostics layer.

---

To compress the feature space from $m$ dimensions to a lower dimension $k$ (where $k < m$), we partition the empirical orthogonal basis matrix $\widehat{\mathbf{P}}$ into two orthogonal subspaces:$$\widehat{\mathbf{P}} = \begin{bmatrix} \widehat{\mathbf{P}}_k & \Big| & \widehat{\mathbf{P}}_{m-k} \end{bmatrix}$$where $\widehat{\mathbf{P}}_k \in \mathbb{R}^{m \times k}$ contains the first $k$ column eigenvectors corresponding to the largest $k$ eigenvalues ($\widehat{\lambda}_1 \ge \dots \ge \widehat{\lambda}_k$). This partitioning allows us to partition $\mathbf{z}\in \mathbb{R}^m$ as
$$\mathbf{z}_{i}=\begin{bmatrix}\mathbf{z}_{i,k} \\\mathbf{z}_{i,m-k}
\end{bmatrix}$$
where $\mathbf{z}_{i,k} \in \mathbb{R}^{k}$ and $\mathbf{z}_{i,m-k} \in \mathbb{R}^{m-k}.$
From
$$\mathbf{z}_i = \widehat{\mathbf{P}}^T (\mathbf{x}_i - \widehat{\boldsymbol{\mu}}_n)$$
we thus have
$$(\mathbf{x}_i - \widehat{\boldsymbol{\mu}}_n)=\widehat{\mathbf{P}}_k \mathbf{z}_{i,k} + \widehat{\mathbf{P}}_{m-k} \mathbf{z}_{i,m-k}.$$

The partitioning of the empirical basis matrix allows us to isolate the deterministic structural dynamics from the unmodeled residual noise. By retaining only the first $k$ principal components, we define the reconstructed feature vector $\widehat{\mathbf{x}}_i \in \mathbb{R}^m$ as the projection of the low-dimensional scores back into the original measurement space:

$$\widehat{\mathbf{x}}_i \triangleq \widehat{\boldsymbol{\mu}}_n + \widehat{\mathbf{P}}_k \mathbf{z}_{i,k}$$

The geometric displacement between the actual observation $\mathbf{x}_i$ and its lower-dimensional reconstruction $\widehat{\mathbf{x}}_i$ defines the residual error vector $\mathbf{e}_i \in \mathbb{R}^m$:

$$\mathbf{e}_i = \mathbf{x}_i - \widehat{\mathbf{x}}_i = \widehat{\mathbf{P}}_{m-k} \mathbf{z}_{i,m-k}$$

Because $\widehat{\mathbf{P}}_k$ and $\widehat{\mathbf{P}}_{m-k}$ span mutually orthogonal subspaces, the total energy of the centered observation vector can be cleanly decomposed via the Pythagorean identity:

$$\|\mathbf{x}_i - \widehat{\boldsymbol{\mu}}_n\|^2 = \|\widehat{\mathbf{P}}_k \mathbf{z}_{i,k}\|^2 + \|\widehat{\mathbf{P}}_{m-k} \mathbf{z}_{i,m-k}\|^2 = \|\mathbf{z}_{i,k}\|^2 + \|\mathbf{z}_{i,m-k}\|^2$$

---

To construct a robust diagnostics layer, we monitor changes in both the principal and residual subspaces. This is achieved using two complementary statistical indicators.

1. Hotelling's $T^2$ Statistic (Principal Subspace)

Hotelling’s $T^2$ statistic measures the Mahalanobis distance within the $k$-dimensional principal subspace. It detects significant shifts or anomalies in the nominal, high-variance operational modes of the structure:

$$T^2_i = \mathbf{z}_{i,k}^T \widehat{\mathbf{\Lambda}}_k^{-1} \mathbf{z}_{i,k} = \sum_{j=1}^k \frac{z_{i,j}^2}{\widehat{\lambda}_j}$$

where $\widehat{\mathbf{\Lambda}}_k = \text{diag}(\widehat{\lambda}_1, \dots, \widehat{\lambda}_k)$. Under the assumption of multivariate normality, a statistically significant inflation in $T^2_i$ indicates that the system has drifted away from its baseline operational equilibrium.

2. The $Q$ Statistic / Squared Prediction Error (Residual Subspace)

The $Q$ statistic (also known as the Squared Prediction Error, SPE) measures the magnitude of the residual vector. It quantifies how well the baseline $k$-dimensional model explains a new observation:

$$Q_i = \|\mathbf{e}_i\|^2 = \mathbf{e}_i^T \mathbf{e}_i = \mathbf{z}_{i,m-k}^T \mathbf{z}_{i,m-k} = \sum_{j=k+1}^m z_{i,j}^2$$

Anomalies flagged by the $Q$ statistic indicate a break in the underlying structural correlation structure. This implies that a new, unmodeled physical phenomenon—such as a localized structural crack or sensor degradation—is projecting energy into the residual noise floor.

---

To optimize the diagnostic layer for dimensionality reduction or noise filtering, we must formally quantify the information captured by a truncated subset of the principal components. Because the trace of a matrix is invariant under cyclic permutations, the total variance of the system is preserved when mapping from the raw feature space to the de-correlated empirical principal space:

$$\text{tr}(\mathbf{S}) = \text{tr}(\widehat{\mathbf{P}}\widehat{\mathbf{\Lambda}}\widehat{\mathbf{P}}^T) = \text{tr}(\widehat{\mathbf{\Lambda}}\widehat{\mathbf{P}}^T\widehat{\mathbf{P}}) = \text{tr}(\widehat{\mathbf{\Lambda}}) = \sum_{j=1}^m \widehat{\lambda}_j$$

By partitioning the basis matrix into $\widehat{\mathbf{P}}_k$ and $\widehat{\mathbf{P}}_{m-k}$ to isolate the first $k$ dominant components, we explicitly divide this total system variance into the principal (explained) and residual (unexplained) subspaces:

* **Marginal and Cumulative Explained Variance:** The proportion of total variance captured within the principal subspace spanned by $\widehat{\mathbf{P}}_k$ is governed by the largest $k$ eigenvalues. Accumulating these ratios yields the cumulative explained variance ratio $\Phi(k)$, which measures the informational content of the low-dimensional score vector $\mathbf{z}_{i,k}$:

$$\Phi(k) = \frac{\sum_{j=1}^k \widehat{\lambda}_j}{\sum_{i=1}^m \widehat{\lambda}_i}$$

* **Residual Unexplained Variance:** Conversely, the information discarded by omitting the remaining $m-k$ low-variance axes represents the residual unexplained variance ratio $\Psi(k)$. This metric explicitly gauges the structural scale of the unmodeled operational subspace spanned by $\widehat{\mathbf{P}}_{m-k}$, tracking the total energy remaining in the residual score vector $\mathbf{z}_{i,m-k}$:

$$\Psi(k) = 1.0 - \Phi(k) = \frac{\sum_{j=k+1}^m \widehat{\lambda}_j}{\sum_{i=1}^m \widehat{\lambda}_i}$$

Operationalizing these metrics using the empirical eigenvalues $\widehat{\lambda}_j$ provides a mathematically rigorous thresholding strategy for our subspace decomposition. By plotting the decay of $\widehat{\lambda}_j$ against $\Psi(k)$, you can isolate the precise point $k$ where the system's true deterministic mechanical behavior—reconstructed via $\widehat{\mathbf{P}}_k \mathbf{z}_{i,k}$—transitions into the unmodeled stochastic white noise absorbed by $\widehat{\mathbf{P}}_{m-k} \mathbf{z}_{i,m-k}$.

# Subspace Decomposition via Latent Factor Modeling

To isolate the underlying physical phenomena driving the system's structural state and decouple variable-specific measurement noise, we map the observed features into a lower-dimensional latent space. While Principal Component Analysis (PCA) merely rotates the empirical coordinate system to maximize total variance, **Factor Analysis (FA)** models the internal correlation structure by partitioning the universal population variance into shared (structural) and independent (stochastic) subspaces.

Let $\mathbf{Z}_i \in \mathbb{R}^m$ be the normalized, zero-mean random variable derived from centering and scaling the $i$-th baseline feature vector $\mathbf{X}_i$:

$$\mathbf{Z}_i = \mathbf{D}^{-1/2}(\mathbf{X}_i - \boldsymbol{\mu})$$

where $\boldsymbol{\mu}$ is the universal population mean and $\mathbf{D} = \text{diag}(\sigma_1^2, \sigma_2^2, \dots, \sigma_m^2)$ is the diagonal matrix containing the population variances of $\mathbf{X}_i$. By construction, the theoretical population covariance matrix of $\mathbf{Z}_i$ collapses to the universal population correlation matrix, $\mathbf{R}$:

$$\mathbf{R} = \mathbb{E}[\mathbf{Z}_i \mathbf{Z}_i^T] \in \mathbb{R}^{m \times m}, \quad \text{with} \quad \mathbb{E}[Z_{i,j}^2] = 1.0 \quad \forall \ j \in \{1, \dots, m\}.$$

## The Latent Factor Postulate

We assume that the $m$ visible features in $\mathbf{Z}_i$ are linearly generated by a heavily compressed subset of $k$ unobservable (latent) physical mechanisms or operational conditions, where $k < m$. Formally, the generative model for the $i$-th realization vector is defined as:

$$\mathbf{Z}_i = \boldsymbol{\Lambda} \mathbf{F}_i + \boldsymbol{\epsilon}_i$$

Where:

* $\mathbf{F}_i = [F_{i,1}, F_{i,2}, \dots, F_{i,k}]^T \in \mathbb{R}^k$ is the random variable containing the **common factors** that govern the global physical state of the structure.
* $\boldsymbol{\Lambda} \in \mathbb{R}^{m \times k}$ is the invariant **factor loading matrix**, where the scalar $\lambda_{j,r}$ quantifies the linear sensitivity of the $j$-th feature to the $r$-th latent structural mode.
* $\boldsymbol{\epsilon}_i = [\epsilon_{i,1}, \epsilon_{i,2}, \dots, \epsilon_{i,m}]^T \in \mathbb{R}^m$ is the random variable representing the **specific errors** or **uniquenesses** unique to each measurement channel.

To ensure mathematical tractability and structurally isolate global physical effects from local disturbances, we impose three orthogonal population constraints:

1. **Orthogonal Common Factors:** The latent structural modes are mutually de-correlated and normalized to unit variance:

$$\mathbb{E}[\mathbf{F}_i] = \mathbf{0}, \quad \mathbb{E}[\mathbf{F}_i \mathbf{F}_i^T] = \mathbf{I}_{k \times k}$$


2. **De-correlated Unique Errors:** The measurement noise or local variance of any given feature is statistically independent of all other features. Consequently, the unique error covariance matrix $\boldsymbol{\Psi}$ must be strictly diagonal:

$$\mathbb{E}[\boldsymbol{\epsilon}_i] = \mathbf{0}, \quad \mathbb{E}[\boldsymbol{\epsilon}_i \boldsymbol{\epsilon}_i^T] = \boldsymbol{\Psi} = \text{diag}(\varphi_1^2, \varphi_2^2, \dots, \varphi_m^2)$$


3. **Subspace Orthogonality:** The global physical mechanics driving the common factors exert zero statistical influence over the local, channel-specific unique errors:

$$\mathbb{E}[\boldsymbol{\epsilon}_i \mathbf{F}_i^T] = \mathbf{0}_{m \times k}.$$

## Covariance Subspace Partitioning

Under the weight of these constraints, the theoretical correlation structure of the system can be decomposed directly. Expanding the expectation of the product $\mathbf{Z}_i\mathbf{F}_i^T$ yields:

$$\mathbb{E}[\mathbf{Z}_i \mathbf{F}_i^T] = \mathbb{E}[(\boldsymbol{\Lambda}\mathbf{F}_i + \boldsymbol{\epsilon}_i)\mathbf{F}_i^T] = \boldsymbol{\Lambda}\mathbb{E}[\mathbf{F}_i\mathbf{F}_i^T] + \mathbb{E}[\boldsymbol{\epsilon}_i\mathbf{F}_i^T] = \boldsymbol{\Lambda}\mathbf{I} + \mathbf{0} = \boldsymbol{\Lambda}$$

Thus, the factor loading matrix $\boldsymbol{\Lambda}$ explicitly maps the cross-correlation between the observable variables and the hidden structural modes. We evaluate the total population correlation matrix $\mathbf{R}$ by expanding the generative formula:

$$\mathbf{R} = \mathbb{E}[\mathbf{Z}_i \mathbf{Z}_i^T] = \mathbb{E}[(\boldsymbol{\Lambda} \mathbf{F}_i + \boldsymbol{\epsilon}_i)(\boldsymbol{\Lambda} \mathbf{F}_i + \boldsymbol{\epsilon}_i)^T]$$

$$\mathbf{R} = \boldsymbol{\Lambda} \mathbb{E}[\mathbf{F}_i \mathbf{F}_i^T] \boldsymbol{\Lambda}^T + \boldsymbol{\Lambda}\mathbb{E}[\mathbf{F}_i \boldsymbol{\epsilon}_i^T] + \mathbb{E}[\boldsymbol{\epsilon}_i \mathbf{F}_i^T]\boldsymbol{\Lambda}^T + \mathbb{E}[\boldsymbol{\epsilon}_i \boldsymbol{\epsilon}_i^T]$$

Utilizing the orthogonality properties ($\mathbf{I}$, $\mathbf{0}$, and $\boldsymbol{\Psi}$), the expression cleanly reduces to the **Fundamental Equation of Factor Analysis**:

$$\mathbf{R} = \boldsymbol{\Lambda} \boldsymbol{\Lambda}^T + \boldsymbol{\Psi}.$$

## Communality vs. Uniqueness

Focusing on the $j$-th diagonal entry of the correlation matrix $\mathbf{R}$, the total unit variance of each normalized variable $Z_{i,j}$ splits into two distinct, interpretable physical spaces:

$$\mathbb{E}[Z_{i,j}^2] = \sum_{r=1}^k \lambda_{j,r}^2 + \varphi_j^2 = 1.0$$

We define these partitioned elements as:

* **Communality ($h_j^2$):** The portion of a sensor's variance explained collectively by the $k$ shared structural factors:

$$h_j^2 \triangleq \sum_{r=1}^k \lambda_{j,r}^2$$


* **Uniqueness ($\varphi_j^2$):** The remaining portion of variance completely isolated to that specific sensor, representing a combination of localized non-structural anomalies and electronic white noise:

$$\varphi_j^2 \triangleq 1.0 - h_j^2 = \mathbb{E}[\epsilon_{i,j}^2]$$



> 🔬 **Diagnostic Subspace Principles:** Because the factor model forces all inter-variable correlation to be processed exclusively through the loading matrix ($\mathbb{E}[Z_{i,j}Z_{i,l}] = \sum_{r=1}^k \lambda_{j,r}\lambda_{l,r}$ for $j \neq l$), tracking these parameters over a sequence of realizations yields localized diagnostic power:
> * If the system undergoes a global structural shift (e.g., a shifting mechanical load path), the shared covariance network breaks, forcing wide-scale realignments in the factor loading matrix $\boldsymbol{\Lambda}$ and the communality spectrum $h_j^2$.
> * Conversely, if a single sensor (feature) degrades, undergoes decalibration, or experiences an isolated local failure, its cross-correlation with the rest of the network collapses. In this scenario, its communality $h_j^2$ will drop precipitously, forcing its individual uniqueness metric $\varphi_j^2$ to spike toward $1.0$, allowing the diagnostic layer to instantly isolate sensor malfunction from true structural damage.
>

## Estimation of the Factors via Conditional Expectation

The latent factors $\mathbf{F}_i$ cannot be directly calculated by rotating the data (the way we do in PCA), we must treat them as unobserved parameters and estimate them. This formulation represents **Thomson's Method** (also widely known as the **Regression Method** for factor scores). It is derived by treating $\mathbf{f}_i$ as the Minimum Mean Squared Error (MMSE) estimator, which mathematically reduces to the **conditional expectation** under multivariate normal assumptions.

Here is the step-by-step rigorous proof of how we get this equation from your setup.

---

1. Joint Subspace Covariance Construction

To find the conditional expectation $\mathbb{E}[\mathbf{F}_i | \mathbf{z}_i]$, we first examine how the observable normalized variables $\mathbf{Z}_i$ and the hidden common factors $\mathbf{F}_i$ vary together. We construct a partitioned joint random vector $\mathbf{Y}_i$:

$$\mathbf{Y}_i = \begin{bmatrix} \mathbf{Z}_i \\ \mathbf{F}_i \end{bmatrix}$$

Because both $\mathbf{Z}_i$ and $\mathbf{F}_i$ have a mean of zero ($\mathbb{E}[\mathbf{Z}_i] = \mathbf{0}$ and $\mathbb{E}[\mathbf{F}_i] = \mathbf{0}$), the joint mean is $\mathbb{E}[\mathbf{Y}_i] = \mathbf{0}$.

Now, let's construct the partitioned joint covariance matrix $\boldsymbol{\Sigma}_{YY}$:

$$\boldsymbol{\Sigma}_{YY} = \mathbb{E}[\mathbf{Y}_i \mathbf{Y}_i^T] = \begin{bmatrix} \mathbb{E}[\mathbf{Z}_i \mathbf{Z}_i^T] & \mathbb{E}[\mathbf{Z}_i \mathbf{F}_i^T] \\ \mathbb{E}[\mathbf{F}_i \mathbf{Z}_i^T] & \mathbb{E}[\mathbf{F}_i \mathbf{F}_i^T] \end{bmatrix}$$

From the foundations of Factor Analysis that was established before, we know the values of these individual blocks:

1. $\mathbb{E}[\mathbf{Z}_i \mathbf{Z}_i^T] = \mathbf{R}$ (The visible correlation matrix)
2. $\mathbb{E}[\mathbf{Z}_i \mathbf{F}_i^T] = \boldsymbol{\Lambda}$ (The factor loading matrix, which we proved earlier)
3. $\mathbb{E}[\mathbf{F}_i \mathbf{Z}_i^T] = (\mathbb{E}[\mathbf{Z}_i \mathbf{F}_i^T])^T = \boldsymbol{\Lambda}^T$
4. $\mathbb{E}[\mathbf{F}_i \mathbf{F}_i^T] = \mathbf{I}_{k \times k}$ (The orthogonal common factor constraint)

Substituting these blocks into our partitioned matrix yields:

$$\boldsymbol{\Sigma}_{YY} = \begin{bmatrix} \mathbf{R} & \boldsymbol{\Lambda} \\ \boldsymbol{\Lambda}^T & \mathbf{I} \end{bmatrix}$$

---

2. Applying the Conditional Expectation Theorem

Under the standard assumption of a Multivariate Normal distribution for the statistical inference layer, the conditional distribution of one partitioned segment given the other is well-defined.

For any joint normal system where $\boldsymbol{\Sigma} = \begin{bmatrix} \boldsymbol{\Sigma}_{11} & \boldsymbol{\Sigma}_{12} \\ \boldsymbol{\Sigma}_{21} & \boldsymbol{\Sigma}_{22} \end{bmatrix}$, the conditional expectation of the second slot given a realization of the first slot ($\mathbf{x}_1$) is given by the classical linear regression projection formula:

$$\mathbb{E}[\mathbf{X}_2 | \mathbf{x}_1] = \boldsymbol{\mu}_2 + \boldsymbol{\Sigma}_{21} \boldsymbol{\Sigma}_{11}^{-1} (\mathbf{x}_1 - \boldsymbol{\mu}_1)$$

Mapping this directly to our factor analysis matrix partitions:

* Slot 1 is our observed sample vector realization: $\mathbf{x}_1 = \mathbf{z}_i$ (with $\boldsymbol{\mu}_1 = \mathbf{0}$)
* Slot 2 is our hidden latent factor: $\mathbf{X}_2 = \mathbf{F}_i$ (with $\boldsymbol{\mu}_2 = \mathbf{0}$)
* The cross-covariance section $\boldsymbol{\Sigma}_{21} = \boldsymbol{\Lambda}^T$
* The observed covariance section $\boldsymbol{\Sigma}_{11} = \mathbf{R}$

Substituting these variables into the projection formula yields the factor score realization $\mathbf{f}_i$:

$$\mathbf{f}_i = \mathbb{E}[\mathbf{F}_i | \mathbf{z}_i] = \mathbf{0} + \boldsymbol{\Lambda}^T \mathbf{R}^{-1} (\mathbf{z}_i - \mathbf{0})$$

$$\mathbf{f}_i = \boldsymbol{\Lambda}^T \mathbf{R}^{-1} \mathbf{z}_i$$

---



Because this estimator is derived via the conditional expectation, it possesses a vital property: it is the **Best Linear Unbiased Predictor (BLUP)**.

Even if the data deviates slightly from strict normality, this specific linear matrix transformation minimizes the mean squared error ($\mathbb{E}[\|\mathbf{F}_i - \mathbf{f}_i\|^2]$) between the hidden, true physical state of the structure and the feature estimates.

# Multi Attribute Value Technique

Let $v_i\triangleq v(X_i)$ be function of the random variable $X_i$. Define
\begin{align*}
w_i&\triangleq \frac{E(v(X_i))}{\sum_{j=1}^m E(v(X_j))},\\
Q&\triangleq \sum_{i=1}^m w_i E(v(X_i))=\frac{1}{\sum_{j=1}^m E(v(X_j))}\sum_{i=1}^m E(v(X_i))^2
\end{align*}
