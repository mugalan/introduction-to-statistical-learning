# Probability Space

A probability space is characterized by a triple: $(\Omega, \mathfrak{U},P)$ where

* $\Omega$ is some set and is the sample space
* $\mathfrak{U}$ is a $\sigma$-algebra on $\Omega$
* $P: \mathfrak{U} \to [0,1]$ is a probability measure satisfying $P(\Omega)=1$ and countable additivity

In this setup:

* Any $A\in \mathfrak{U}$ represents an event.
* The probability of the event $A\in \mathfrak{U}$ happening is give by the measure ('size') of $A$ defined by
$$P(A)\triangleq \int_A dP.$$

---

If $\Omega \subseteq \mathbb{R}^n$ and $\mathfrak{U}$ is Lebesgue-measurable, and $P\ll\lambda$ then
$$P(A)\triangleq \int_A dP=\int_{A}f(\omega)\,d\lambda (\omega).$$
Here $f(\omega)$ is called the dinsity of $P$ with respect to $\lambda$ and is formally given by the Radon-Nikodym derivative
$$f\triangleq \dfrac{dP}{d\lambda}.$$

---

In the special case where $\Omega$ is discrete, one may take the reference measure to be the counting measure $\#$. If $P\ll \#$, then the Radon--Nikodym derivative
$$
p\triangleq \frac{dP}{d\#}
$$
is the probability mass function of $P$. Hence, for any event $A\subseteq \Omega$,
$$
P(A)=\int_A dP=\int_A p\,d\#
=\sum_{\omega\in A} p(\omega).
$$
In particular,
$$
p(\omega)=P(\{\omega\}),
\qquad
\sum_{\omega\in\Omega}p(\omega)=1.
$$
Thus, in the discrete case, integration with respect to counting measure reduces to summation, and the density of $P$ with respect to counting measure is precisely the probability mass function.

# Random Variable

Consider a probability space $(\Omega,\mathfrak U,P)$ and a measurable space
$(E, \mathfrak{E})$. A random variable is a measurable map
$$
X:\Omega\to E,
$$
meaning that
$$
X^{-1}(B)\in\mathfrak{U}
\qquad \text{for every } B\in\mathfrak{E}.
$$
Equivalently, the $\sigma$-algebra generated by $X$,
$$
\sigma(X)\triangleq \{X^{-1}(B):B\in\mathfrak{E}\},
$$
satisfies
$$
\sigma(X)\subseteq \mathfrak U.
$$

# The Law and the Distribution of a Random Variable

## The Law of $X$

Once we have a measurable function $X$, we can use it to "transport" the probability measure from our original space to the target space. This transported measure is called the **Law** of the Variable.


If $X : (\Omega, \mathfrak{A}) \to (E, \mathfrak{E})$ is a random Variable and $P$ is a probability measure on $(\Omega, \mathfrak{A})$, the **law of $X$ under $P$** (denoted $P_X$ or $\mathfrak{L}_X$) is a probability measure on the target space $(E, \mathfrak{E})$ defined by:

$$\forall B \in \mathfrak{E}, \quad P_X(B) \triangleq P(X^{-1}(B)) = P(\{\omega \in \Omega : X(\omega) \in B\})$$

In standard notation, we usually write this as:


$$P_X(B) = P(X \in B)=\int_{X^{-1}(B)}dP, \qquad \forall B\in \mathfrak{E}.$$

## The "Push-forward" Mechanism

Think of $X$ as a map. $P$ tells us how much "mass" is at each point in $\Omega$. The Law $P_X$ is simply the result of "pushing" all that mass forward onto the points in $E$.
The push-forwad measure on $E$ is therefore characterized by
$$P_X(B) = P(X \in B), \qquad \forall B\in \mathfrak{E}$$
which can equivalently be given by
$$\int_{X^{-1}(B)}dP=\int_BdP_X, \qquad \forall B\in \mathfrak{E}.$$

## Pushforward Rule

For every measurable function $g:E\to\mathbb R$,
$$
\int_\Omega g(X)\,dP=\int_E g\,dP_X,
$$
whenever the integrals are well defined.


## Density of a Random Variable

Similar to $P$ on $\Omega$, in order to represent $P_X$ as a continuous distribution, we compare it to a reference measure on the target space—typically the Lebesgue measure $\lambda$ on $(E, \mathfrak{E})$.
If $P_X \ll \lambda$ then by the Radon-Nikodym Theorem, there exists a unique (up to a set of $\lambda$-measure zero), non-negative, and measurable function $f_X: E \to [0, \infty)$ such that for every $B \in \mathfrak{E}$:
$$P(X \in B) = P_X(B) = \int_{B} dP_X=\int_{B} f_X(x) \, d\lambda(x)$$

The function $f_X = \frac{dP_X}{d\lambda}$ is the probability density function (PDF) of the random variable $X$.


## The Joint Probability Density

Let $X:(\Omega,\mathfrak A)\to(\mathscr X,\mathfrak U_X)$ and
$Y:(\Omega,\mathfrak A)\to(\mathscr Y,\mathfrak U_Y)$ be two random variables
defined on the same probability space $(\Omega,\mathfrak A,P)$. They may be
viewed together as a single random ordered pair
$$
Z=(X,Y):\Omega\to\mathscr X\times\mathscr Y.
$$

The map $Z=(X,Y)$ is measurable with respect to the product
$\sigma$-algebra $\mathfrak U_X\otimes\mathfrak U_Y$. The joint law, or
joint distribution, of $X$ and $Y$, denoted $P_{X,Y}$, is the probability
measure on
$$
(\mathscr X\times\mathscr Y,\mathfrak U_X\otimes\mathfrak U_Y)
$$
defined by
$$
P_{X,Y}(B)
=
P\big((X,Y)\in B\big)
=
P\big((X,Y)^{-1}(B)\big),
\qquad B\in \mathfrak U_X\otimes\mathfrak U_Y.
$$


If $\mathscr X\subseteq\mathbb R^m$, $\mathscr Y\subseteq\mathbb R^n$, and
the joint law $P_{X,Y}$ is absolutely continuous with respect to the
Lebesgue measure on $\mathscr X\times\mathscr Y$, then there exists a
measurable function $f_{X,Y}:\mathscr X\times\mathscr Y\to[0,\infty)$ such
that
$$
P\big((X,Y)\in B\big)
=
P_{X,Y}(B)
=
\int_B dP_{X,Y}
=
\int_B f_{X,Y}(x,y)\,d(x,y),
$$
for every $B\in\mathfrak U_X\otimes\mathfrak U_Y$.
When $\mathscr X\subseteq\mathbb R$ and $\mathscr Y\subseteq\mathbb R$, this
is often written as
$$
P\big((X,Y)\in B\big)
=
\iint_B f_{X,Y}(x,y)\,dx\,dy.
$$
The function $f_{X,Y}$ is called the joint probability density function of
$X$ and $Y$.

## The Marginal Probability Density

The **marginal law** of one variable is obtained from the joint law by
projecting onto the corresponding coordinate. In measure-theoretic terms,
if $P_{X,Y}$ is the joint law of $(X,Y)$, then the marginal law of $X$
is the pushforward of $P_{X,Y}$ under the coordinate projection
$$
\pi_X:\mathscr X\times \mathscr Y\to \mathscr X,
\qquad
\pi_X(x,y)=x.
$$
Thus, for every measurable set $A\in\mathfrak U_X$,
$$
P_X(A)
=
P_{X,Y}(A\times \mathscr Y).
$$
Similarly, for every measurable set $C\in\mathfrak U_Y$,
$$
P_Y(C)
=
P_{X,Y}(\mathscr X\times C).
$$

If $X$ and $Y$ are jointly continuous with joint density $f_{X,Y}$,
then the marginal density of $X$ is obtained by integrating out $Y$:
$$
f_X(x)
=
\int_{\mathscr Y} f_{X,Y}(x,y)\,dy.
$$
When $\mathscr Y=\mathbb R$, this becomes
$$
f_X(x)
=
\int_{-\infty}^{\infty} f_{X,Y}(x,y)\,dy.
$$

Similarly,
$$
f_Y(y)
=
\int_{\mathscr X} f_{X,Y}(x,y)\,dx.
$$

In the discrete case we have:
$$p_X(x) = \sum_{y \in \text{Img}(Y)} p_{X,Y}(x, y)$$
and
$$p_Y(y) = \sum_{x \in \text{Img}(X)} p_{X,Y}(x, y).$$

# The Expectation and Covariance of a Random Variable

In the case where $E=\mathbb{R}^n$, the **expectation** (or mean) of $X$ is defined as the integral of $X$ with respect to the probability measure $P$:

$$\mathbb E[X] = \int_{\Omega} X(\omega) \, dP(\omega)=\int_{\mathbb{R}^n}x\,dP_{X}(x)$$

The **Covariance** is defined as:

$$\mathrm{Cov}(X) = E\left[ (X - \mathbb E[X])(X - \mathbb E[X])^T \right] = \int_{\mathbb{R}^n} (x - \mathbb E[X])(x - \mathbb E[X])^T \, dP_X$$

# Independence of Random Variables

Let $(\Omega,\mathfrak A,P)$ be a probability space. Let
$X_1,X_2,\dots,X_n$ be random variables, where each
$$
X_i:(\Omega,\mathfrak A)\to(E_i,\mathfrak E_i)
$$
is measurable.

The random variables $X_1,X_2,\dots,X_n$ are said to be independent if, for every choice of measurable sets
$B_i\in\mathfrak E_i$, $i=1,\dots,n$, we have
$$
P(X_1\in B_1,\dots,X_n\in B_n)
=
\prod_{i=1}^n P(X_i\in B_i).
$$
Equivalently,
$$
P\left(\bigcap_{i=1}^n X_i^{-1}(B_i)\right)
=
\prod_{i=1}^n P\left(X_i^{-1}(B_i)\right).
$$

If the joint law of $(X_1,\dots,X_n)$ is absolutely continuous with respect to a product reference measure, and if each marginal law has density $f_{X_i}$, then independence is equivalent to the factorization
$$
f_{X_1,\dots,X_n}(x_1,\dots,x_n)
=
\prod_{i=1}^n f_{X_i}(x_i)
$$
almost everywhere.

# Multivariate Central Limit Thoerem

Let
$$
X_1,X_2,\dots
$$
be i.i.d. random variables in $\mathbb R^d$ with
$$
\mathbb E[X_1]=\mu\in\mathbb R^d,
\qquad
\operatorname{Cov}(X_1)=\Sigma\in\mathbb R^{d\times d},
$$
where $\Sigma$ is finite. Define
$$
\bar X_n=\frac1n\sum_{i=1}^n X_i.
$$
Then the multivariate central limit theorem states that
$$
\sqrt n(\bar X_n-\mu)
\xrightarrow{d}
\mathscr N_d(0,\Sigma).
$$


Equivalently, for large $n$,
$$
\bar X_n
\approx
\mathscr N_d\left(\mu,\frac{1}{n}\Sigma\right).
$$

# Conditional Expectation

## Definition of $\mathbb{E}[X\mid Y]$

Let $(\Omega, \mathfrak{A}, P)$ be a probability space.

* Let $X: \Omega \to \mathbb{R}$ be a random variable that is integrable ($X \in L^1$).
* Let $Y: \Omega \to \mathbb{R}$ be another random variable.
* Let $\sigma(Y)$ be the **sub-$\sigma$-algebra** generated by $Y$.

**The Existence: Conditional Expectation**

The conditional expectation of $X$ given $Y$, denoted $\mathbb E[X|Y]$ (or more formally $\mathbb E[X|\sigma(Y)]$), is defined as the unique (almost surely) random variable $Z$ that satisfies two conditions:

1. **Measurability:** $Z$ is measurable with respect to $\sigma(Y)$.
That is by the Doob-Dynkin lemma $Z$ must be a function of $Y$:
$$Z = g(Y).$$
2. **The Integral Property (Partial Averaging):** For every set $G \in \sigma(Y)$:

$$\int_G Z \, dP = \int_G X \, dP$$



*(Translation: On any set of outcomes that $Y$ can distinguish, the "average" of our estimate $Z$ must match the "average" of the actual $X$.)*


## The Joint Probability ($P(X\in B\mid Y\in A)$)

Let
$$
Z_B\triangleq \mathbb E[\mathbb 1_{\{X\in B\}}\mid Y].
$$

Then it follows that

$$
P(X\in B,\;Y\in A)=\int_{Y^{-1}(A)} Z_B\,dP.
$$

---

$Z_B$ is measurable with respect to $\sigma(Y)$.
Hence from the Doob-Dynkin Lemma $Z_B$ must be a function of $Y$. We can write $Z_B = g_B(Y)$ and then the above expression can be written as:

$$
P(X\in B,\;Y\in A)=\int_{Y^{-1}(A)} g_B(Y)\,dP.
$$

Now applying the pushforward rule:
$$
\int_{Y^{-1}(A)} g_B(Y)\,dP=\int_A g_B(y)\,dP_Y(y),
$$

we have

$$
P(X\in B,\;Y\in A)=\int_A g_B(y)\,dP_Y(y).
$$

If $Y$ has density $f_Y$, so that

$$
dP_Y(y)=f_Y(y)\,dy,
$$

then this becomes

$$
P(X\in B,\;Y\in A)=\int_A g_B(y)\,f_Y(y)\,dy.
$$

In the discrete setting we have:
$$
P(X\in B,\;Y\in A)=\sum_{y\in A} g_B(y)\,P(Y=y).
$$

---

This means that the function $g_B(y)$ satisfies:

$$g_B(y)=P(X\in B\mid Y=y),$$
holds only $P_Y$-almost everywhere.

## The conditional Probability, $P(X\in B\mid Y=y)$

In measure theory, the "probability" of an event is simply the **measure of the set** representing that event. Thus the expression
$$g_B(y)=P(X\in B\mid Y=y)$$
indicates that for a fixed $y$ the map
$$
P_{X \mid Y=y}:\mathfrak{U} \mapsto [0,1]
$$
where $P_{X \mid Y=y}(B)$ for any $B\in \mathfrak{U}$ is a probability measure. If $P_{X \mid Y=y}$ is absolutely continuous with the length measure on $\mathbb{R}$ the Radon-Nikodym theorem tells us that there existis a unique density function $f_{X|Y}$ that satisfies
$$
\dfrac{dP_{X \mid Y=y}}{dx}(x)=f_{X|Y}(x\mid y).
$$

Thus we have the expression
$$
g_B(y)=P_{X \mid Y=y}(B)
=\int_B dP_{X \mid Y=y}=\int_B f_{X|Y}(x \mid y) \, dx.
$$
indicating that $g_B(y)$ gives the conditional probality of $X\in B$ given that $Y=y$. That is
$$
g_B(y)\triangleq P(X\in B\mid Y=y)=\int_B f_{X|Y}(x \mid y) \, dx.
$$

## The Product Rule of the Conditional Density

### Continuous Case

Substituting the above expression, $
g_B(y)=\int_B f_{X|Y}(x \mid y) \, dx
$, into the integral $
P(X\in B,\;Y\in A)=\int_A g_B(y)\,f_Y(y)\,dy
$ we have:
$$P(X \in B, Y \in A)
= \int_A \int_B f_{X|Y}(x \mid y) \, f_Y(y)\,dx \, dy.
$$

Since $P(X \in B, Y \in A) = \iint_{A \times B} f_{X,Y}(x,y) \, dx \, dy$ where $f_{X,Y}(x,y)$ is the joint probability density,
we obtain the product rule

$$
f_{X,Y}(x,y)=f_{X|Y}(x\mid y)f_Y(y)
$$
for almost every $(x,y)$.
The equality holds for the joint density $f_{X,Y}$ by the uniqueness of the Radon-Nikodym derivative.

### Discrete Case

The relationship $f_{X,Y}(x,y) = f_{X|Y}(x|y)f_Y(y)$ represents the Product Rule of probability for the continuous case. In the discrete case this is expressed as
$$p_{X,Y}(x,y) = p_{X|Y}(x|y)p_Y(y),$$
where
$$p_{X|Y}(x|y)=P(X=x \mid Y=y).$$

$$P(X \in B, Y \in A) = \sum_{y \in A} P(X \in B \mid Y=y) P(Y=y)$$

## The Disintegration Theorem

Let $(X, Y): \Omega \to \mathscr{X} \times \mathscr{Y}$ be the joint map. By the **Pushforward Rule** we cans see that
$$\int_{Y^{-1}(A)} X \, dP = \int_A \left( \int_{\mathscr{X}} x \, dP_{X \mid Y=y}(x)\right) dP_Y(y)= \int_A \left( \int_{\mathscr{X}} x \, f_{X \mid Y=y}(x\mid y) \,dx\right) f_Y(y)dy.$$

In the discrete case the above result is expressed as:


$$\sum_{\omega \in Y^{-1}(A)} X(\omega)\, P(\{\omega\}) =
\sum_{y\in A}\sum_{x\in \mathscr{X}}xp_{X\mid Y}(x\mid y) \, P(Y=y).$$

## The Conditional Expectation Evaluated at $Y=y$

We apply the definition of conditional expectation for a general random variable $X$ and using the Doob-Dynkin lemma we have for any $A\in \mathscr{Y}$:

$$
\int_{Y^{-1}(A)} X \, dP =
\int_{Y^{-1}(A)} \mathbb E[X \mid Y] \, dP =\int_A \mathbb E[X \mid Y=y] \, dP_Y(y).
$$

Note that in the previous section we have shown that the Disintegration Theorem gives
$$
\int_{Y^{-1}(A)} X \, dP = \int_A \left( \int_{\mathscr{X}} x \, dP_{X \mid Y=y}(x) \right) dP_Y(y).
$$

Thus we see that
$$\mathbb{E}[X\mid Y=y]= \int_{\mathscr{X}} x \, dP_{X \mid Y=y}(x)=
\int_{\mathscr{X}} x f_{X|Y}(x\mid y)dx.
$$
The first term, $\mathbb{E}[X|Y=y]$, represents the evaluation of the $\sigma(Y)$-measurable random variable $g(Y)=\mathbb{E}[X\mid Y]$ at $Y=y$.


---

In the discrete case,
the partial expectation over the event $\{Y \in A\}$ simplifies to the discrete weighted sum:

$$\sum_{\omega \in Y^{-1}(A)} X(\omega)\, P(\{\omega\}) =
\sum_{y\in A}\mathbb E[X \mid Y=y] \, P(Y=y),$$

and the factor $g(y)$ becomes:
$$g(y) = \mathbb{E}[X \mid Y=y] = \sum_{x \in \mathscr{X}} x \, p_{X\mid Y}(x \mid y),$$
where $p_{X\mid Y}(x \mid y) = P(X=x \mid Y=y)$ is the conditional probability mass function.

## Conditional Random Variable, $X \mid Y=y$

In the standard framework of probability, we often talk about the distribution of $X$ across the entire sample space. However, once we observe $Y=y$, the "global" distribution of $X$ becomes obsolete; we need a mathematical object that represents our updated knowledge of $X$ within this specific "slice" of reality.


While $X$ is a mapping from the sample space $\Omega$ to $\mathscr{Y}$, the **conditional random variable** $X \mid Y=y$ (read as "$X$ given $Y=y$") is defined as a random variable whose law is the **Regular Conditional Law** $P_{X \mid Y=y}$.

---

Unlike the original $X$, which is a single map defined on $(\Omega, \mathfrak{A}, P)$, the conditional variable $X \mid Y=y$ can be viewed as a **family of random variables** indexed by the possible outcomes of $Y$ and satisifes:

* **Realization:** If $Y$ is fixed at $y$, $X \mid Y=y$ behaves like a standard random variable with its own mean, variance, and moments, all of which are now functions of $y$.
* **The Density:** The function $f_{X|Y}(x \mid y)$ serves as the probability density function (PDF) for this variable. Just as we write $X \sim f_X(x)$, we now write:
$$(X \mid Y=y) \sim f_{X|Y}(x \mid y).$$
* **The Expectation:** The "best guess" for the value of this conditional random variable is the conditional mean that was derived earlier:
$$\mathbb{E}[(X \mid Y=y)] \triangleq \int_{\mathscr{X}} x f_{X|Y}(x \mid y) \, dx=\mathbb{E}[X \mid Y=y].$$

---

The relationship between the global random variable $X$ and the conditional family $X \mid Y=y$ is bridged by the **Law of Total Probability**. The global distribution is the mixture of these conditional variables, weighted by the marginal distribution of $Y$:
$$f_X(x) = \int_{\mathscr{Y}} f_{X|Y}(x \mid y) f_Y(y) \, dy.$$

In essence, $X \mid Y=y$ is the version of $X$ that lives exclusively on the "slice" of the universe where $Y$ has already happened and its value is known.

## Relationship between $g_B(y)$ and $g(y)$

The best way to understand the difference is to carefully examine how they are formally defined.

For a fixed $y$ (where $f_Y(y) > 0$), the mapping $g_B(y)$ is defined for any Borel set $B \in \mathfrak{B}(\mathscr{X})$ as:

$$g_B(y) \triangleq P(X \in B \mid Y=y) = \int_B f_{X|Y}(x \mid y) \, dx = \int_B dP_{X|Y=y}(x).$$

This demonstrates that $g_B(y)$, for a fixed $y$, evaluates the conditional probability measure $P_{X|Y=y}(B)$ over a given set $B$ in the state space of $X$.

On the other hand, $g(y)$ is defined as:

$$g(y) \triangleq \mathbb{E}[X \mid Y=y] = \int_{\mathscr{X}} x \, f_{X|Y}(x \mid y) \, dx = \int_{\mathscr{X}} x \, dP_{X|Y=y}(x).$$

This demonstrates that $g(y)$, for a fixed $y$, computes the functional average (the integration of the identity map $X(x)=x$) over the entire domain $\mathscr{X}$ with respect to that same conditional measure $P_{X|Y=y}$.

Thus, they measure two distinct mathematical objects: $g_B(y)$ evaluates a localized **probability measure**, whereas $g(y)$ calculates a **numerical expectation** under that measure.

However, they are related in a very subtle manner through the concept of a **regular conditional distribution**.

For a fixed $y$, $P_{X|Y=y}$ behaves as a valid probability measure on $(\mathscr{X}, \mathfrak{B}(\mathscr{X}))$ with density $f_{X|Y}(x \mid y)$. We can define an abstract **conditional random variable**, denoted $(X \mid Y=y)$, whose probability distribution is exactly given by this measure.

Importantly, this conditional random variable $(X \mid Y=y)$ is a completely different mathematical object than the $\sigma(Y)$-measurable random variable $\mathbb{E}[X \mid Y]$. However, computing the standard expectation of this localized variable yields:

$$\mathbb{E}[(X \mid Y=y)] \triangleq \int_{\mathscr{X}} x \, f_{X|Y}(x \mid y) \, dx = \mathbb{E}[X \mid Y=y] = g(y).$$


## Properties of $\mathbb{E}[X\mid Y]$

### Law of total expectation

$$\mathbb E[X] = \mathbb E[\mathbb E[X | Y]]$$

### Conditional Variance ($\mathrm{Var}(X|Y)$)

The conditional variance, $\mathrm{Var}(X|Y)$, is the variance of $X$ given a specific outcome of $Y$. Formally, it is a $\sigma(Y)$-measurable random variable defined as:
$$\mathrm{Var}(X|Y) \triangleq \mathbb E[X^2 | Y] - (\mathbb E[X|Y])^2$$
It measures the **residual uncertainty** of $X$ that remains after $Y$ is known. If you were to observe $Y=y$, the conditional variance tells you the spread of the remaining probability distribution of $X$.

### Law of Total Variance

For any two random variables $X$ and $Y$ with finite variance, the total variance of $X$ can be split into two distinct parts:

$$\mathrm{Var}(X) = \mathbb E[\mathrm{Var}(X|Y)] + \mathrm{Var}(\mathbb E[X|Y])$$

* **$\mathrm{Var}(X)$**: Total Uncertainty (The "Total Spread").
* **$\mathbb E[\mathrm{Var}(X|Y)]$**: **Unexplained Variance** (The "Average Spread of the Slices").
* **$\mathrm{Var}(\mathbb E[X|Y])$**: **Explained Variance** (The "Spread of the Means").

The **Explained Variance** represents the part of $X$'s variance that is "accounted for" or explained by the linear relationship with $Y$. **Unexplained Variance** represents the remaining uncertainty in $X$ after $Y$ is known, often referred to as the estimation error or residual variance.

### Optimality of the Conditional Estimate

Let $X$ be a random variable with $\mathbb{E}[X^2] < \infty$. Let $Y$ is a $\sigma(X)$-measurable random variable and $Z$ be a $\sigma(Y)$-measurable random variable with $\mathbb{E}[Z^2] < \infty$. Then
\begin{align*}
\mathbb{E}[(X - \mathbb{E}[X \mid Y])^2] &\leq \mathbb{E}[(X - Z)^2].
\end{align*}

>This result proves that $\mathbb{E}[X \mid Y]$ is the **optimal estimate** of $X$ given the information in $Y$, in the sense that it is the unique minimizer of the **Mean Square Error (MSE)**. Any other $\sigma(Y)$-measurable estimate $Z$ will necessarily result in a larger (or equal) expected squared error.

---

Using the Law of Iterated Expectations we have:

\begin{align*}
\mathbb{E}[(X - \mathbb{E}[X \mid Y])^2] &= \mathbb{E}[\text{Var}(X \mid Y)]
\end{align*}

## Conditioning of Multivariate Gaussian Random Variables

Given a joint multivariate normal distribution
$$Z = \begin{bmatrix} X \\ Y \end{bmatrix} \sim \mathscr{N} \left( \begin{bmatrix} \mu_X \\ \mu_Y \end{bmatrix}, \begin{bmatrix} \Sigma_{XX} & \Sigma_{XY} \\ \Sigma_{YX} & \Sigma_{YY} \end{bmatrix} \right)$$ the "best predictor" for all these methods is the conditional expectation:

$$\mathbb{E}[X|Y] = \mu_X + \Sigma_{XY}\Sigma_{YY}^{-1}(Y - \mu_Y).$$

---

In the context of Bayesian inference, filtering, and machine learning, here is the breakdown of the components of the above expression using the Bayesian terminology:

1. $\mu_X$ (The Prior Mean):
* **What it is:** This is your "best guess" for $X$ if you have zero information about $Y$.
* **Role:** It acts as the starting point (the baseline). In a **Kalman Filter**, this would be the "predicted state" before the measurement comes in. In **Gaussian Process Regression**, this is often set to zero or a constant mean function.

2. $\Sigma_{XX}$ (The Prior Covariance):
* **What it is:** This represents your uncertainty about $X$ before observing $Y$.
* **Role:** It defines the "search space." If $\Sigma_{XX}$ is very large, it means your prior is weak, and the model will rely more heavily on the new data $Y$. In terms of the **Law of Total Variance**, this is the "Total Variance" that you are trying to reduce.

3. $\Sigma_{XY}$ (The "Information Bridge"):
* **What it is:** The cross-covariance between the latent variable $X$ and the observable $Y$.
* **Role:** This is arguably the most important part of the formula. It defines how much $X$ and $Y$ are "entangled."
    * If $\Sigma_{XY} = 0$, the variables are independent, the formula collapses to $\mathbb{E}[X|Y] = \mu_X$, and the data $Y$ tells you nothing.
    * As $\Sigma_{XY}$ increases, the "bridge" becomes stronger, allowing more information to "leak" from $Y$ to $X$.

4. The "Update" Term: $\Sigma_{XY}\Sigma_{YY}^{-1}(Y - \mu_Y)$:
This whole block is the **Bayesian Update** (or the **Correction**):
* **$(Y - \mu_Y)$:** This is known as the **"Innovation"** or **"Residual."** It is the "surprise"—the difference between what you actually saw ($Y$) and what you expected to see ($\mu_Y$).
* **$\Sigma_{XY}\Sigma_{YY}^{-1}$:** This is the **"Kalman Gain"** or the **"Regression Coefficient."** It dictates exactly how much of that "surprise" should be applied to your prior mean to get your new, better estimate.

5. The Mean Square Error:
This is rhe **Conditional Covariance Matrix**, denoted as $\Sigma_{X|Y}$
$$MSE = \Sigma_{X|Y} = \Sigma_{XX} - \Sigma_{XY}\Sigma_{YY}^{-1}\Sigma_{YX}$$
Here
* **$\Sigma_{XX}$ (Prior Uncertainty):** This is the total variance/uncertainty you have about $X$ before you see any data. It represents the "starting" error.
* **$\Sigma_{XY}\Sigma_{YY}^{-1}\Sigma_{YX}$ (Explained Variance):** This is the amount of uncertainty that is removed or "explained" by observing $Y$.
    * $\Sigma_{XY}$ is the connection between the variables.
    * $\Sigma_{YY}^{-1}$ accounts for the noise/scale of the observations.
* **The Subtraction:** The MSE is what remains of your prior uncertainty after you subtract the information "leaked" by $Y$.
* **The "Optimality" Identity:**
The conditional expectation $\mathbb{E}[X|Y]$ is the unique estimator that achieves this minimum possible error. Any other prediction method will result in a larger MSE:
$$\mathbb{E}[(X - \text{estimate})^2] \ge \mathbb{E}[(X - \mathbb{E}[X|Y])^2]$$




# Applications of Conditional Expectation

## The Kalman Filter

Consider the linear Gaussian process:
\begin{align*}
x_k &= A_{k-1}\,x_{k-1} + G_{k-1}\,w_{k-1}, \\
y_k &= H_k\,x_k + z_k,
\end{align*}
where $w_k \sim \mathscr{N}(0,\Sigma_p)$ and $z_k \sim \mathscr{N}(0,\Sigma_m)$ are mutually independent white noise sequences, also independent of the initial state $x_0 \sim \mathscr{N}(m_0, P_0)$.
Here $y_k$ denotes the random variable representing the measurement at time step $k$, while $y_k^{\mathrm{obs}}$ denotes its observed numerical realization.

---

**Prediction Step (Time Update):**

Define the filter model
\begin{align*}
x^{-}_k &= A_{k-1}\,x^{+}_{k-1} + G_{k-1}\,w_{k-1}, \\
y^{-}_k &= H_k\,x^{-}_k + z_k,
\end{align*}
where
$$x^{+}_{k-1} \sim \mathscr{N}(m_{k-1}, P_{k-1}),$$


Because the filter equation is linear and the process noise is Gaussian, the predicted state $x^{-}_k$ is also Gaussian. Taking the expectation of the state equation yields the predicted mean:


$$m_k^- \triangleq \mathbb{E}[x^{-}_k] = A_{k-1}\mathbb{E}[x^{+}_{k-1}] + G_{k-1}\mathbb{E}[w_{k-1}] = A_{k-1}m_{k-1},$$


since $\mathbb{E}[w_{k-1}] = 0$.

The predicted covariance $P_k^-$ is computed by applying the variance operator to the state equation:
\begin{align*}
P_k^- &\triangleq \text{Var}(x^{-}_k) \\
&= A_{k-1}\text{Var}(x^{+}_{k-1})A_{k-1}^T + G_{k-1}\text{Var}(w_{k-1})G_{k-1}^T \\
&= A_{k-1}P_{k-1}A_{k-1}^T + G_{k-1}\Sigma_p G_{k-1}^T.
\end{align*}
Thus, prior to incorporating the new measurement, our belief of the state is characterized by the prior distribution:


$$x_k^- \sim \mathscr{N}(m_k^-, P_k^-).$$

---

**Measurement Prediction:**

Prior to its physical realization, the upcoming measurement at time step $k$ is treated as the random variable $y_k$. Based on our prior state belief, its expected value is:


$$\mathbb{E}[y^{-}_k] = H_k \mathbb{E}[x_k^-] + \mathbb{E}[z_k] = H_k m_k^-,$$


and its variance is:


$$\text{Var}(y^{-}_k) = H_k \text{Var}(x_k^-) H_k^T + \text{Var}(z_k) = H_k P_k^- H_k^T + \Sigma_m.$$

---

**Joint Prior Distribution:**

The cross-covariance between the predicted state $x_k^-$ and the anticipated measurement random variable $y_k$ is evaluated as:


$$\text{Cov}(x_k^-, y^{-}_k) = \text{Cov}(x_k^-, H_k x_k^- + z_k) = P_k^- H_k^T.$$

Therefore, the joint distribution of the predicted state and the upcoming measurement is a block-structured multivariate Gaussian:
$$\begin{bmatrix} x_k^- \\ y^{-}_k \end{bmatrix} \sim \mathscr{N}\left(
\begin{bmatrix} m_k^- \\ H_k m_k^- \end{bmatrix},
\begin{bmatrix} P_k^- & P_k^- H_k^T \\ H_k P_k^- & H_k P_k^- H_k^T + \Sigma_m \end{bmatrix}
\right).$$

---

**Measurement Update (Correction Step):**

At time step $k$, a physical measurement is observed, causing the random variable to take a specific numerical realization: $y_k = y^{\mathrm{obs}}_{k}$.

Applying the standard conditioning properties of multivariate normal distributions, we update our prior belief of the state by slicing the joint distribution at the realization $y^{\mathrm{obs}}_{k}$. This yields the posterior state distribution:


$$x^{+}_k\triangleq (x^{-}_k \mid y^{-}_k = y^{\mathrm{obs}}_{k}) \sim \mathscr{N}(m_k, P_k),$$


where the updated mean $m_k$ and updated covariance $P_k$ are given by:
\begin{align*}
K_k &\triangleq P_k^- H_k^T (H_k P_k^- H_k^T + \Sigma_m)^{-1}, \\
m_k &= m_k^- + K_k (y^{\mathrm{obs}}_{k} - H_k m_k^-), \\
P_k &= (I - K_k H_k) P_k^-.
\end{align*}

Here, $K_k$ is the Kalman Gain, and
$$ m_k=\mathbb{E}[x^{-}_k \mid y^{-}_k = y^{\mathrm{obs}}_{k}]$$ acts as the updated state estimate, and
$$P_k=\text{Var}(x^{-}_k \mid y^{-}_k = y^{\mathrm{obs}}_{k})$$ is the new posterior error covariance matrix used to seed the next recursive time-step.

Note that $x^{+}_k$ denotes an abstract random variable distributed according to the conditional law of $x^{-}_k$ given the observed value $y^{\mathrm{obs}}_{k}$.

## Gaussian Process Regression

Let ${G}$ be some set. Consider a $1$-dimensional Gaussian process $X_g\sim \mathscr{N}(\mu_g,\kappa(g,g))$ where $g\in G$ and $\kappa:G\times G \mapsto \mathbb{R}$ is the covariance (also known as the kernel)
$$\mathbb{E}\left((X_g-\mu_g)(X_{g'}-\mu_{g'})\right)=\kappa(g,g').$$

Let $Y_g$ be another stochastic process that satisfies
\begin{align}
Y_g&=X_g+\nu_g,
\end{align}
where $\nu_g\sim \mathscr{N}(\mu,\sigma_m^2)$.
Assume that $\nu_g\perp X_{g'}$ for all $g,g'\in G$ and that
$$
\mathrm{Cov}(\nu_g,\nu_{g'})= \sigma_m^2 \delta_{gg'}
$$

This represents a noisy sampling of $X_g$. If one samples the stochastic process $Y_g$ at different $g$ and have the set of observations ${y}_{n}\triangleq [y_{g_1},y_{g_2},\cdots,y_{g_n}]^T$, the Gaussian Process (GP) regression is the problem of estimating $X_g$ given the observations $y_n$.


Consider the multidimensional random variable
$\mathscr{Y}_n\triangleq[Y_{g_1},Y_{g_2},\cdots,Y_{g_n}]^T$ and $\mathscr{X}_n\triangleq[X_{g_1},X_{g_2},\cdots,X_{g_n}]^T$. Then
$$\mathscr{Y}_n\sim  \mathscr{N}(\mu_{\mathscr{Y}_n},K_n+\sigma_m^2I)$$
where, $\mu_{\mathscr{Y}_n}\triangleq[\mu_{g_1}+\mu,\mu_{g_2}+\mu,\cdots,\mu_{g_n}+\mu]^T$ and $K_n$ is the $n\times n$ covariance matrix
\begin{align*}
K_n&\triangleq \mathbb{E}\left((\mathscr{X}_n-\mu_{\mathscr{X}_n})(\mathscr{X}_n-\mu_{\mathscr{X}_n})^T\right)=
\begin{bmatrix}
\kappa(g_1,g_1) & \kappa(g_1,g_2) & \cdots & \kappa(g_1,g_n)\\
\vdots & \vdots & \ddots &\vdots\\
\kappa(g_n,g_1) & \kappa(g_n,g_2) & \cdots & \kappa(g_n,g_n)
\end{bmatrix},
\end{align*}
where
$\mu_{\mathscr{X}_n}\triangleq[\mu_{g_1},\mu_{g_2},\cdots,\mu_{g_n}]^T.$
Let
$$
k\triangleq
\begin{bmatrix}\kappa(g_1,g)&\kappa(g_2,g) & \cdots &\kappa(g_n,g)
\end{bmatrix}^T,
$$
then we have that
$$
\mathrm{Cov}(\mathscr{Y}_n,X_g)=k
$$
and hence that
\begin{align*}
\begin{bmatrix}
X_{g}\\ \mathscr{Y}_n
\end{bmatrix}
\sim\mathscr{N}\left(\begin{bmatrix}  \mu_{g} \\ \mu_{\mathscr{Y}_n}
\end{bmatrix},\begin{bmatrix} \kappa(g,g) & k^T \\ k & K_n+\sigma^2_m
I \end{bmatrix}\right).
\end{align*}

Thus the Gaussian Process (GP) regression problem boils down to finding $\mathbb{E}[X_{g}|\mathscr{Y}_n=y_n]$.


## Linear Regression

Consider a set of i.i.d. random variables
$$X_1,X_2,\cdots,X_n$$
where each $X_i$ is a $q$-dimensional random variable:
$$
X_i\sim \mathscr{N}(\mu_X, \Sigma_X).
$$
The $m$-dimensional random variable $Y_i$ is defined by
$$
Y_i=\beta X_i + \beta_0 + \nu_i
$$
where $\beta$ is a $m\times q$ constant matrix, $\beta_0$ is $m\times 1$ constant matrix and $\nu_i\sim \mathscr{N}(0,\Sigma_\nu)$.

Assuming that $X_i \perp \nu_i$ we have
$$
Y_i
\sim
\mathscr N
\Big(
\beta \mu_X+\beta_0,
\beta \Sigma_X\beta^T+\Sigma_\nu
\Big).
$$

The joint distribution of ${X}_i$ and $Y_i$ is

$$
\begin{bmatrix}
Y_i\\ X_i
\end{bmatrix}
\sim
\mathscr N
\Big(
\begin{bmatrix}
\beta \mu_X+\beta_0\\
\mu_X
\end{bmatrix},
\begin{bmatrix}
\beta \Sigma_X\beta^T+\Sigma_\nu & \beta \Sigma_X\\
\Sigma_X\beta^T & \Sigma_X
\end{bmatrix}
\Big).
$$


From the above we have we easily see that

$$
\mathbb E[Y_i\mid X_i=x_i]
=
\beta x_i+\beta_0,
$$
and
$$
\operatorname{Var}(Y_i\mid X_i=x_i)
=
\Sigma_\nu.
$$
From the above formulation we see that
$$
Y_i\mid X_i=x_i
\sim
\mathscr N(\beta x_i+\beta_0,\Sigma_\nu).
$$

# Data as realizations of random variables

## The data generation process

Let each row of an industrial dataset be modeled as a $m$-dimensional matrix valued **random variable:**  
$$\mathbf{X}_i = [X_{i,1}, X_{i,2}, \dots, X_{i,m}]^T,$$ that is governed by an underlying, unknown multivariate probability distribution. Each individual random variable $\mathbf{X}_i$ is characterized by its first two theoretical moments:

* **The Population Mean:** $\boldsymbol{\mu}_i = \mathbb{E}[\mathbf{X}_i] \in \mathbb{R}^{m}$
* **The Population Covariance Matrix:** $\boldsymbol{\Sigma}_i = \mathbb{E}[(\mathbf{X}_i - \boldsymbol{\mu}_i)(\mathbf{X}_i - \boldsymbol{\mu}_i)^T] \in \mathbb{R}^{m \times m}$

When collecting a dataset of size $n$, we observe a finite sequence of these random matrices, denoted as $\{\mathbf{X}_i\}_{i=1}^n$.
The actual dataset processed by a computer is the matrix of empirical **realizations**—the concrete numerical values extracted from the system.

---

Formally, we define the column matrix
$$\mathbf{x}_i = [x_{i,1}, x_{i,2}, \dots, x_{i,m}]^T \in \mathbb{R}^{m}$$
as the observed realization of the random variable $\mathbf{X}_i$. Mapping the $j$-th feature of the $i$-th sample as the scalar $x_{i,j}$ yields the structured $n \times m$ observed data matrix $\mathbf{X}_{\text{obs}}$:

$$\mathbf{X}_{\text{obs}} = \begin{bmatrix}
x_{1,1} & x_{1,2} & \dots & x_{1,m} \\
x_{2,1} & x_{2,2} & \dots & x_{2,m} \\
\vdots & \vdots & \ddots & \vdots \\
x_{n,1} & x_{n,2} & \dots & x_{n,m}
\end{bmatrix} = \begin{bmatrix} \mathbf{x}_1^T \\ \mathbf{x}_2^T \\ \vdots \\ \mathbf{x}_n^T \end{bmatrix}.$$

---

To justify the application of classical multivariate statistical inference, we impose the foundational assumption that the sequence of random varaibles $\{X_i\}_{i=1}^n$ is **independent and identically distributed (i.i.d.)**.

1. **Identical Distribution:** Every random variable $\mathbf{X}_i$ possesses the exact same probability law. Consequently, the distinct population moments collapse into a single, universal system profile:

$$\boldsymbol{\mu}_i = \boldsymbol{\mu} \quad \text{and} \quad \boldsymbol{\Sigma}_i = \boldsymbol{\Sigma} \quad \forall \ i \in \{1, 2, \dots, n\}$$

2. **Mutual Independence:** The joint probability distribution of the sequence collapses into the product of their individual distributions, written as $\mathbf{X}_1 \perp \mathbf{X}_2 \perp \dots \perp \mathbf{X}_n$. This guarantees that no combination of rows carries hidden, collective information about any other row, meaning the physical state captured by any subset of observations exerts zero statistical influence over the rest.

---

Mutual Independence implies that the joint Probability Density Function (PDF), denoted by $f_{\mathbf{X}_1, \mathbf{X}_2, \dots, \mathbf{X}_n}$ satisfies:

$$f_{\mathbf{X}_1, \mathbf{X}_2, \dots, \mathbf{X}_n}(\mathbf{x}_1, \mathbf{x}_2, \dots, \mathbf{x}_n) = \prod_{i=1}^{n} f_{\mathbf{X}}(\mathbf{x}_i) \quad \forall \mathbf{x}_i \in \mathbb{R}^m$$

where $f_{\mathbf{X}}(\mathbf{x}_i)$ represents the identical marginal PDF evaluated at the $i$-th realization.

## Empirical Mean, Covariance and the Pearson Correlation

The first and second empirical moments calculated directly from the observed realizations of $\mathbf{X}_i \in \mathbb{R}^{m}$ are:

* **The Empirical Mean ($\widehat{\boldsymbol{\mu}}_n \in \mathbb{R}^{m \times 1}$):** The sample average mapping the spatial center of gravity of the observed feature space.

$$\widehat{\boldsymbol{\mu}}_n = \frac{1}{n} \sum_{i=1}^n \mathbf{X}_i $$


* **The Empirical Covariance Matrix ($\mathbf{S} \in \mathbb{R}^{m \times m}$):** Utilizing Bessel's correction ($n-1$) to ensure an unbiased estimator ($\mathbb{E}[\mathbf{S}] = \boldsymbol{\Sigma}$) is
$$\mathbf{S} = \frac{1}{n-1} \sum_{i=1}^n (\mathbf{X}_i - \widehat{\boldsymbol{\mu}}_n)(\mathbf{X}_i - \widehat{\boldsymbol{\mu}}_n)^T.$$

* **Pearson Correlation Matrix ($\mathbf{R}$):** Standardizes the sample covariance matrix, $\mathbf{S}$, between $[-1, 1]$.
$$\mathbf{R} = \mathbf{D}^{-1/2} \mathbf{S} \mathbf{D}^{-1/2}$$
where
$$\mathbf{D} = \text{diag}(\mathbf{S}) = \begin{bmatrix} s_1^2 & 0 & \dots & 0 \\ 0 & s_2^2 & \dots & 0 \\ \vdots & \vdots & \ddots & \vdots \\ 0 & 0 & \dots & s_m^2 \end{bmatrix}.$$
Explicitly writtent down the $jk$ matrix element of $\mathbf{R}$ is
$$r_{jk} = \frac{s_{jk}}{s_j \cdot s_k} = \frac{\sum_{i=1}^n (X_{i,j} - \widehat{\mu}_j)(X_{i,k} - \widehat{\mu}_k)}{\sqrt{\sum_{i=1}^n (X_{i,j} - \widehat{\mu}_j)^2} \sqrt{\sum_{i=1}^n (X_{i,k} - \widehat{\mu}_k)^2}}.$$



## The Asymptotic Consequence: The Multivariate Central Limit Theorem

The **Lindeberg-Lévy Multivariate Central Limit Theorem (CLT)**.


$$\sqrt{n}\left(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu}\right) \xrightarrow{d} \mathscr{N}\left(\mathbf{0}, \boldsymbol{\Sigma}\right)$$

and thus

$$\widehat{\boldsymbol{\mu}}_n \sim \mathscr{N}\left(\boldsymbol{\mu}, \frac{1}{n}\boldsymbol{\Sigma}\right)$$

---

In practical applications, the true population parameters $\boldsymbol{\mu}$ and $\boldsymbol{\Sigma}$ remain latent and unobservable. To operationalize the CLT result, we invoke **Slutsky's Theorem**:
$$
\widehat{\boldsymbol{\mu}}_n
\approx
\mathscr{N}\left(
\boldsymbol{\mu},
\frac{1}{n}\mathbf{S}
\right).
$$

### Maximum Likelyhood Estimation

We can also approximate the latent population distribution $\mathscr{N}(\boldsymbol{\mu}, \boldsymbol{\Sigma})$ by finding the parameters that maximize the joint probability of observing our collected dataset. Assuming the realizations $\mathbf{x}_i$ are mutually independent, the joint log-likelihood function is defined as:

$$\ell(\boldsymbol{\mu}, \boldsymbol{\Sigma} \;|\; \mathbf{X}_{\text{obs}}) = -\frac{nm}{2}\ln(2\pi) - \frac{n}{2}\ln|\boldsymbol{\Sigma}| - \frac{1}{2}\sum_{i=1}^n (\mathbf{x}_i - \boldsymbol{\mu})^T \boldsymbol{\Sigma}^{-1} (\mathbf{x}_i - \boldsymbol{\mu})$$

Maximizing this function by taking its vector and matrix derivatives with respect to $\boldsymbol{\mu}$ and $\boldsymbol{\Sigma}^{-1}$ yields the unique, mathematically optimal Maximum Likelihood Estimators (MLE):

$$\widehat{\boldsymbol{\mu}}_{\text{MLE}} = \frac{1}{n} \sum_{i=1}^n \mathbf{x}_i = \widehat{\boldsymbol{\mu}}_n$$

$$\widehat{\boldsymbol{\Sigma}}_{\text{MLE}} = \frac{1}{n} \sum_{i=1}^n (\mathbf{x}_i - \widehat{\boldsymbol{\mu}}_n)(\mathbf{x}_i - \widehat{\boldsymbol{\mu}}_n)^T$$

While $\widehat{\boldsymbol{\mu}}_{\text{MLE}}$ is perfectly identical to our empirical mean vector $\widehat{\boldsymbol{\mu}}_n$, the pure maximum likelihood matrix $\widehat{\boldsymbol{\Sigma}}_{\text{MLE}}$ is inherently biased in finite samples ($\mathbb{E}[\widehat{\boldsymbol{\Sigma}}_{\text{MLE}}] = \frac{n-1}{n}\boldsymbol{\Sigma}$) because it measures variations against the calculated sample mean rather than the true population mean.

To resolve this systematic underestimation, we apply Bessel's correction by scaling the MLE result by $\frac{n}{n-1}$. This substitutes the biased MLE matrix with the unbiased empirical sample covariance matrix $\mathbf{S}$:

$$\mathbf{S} = \frac{n}{n-1}\widehat{\boldsymbol{\Sigma}}_{\text{MLE}} = \frac{1}{n-1} \sum_{i=1}^n (\mathbf{x}_i - \widehat{\boldsymbol{\mu}}_n)(\mathbf{x}_i - \widehat{\boldsymbol{\mu}}_n)^T$$

Consequently, the operational micro-scale model
$$\mathbf{X}_i \sim \mathscr{N}(\widehat{\boldsymbol{\mu}}_n, \mathbf{S})$$
is fundamentally an **unbiased Maximum Likelihood framework**. This connection guarantees that the geometric boundary established for real-time anomaly detection layer is both asymptotically optimal under likelihood theory and mathematically guarded against small-sample estimation bias.

# Principal Component Analysis (PCA)

To exploit the structure of the empirical covariance matrix for structural health monitoring, we map the raw random variables into an orthogonal, linearly uncorrelated coordinate space. This is achieved via Principal Component Analysis (PCA) derived directly from the population moments.

Consider a zero-mean random varaible $\widetilde{\mathbf{X}}_i = (\mathbf{X}_i - \boldsymbol{\mu}) \in \mathbb{R}^{m}$, representing the centered baseline features. Its population covariance matrix $\boldsymbol{\Sigma} \in \mathbb{R}^{m \times m}$ is symmetric and positive-semidefinite:

$$\boldsymbol{\Sigma} = \mathbb{E}[\widetilde{\mathbf{X}}_i \widetilde{\mathbf{X}}_i^T]$$

By the Spectral Theorem (or the Singular Value Decomposition of a symmetric matrix), $\boldsymbol{\Sigma}$ can be factored into its orthogonal eigenvectors and real, non-negative eigenvalues:

$$\boldsymbol{\Sigma} = \mathbf{P} \mathbf{\Lambda} \mathbf{P}^T$$

Where:

* $\mathbf{P} = [\mathbf{p}_1, \mathbf{p}_2, \dots, \mathbf{p}_m] \in \mathbb{R}^{m \times m}$ is an orthogonal matrix ($\mathbf{P}^T\mathbf{P} = \mathbf{P}\mathbf{P}^T = \mathbf{I}$) containing the principal component loading vectors.
* $\mathbf{\Lambda} = \text{diag}(\lambda_1, \lambda_2, \dots, \lambda_m) \in \mathbb{R}^{m \times m}$ is a diagonal matrix containing sorted eigenvalues ($\lambda_1 \ge \lambda_2 \ge \dots \ge \lambda_m \ge 0$), representing the population variance along each principal axis.

---

We define the transformed random varable $\mathbf{Z}_i \in \mathbb{R}^{m \times 1}$ as the projection of the centered observations onto the orthogonal eigenvector basis:

$$\mathbf{Z}_i \triangleq \mathbf{P}^T \widetilde{\mathbf{X}}_i$$

We see that:

$$\mathbb{E}[\mathbf{Z}_i \mathbf{Z}_i^T] = \mathbf{\Lambda}$$

Because $\mathbf{\Lambda}$ is strictly diagonal, the cross-covariances between distinct principal components are exactly zero ($\mathbb{E}[Z_{i,j}Z_{i,k}] = 0$ for all $j \neq k$).

---

In practical applications, the latent matrix $\boldsymbol{\Sigma}$ is unknown. We operationalize PCA by substituting the consistent, unbiased empirical sample covariance matrix $\mathbf{S}$ derived via our Maximum Likelihood framework:

$$\mathbf{S} = \widehat{\mathbf{P}} \widehat{\mathbf{\Lambda}} \widehat{\mathbf{P}}^T$$

The observed realizations are centered using the empirical mean vector $\widehat{\boldsymbol{\mu}}_n$ and projected into the empirical principal space to obtain sample scores $\mathbf{z}_i$:

$$\mathbf{z}_i = \widehat{\mathbf{P}}^T (\mathbf{x}_i - \widehat{\boldsymbol{\mu}}_n).$$
Or equivalently
$$\mathbf{x}_i =\widehat{\boldsymbol{\mu}}_n + \widehat{\mathbf{P}}\mathbf{z}_i .$$

This operational mapping guarantees that the sample covariance of the transformed data matrix collapses to the diagonal matrix $\widehat{\boldsymbol{\Lambda}}$, establishing a variance-optimal orthogonal coordinate system for isolating linearly uncorrelated diagnostic modes in the diagnostics layer.

---

To compress the feature space from $m$ dimensions to a lower dimension $k$ (where $k < m$), we partition the empirical orthogonal basis matrix $\widehat{\mathbf{P}}$ into two orthogonal subspaces:$$\widehat{\mathbf{P}} = \begin{bmatrix} \widehat{\mathbf{P}}_k & \Big| & \widehat{\mathbf{P}}_{m-k} \end{bmatrix}$$where $\widehat{\mathbf{P}}_k \in \mathbb{R}^{m \times k}$ contains the first $k$ column eigenvectors corresponding to the largest $k$ eigenvalues ($\widehat{\lambda}_1 \ge \dots \ge \widehat{\lambda}_k$). This partitioning allows us to partition $\mathbf{z}\in \mathbb{R}^m$ as
$$\mathbf{z}_{i}=\begin{bmatrix}\mathbf{z}_{i,k} \\\mathbf{z}_{i,m-k}
\end{bmatrix}$$
where $\mathbf{z}_{i,k} \in \mathbb{R}^{k}$ and $\mathbf{z}_{i,m-k} \in \mathbb{R}^{m-k}.$
From
$$\mathbf{z}_i = \widehat{\mathbf{P}}^T (\mathbf{x}_i - \widehat{\boldsymbol{\mu}}_n)$$
we thus have
$$(\mathbf{x}_i - \widehat{\boldsymbol{\mu}}_n)=\widehat{\mathbf{P}}_k \mathbf{z}_{i,k} + \widehat{\mathbf{P}}_{m-k} \mathbf{z}_{i,m-k}.$$

The partitioning of the empirical basis matrix allows us to isolate the deterministic structural dynamics from the unmodeled residual noise. By retaining only the first $k$ principal components, we define the reconstructed feature vector $\widehat{\mathbf{x}}_i \in \mathbb{R}^m$ as the projection of the low-dimensional scores back into the original measurement space:

$$\widehat{\mathbf{x}}_i \triangleq \widehat{\boldsymbol{\mu}}_n + \widehat{\mathbf{P}}_k \mathbf{z}_{i,k}$$

The geometric displacement between the actual observation $\mathbf{x}_i$ and its lower-dimensional reconstruction $\widehat{\mathbf{x}}_i$ defines the residual error vector $\mathbf{e}_i \in \mathbb{R}^m$:

$$\mathbf{e}_i = \mathbf{x}_i - \widehat{\mathbf{x}}_i = \widehat{\mathbf{P}}_{m-k} \mathbf{z}_{i,m-k}$$

Because $\widehat{\mathbf{P}}_k$ and $\widehat{\mathbf{P}}_{m-k}$ span mutually orthogonal subspaces, the total energy of the centered observation vector can be cleanly decomposed via the Pythagorean identity:

$$\|\mathbf{x}_i - \widehat{\boldsymbol{\mu}}_n\|^2 = \|\widehat{\mathbf{P}}_k \mathbf{z}_{i,k}\|^2 + \|\widehat{\mathbf{P}}_{m-k} \mathbf{z}_{i,m-k}\|^2 = \|\mathbf{z}_{i,k}\|^2 + \|\mathbf{z}_{i,m-k}\|^2$$


# Factor Analysis

To isolate the underlying physical phenomena driving the system's structural state and decouple variable-specific measurement noise, we map the observed features into a lower-dimensional latent space. While Principal Component Analysis (PCA) merely rotates the empirical coordinate system to maximize total variance, **Factor Analysis (FA)** models the internal correlation structure by partitioning the universal population variance into shared (structural) and independent (stochastic) subspaces.

Let $\mathbf{Z}_i \in \mathbb{R}^m$ be the normalized, zero-mean random variable derived from centering and scaling the $i$-th baseline feature vector $\mathbf{X}_i$:

$$\mathbf{Z}_i = \mathbf{D}^{-1/2}(\mathbf{X}_i - \boldsymbol{\mu})$$

where $\boldsymbol{\mu}$ is the universal population mean and $\mathbf{D} = \text{diag}(\sigma_1^2, \sigma_2^2, \dots, \sigma_m^2)$ is the diagonal matrix containing the population variances of $\mathbf{X}_i$. By construction, the theoretical population covariance matrix of $\mathbf{Z}_i$ collapses to the universal population correlation matrix, $\mathbf{R}$:

$$\mathbf{R} = \mathbb{E}[\mathbf{Z}_i \mathbf{Z}_i^T] \in \mathbb{R}^{m \times m}, \quad \text{with} \quad \mathbb{E}[Z_{i,j}^2] = 1.0 \quad \forall \ j \in \{1, \dots, m\}.$$

## The Latent Factor Postulate

We assume that the $m$ visible features in $\mathbf{Z}_i$ are linearly generated by a heavily compressed subset of $k$ unobservable (latent) physical mechanisms or operational conditions, where $k < m$. Formally, the generative model for the $i$-th realization vector is defined as:

$$\mathbf{Z}_i = \boldsymbol{\Lambda} \mathbf{F}_i + \boldsymbol{\epsilon}_i$$

Where:

* $\mathbf{F}_i = [F_{i,1}, F_{i,2}, \dots, F_{i,k}]^T \in \mathbb{R}^k$ is the random variable containing the **common factors** that govern the global physical state of the structure.
* $\boldsymbol{\Lambda} \in \mathbb{R}^{m \times k}$ is the invariant **factor loading matrix**, where the scalar $\lambda_{j,r}$ quantifies the linear sensitivity of the $j$-th feature to the $r$-th latent structural mode.
* $\boldsymbol{\epsilon}_i = [\epsilon_{i,1}, \epsilon_{i,2}, \dots, \epsilon_{i,m}]^T \in \mathbb{R}^m$ is the random variable representing the **specific errors** or **uniquenesses** unique to each measurement channel.

To ensure mathematical tractability and structurally isolate global physical effects from local disturbances, we impose three orthogonal population constraints:

1. **Orthogonal Common Factors:** The latent structural modes are mutually de-correlated and normalized to unit variance:

$$\mathbb{E}[\mathbf{F}_i] = \mathbf{0}, \quad \mathbb{E}[\mathbf{F}_i \mathbf{F}_i^T] = \mathbf{I}_{k \times k}$$


2. **De-correlated Unique Errors:** The measurement noise or local variance of any given feature is statistically independent of all other features. Consequently, the unique error covariance matrix $\boldsymbol{\Psi}$ must be strictly diagonal:

$$\mathbb{E}[\boldsymbol{\epsilon}_i] = \mathbf{0}, \quad \mathbb{E}[\boldsymbol{\epsilon}_i \boldsymbol{\epsilon}_i^T] = \boldsymbol{\Psi} = \text{diag}(\varphi_1^2, \varphi_2^2, \dots, \varphi_m^2)$$


3. **Subspace Orthogonality:** The global physical mechanics driving the common factors exert zero statistical influence over the local, channel-specific unique errors:

$$\mathbb{E}[\boldsymbol{\epsilon}_i \mathbf{F}_i^T] = \mathbf{0}_{m \times k}.$$

Utilizing the orthogonality properties ($\mathbf{I}$, $\mathbf{0}$, and $\boldsymbol{\Psi}$), the expression cleanly reduces to the **Fundamental Equation of Factor Analysis**:

$$\mathbf{R} = \boldsymbol{\Lambda} \boldsymbol{\Lambda}^T + \boldsymbol{\Psi}.$$

Thus factor analysis reduces to a marix factorization problem.

### Communality vs. Uniqueness

Focusing on the $j$-th diagonal entry of the correlation matrix $\mathbf{R}$, the total unit variance of each normalized variable $Z_{i,j}$ splits into two distinct, interpretable physical spaces:

$$\mathbb{E}[Z_{i,j}^2] = \sum_{r=1}^k \lambda_{j,r}^2 + \varphi_j^2 = 1.0$$

We define these partitioned elements as:

* **Communality ($h_j^2$):** The portion of a sensor's variance explained collectively by the $k$ shared structural factors:

$$h_j^2 \triangleq \sum_{r=1}^k \lambda_{j,r}^2$$


* **Uniqueness ($\varphi_j^2$):** The remaining portion of variance completely isolated to that specific sensor, representing a combination of localized non-structural anomalies and electronic white noise:

$$\varphi_j^2 \triangleq 1.0 - h_j^2 = \mathbb{E}[\epsilon_{i,j}^2]$$



### Estimating the Loading Matrix $\boldsymbol{\Lambda}$ via Maximum Likelihood Optimization:

Given our underlying generative model where $\mathbf{Z}_i \sim \mathscr{N}(\mathbf{0}, \boldsymbol{\Sigma})$ with the structured covariance $\boldsymbol{\Sigma} = \boldsymbol{\Lambda}\boldsymbol{\Lambda}^T + \boldsymbol{\Psi}$, the objective function is derived by minimizing the structural discrepancy between the fitted model covariance ($\boldsymbol{\Sigma}$) and the empirical correlation matrix ($\mathbf{R}$).

The log-likelihood objective function to be minimized with respect to $\boldsymbol{\Lambda}$ and $\boldsymbol{\Psi}$ is formulated as:

$$\mathscr{L}(\boldsymbol{\Lambda}, \boldsymbol{\Psi}) = \ln|\boldsymbol{\Lambda}\boldsymbol{\Lambda}^T + \boldsymbol{\Psi}| + \text{Tr}\left[ \mathbf{R} (\boldsymbol{\Lambda}\boldsymbol{\Lambda}^T + \boldsymbol{\Psi})^{-1} \right] - \ln|\mathbf{R}| - m$$

### Estimation of the Factors via Conditional Expectation

1. Joint Subspace Covariance Construction

To find the conditional expectation $\mathbb{E}[\mathbf{F}_i | \mathbf{z}_i]$, we first examine how the observable normalized variables $\mathbf{Z}_i$ and the hidden common factors $\mathbf{F}_i$ vary together. We construct a partitioned joint random vector $\mathbf{Y}_i$:

$$\mathbf{Y}_i = \begin{bmatrix} \mathbf{F}_i \\ \mathbf{Z}_i \end{bmatrix}$$

Because both $\mathbf{Z}_i$ and $\mathbf{F}_i$ have a mean of zero ($\mathbb{E}[\mathbf{Z}_i] = \mathbf{0}$ and $\mathbb{E}[\mathbf{F}_i] = \mathbf{0}$), the joint mean is $\mathbb{E}[\mathbf{Y}_i] = \mathbf{0}$.

Now, let's construct the partitioned joint covariance matrix $\boldsymbol{\Sigma}_{YY}$:

$$\boldsymbol{\Sigma}_{YY} = \mathbb{E}[\mathbf{Y}_i \mathbf{Y}_i^T] = \begin{bmatrix} \mathbb{E}[\mathbf{F}_i \mathbf{F}_i^T] & \mathbb{E}[\mathbf{F}_i \mathbf{Z}_i^T] \\ \mathbb{E}[\mathbf{Z}_i \mathbf{F}_i^T] & \mathbb{E}[\mathbf{Z}_i \mathbf{Z}_i^T] \end{bmatrix}$$

From the foundations of Factor Analysis that was established before, we know the values of these individual blocks:

1. $\mathbb{E}[\mathbf{Z}_i \mathbf{Z}_i^T] = \mathbf{R}$ (The visible correlation matrix)
2. $\mathbb{E}[\mathbf{Z}_i \mathbf{F}_i^T] = \boldsymbol{\Lambda}$ (The factor loading matrix, which we proved earlier)
3. $\mathbb{E}[\mathbf{F}_i \mathbf{Z}_i^T] = (\mathbb{E}[\mathbf{Z}_i \mathbf{F}_i^T])^T = \boldsymbol{\Lambda}^T$
4. $\mathbb{E}[\mathbf{F}_i \mathbf{F}_i^T] = \mathbf{I}_{k \times k}$ (The orthogonal common factor constraint)

Substituting these blocks into our partitioned matrix yields:

$$\boldsymbol{\Sigma}_{YY} = \begin{bmatrix} \mathbf{I} & \boldsymbol{\Lambda}^T \\ \boldsymbol{\Lambda} & \mathbf{R} \end{bmatrix}$$

---

2. Applying the Conditional Expectation Theorem

Under the standard assumption of a Multivariate Normal distribution for the statistical inference layer, the conditional distribution of one partitioned segment given the other is well-defined.

For any joint normal system where $\boldsymbol{\Sigma} = \begin{bmatrix} \boldsymbol{\Sigma}_{11} & \boldsymbol{\Sigma}_{12} \\ \boldsymbol{\Sigma}_{21} & \boldsymbol{\Sigma}_{22} \end{bmatrix}$, the conditional expectation of the second slot given a realization of the first slot ($\mathbf{x}_1$) is given by the classical linear regression projection formula:

$$\mathbb{E}[\mathbf{X}_1 | \mathbf{X}_2] = \boldsymbol{\mu}_1 + \boldsymbol{\Sigma}_{12} \boldsymbol{\Sigma}_{22}^{-1} (\mathbf{X}_2 - \boldsymbol{\mu}_2)$$

Mapping this directly to our factor analysis matrix partitions:

* Slot 1 is our hidden latent factor: $\mathbf{X}_1 = \mathbf{F}_i$ (with $\boldsymbol{\mu}_1 = \mathbf{0}$)
* Slot 2 is our observed sample realization: $\mathbf{X}_2 = \mathbf{Z}_i$ (with $\boldsymbol{\mu}_2 = \mathbf{0}$)
* The cross-covariance section $\boldsymbol{\Sigma}_{12} = \boldsymbol{\Lambda}^T$
* The observed covariance section $\boldsymbol{\Sigma}_{22} = \mathbf{R}$

Substituting these variables into the projection formula yields the factor score realization $\mathbf{f}_i$:

$$\mathbf{f}_i = \mathbb{E}[\mathbf{F}_i | \mathbf{Z}_i] = \mathbf{0} + \boldsymbol{\Lambda}^T \mathbf{R}^{-1} (\mathbf{Z}_i - \mathbf{0})$$

$$\mathbf{f}_i = \boldsymbol{\Lambda}^T \mathbf{R}^{-1} \mathbf{Z}_i$$


Because this estimator is derived via the conditional expectation, it possesses a vital property: it is the **Best Linear Predictor**.

# Hypothesis Testing

## Overview


Hypothesis testing operates on a philosophy similar to "innocent until proven guilty." We **start by assuming a baseline state of no effect** and **see if our data provides enough evidence to reject that assumption**.

* **The Null Hypothesis ($H_0$):** The status quo. It states that there is no effect, no difference, or no relationship. For example, $\mu_1 = \mu_2$ (the means of two groups are equal).
* **The Alternative Hypothesis ($H_a$ or $H_1$):** The claim you are actually trying to prove. It states that there *is* an effect or difference (e.g., $\mu_1 \neq \mu_2$).



## The Measure Theoretic Definition

Let $(\mathscr X,\mathfrak X)$ be a measurable sample space. A statistical
model is a family of probability measures
$$
\{P_\theta:\theta\in\Theta\}
$$
on $(\mathscr X,\mathfrak X)$, where $\Theta$ is the parameter space.
For each fixed $\theta\in\Theta$, the triple
$$
(\mathscr X,\mathfrak X,P_\theta)
$$
is a probability space. The statistical model is fixed by the data generation process.

---

Suppose that the parameter space $\Theta$ contains two disjoint subsets
$\Theta_0$ and $\Theta_1$, where
$$
\Theta_0,\Theta_1\subseteq \Theta,
\qquad
\Theta_0\cap \Theta_1=\varnothing.
$$

The null and alternative hypotheses are defined by
$$
H_0:\theta\in\Theta_0,
\qquad
H_1:\theta\in\Theta_1.
$$

---

A **hypothesis test** is determined by a measurable rejection region
$R\in\mathfrak X$. Let
$$
\varphi(x)=\mathbf 1_R(x).
$$

After observing data $X=x$, we reject $H_0$ precisely when

$$
x\in R.
$$
Thus the hypothesis test is
$$
\varphi(x)=
\begin{cases}
1, & \text{reject }H_0,\\
0, & \text{do not reject }H_0.
\end{cases}
$$

---

The **power function** of the test is

$$
\pi(\theta)
=
P_\theta(X\in R)
=
\mathbb E_\theta[\varphi(X)].
$$
It gives the probability that the test rejects $H_0$ when the true parameter is $\theta$.

---

The **size** of the test is defined as

$$
\sup_{\theta\in\Theta_0}\pi(\theta)=\sup_{\theta\in\Theta_0}P_\theta(X\in R).
$$

A test is said to have **level** $\alpha$ if

$$
\sup_{\theta\in\Theta_0}P_\theta(X\in R)\leq \alpha.
$$



---

The Type I error probability is defined as $\alpha_I(\theta)=\pi(\theta)$ for $\theta\in \Theta_0$. That is

$$
\alpha_{I}(\theta)=\pi(\theta)=
P_\theta(X\in R)
=
P_\theta(\text{reject }H_0)
\qquad \mathrm{for}\quad \theta \in \Theta_0.
$$
This is the probability that the test rejects $H_0$ when $H_0$ is true  (i.e. $\theta\in \Theta_0$).


The size of the test is the worst-case Type I error probability over the null parameter set:

$$
\sup_{\theta\in\Theta_0}\alpha_{I}(\theta)
=
\sup_{\theta\in\Theta_0}P_\theta(X\in R).
$$

---



The Type II error probability is

$$
\beta(\theta)
=
P_\theta(X\notin R)
=
1-\pi(\theta)\qquad \mathrm{for} \quad\theta\in\Theta_1.
$$
This is the probability that the test fails to reject $H_0$ when $H_0$ is false (i.e. $\theta\in \Theta_1$).

---

In summary, a hypothesis test is formally a measurable decision rule

$$
\varphi:\mathscr X\to\{0,1\}
$$

that determines, from the observed data, whether or not to reject the null hypothesis $H_0$.



## The Decision Rule


Suppose that a hypothesis test is given by a nested family of rejection regions

$$
\{R_\gamma:\gamma\in(0,1)\},
$$

where each $R_\gamma\in\mathfrak X$, and where

$$
\gamma_1\leq \gamma_2
\qquad\Longrightarrow\qquad
R_{\gamma_1}\subseteq R_{\gamma_2}.
$$

The corresponding level-$\gamma$ test is

$$
\varphi_\gamma(x)=\mathbf 1_{R_\gamma}(x).
$$

After observing data $X=x$, the $p$-value is is calibrated under the null family $\{P_\theta | \theta \in \Theta_0\}$, as

$$
p(x)
=
\inf\{\gamma\in(0,1):x\in R_\gamma\}.
$$

Equivalently,

$$
p(x)
=
\inf\{\gamma\in(0,1):\varphi_\gamma(x)=1\}.
$$

Thus, the $p$-value is the smallest significance level $\gamma$ at which the observed data would lead to rejection of $H_0$.


A valid $p$-value satisfies

$$
P_\theta(p(X)\leq \gamma)\leq \gamma
\qquad
\text{for every } \theta\in\Theta_0.
$$

Therefore for a pre-determinded threshold significant level $\alpha$:

* If $p\text{-value} \leq \alpha$: We **reject $H_0$**. The observed data are sufficiently unlikely under the null model, according to the chosen test statistic.
* If $p\text{-value} > \alpha$: We **fail to reject $H_0$**. We do *not* "accept" the null; we simply state that we lack sufficient evidence to disprove it.

Hence the decision rule can be written as

$$
\text{Reject }H_0
\qquad\Longleftrightarrow\qquad
p(x)\leq \alpha.
$$

---

Typically, one does not start by specifying the rejection region $R$ directly. Instead, one first chooses a test statistic

$$
T:\mathscr X\to\mathbb R,
$$

where larger values of $T(X)$ provide stronger evidence against $H_0$. For a pre-determined significance level $\alpha$, the rejection region is then induced by the statistic as

$$
R_\alpha
=\{x\in\mathscr X:T(x)\geq c_\alpha\},
$$

where the critical value $c_\alpha$ is chosen so that

$$
\sup_{\theta\in\Theta_0}P_\theta(T(X)\geq c_\alpha)\leq \alpha.
$$

Equivalently,

$$
R_\alpha
=
T^{-1}\bigl([c_\alpha,\infty)\bigr).
$$

Thus, the statistic $T(X)$, together with the critical value $c_\alpha$, determines the rejection region (R_\alpha).

For a simple null hypothesis $\Theta_0={\theta_0}$, the corresponding $p$-value is

$$
p(x)=P_{\theta_0}\left(T(X)\geq T(x)\right).
$$

For a composite null hypothesis, a conservative valid (p)-value is

$$
p(x)=\sup_{\theta\in\Theta_0}
P_\theta\left(T(X)\geq T(x)\right).
$$




## The Standard Roadmap

Every hypothesis test follows these exact five steps:

1. **State the hypotheses** ($H_0$ and $H_a$).
2. **Formulate an analysis plan** (choose the right test—like a t-test, Chi-Square, or ANOVA—based on your data type and assumptions).
3. **Analyze sample data** (calculate the test statistic and $p$-value).
4. **Apply the decision rule** (compare the $p$-value to $\alpha$).
5. **Interpret the results** in the context of your original real-world problem.


## A/B Testing

A/B testing is simply the tech industry's and business world's name for a two-sample hypothesis test.

**A/B testing** is simply the tech industry’s and business world's name for a **two-sample hypothesis test**.

It is a controlled experiment where you split your audience into two groups to compare two versions of something (Version A and Version B) to see which one performs better based on a specific metric.

While academic statisticians talk about "treatment vs. control groups," data scientists and product managers talk about "A/B testing."

---

**Example**

Imagine you run an e-commerce website, and you want to know if changing your "Buy Now" button from **Blue** to **Green** will increase sales.

* **Group A (The Control):** $50\%$ of your website visitors see the original **Blue** button.
* **Group B (The Variation):** The other $50\%$ of visitors see the new **Green** button.

You let the experiment run for a few weeks, track how many people click the button in each group, and then use your hypothesis testing machinery to analyze the results.


# Entropy

## Overview

Entropy began in thermodynamics as a measure of unavailable or wasted energy, but in the 20th century it was reinterpreted as a measure of information and uncertainty.

In 1948, Claude Shannon showed that the uncertainty in a discrete probability distribution is measured by

$$
H(X)=-\sum_{j=1}^k p_j\ln p_j.
$$

This quantity, called information entropy, measures the average information gained when uncertainty is removed. Events that are certain carry no new information, while unlikely events carry more information.

The same mathematical form had already appeared in statistical mechanics through Boltzmann and Gibbs:

$$
S=-k_B\sum_j p_j\ln p_j.
$$

This connection was deeply clarified by E. T. Jaynes in 1957. Jaynes argued that entropy is not only a physical concept but also a principle of inference: given limited information, one should choose the probability distribution that maximizes entropy subject to the known constraints.


## Shannon Entropy (Discrete)

For a discrete random variable $X$, Shannon entropy is defined as:
$$H(X) = -\sum_{i} P(x_i) \ln P(x_i)$$

* **Range:** $[0, \infty)$
* **Why it can't be negative:** Since $0 \le P(x_i) \le 1$, the term $\ln P(x_i)$ is always $\le 0$. Multiplying by the negative sign in the formula ensures $H(X)$ is always non-negative.
* **Physical Meaning:** It represents the average number of bits (or nats) needed to identify an outcome. You can't need "negative bits" to describe something.
* **Minimum Value:** $0$ (occurs when one outcome has probability $1$; there is no uncertainty).


## Differential Entropy (Continuous)

For a continuous random variable $X$ with PDF $f(x)$, differential entropy, with respect to the Lebesgue measure, is defined as:
$$h(X) = -\int_{-\infty}^{\infty} f(x) \ln f(x) dx$$

* **Range:** $(-\infty, \infty)$
* **Why it can be negative:** Unlike probabilities, a probability **density** $f(x)$ can be much greater than $1$.
    > **Example:** If $X$ is uniform on $[0, 0.1]$, then $f(x) = 10$. The $\ln(10)$ term is positive, so $-\int 10 \ln(10) dx$ becomes **negative**.
* **Physical Meaning:** It does **not** represent the absolute number of bits to describe the value (which would be infinite for any continuous variable). Instead, it describes the volume of the support relative to the Lebesgue measure.
* **The Delta Limit:** As you saw in the previous plot, as a distribution narrows (like the Dirac Delta), $f(x) \to \infty$, which forces $h(X) \to -\infty$.

## Conditional Entropy

### Conditional Differential Entropy, $(h(X\mid Y))$


The conditional differential entropy $h(X \mid Y)$ measures the amount of uncertainty remaining in the random variable $X$, **on average**, after the random variable $Y$ has been observed.

Mathematically, it is the expected value (over all possible outcomes of $Y$) of the differential entropy of the conditional random variable $X \mid Y = y$.

Recall that:
1. For a fixed $y$, the random variable $X \mid Y=y$ has the law $P_{X \mid Y=y}$ and the PDF $f_{X \mid Y}(x \mid y)$.
2. The differential entropy of this specific conditional random variable is:
   $$h(X \mid Y = y) = -\int_{\mathscr{X}} f_{X \mid Y}(x \mid y) \ln f_{X \mid Y}(x \mid y) \, dx$$
3. The conditional entropy $h(X \mid Y)$ is defined as the expected value of $h(X \mid Y = y)$ across all possible values of the conditioning random variable $Y$:
   $$h(X \mid Y) \triangleq \mathbb{E}_Y [ h(X \mid Y = Y) ] = \int_{\mathscr{Y}} h(X \mid Y = y) f_Y(y) \, dy$$


Substituting the expression for $h(X \mid Y = y)$ into the expectation gives the complete formula for the conditional differential entropy:

$$h(X \mid Y) = -\int_{\mathscr{Y}} \left( \int_{\mathscr{X}} f_{X \mid Y}(x \mid y) \ln f_{X \mid Y}(x \mid y) \, dx \right) f_Y(y) \, dy$$

Since $f_{X,Y}(x, y) = f_{X \mid Y}(x \mid y) f_Y(y)$, we can merge the terms into a single joint expectation:

$$h(X \mid Y) = -\int_{\mathscr{Y}} \int_{\mathscr{X}} f_{X,Y}(x, y) \ln f_{X \mid Y}(x \mid y) \, dx \, dy$$



### Conditional Shannon Entropy, $(H(X\mid Y))$

The conditional entropy $H(X \mid Y)$ measures the amount of uncertainty remaining in the discrete random variable $X$, **on average**, after the discrete random variable $Y$ has been observed.

Mathematically, it is the expected value (over all possible outcomes of $Y$) of the entropy of the conditional random variable $X \mid Y = y$.


1.  For a fixed $y$, the random variable $X \mid Y=y$ follows the conditional probability mass function (PMF) $P(x \mid y)$.
2.  The entropy of this specific conditional random variable is:
$$H(X \mid Y = y) = -\sum_{x \in \mathscr{X}} P(x \mid y) \ln P(x \mid y)$$
3.  The conditional entropy $H(X \mid Y)$ is defined as the expected value of $H(X \mid Y = y)$ across all possible values of the conditioning random variable $Y$:
$$H(X \mid Y) \triangleq \mathbb{E}_Y [ H(X \mid Y = Y) ] = \sum_{y \in \mathscr{Y}} P(y) H(X \mid Y = y)$$

Substituting the expression for $H(X \mid Y = y)$ into the expectation gives the full formula for discrete conditional entropy:

$$H(X \mid Y) = -\sum_{y \in \mathscr{Y}} P(y) \sum_{x \in \mathscr{X}} P(x \mid y) \ln P(x \mid y)$$

Since the joint probability is defined as $P(x, y) = P(x \mid y) P(y)$, we can simplify the terms into a single joint expectation:

$$H(X \mid Y) = -\sum_{x \in \mathscr{X}, y \in \mathscr{Y}} P(x, y) \ln P(x \mid y)$$

> **Note:** Unlike differential entropy, which can be negative, discrete conditional entropy $H(X \mid Y)$ is always non-negative ($H(X \mid Y) \geq 0$).

## Entropy of the Conditional Expectation, $(h(\mathbb{E}(X \mid Y)))$

Recall that $\mathbb{E}[X \mid Y] = g(Y)$ is a *random variable* that is a deterministic function of $Y$.

When we evaluate $h(\mathbb{E}(X \mid Y))$, we are calculating the differential entropy of this **new random variable**, $Z = g(Y) = \mathbb{E}[X \mid Y]$.

Because $Z = g(Y)$ is a function of $Y$, it lives on its own state space $\mathscr{Z} \subseteq \mathbb{R}^n$. To compute its entropy, we must determine its own probability density function, which we will call $f_Z(z)$.


The density $f_Z(z)$ is obtained by mapping the marginal distribution of $Y$ through the function $g(y) = \int_{\mathscr{X}} x f_{X \mid Y}(x \mid y) \, dx$.

If $g(y)$ is a smooth, invertible function, the density of $Z$ can be found via the standard transformation of variables using the Jacobian determinant:

$$f_Z(z) = f_Y(g^{-1}(z)) \left| \det \frac{\partial g^{-1}(z)}{\partial z} \right|$$

Once this density is established, the differential entropy $h(\mathbb{E}(X \mid Y))$ is evaluated as:

$$h(\mathbb{E}(X \mid Y)) = h(Z) = -\int_{\mathscr{Z}} f_Z(z) \ln f_Z(z) \, dz.$$

## Mutual Information,$I(X;Y)$

### For Continuous Random Variables

Mutual Information measures **how far the joint distribution $P_{X,Y}$ is from the case where $X$ and $Y$ are independent** ($P_X \otimes P_Y$).

$$I(X;Y) \triangleq D_{KL}(P_{X,Y} \, \| \, P_X P_Y)$$

Using the Radon-Nikodym derivative (the ratio of densities), this is defined as:


$$I(X;Y) = \int_{\mathfrak{X} \times \mathfrak{Y}} f_{X,Y}(x,y) \ln \left( \frac{f_{X,Y}(x,y)}{f_X(x) f_Y(y)} \right) dx \, dy$$

#### **Relationship Between Mutual Information and Conditional Entropy**

Starting from mutual information,

$$
I(X;Y)=\iint f_{X,Y}(x,y)\ln\frac{f_{X,Y}(x,y)}{f_X(x)f_Y(y)}\,dx\,dy,
$$

use

$$
f_{X,Y}(x,y)=f_{X|Y}(x|y)f_Y(y)
$$

to get

$$
I(X;Y)=\iint f_{X,Y}(x,y)\ln\frac{f_{X|Y}(x|y)}{f_X(x)}\,dx\,dy.
$$

Then

$$
I(X;Y)
=\iint f_{X,Y}(x,y)\ln f_{X|Y}(x|y)\,dx\,dy
=\iint f_{X,Y}(x,y)\ln f_X(x)\,dx\,dy
-h(X|Y)+h(X).
$$

Hence

$$
I(X;Y)=h(X)-h(X|Y).
$$

Interestingly, even though $h(X)$ can be negative, Mutual Information $I(X;Y)$ is always non-negative ($I(X;Y) \ge 0$), even in the continuous case. This is because $I(X;Y)$ measures the reduction in uncertainty. Even if the starting uncertainty is negative, knowing $Y$ can only make you "more certain" (more negative) or leave you where you are. The difference—the information gained—remains a positive quantity.Is the idea of "negative uncertainty" starting to click, or does it still feel a bit like mathematical sorcery?

---

**Summary**

> Mutual Information is the **"Universal Information Gain."** It is the reduction in the entropy of $X$ achieved by observing $Y$. In other words it measures the percentage of **total uncertainty** (entropy) removed.




### For Discrete Random Variables

$I(X;Y)$ measures the statistical dependence between two discrete variables by quantifying how much the joint distribution $P(x,y)$ deviates from independence $P(x)P(y)$.

$$I(X;Y) = D_{KL}(P(x,y) \| P(x)P(y)) = \sum_{x,y} P(x,y) \ln \left( \frac{P(x,y)}{P(x)P(y)} \right)$$

#### **Relationship Between Mutual Information and Conditional Entropy**

By substituting $P(x,y) = P(x|y)P(y)$, we can decompose the formula:

$$I(X;Y) = \sum_{x,y} P(x,y) \ln P(x|y) - \sum_{x,y} P(x,y) \ln P(x)$$
$$I(X;Y) = -H(X|Y) + H(X)$$

Resulting in the fundamental identity:
$$I(X;Y) = H(X) - H(X|Y)$$

---

**Summary:**

* **Information Gain:** $I(X;Y)$ is the reduction in uncertainty of $X$ due to the knowledge of $Y$.
* **Non-negativity:** Unlike differential entropy, both $H(X)$ and $I(X;Y)$ are always $\geq 0$ in the discrete case.
* **Symmetry:** It captures the information shared between variables, such that $I(X;Y) = I(Y;X)$.

## Joint Entropy and Conditional Entropy

1. **Differential Entropy (Continuous Case):**
For continuous random variables with a joint density $f(x,y)$, the relationship is mathematically identical because the joint density follows the same product rule:
$$f(x,y) = f(x)f(y|x).$$
* The differential entropy is defined via the Lebesgue integral:
$$h(X,Y) = -\int \int f(x,y) \ln f(x,y) \, dx dy.$$
* Just as in the discrete case, the linearity of the integral and the property of the logarithm allow the joint entropy to be split into the marginal and conditional components:
$$h(X,Y) = h(X) + h(Y|X).$$

2. **Shannon Entropy (Discrete Case):**
For discrete random variables, the joint probability can be decomposed using the product rule:
$$P(X, Y) = P(X)P(Y|X).$$
* The joint entropy is defined as:
$$H(X,Y) = -\sum_{x,y} p(x,y) \ln p(x,y).$$
* By substituting the product rule into the logarithm:
$$\ln p(x,y) = \ln p(x) + \ln p(y|x)$$.
* Distributing the summation and expectation leads to the identity:
$$H(X,Y) = H(X) + H(Y|X).$$

# Maximum Entropy Inference

Let us denote by $\Omega$ the sample space consisting of all possible outcomes of a system. We will model the micro-state of a system by a random variable:
$$X:\Omega \to \mathscr{X},$$
where $\mathscr{X}$ is some discrete set of possibly infinitely many labels. That is $\mathscr{X}=\{x_1,x_2,\cdots x_k\}$ where $k$ can be $\infty$. Each $x_i\in \mathscr{X}$ denotes a unique micro-state of the system.

Let $(\mathscr{X},\mathfrak{P}(\mathscr{X}),P)$ be a probability model defined on $\mathscr{X}$ and the $\sigma$-algebra of the powerset of $\mathscr{X}$.
Denote by $P(\{x=x_j\})=p_j$ the probability of the system being in the microstate $x_j$.

A measurement or macroscopic observation of the system corresponds to a random variable $f_i:\mathscr{X}\to \mathbb{R}$. That is we assume that such observations depend only on the micro-state of the system.

To find the most unbiased probability distribution, we utilize **Jaynes' Principle of Maximum Entropy**. Claude Shannon demonstrated that the unique mathematical measure of uncertainty associated with a discrete probability distribution is the information entropy:


$$H(p_1, \dots, p_k) = -\sum_{j=1}^k p_j \ln p_j$$



Because $H$ is a strictly concave function and the constraints are linear equations, the optimization problem possesses a unique global maximum. We find this maximum distribution $\{p_j^*\}$ by constructing the Lagrangian function $\mathscr{L}$:


$$\mathscr{L}(\{p_j\}, \gamma, \{\lambda_i\}) = -\sum_{j=1}^k p_j \ln p_j - \gamma \left(\sum_{j=1}^k p_j - 1\right) - \sum_{i=1}^m \lambda_i \left(\sum_{j=1}^k p_j f_i(x_j) - \langle{f}_i(x)\rangle\right)$$

Since the Lagrange multipliers decouple the constraints, we can vary the probabilities $p_j$ independently. Setting the partial derivatives with respect to each $p_j$ to zero yields:


$$\frac{\partial \mathscr{L}}{\partial p_j} = -(1 + \ln p_j) - \gamma - \sum_{i=1}^m \lambda_i f_i(x_j) = 0$$

Solving for $p_j$ gives the generalized Boltzmann distribution:


$$p_j^* = \exp\left(-(1+\gamma) - \sum_{i=1}^m \lambda_i f_i(x_j)\right)$$

Applying the normalization constraint $\sum p_j^* = 1$, we define the **Partition Function**, $Z$:


$$Z(\lambda_1, \dots, \lambda_m) \triangleq e^{1+\gamma} = \sum_{j=1}^k \exp\left(-\sum_{i=1}^m \lambda_i f_i(x_j)\right)$$

Thus, the canonical maximum entropy distribution is uniquely determined by:


$$p_j^* = \frac{1}{Z} \exp\left(-\sum_{i=1}^m \lambda_i f_i(x_j)\right)$$

From this explicit form of $p_j^*$, several rigorous identities follow directly:

1. **Expectation Values:**

$$\frac{\partial \ln Z}{\partial \lambda_i} = \frac{1}{Z} \frac{\partial Z}{\partial \lambda_i} = -\sum_{j=1}^k f_i(x_j) \frac{1}{Z}\exp\left(-\sum_{l=1}^m \lambda_l f_l(x_j)\right) = -\langle f_i(x)\rangle$$


2. **Fluctuations (Variance):**

$$\frac{\partial^2 \ln Z}{\partial \lambda_i^2} = \langle f_i(x)\rangle^2 - \langle f_i(x)^2\rangle = \operatorname{Var}(f_i(x))$$


3. **Maximized Entropy :**

$$H = - \sum_{j=1}^k p_j^* \ln p_j^* = -k_B \sum_{j=1}^k p_j^* \left(-\ln Z - \sum_{i=1}^m \lambda_i f_i(x_j)\right) = \ln Z + k_B \sum_{i=1}^m \lambda_i \bar{f}_i$$



If the microstate properties $f_i(x_j, \alpha_1, \dots, \alpha_r)$ depend smoothly on external parameters $\alpha$, the total differential of the entropy expands cleanly to:


$$dH =  \sum_{i=1}^m \lambda_i d\langle f_i(x)\rangle + k_B \sum_{r} \frac{\partial \ln Z}{\partial \alpha_r} d\alpha_r$$



## Logistic Regression

Suppose we have a training dataset consisting of $N$ observations, $D = \{(\mathbf{x}^{(k)}, y^{(k)})\}_{k=1}^N$. We define a set of joint feature functions $\mathbf{f}(\mathbf{x}, Y)\in \mathbb{R}^m$ that capture the interactions between the input features and the output classes.
From our training data, we can calculate the empirical joint distribution $$\widetilde{P}(\mathbf{x}, Y) = \frac{1}{N} \sum_{k=1}^N \delta(\mathbf{x} - \mathbf{x}^{(k)})\delta(Y - y^{(k)}).$$
The macroscopic facts available to us are the empirical expectation values of our feature functions:

$$\mathbb{E}_{\widetilde{P}}[\mathbf{f}(\mathbf{x}, Y)] = \sum_{\mathbf{x}, Y} \mathbf{f}(\mathbf{x}, Y)\widetilde{P}(\mathbf{x}, Y)  = \frac{1}{N} \sum_{k=1}^N \mathbf{f}(\mathbf{x}^{(k)}, y^{(k)}).$$
This is the sample expectation of $\mathbf{f}(\mathbf{x}, Y)$.

Our model's distribution $P(\mathbf{x}, Y) = \widetilde{P}(\mathbf{x}) P(Y \mid \mathbf{x})$ must match these macroscopic constraints exactly. Therefore, for each feature dimension $i$, we require:

$$\mathbb{E}_{P}[f_i(\mathbf{x}, Y)] = \mathbb{E}_{\widetilde{P}}[f_i(\mathbf{x}, Y)] \implies \sum_{\mathbf{x}} \widetilde{P}(\mathbf{x}) \sum_{c=1}^C P(Y=c \mid \mathbf{x}) f_i(\mathbf{x}, c) = \frac{1}{N} \sum_{k=1}^N f_i(\mathbf{x}^{(k)}, y^{(k)}).$$

To find the least-biased model, we maximize the **Conditional Shannon Entropy** $H(Y \mid X)$, which measures the average uncertainty remaining in the label $Y$ given that we observe $\mathbf{x}$.

By definition
$$\begin{align*}
H(Y \mid X) &= -\sum_{\mathbf{x}} \sum_{c=1}^C \widetilde{P}(\mathbf{x}, c) \ln P(Y=c \mid \mathbf{x})
\end{align*}$$

The Entropy of $(Y \mid X=\mathbf{x})$ is:
$$H(Y \mid X=\mathbf{x}) = -\sum_{c=1}^C P(Y=c \mid \mathbf{x}) \ln P(Y=c \mid \mathbf{x}).$$
Thus the conditionaly entropy is:
$$\begin{align*}
H(Y \mid X) &= -\sum_{\mathbf{x}} \widetilde{P}(\mathbf{x}) \sum_{c=1}^C P(Y=c \mid \mathbf{x}) \ln P(Y=c \mid \mathbf{x}) \\
&= -\sum_{\mathbf{x}} \sum_{c=1}^C \widetilde{P}(\mathbf{x}, c) \ln P(Y=c \mid \mathbf{x})
\end{align*}$$

Following the discrete MaxEnt framework established previously, we construct the Lagrangian functional $\mathscr{L}$ to maximize $H(Y \mid X)$ subject to our empirical expectation constraints and the fundamental constraint that probabilities must sum to 1 for every individual input $\mathbf{x}$:

$$\begin{align*}
\mathscr{L}(P, {\gamma(\mathbf{x})}, \mathbf{w}^c) = &-\sum_{\mathbf{x}} \widetilde{P}(\mathbf{x}) \sum_{c=1}^C P(Y=c \mid \mathbf{x}) \ln P(Y=c \mid \mathbf{x}) \\
&- \sum_{\mathbf{x}} \gamma(\mathbf{x}) \left( \sum_{c=1}^C P(Y=c \mid \mathbf{x}) - 1 \right) \\
&- \mathbf{w}^c \cdot \left( \sum_{\mathbf{x}} \widetilde{P}(\mathbf{x}) \sum_{c=1}^C P(Y=c \mid \mathbf{x}) \mathbf{f}(\mathbf{x}, c) - \mathbb{E}_{\widetilde{P}}[\mathbf{f}(\mathbf{x}, Y)] \right)
\end{align*}$$

Here, $\mathbf{w}^c = [w^c_1, w^c_2, \dots, w^c_m]^T$ represents the vector of Lagrange multipliers corresponding to our feature expectations, and $\gamma(\mathbf{x})$ represents the point-wise multipliers ensuring local probability normalization.


## MaxEnt Solution

We optimize this functional by taking the partial derivative of $\mathscr{L}$ with respect to the conditional probability of a specific class $c$ given a specific input $\mathbf{x}$, and setting it equal to zero:

$$\frac{\partial \mathscr{L}}{\partial P(Y=c \mid \mathbf{x})} = -\widetilde{P}(\mathbf{x}) \left( 1 + \ln P(Y=c \mid \mathbf{x}) \right) - \gamma(\mathbf{x}) - \widetilde{P}(\mathbf{x}) \left( \mathbf{w}^c \cdot \mathbf{f}(\mathbf{x}, c) \right) = 0$$

Dividing through by the empirical data probability $\widetilde{P}(\mathbf{x})$ (on the support of $\widetilde{P}$) and isolating the log probability term gives:

$$\ln P(Y=c \mid \mathbf{x}) = -\left(1 + \frac{\gamma(\mathbf{x})}{\widetilde{P}(\mathbf{x})}\right) - \mathbf{w}^c \cdot \mathbf{f}(\mathbf{x}, c)$$

Exponentiating both sides reveals the canonical exponential shape of the MaxEnt solution:

$$P^*(Y=c \mid \mathbf{x}) = \exp\left(-\left(1 + \frac{\gamma(\mathbf{x})}{\widetilde{P}(\mathbf{x})}\right)\right) \exp\left( -\mathbf{w}^c \cdot \mathbf{f}(\mathbf{x}, c) \right)$$

By enforcing the structural axiom of total probability
$$\sum_{l=1}^C P^*(Y=l \mid \mathbf{x}) = 1,$$
we eliminate the local multiplier $\gamma(\mathbf{x})$ and define the conditional partition function $Z(\mathbf{x})$:

$$Z(\mathbf{x}) \triangleq \exp\left(1 + \frac{\gamma(\mathbf{x})}{\widetilde{P}(\mathbf{x})}\right) = \sum_{l=1}^C \exp\left( -\mathbf{w}^l \cdot \mathbf{f}(\mathbf{x}, l) \right)$$

This yields the definitive conditional probability distribution:

$$P^*(Y=c \mid \mathbf{x}) = \frac{\exp\left( -\mathbf{w}^c \cdot \mathbf{f}(\mathbf{x}, c) \right)}{\sum_{l=1}^C \exp\left( -\mathbf{w}^l \cdot \mathbf{f}(\mathbf{x}, l) \right)}$$

Since the Lagrange multipliers, $\mathbf{w}$, are arbitray constants without loss of generality we can write:
$$P^*(Y=c \mid \mathbf{x}) = \frac{\exp\left( \mathbf{w}^c \cdot \mathbf{f}(\mathbf{x}, c) \right)}{\sum_{l=1}^C \exp\left( \mathbf{w}^l \cdot \mathbf{f}(\mathbf{x}, l) \right)}$$

To map this directly to the standard presentation of multi-class classification found in machine learning textbooks, we can choose the specific joint feature functions:

$$\mathbf{f}(\mathbf{x}, c) = \mathbb{I}(Y = c) \mathbf{x}$$

Where $\mathbb{I}$ is the indicator function that equals 1 if the statement is true and 0 otherwise. This decouples the global weight vector $\mathbf{w}$ into class-specific weight vectors $\mathbf{w}_c$ for each category. Substituting this choice back into our general MaxEnt solution yields:


$$P^*(Y=c \mid \mathbf{x}) = \frac{\exp\left( \mathbf{w}_c \cdot \mathbf{x} \right)}{\sum_{l=1}^C \exp\left( \mathbf{w}_l \cdot \mathbf{x} \right)}$$

This is precisely the **Softmax Regression** or **Multiclass Logistic Regression** model.
