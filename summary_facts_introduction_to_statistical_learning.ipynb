# Probability Space

A probability space is characterized by a triple: $(\Omega, \mathfrak{U},P)$ where

* $\Omega$ is some set and is the sample space
* $\mathfrak{U}$ is a $\sigma$-algebra on $\Omega$
* $P: \mathfrak{U} \to [0,1]$ is a probability measure satisfying $P(\Omega)=1$ and countable additivity

In this setup:

* Any $A\in \mathfrak{U}$ represents an event.
* The probability of the event $A\in \mathfrak{U}$ happening is give by the measure ('size') of $A$ defined by
$$P(A)\triangleq \int_A dP.$$

---

If $\Omega \subseteq \mathbb{R}^n$ and $\mathfrak{U}$ is Lebesgue-measurable, and $P\ll\lambda$ then
$$P(A)\triangleq \int_A dP=\int_{A}f(\omega)\,d\lambda (\omega).$$
Here $f(\omega)$ is called the dinsity of $P$ with respect to $\lambda$ and is formally given by the Radon-Nikodym derivative
$$f\triangleq \dfrac{dP}{d\lambda}.$$

---

In the special case where $\Omega$ is discrete, one may take the reference measure to be the counting measure $\#$. If $P\ll \#$, then the Radon--Nikodym derivative
$$
p\triangleq \frac{dP}{d\#}
$$
is the probability mass function of $P$. Hence, for any event $A\subseteq \Omega$,
$$
P(A)=\int_A dP=\int_A p\,d\#
=\sum_{\omega\in A} p(\omega).
$$
In particular,
$$
p(\omega)=P(\{\omega\}),
\qquad
\sum_{\omega\in\Omega}p(\omega)=1.
$$
Thus, in the discrete case, integration with respect to counting measure reduces to summation, and the density of $P$ with respect to counting measure is precisely the probability mass function.

# Random Variable

Consider a probability space $(\Omega,\mathfrak U,P)$ and a measurable space
$(E, \mathfrak{E})$. A random variable is a measurable map
$$
X:\Omega\to E,
$$
meaning that
$$
X^{-1}(B)\in\mathfrak{E}
\qquad \text{for every } B\in\mathfrak{E}.
$$
Equivalently, the $\sigma$-algebra generated by $X$,
$$
\sigma(X)\triangleq \{X^{-1}(B):B\in\mathfrak{E}\},
$$
satisfies
$$
\sigma(X)\subseteq \mathfrak U.
$$

# The Law and the Distribution of a Random Variable

## The Law of $X$

Once we have a measurable function $X$, we can use it to "transport" the probability measure from our original space to the target space. This transported measure is called the **Law** of the Variable.


If $X : (\Omega, \mathfrak{A}) \to (E, \mathfrak{E})$ is a random Variable and $P$ is a probability measure on $(\Omega, \mathfrak{A})$, the **law of $X$ under $P$** (denoted $P_X$ or $\mathfrak{L}_X$) is a probability measure on the target space $(E, \mathfrak{E})$ defined by:

$$\forall B \in \mathfrak{E}, \quad P_X(B) \triangleq P(X^{-1}(B)) = P(\{\omega \in \Omega : X(\omega) \in B\})$$

In standard notation, we usually write this as:


$$P_X(B) = P(X \in B)=\int_{X^{-1}(B)}dP, \qquad \forall B\in \mathfrak{E}.$$

## The "Push-forward" Mechanism

Think of $X$ as a map. $P$ tells us how much "mass" is at each point in $\Omega$. The Law $P_X$ is simply the result of "pushing" all that mass forward onto the points in $E$.
The push-forwad measure on $E$ is therefore characterized by
$$P_X(B) = P(X \in B), \qquad \forall B\in \mathfrak{E}$$
which can equivalently be given by
$$\int_{X^{-1}(B)}dP=\int_BdP_X, \qquad \forall B\in \mathfrak{E}.$$

## Pushforward Rule

For every measurable function $g:E\to\mathbb R$,
$$
\int_\Omega g(X)\,dP=\int_E g\,dP_X,
$$
whenever the integrals are well defined.


## Density of a Random Variable

Similar to $P$ on $\Omega$, in order to represent $P_X$ as a continuous distribution, we compare it to a reference measure on the target space—typically the Lebesgue measure $\lambda$ on $(E, \mathfrak{E})$.
If $P_X \ll \lambda$ then by the Radon-Nikodym Theorem, there exists a unique (up to a set of $\lambda$-measure zero), non-negative, and measurable function $f_X: E \to [0, \infty)$ such that for every $B \in \mathfrak{E}$:
$$P(X \in B) = P_X(B) = \int_{B} dP_X=\int_{B} f_X(x) \, d\lambda(x)$$

The function $f_X = \frac{dP_X}{d\lambda}$ is the probability density function (PDF) of the random variable $X$.


## The Joint Probability Density

Let $X:(\Omega,\mathfrak A)\to(\mathscr X,\mathfrak U_X)$ and
$Y:(\Omega,\mathfrak A)\to(\mathscr Y,\mathfrak U_Y)$ be two random variables
defined on the same probability space $(\Omega,\mathfrak A,P)$. They may be
viewed together as a single random ordered pair
$$
Z=(X,Y):\Omega\to\mathscr X\times\mathscr Y.
$$

The map $Z=(X,Y)$ is measurable with respect to the product
$\sigma$-algebra $\mathfrak U_X\otimes\mathfrak U_Y$. The joint law, or
joint distribution, of $X$ and $Y$, denoted $P_{X,Y}$, is the probability
measure on
$$
(\mathscr X\times\mathscr Y,\mathfrak U_X\otimes\mathfrak U_Y)
$$
defined by
$$
P_{X,Y}(B)
=
P\big((X,Y)\in B\big)
=
P\big((X,Y)^{-1}(B)\big),
\qquad B\in \mathfrak U_X\otimes\mathfrak U_Y.
$$


If $\mathscr X\subseteq\mathbb R^m$, $\mathscr Y\subseteq\mathbb R^n$, and
the joint law $P_{X,Y}$ is absolutely continuous with respect to the
Lebesgue measure on $\mathscr X\times\mathscr Y$, then there exists a
measurable function $f_{X,Y}:\mathscr X\times\mathscr Y\to[0,\infty)$ such
that
$$
P\big((X,Y)\in B\big)
=
P_{X,Y}(B)
=
\int_B dP_{X,Y}
=
\int_B f_{X,Y}(x,y)\,d(x,y),
$$
for every $B\in\mathfrak U_X\otimes\mathfrak U_Y$.
When $\mathscr X\subseteq\mathbb R$ and $\mathscr Y\subseteq\mathbb R$, this
is often written as
$$
P\big((X,Y)\in B\big)
=
\iint_B f_{X,Y}(x,y)\,dx\,dy.
$$
The function $f_{X,Y}$ is called the joint probability density function of
$X$ and $Y$.

## The Marginal Probability Density

The **marginal law** of one variable is obtained from the joint law by
projecting onto the corresponding coordinate. In measure-theoretic terms,
if $P_{X,Y}$ is the joint law of $(X,Y)$, then the marginal law of $X$
is the pushforward of $P_{X,Y}$ under the coordinate projection
$$
\pi_X:\mathscr X\times \mathscr Y\to \mathscr X,
\qquad
\pi_X(x,y)=x.
$$
Thus, for every measurable set $A\in\mathfrak U_X$,
$$
P_X(A)
=
P_{X,Y}(A\times \mathscr Y).
$$
Similarly, for every measurable set $C\in\mathfrak U_Y$,
$$
P_Y(C)
=
P_{X,Y}(\mathscr X\times C).
$$

If $X$ and $Y$ are jointly continuous with joint density $f_{X,Y}$,
then the marginal density of $X$ is obtained by integrating out $Y$:
$$
f_X(x)
=
\int_{\mathscr Y} f_{X,Y}(x,y)\,dy.
$$
When $\mathscr Y=\mathbb R$, this becomes
$$
f_X(x)
=
\int_{-\infty}^{\infty} f_{X,Y}(x,y)\,dy.
$$

Similarly,
$$
f_Y(y)
=
\int_{\mathscr X} f_{X,Y}(x,y)\,dx.
$$

In the discrete case we have:
$$p_X(x) = \sum_{y \in \text{Img}(Y)} p_{X,Y}(x, y)$$
and
$$p_Y(y) = \sum_{x \in \text{Img}(X)} p_{X,Y}(x, y).$$

# The Expectation and Covariance of a Random Variable

In the case where $E=\mathbb{R}^n$, the **expectation** (or mean) of $X$ is defined as the integral of $X$ with respect to the probability measure $P$:

$$\mathbb E[X] = \int_{\Omega} X(\omega) \, dP(\omega)=\int_{\mathbb{R}^n}x\,dP_{X}(x)$$

The **Covariance** is defined as:

$$\mathrm{Cov}(X) = E\left[ (X - \mathbb E[X])(X - \mathbb E[X])^T \right] = \int_{\mathbb{R}^n} (x - \mathbb E[X])(x - \mathbb E[X])^T \, dP_X$$

# Independence of Random Variables

Let $(\Omega,\mathfrak A,P)$ be a probability space. Let
$X_1,X_2,\dots,X_n$ be random variables, where each
$$
X_i:(\Omega,\mathfrak A)\to(E_i,\mathfrak E_i)
$$
is measurable.

The random variables $X_1,X_2,\dots,X_n$ are said to be independent if, for every choice of measurable sets
$B_i\in\mathfrak E_i$, $i=1,\dots,n$, we have
$$
P(X_1\in B_1,\dots,X_n\in B_n)
=
\prod_{i=1}^n P(X_i\in B_i).
$$
Equivalently,
$$
P\left(\bigcap_{i=1}^n X_i^{-1}(B_i)\right)
=
\prod_{i=1}^n P\left(X_i^{-1}(B_i)\right).
$$

If the joint law of $(X_1,\dots,X_n)$ is absolutely continuous with respect to a product reference measure, and if each marginal law has density $f_{X_i}$, then independence is equivalent to the factorization
$$
f_{X_1,\dots,X_n}(x_1,\dots,x_n)
=
\prod_{i=1}^n f_{X_i}(x_i)
$$
almost everywhere.

# Multivariate Central Limit Thoerem

Let
$$
X_1,X_2,\dots
$$
be i.i.d. random variables in $\mathbb R^d$ with
$$
\mathbb E[X_1]=\mu\in\mathbb R^d,
\qquad
\operatorname{Cov}(X_1)=\Sigma\in\mathbb R^{d\times d},
$$
where $\Sigma$ is finite. Define
$$
\bar X_n=\frac1n\sum_{i=1}^n X_i.
$$
Then the multivariate central limit theorem states that
$$
\sqrt n(\bar X_n-\mu)
\xrightarrow{d}
\mathscr N_d(0,\Sigma).
$$


Equivalently, for large $n$,
$$
\bar X_n
\approx
\mathscr N_d\left(\mu,\frac{1}{n}\Sigma\right).
$$

# Conditional Expectation

## Definition of $\mathbb{E}[X\mid Y]$

Let $(\Omega, \mathfrak{A}, P)$ be a probability space.

* Let $X: \Omega \to \mathbb{R}$ be a random variable that is integrable ($X \in L^1$).
* Let $Y: \Omega \to \mathbb{R}$ be another random variable.
* Let $\sigma(Y)$ be the **sub-$\sigma$-algebra** generated by $Y$.

**The Existence: Conditional Expectation**

The conditional expectation of $X$ given $Y$, denoted $\mathbb E[X|Y]$ (or more formally $\mathbb E[X|\sigma(Y)]$), is defined as the unique (almost surely) random variable $Z$ that satisfies two conditions:

1. **Measurability:** $Z$ is measurable with respect to $\sigma(Y)$.
That is by the Doob-Dynkin lemma $Z$ must be a function of $Y$:
$$Z = g(Y).$$
2. **The Integral Property (Partial Averaging):** For every set $G \in \sigma(Y)$:

$$\int_G Z \, dP = \int_G X \, dP$$



*(Translation: On any set of outcomes that $Y$ can distinguish, the "average" of our estimate $Z$ must match the "average" of the actual $X$.)*


## The Joint Probability ($P(X\in B\mid Y\in A)$)

Let
$$
Z_B\triangleq \mathbb E[\mathbb 1_{\{X\in B\}}\mid Y].
$$

Then it follows that

$$
P(X\in B,\;Y\in A)=\int_{Y^{-1}(A)} Z_B\,dP.
$$

---

$Z_B$ is measurable with respect to $\sigma(Y)$.
Hence from the Doob-Dynkin Lemma $Z_B$ must be a function of $Y$. We can write $Z_B = g_B(Y)$ and then the above expression can be written as:

$$
P(X\in B,\;Y\in A)=\int_{Y^{-1}(A)} g_B(Y)\,dP.
$$

Now applying the pushforward rule:
$$
\int_{Y^{-1}(A)} g_B(Y)\,dP=\int_A g_B(y)\,dP_Y(y),
$$

we have

$$
P(X\in B,\;Y\in A)=\int_A g_B(y)\,dP_Y(y).
$$

If $Y$ has density $f_Y$, so that

$$
dP_Y(y)=f_Y(y)\,dy,
$$

then this becomes

$$
P(X\in B,\;Y\in A)=\int_A g_B(y)\,f_Y(y)\,dy.
$$

In the discrete setting we have:
$$
P(X\in B,\;Y\in A)=\sum_{y\in A} g_B(y)\,P(Y=y).
$$

---

This means that the function $g_B(y)$ satisfies:

$$g_B(y)=P(X\in B\mid Y=y),$$
holds only $P_Y$-almost everywhere.

## The conditional Probability, $P(X\in B\mid Y=y)$

In measure theory, the "probability" of an event is simply the **measure of the set** representing that event. Thus the expression
$$g_B(y)=P(X\in B\mid Y=y)$$
indicates that for a fixed $y$ the map
$$
P_{X \mid Y=y}:\mathfrak{U} \mapsto [0,1]
$$
where $P_{X \mid Y=y}(B)$ for any $B\in \mathfrak{U}$ is a probability measure. If $P_{X \mid Y=y}$ is absolutely continuous with the length measure on $\mathbb{R}$ the Radon-Nikodym theorem tells us that there existis a unique density function $f_{X|Y}$ that satisfies
$$
\dfrac{dP_{X \mid Y=y}}{dx}(x)=f_{X|Y}(x\mid y).
$$

Thus we have the expression
$$
g_B(y)=P_{X \mid Y=y}(B)
=\int_B dP_{X \mid Y=y}=\int_B f_{X|Y}(x \mid y) \, dx.
$$
indicating that $g_B(y)$ gives the conditional probality of $X\in B$ given that $Y=y$. That is
$$
g_B(y)\triangleq P(X\in B\mid Y=y)=\int_B f_{X|Y}(x \mid y) \, dx.
$$

## The Product Rule of the Conditional Density

### Continuous Case

Substituting the above expression, $
g_B(y)=\int_B f_{X|Y}(x \mid y) \, dx
$, into the integral $
P(X\in B,\;Y\in A)=\int_A g_B(y)\,f_Y(y)\,dy
$ we have:
$$P(X \in B, Y \in A)
= \int_A \int_B f_{X|Y}(x \mid y) \, f_Y(y)\,dx \, dy.
$$

Since $P(X \in B, Y \in A) = \iint_{A \times B} f_{X,Y}(x,y) \, dx \, dy$ where $f_{X,Y}(x,y)$ is the joint probability density,
we obtain the product rule

$$
f_{X,Y}(x,y)=f_{X|Y}(x\mid y)f_Y(y)
$$
for almost every $(x,y)$.
The equality holds for the joint density $f_{X,Y}$ by the uniqueness of the Radon-Nikodym derivative.

### Discrete Case

The relationship $f_{X,Y}(x,y) = f_{X|Y}(x|y)f_Y(y)$ represents the Product Rule of probability for the continuous case. In the discrete case this is expressed as
$$p_{X,Y}(x,y) = p_{X|Y}(x|y)p_Y(y),$$
where
$$p_{X|Y}(x|y)=P(X=x \mid Y=y).$$

$$P(X \in B, Y \in A) = \sum_{y \in A} P(X \in B \mid Y=y) P(Y=y)$$

## The Disintegration Theorem

Let $(X, Y): \Omega \to \mathscr{X} \times \mathscr{Y}$ be the joint map. By the **Pushforward Rule** we cans see that
$$\int_{Y^{-1}(A)} X \, dP = \int_A \left( \int_{\mathscr{X}} x \, dP_{X \mid Y=y}(x)\right) dP_Y(y)= \int_A \left( \int_{\mathscr{X}} x \, f_{X \mid Y=y}(x\mid y) \,dx\right) f_Y(y)dy.$$

In the discrete case the above result is expressed as:


$$\sum_{\omega \in Y^{-1}(A)} X(\omega)\, P(\{\omega\}) =
\sum_{y\in A}\sum_{x\in \mathscr{X}}xp_{X\mid Y}(x\mid y) \, P(Y=y).$$

## The Conditional Expectation Evaluated at $Y=y$

We apply the definition of conditional expectation for a general random variable $X$ and using the Doob-Dynkin lemma we have for any $A\in \mathscr{Y}$:

$$
\int_{Y^{-1}(A)} X \, dP =
\int_{Y^{-1}(A)} \mathbb E[X \mid Y] \, dP =\int_A \mathbb E[X \mid Y=y] \, dP_Y(y).
$$

Note that in the previous section we have shown that the Disintegration Theorem gives
$$
\int_{Y^{-1}(A)} X \, dP = \int_A \left( \int_{\mathscr{X}} x \, dP_{X \mid Y=y}(x) \right) dP_Y(y).
$$

Thus we see that
$$\mathbb{E}[X\mid Y=y]= \int_{\mathscr{X}} x \, dP_{X \mid Y=y}(x)=
\int_{\mathscr{X}} x f_{X|Y}(x\mid y)dx.
$$
The first term, $\mathbb{E}[X|Y=y]$, represents the evaluation of the $\sigma(Y)$-measurable random variable $g(Y)=\mathbb{E}[X\mid Y]$ at $Y=y$.


---

In the discrete case,
the partial expectation over the event $\{Y \in A\}$ simplifies to the discrete weighted sum:

$$\sum_{\omega \in Y^{-1}(A)} X(\omega)\, P(\{\omega\}) =
\sum_{y\in A}\mathbb E[X \mid Y=y] \, P(Y=y),$$

and the factor $g(y)$ becomes:
$$g(y) = \mathbb{E}[X \mid Y=y] = \sum_{x \in \mathscr{X}} x \, p_{X\mid Y}(x \mid y),$$
where $p_{X\mid Y}(x \mid y) = P(X=x \mid Y=y)$ is the conditional probability mass function.

## Conditional Random Variable, $X \mid Y=y$

In the standard framework of probability, we often talk about the distribution of $X$ across the entire sample space. However, once we observe $Y=y$, the "global" distribution of $X$ becomes obsolete; we need a mathematical object that represents our updated knowledge of $X$ within this specific "slice" of reality.


While $X$ is a mapping from the sample space $\Omega$ to $\mathscr{Y}$, the **conditional random variable** $X \mid Y=y$ (read as "$X$ given $Y=y$") is defined as a random variable whose law is the **Regular Conditional Law** $P_{X \mid Y=y}$.

---

Unlike the original $X$, which is a single map defined on $(\Omega, \mathfrak{A}, P)$, the conditional variable $X \mid Y=y$ can be viewed as a **family of random variables** indexed by the possible outcomes of $Y$ and satisifes:

* **Realization:** If $Y$ is fixed at $y$, $X \mid Y=y$ behaves like a standard random variable with its own mean, variance, and moments, all of which are now functions of $y$.
* **The Density:** The function $f_{X|Y}(x \mid y)$ serves as the probability density function (PDF) for this variable. Just as we write $X \sim f_X(x)$, we now write:
$$(X \mid Y=y) \sim f_{X|Y}(x \mid y).$$
* **The Expectation:** The "best guess" for the value of this conditional random variable is the conditional mean that was derived earlier:
$$\mathbb{E}[(X \mid Y=y)] \triangleq \int_{\mathscr{X}} x f_{X|Y}(x \mid y) \, dx=\mathbb{E}[X \mid Y=y].$$

---

The relationship between the global random variable $X$ and the conditional family $X \mid Y=y$ is bridged by the **Law of Total Probability**. The global distribution is the mixture of these conditional variables, weighted by the marginal distribution of $Y$:
$$f_X(x) = \int_{\mathscr{Y}} f_{X|Y}(x \mid y) f_Y(y) \, dy.$$

In essence, $X \mid Y=y$ is the version of $X$ that lives exclusively on the "slice" of the universe where $Y$ has already happened and its value is known.

## Relationship between $g_B(y)$ and $g(y)$

The best way to understand the difference is to carefully examine how they are formally defined.

For a fixed $y$ (where $f_Y(y) > 0$), the mapping $g_B(y)$ is defined for any Borel set $B \in \mathfrak{B}(\mathscr{X})$ as:

$$g_B(y) \triangleq P(X \in B \mid Y=y) = \int_B f_{X|Y}(x \mid y) \, dx = \int_B dP_{X|Y=y}(x).$$

This demonstrates that $g_B(y)$, for a fixed $y$, evaluates the conditional probability measure $P_{X|Y=y}(B)$ over a given set $B$ in the state space of $X$.

On the other hand, $g(y)$ is defined as:

$$g(y) \triangleq \mathbb{E}[X \mid Y=y] = \int_{\mathscr{X}} x \, f_{X|Y}(x \mid y) \, dx = \int_{\mathscr{X}} x \, dP_{X|Y=y}(x).$$

This demonstrates that $g(y)$, for a fixed $y$, computes the functional average (the integration of the identity map $X(x)=x$) over the entire domain $\mathscr{X}$ with respect to that same conditional measure $P_{X|Y=y}$.

Thus, they measure two distinct mathematical objects: $g_B(y)$ evaluates a localized **probability measure**, whereas $g(y)$ calculates a **numerical expectation** under that measure.

However, they are related in a very subtle manner through the concept of a **regular conditional distribution**.

For a fixed $y$, $P_{X|Y=y}$ behaves as a valid probability measure on $(\mathscr{X}, \mathfrak{B}(\mathscr{X}))$ with density $f_{X|Y}(x \mid y)$. We can define an abstract **conditional random variable**, denoted $(X \mid Y=y)$, whose probability distribution is exactly given by this measure.

Importantly, this conditional random variable $(X \mid Y=y)$ is a completely different mathematical object than the $\sigma(Y)$-measurable random variable $\mathbb{E}[X \mid Y]$. However, computing the standard expectation of this localized variable yields:

$$\mathbb{E}[(X \mid Y=y)] \triangleq \int_{\mathscr{X}} x \, f_{X|Y}(x \mid y) \, dx = \mathbb{E}[X \mid Y=y] = g(y).$$


## Properties of $\mathbb{E}[X\mid Y]$

### Law of total expectation

$$\mathbb E[X] = \mathbb E[\mathbb E[X | Y]]$$

### Conditional Variance ($\mathrm{Var}(X|Y)$)

The conditional variance, $\mathrm{Var}(X|Y)$, is the variance of $X$ given a specific outcome of $Y$. Formally, it is a $\sigma(Y)$-measurable random variable defined as:
$$\mathrm{Var}(X|Y) \triangleq \mathbb E[X^2 | Y] - (\mathbb E[X|Y])^2$$
It measures the **residual uncertainty** of $X$ that remains after $Y$ is known. If you were to observe $Y=y$, the conditional variance tells you the spread of the remaining probability distribution of $X$.

### Law of Total Variance

For any two random variables $X$ and $Y$ with finite variance, the total variance of $X$ can be split into two distinct parts:

$$\mathrm{Var}(X) = \mathbb E[\mathrm{Var}(X|Y)] + \mathrm{Var}(\mathbb E[X|Y])$$

* **$\mathrm{Var}(X)$**: Total Uncertainty (The "Total Spread").
* **$\mathbb E[\mathrm{Var}(X|Y)]$**: **Unexplained Variance** (The "Average Spread of the Slices").
* **$\mathrm{Var}(\mathbb E[X|Y])$**: **Explained Variance** (The "Spread of the Means").

The **Explained Variance** represents the part of $X$'s variance that is "accounted for" or explained by the linear relationship with $Y$. **Unexplained Variance** represents the remaining uncertainty in $X$ after $Y$ is known, often referred to as the estimation error or residual variance.

### Optimality of the Conditional Estimate

Let $X$ be a random variable with $\mathbb{E}[X^2] < \infty$. Let $Y$ is a $\sigma(X)$-measurable random variable and $Z$ be a $\sigma(Y)$-measurable random variable with $\mathbb{E}[Z^2] < \infty$. Then
\begin{align*}
\mathbb{E}[(X - \mathbb{E}[X \mid Y])^2] &\leq \mathbb{E}[(X - Z)^2].
\end{align*}

>This result proves that $\mathbb{E}[X \mid Y]$ is the **optimal estimate** of $X$ given the information in $Y$, in the sense that it is the unique minimizer of the **Mean Square Error (MSE)**. Any other $\sigma(Y)$-measurable estimate $Z$ will necessarily result in a larger (or equal) expected squared error.

---

Using the Law of Iterated Expectations we have:

\begin{align*}
\mathbb{E}[(X - \mathbb{E}[X \mid Y])^2] &= \mathbb{E}[\text{Var}(X \mid Y)]
\end{align*}

## Conditioning of Multivariate Gaussian Random Variables

Given a joint multivariate normal distribution
$$Z = \begin{bmatrix} X \\ Y \end{bmatrix} \sim \mathscr{N} \left( \begin{bmatrix} \mu_X \\ \mu_Y \end{bmatrix}, \begin{bmatrix} \Sigma_{XX} & \Sigma_{XY} \\ \Sigma_{YX} & \Sigma_{YY} \end{bmatrix} \right)$$ the "best predictor" for all these methods is the conditional expectation:

$$\mathbb{E}[X|Y] = \mu_X + \Sigma_{XY}\Sigma_{YY}^{-1}(Y - \mu_Y).$$

---

In the context of Bayesian inference, filtering, and machine learning, here is the breakdown of the components of the above expression using the Bayesian terminology:

1. $\mu_X$ (The Prior Mean):
* **What it is:** This is your "best guess" for $X$ if you have zero information about $Y$.
* **Role:** It acts as the starting point (the baseline). In a **Kalman Filter**, this would be the "predicted state" before the measurement comes in. In **Gaussian Process Regression**, this is often set to zero or a constant mean function.

2. $\Sigma_{XX}$ (The Prior Covariance):
* **What it is:** This represents your uncertainty about $X$ before observing $Y$.
* **Role:** It defines the "search space." If $\Sigma_{XX}$ is very large, it means your prior is weak, and the model will rely more heavily on the new data $Y$. In terms of the **Law of Total Variance**, this is the "Total Variance" that you are trying to reduce.

3. $\Sigma_{XY}$ (The "Information Bridge"):
* **What it is:** The cross-covariance between the latent variable $X$ and the observable $Y$.
* **Role:** This is arguably the most important part of the formula. It defines how much $X$ and $Y$ are "entangled."
    * If $\Sigma_{XY} = 0$, the variables are independent, the formula collapses to $\mathbb{E}[X|Y] = \mu_X$, and the data $Y$ tells you nothing.
    * As $\Sigma_{XY}$ increases, the "bridge" becomes stronger, allowing more information to "leak" from $Y$ to $X$.

4. The "Update" Term: $\Sigma_{XY}\Sigma_{YY}^{-1}(Y - \mu_Y)$:
This whole block is the **Bayesian Update** (or the **Correction**):
* **$(Y - \mu_Y)$:** This is known as the **"Innovation"** or **"Residual."** It is the "surprise"—the difference between what you actually saw ($Y$) and what you expected to see ($\mu_Y$).
* **$\Sigma_{XY}\Sigma_{YY}^{-1}$:** This is the **"Kalman Gain"** or the **"Regression Coefficient."** It dictates exactly how much of that "surprise" should be applied to your prior mean to get your new, better estimate.

5. The Mean Square Error:
This is rhe **Conditional Covariance Matrix**, denoted as $\Sigma_{X|Y}$
$$MSE = \Sigma_{X|Y} = \Sigma_{XX} - \Sigma_{XY}\Sigma_{YY}^{-1}\Sigma_{YX}$$
Here
* **$\Sigma_{XX}$ (Prior Uncertainty):** This is the total variance/uncertainty you have about $X$ before you see any data. It represents the "starting" error.
* **$\Sigma_{XY}\Sigma_{YY}^{-1}\Sigma_{YX}$ (Explained Variance):** This is the amount of uncertainty that is removed or "explained" by observing $Y$.
    * $\Sigma_{XY}$ is the connection between the variables.
    * $\Sigma_{YY}^{-1}$ accounts for the noise/scale of the observations.
* **The Subtraction:** The MSE is what remains of your prior uncertainty after you subtract the information "leaked" by $Y$.
* **The "Optimality" Identity:**
The conditional expectation $\mathbb{E}[X|Y]$ is the unique estimator that achieves this minimum possible error. Any other prediction method will result in a larger MSE:
$$\mathbb{E}[(X - \text{estimate})^2] \ge \mathbb{E}[(X - \mathbb{E}[X|Y])^2]$$




# Data as realizations of random variables

## The data generation process

Let each row of an industrial dataset be modeled as a $m$-dimensional matrix valued **random variable:**  
$$\mathbf{X}_i = [X_{i,1}, X_{i,2}, \dots, X_{i,m}]^T,$$ that is governed by an underlying, unknown multivariate probability distribution. Each individual random variable $\mathbf{X}_i$ is characterized by its first two theoretical moments:

* **The Population Mean:** $\boldsymbol{\mu}_i = \mathbb{E}[\mathbf{X}_i] \in \mathbb{R}^{m}$
* **The Population Covariance Matrix:** $\boldsymbol{\Sigma}_i = \mathbb{E}[(\mathbf{X}_i - \boldsymbol{\mu}_i)(\mathbf{X}_i - \boldsymbol{\mu}_i)^T] \in \mathbb{R}^{m \times m}$

When collecting a dataset of size $n$, we observe a finite sequence of these random matrices, denoted as $\{\mathbf{X}_i\}_{i=1}^n$.
The actual dataset processed by a computer is the matrix of empirical **realizations**—the concrete numerical values extracted from the system.

---

Formally, we define the column matrix
$$\mathbf{x}_i = [x_{i,1}, x_{i,2}, \dots, x_{i,m}]^T \in \mathbb{R}^{m}$$
as the observed realization of the random variable $\mathbf{X}_i$. Mapping the $j$-th feature of the $i$-th sample as the scalar $x_{i,j}$ yields the structured $n \times m$ observed data matrix $\mathbf{X}_{\text{obs}}$:

$$\mathbf{X}_{\text{obs}} = \begin{bmatrix}
x_{1,1} & x_{1,2} & \dots & x_{1,m} \\
x_{2,1} & x_{2,2} & \dots & x_{2,m} \\
\vdots & \vdots & \ddots & \vdots \\
x_{n,1} & x_{n,2} & \dots & x_{n,m}
\end{bmatrix} = \begin{bmatrix} \mathbf{x}_1^T \\ \mathbf{x}_2^T \\ \vdots \\ \mathbf{x}_n^T \end{bmatrix}.$$

---

To justify the application of classical multivariate statistical inference, we impose the foundational assumption that the sequence of random varaibles $\{X_i\}_{i=1}^n$ is **independent and identically distributed (i.i.d.)**.

1. **Identical Distribution:** Every random variable $\mathbf{X}_i$ possesses the exact same probability law. Consequently, the distinct population moments collapse into a single, universal system profile:

$$\boldsymbol{\mu}_i = \boldsymbol{\mu} \quad \text{and} \quad \boldsymbol{\Sigma}_i = \boldsymbol{\Sigma} \quad \forall \ i \in \{1, 2, \dots, n\}$$

2. **Mutual Independence:** The joint probability distribution of the sequence collapses into the product of their individual distributions, written as $\mathbf{X}_1 \perp \mathbf{X}_2 \perp \dots \perp \mathbf{X}_n$. This guarantees that no combination of rows carries hidden, collective information about any other row, meaning the physical state captured by any subset of observations exerts zero statistical influence over the rest.

---

Mutual Independence implies that the joint Probability Density Function (PDF), denoted by $f_{\mathbf{X}_1, \mathbf{X}_2, \dots, \mathbf{X}_n}$ satisfies:

$$f_{\mathbf{X}_1, \mathbf{X}_2, \dots, \mathbf{X}_n}(\mathbf{x}_1, \mathbf{x}_2, \dots, \mathbf{x}_n) = \prod_{i=1}^{n} f_{\mathbf{X}}(\mathbf{x}_i) \quad \forall \mathbf{x}_i \in \mathbb{R}^m$$

where $f_{\mathbf{X}}(\mathbf{x}_i)$ represents the identical marginal PDF evaluated at the $i$-th realization.

## Empirical Mean, Covariance and the Pearson Correlation

The first and second empirical moments calculated directly from the observed realizations of $\mathbf{X}_i \in \mathbb{R}^{m}$ are:

* **The Empirical Mean ($\widehat{\boldsymbol{\mu}}_n \in \mathbb{R}^{m \times 1}$):** The sample average mapping the spatial center of gravity of the observed feature space.

$$\widehat{\boldsymbol{\mu}}_n = \frac{1}{n} \sum_{i=1}^n \mathbf{X}_i $$


* **The Empirical Covariance Matrix ($\mathbf{S} \in \mathbb{R}^{m \times m}$):** Utilizing Bessel's correction ($n-1$) to ensure an unbiased estimator ($\mathbb{E}[\mathbf{S}] = \boldsymbol{\Sigma}$) is
$$\mathbf{S} = \frac{1}{n-1} \sum_{i=1}^n (\mathbf{X}_i - \widehat{\boldsymbol{\mu}}_n)(\mathbf{X}_i - \widehat{\boldsymbol{\mu}}_n)^T.$$

* **Pearson Correlation Matrix ($\mathbf{R}$):** Standardizes the sample covariance matrix, $\mathbf{S}$, between $[-1, 1]$.
$$\mathbf{R} = \mathbf{D}^{-1/2} \mathbf{S} \mathbf{D}^{-1/2}$$
where
$$\mathbf{D} = \text{diag}(\mathbf{S}) = \begin{bmatrix} s_1^2 & 0 & \dots & 0 \\ 0 & s_2^2 & \dots & 0 \\ \vdots & \vdots & \ddots & \vdots \\ 0 & 0 & \dots & s_m^2 \end{bmatrix}.$$
Explicitly writtent down the $jk$ matrix element of $\mathbf{R}$ is
$$r_{jk} = \frac{s_{jk}}{s_j \cdot s_k} = \frac{\sum_{i=1}^n (x_{i,j} - \widehat{\mu}_j)(x_{i,k} - \widehat{\mu}_k)}{\sqrt{\sum_{i=1}^n (x_{i,j} - \widehat{\mu}_j)^2} \sqrt{\sum_{i=1}^n (x_{i,k} - \widehat{\mu}_k)^2}}.$$
