# 1. Foundations and Applications of the Maximum Entropy Principle

The **Principle of Maximum Entropy (MaxEnt)** represents a profound paradigm shift in how we interpret probability and statistical modeling. Historically, statistical mechanics (developed by Boltzmann and Gibbs) treated probabilities as physical frequencies of microscopic states. However, in 1957, physicist **Edwin T. Jaynes** realized that statistical mechanics could be entirely re-founded as a pure branch of statistical inference.

Jaynes argued that a probability distribution is not an inherent property of a physical system, but rather a representation of our state of knowledge (or lack thereof) about that system.

If we are tasked with assigning a probability distribution over a state space given only a limited set of macroscopic measurements, how do we proceed without introducing arbitrary bias? Jaynes provided an elegant answer:

> **The Principle of Maximum Entropy:** The most rational, least-biased probability distribution is the one that maximizes our uncertainty (entropy) subject to the constraints of our known data.

Specifically we seek a discrete probability distribution $p_i = P(X = x_i)$ for $i = 1, \dots, k$ that maximizes the **Shannon Entropy** $H(P)$:

$$\max_{p_1, \dots, p_k} H(P) = -\sum_{i=1}^{k} p_i \ln p_i$$

Subject to the normalization constraint and $m$ macrostate measurement constraints:

$$\sum_{i=1}^{k} p_i = 1 \quad \text{and} \quad \sum_{i=1}^{k} p_i f_j(x_i) = \langle f_j \rangle \quad (j = 1, \dots, m)$$



---



In data science and machine learning, this exact framework is known as the **Maximum Entropy Classifier**, which is mathematically identical to **Multiclass Logistic Regression** (or the Softmax Classifier).

Suppose we want to predict a categorical label $Y \in \{1, \dots, C\}$ given an input feature array $\mathbf{x}$. We extract data features $\mathbf{f}(\mathbf{x}, c)$ from our training dataset and measure their historical empirical averages.

To build an unbiased model, we maximize the conditional Shannon entropy $H(Y|\mathbf{x})$ subject to matching those observed averages. Mapping the optimization directly to our MaxEnt derivation reveals that the model weights $\mathbf{w}$ behave precisely like the physical Lagrange multipliers $\boldsymbol{\lambda}$:

$$P(Y = c \mid \mathbf{x}) = \frac{\exp\left( \mathbf{w}_c \cdot \mathbf{f}(\mathbf{x}) \right)}{\sum_{l=1}^C \exp\left( \mathbf{w}_l \cdot \mathbf{f}(\mathbf{x}) \right)}$$

This is the standard **Softmax** activation function used globally in neural networks and linear classifiers.

When a machine learning algorithm optimizes its weights using cross-entropy loss (Gradient Descent), it modifies $\mathbf{w}$ until the model's predicted feature averages perfectly match the real world's macroscopic data averages. At that exact point, the model reaches a state of informational "equilibrium."

---


When transitioning from a discrete set of microstates to a continuous state space $X \in \mathbb{R}^d$, replacing the summation with an integral ($-\int p(x) \ln p(x) dx$) creates a mathematical flaw known as differential entropy. Differential entropy is highly problematic because it changes when you change coordinates, and it can become infinitely negative.

To preserve mathematical rigor, Jaynes introduced an explicit baseline reference density $m(x)$, which represents the invariant measure (or prior weight) of the space. The continuous **Shannon-Jaynes entropy** functional is defined as:

$$H_c[p] \triangleq -\int_{\mathbb{R}^d} p(x) \ln \left( \frac{p(x)}{m(x)} \right) dx$$

We maximize this functional subject to continuous expectation constraints:

$$\int_{\mathbb{R}^d} p(x) \, dx = 1 \quad \text{and} \quad \int_{\mathbb{R}^d} p(x) f_i(x) \, dx = \langle f_i(x)\rangle \quad (i=1, \dots, m)$$


A powerful demonstration of the continuous MaxEnt principle explains why the **Normal (Gaussian) Distribution** appears universally in nature and statistics: it is the single most disordered, least-biased distribution possible for a given mean and variance.

Let our domain be $X \in (-\infty, \infty)$ with a completely uniform baseline measure $m(x) = 1$. We enforce two explicit macroscopic constraints from our data measurements:

1. **The Mean:** $\mathbb{E}[X] = \int_{-\infty}^{\infty} p(x) \cdot x \, dx = \mu$
2. **The Variance:** $\mathbb{E}[(X-\mu)^2] = \int_{-\infty}^{\infty} p(x) \cdot (x - \mu)^2 \, dx = \sigma^2$

Plugging these constraints directly into our continuous MaxEnt equation gives:

$$p^*(x) = \frac{1}{Z} \exp\left( -\lambda_1 x - \lambda_2 (x - \mu)^2 \right)$$

Evaluating the definitive Gaussian integrals to isolate our normalization constant $Z$ and the scaling multipliers $\lambda_1$ and $\lambda_2$ resolves cleanly to:

$$\lambda_1 = 0, \quad \lambda_2 = \frac{1}{2\sigma^2}, \quad Z = \sqrt{2\pi\sigma^2}$$

Substituting these parameters back into the distribution yields the classic probability density function of the Normal Distribution:

$$p^*(x) = \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left( -\frac{(x - \mu)^2}{2\sigma^2} \right)$$

This exposes a fundamental truth about data science and physics: if you model a continuous system knowing only its mean and variance, assuming any distribution other than a Gaussian actively injects unproven, biased assumptions into your calculations.

# 2. Maximum Entropy Inference Through Statistical Mechanics

## 2.1. Overview

In Thermodynamics We have seen that that there exists many micro-states that correspond to a given macro-state. A macro-state is characterized by some average behavior of the collection of particles. Variables such as the number of particles, energy, temperature, pressure, and volume are some quantities that represent this average behavior and hence help characterize the macroscopic state. Given such information what is the best estimate that we can make of any other unknown property of the system? What we mean by best here is that those estimates must be unbiased. We will see that maximizing entropy gives us such an unbiased estimate. That is we shall see that by starting with entropy as a fundamental property and applying notions of statistical inference allows us to recover the principles of statistical mechanics without having to invoke any additional assumptions other than the classically known laws of nature.

## 2.2. Micro-State of a Thermodynamic System

Let us denote by $\Omega$ the set made up of all possible situations that a thermodynamic system can exist. That is $\Omega$ is the sample space consisting of all possible microscopical physical configurations of the system.


Since there is practically no way of inferring the microscopical physical configuration of a system nor the exact influence the surrounding has on the system the best we can do is to assume that the microscopical physical configuration of the system occurs according to some random process.

We will model the micro-state of a system by a random variable:
$$X:\Omega \to \mathscr{X},$$
where $\mathscr{X}$ is some discrete set of possibly infinitely many labels. That is $\mathscr{X}=\{x_1,x_2,\cdots x_k\}$ where $k$ can be $\infty$. Each $x_i\in \mathscr{X}$ denotes a unique micro-state of the system.

Let $(\mathscr{X},\mathfrak{P}(\mathscr{X}),P)$ be a probability model defined on $\mathscr{X}$ and the $\sigma$-algebra of the powerset of $\mathscr{X}$.
Denote by $P(\{x=x_j\})=p_j$ the probability of the system being in the microstate $x_j$.

A measurement or macroscopic observation of the system corresponds to a random variable $f_i:\mathscr{X}\to \mathbb{R}$. That is we assume that such observations depend only on the micro-state of the system.
For a given system the number of particles and the energy of the system are some such random variables.  Depending on what information (the measurements) we have about the system we can classify systems into three distinct groups.

* **Open Systems:** Systems that can exchange both particles and energy with the surrounding. Both the energy and the number of particles of the system can change and can be measured. That is the measurements we have are the number of particles $N(X)$ and the Energy $E(X)$.
* **Closed Systems:** Systems that can exchange \textit{only energy} with the surrounding. The energy of the system can change $E(X)$ while the number of particles $N$ remains fixed. Thus the only measurement we have is the energy $E(X)$.

* **Isolated Systems:** Systems that *can not* exchange particles nor energy with the surrounding. We have no information of the system other than that the energy $E$ and the number of particles $N$ remain fixed.

## 2.3. Given Macroscopic Measurements What Can We Say About the System?

In each macroscopic circumstance, we seek the best estimate of the system's microstate given limited information. Specifically, if our only data consists of the expectation values of a set of $m$ functions, $\mathbb{E}[f_i(X)] \triangleq \bar{f}_i\triangleq\langle{f}_i(x)\rangle$, what is the most rational expectation value for an unmeasured function $g(X)$?

We know the expectation values of $f_i(X)$, but the underlying probability distribution $\{p_j\}_{j=1}^k$ over the microstates $x_j$ is unknown. The known constraints are:


$$\mathbb{E}[f_i(X)] = \sum_{j=1}^k p_j f_i(x_j) \triangleq \langle{f}_i(x)\rangle, \quad \text{for } i=1, \dots, m$$

$$\sum_{j=1}^k p_j = 1, \quad \text{where } p_j \ge 0.$$



>The problem we need to solve is to find the **most unbiased probability distribution $\{p_j\}_{j=1}^k$** given that $\mathbb{E}[{f}_i(X)]$ is known. Intuitively the most unbiased distribution should be the one with the most uncertainty.


## 2.4. Maximum Entropy Inference for Discrete Probability Spaces


To find the most unbiased probability distribution, we utilize **Jaynes' Principle of Maximum Entropy**. Claude Shannon demonstrated that the unique mathematical measure of uncertainty associated with a discrete probability distribution is the information entropy:


$$H(p_1, \dots, p_k) = -\sum_{j=1}^k p_j \ln p_j$$


*Note: In a physical context, this is scaled to thermodynamic entropy via $S = k_B H$.*

Because $H$ is a strictly concave function and the constraints are linear equations, the optimization problem possesses a unique global maximum. We find this maximum distribution $\{p_j^*\}$ by constructing the Lagrangian function $\mathscr{L}$:


$$\mathscr{L}(\{p_j\}, \gamma, \{\lambda_i\}) = -\sum_{j=1}^k p_j \ln p_j - \gamma \left(\sum_{j=1}^k p_j - 1\right) - \sum_{i=1}^m \lambda_i \left(\sum_{j=1}^k p_j f_i(x_j) - \langle{f}_i(x)\rangle\right)$$

Since the Lagrange multipliers decouple the constraints, we can vary the probabilities $p_j$ independently. Setting the partial derivatives with respect to each $p_j$ to zero yields:


$$\frac{\partial \mathscr{L}}{\partial p_j} = -(1 + \ln p_j) - \gamma - \sum_{i=1}^m \lambda_i f_i(x_j) = 0$$

Solving for $p_j$ gives the generalized Boltzmann distribution:


$$p_j^* = \exp\left(-(1+\gamma) - \sum_{i=1}^m \lambda_i f_i(x_j)\right)$$

Applying the normalization constraint $\sum p_j^* = 1$, we define the **Partition Function**, $Z$:


$$Z(\lambda_1, \dots, \lambda_m) \triangleq e^{1+\gamma} = \sum_{j=1}^k \exp\left(-\sum_{i=1}^m \lambda_i f_i(x_j)\right)$$

Thus, the canonical maximum entropy distribution is uniquely determined by:


$$p_j^* = \frac{1}{Z} \exp\left(-\sum_{i=1}^m \lambda_i f_i(x_j)\right)$$



### 2.4.1 The Maximum of Binary Entropy $H_b(p)$

The entropy of a binary system (like a coin flip with success probability $p$ and failure probability $1-p$) is given by:


$$H_b(p) = -p \ln p - (1-p) \ln(1-p)$$

The below figure shows that the uncertainty is maximized at a perfect 50/50 split ($p = 0.5$), we use standard calculus to find where the first derivative equals zero, and verify it is a maximum using the second derivative.

```python
import numpy as np
import plotly.graph_objects as go

# 1. Generate probabilities p from 0 to 1 (excluding absolute 0 and 1 to prevent log(0) errors)
p = np.linspace(1e-7, 1 - 1e-7, 500)

# 2. Compute binary entropy H_b(p) in nats
H_b = -p * np.log(p) - (1 - p) * np.log(1 - p)

# 3. Create the interactive Plotly figure
fig = go.Figure()

# Plot the continuous Binary Entropy curve
fig.add_trace(
    go.Scatter(
        x=p,
        y=H_b,
        mode='lines',
        name='Binary Entropy $H_b(p)$',
        line=dict(color='#1f77b4', width=3)
    )
)

# Highlight the maximum uncertainty point (p = 0.5, H = ln(2))
fig.add_trace(
    go.Scatter(
        x=[0.5],
        y=[-0.5 * np.log(0.5) - 0.5 * np.log(0.5)],
        mode='markers+text',
        name='Maximum Uncertainty',
        marker=dict(color='red', size=10, symbol='circle'),
        text=['Max Entropy: $p=0.5, H=\\ln(2)$'],
        textposition='bottom center'
    )
)

# 4. Refine layout options
fig.update_layout(
    title={
        'text': "Binary Entropy Function $H_b(p) = -p \\ln p - (1-p) \\ln(1-p)$",
        'y':0.9,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'
    },
    xaxis_title="Probability $p$ (Success)",
    yaxis_title="Entropy $H_b(p)$ (nats)",
    xaxis=dict(range=[0, 1], tickvals=[0, 0.2, 0.4, 0.5, 0.6, 0.8, 1.0]),
    yaxis=dict(range=[0, 0.8]),
    template="plotly_white",
    hovermode="x unified",
    width=800,
    height=500,
    showlegend=False
)

# Render the interactive plot
fig.show()
```

## 2.5. Recovering Thermodynamic Identities

From this explicit form of $p_j^*$, several rigorous identities follow directly:

1. **Expectation Values:**

$$\frac{\partial \ln Z}{\partial \lambda_i} = \frac{1}{Z} \frac{\partial Z}{\partial \lambda_i} = -\sum_{j=1}^k f_i(x_j) \frac{1}{Z}\exp\left(-\sum_{l=1}^m \lambda_l f_l(x_j)\right) = -\langle f_i(x)\rangle$$


2. **Fluctuations (Variance):**

$$\frac{\partial^2 \ln Z}{\partial \lambda_i^2} = \langle f_i(x)\rangle^2 - \langle f_i(x)^2\rangle = \operatorname{Var}(f_i(x))$$


3. **Maximized Thermodynamic Entropy ($S = k_B H$):**

$$S = -k_B \sum_{j=1}^k p_j^* \ln p_j^* = -k_B \sum_{j=1}^k p_j^* \left(-\ln Z - \sum_{i=1}^m \lambda_i f_i(x_j)\right) = k_B \ln Z + k_B \sum_{i=1}^m \lambda_i \bar{f}_i$$



If the microstate properties $f_i(x_j, \alpha_1, \dots, \alpha_r)$ depend smoothly on external parameters $\alpha$, the total differential of the entropy expands cleanly to:


$$dS = k_B \sum_{i=1}^m \lambda_i d\langle f_i(x)\rangle + k_B \sum_{r} \frac{\partial \ln Z}{\partial \alpha_r} d\alpha_r$$

This framework cleanly derives the Micro-canonical, Canonical, and Grand Canonical ensembles depending strictly on which macrostate averages $\langle f_i(x)\rangle$ are preserved as constraints.

If both the number of particles and the total energy of the particles remains fixed in a system (an isolated system) then the resulting maximum entropy probability distribution is simply the uniform distribution of $p_j^*(x_j)=1/\Omega$ where $Z=\Omega$ is the total number of micro-states. This distribution is referred to as the **Micro-canonical Ensemble**.
On the other hand if only the number of particles of the system is fixed (a closed system) and the system can exchange energy with the surrounding through heat interactions then
picking $f$ to be the energy results in the maximum entropy probability distribution known as the **Canonical Ensemble**. Similarly if the system interacts with the surrounding by exchanging both particles and energy (open systems) then by picking $f_1$ to be the energy and $f_i$ to be the number of particles of the $i^{\mathrm{th}}$ type results in the probability distribution what is knows as the **Grand Canonical Ensemble**. Below we will derive these results and take a brief look at the thermodynamics of these three ensembles. We will start with the Canonical Ensemble then look at the more general Grand Canonical Ensemble and finally look at the Micro-canonical Ensemble.

## 2.6. Thermodynamics of the Canonical Ensemble


Let us consider a closed system that is thermally interacting with a reservoir (or thermal bath). By definition, a closed system can exchange energy—but not particles—with its surroundings. A thermal bath is modeled as a macroscopic system with an exceptionally large number of degrees of freedom, meaning its intensive properties remain invariant under energy exchange with the small system.

The energy eigenvalues of the system's microstates, denoted as $E_j(V)$, depend explicitly on the external parameter volume $V$. If our macroscopic measurement is restricted strictly to the internal average energy $U \triangleq \mathbb{E}[E]$, we can apply the Principle of Maximum Entropy. Setting $f_1(x_j) = E_j(V)$, the canonical partition function is given by:

$$Z(\lambda, V) = \sum_{j} e^{-\lambda E_j(V)}$$

The maximum entropy probability distribution of finding the system in the $j^{\text{th}}$ microstate is:

$$p_j^* = \frac{1}{Z} e^{-\lambda E_j(V)}$$

This statistical ensemble is known as the **Canonical Ensemble**.

The Lagrange multiplier $\lambda$ enforces the constraint on the average internal energy. Because it is a constant at equilibrium, it characterizes the invariant thermal properties of the reservoir. By defining the absolute temperature $T$ of the thermal bath as:

$$T \triangleq \frac{1}{k_B \lambda}$$

we recover the foundational laws of classical thermodynamics.

### 2.6.1 Thermodynamic Identities and Potentials

Substituting the thermodynamic definition of temperature into our maximum entropy framework yields the following core relations:

$$\begin{align*}
Z(T,V) &= \sum_{j} e^{-\frac{E_j(V)}{k_BT}} \\
p_j^* &= \frac{1}{Z} e^{-\frac{E_j(V)}{k_BT}} \\
S(U,T,V) &= k_B \ln Z + \frac{U}{T} \\
\left(\frac{\partial S}{\partial U}\right)_V &= \frac{1}{T}
\end{align*}$$

If we define the **Helmholtz Free Energy** as $F(T,V) \triangleq -k_B T \ln Z$, it serves as the characteristic thermodynamic potential for the canonical ensemble. Rearranging the expression for entropy shows its direct relationship to internal energy:

$$\begin{align*}
S &= -\left(\frac{\partial F}{\partial T}\right)_V \
U &= F + TS
\end{align*}$$

> **Key Takeaway:** The Helmholtz free energy $F(T,V)$ is a bridging function. Analytically evaluating $Z$ completely determines both the entropy $S$ and the internal energy $U$ of a closed system.

---

**The First Law of Thermodynamics:**

Taking the total differential of the entropy $S(T,V)$ using our structural identities reveals:

$$dS = \frac{1}{T}\left(dU - \left\langle \frac{\partial E}{\partial V} \right\rangle dV\right) = \frac{1}{T}\left(dU + k_B T \frac{\partial \ln Z}{\partial V} dV\right) = \frac{1}{T}\left(dU - \left(\frac{\partial F}{\partial V}\right)_T dV\right)$$

Rearranging this differential equation isolates the internal energy change:

$$dU = T dS + \left(\frac{\partial F}{\partial V}\right)_T dV$$

The first law of thermodynamics (conservation of energy) states that for a reversible process in a closed system, $dU = \delta Q - \delta W = \delta Q - P dV$, where $P$ is the macroscopic pressure. Comparing these terms yields:

$$\begin{align*}
\delta Q &= T dS \
P &\triangleq -\left(\frac{\partial F}{\partial V}\right)_T = -\left\langle \frac{\partial E}{\partial V} \right\rangle = T\left(\frac{\partial S}{\partial V}\right)_T - \left(\frac{\partial U}{\partial V}\right)_T
\end{align*}$$

Furthermore, we observe that the expected microscopic work matches the classical thermodynamic work:

$$\langle dE \rangle = \delta W = P dV$$

---

**Energy Fluctuations and Heat Capacity**

The constant-volume heat capacity of a closed system is defined as:

$$C_V = \left(\frac{\partial U}{\partial T}\right)_V = T\left(\frac{\partial S}{\partial T}\right)_V$$

Using our earlier MaxEnt identity for the variance of a constrained function ($\operatorname{Var}(f) = \frac{\partial^2 \ln Z}{\partial \lambda^2}$), we map the energy variance of the ensemble using the parameter $\lambda$:

$$\langle E^2 \rangle - \langle E \rangle^2 = \frac{\partial^2 \ln Z}{\partial \lambda^2} = -\frac{\partial^2 (\lambda F)}{\partial \lambda^2} = -2\frac{\partial F}{\partial \lambda} - \lambda \frac{\partial^2 F}{\partial \lambda^2}$$

Transforming derivatives back from $\lambda \to T$ via the chain rule ($\frac{\partial}{\partial \lambda} = -k_B T^2 \frac{\partial}{\partial T}$) simplifies the expression:

$$\langle E^2 \rangle - \langle E \rangle^2 = -k_B T^3 \left(\frac{\partial^2 F}{\partial T^2}\right)_V = k_B T^2 \left[ T \left(\frac{\partial S}{\partial T}\right)_V \right] = k_B T^2 C_V$$

Isolating the heat capacity reveals a profound connection between macroscopic thermodynamics and microscopic statistical fluctuations:

$$C_V = \frac{1}{k_B T^2} \left( \langle E^2 \rangle - \langle E \rangle^2 \right)$$

> **Fluctuation-Dissipation Note:** The heat capacity—a macroscopic measure of a system's thermal responsiveness—is directly proportional to the variance of the energy distribution among its microscopic states.



### 2.6.2 Example: The quantum Harmonic Oscillator

To illustrate the canonical framework, consider a single quantum harmonic oscillator of angular frequency $\omega$ in thermal equilibrium with a heat bath at temperature $T$.

The quantum mechanical energy eigenvalues are given by:

$$E_j = \hbar \omega \left(j + \frac{1}{2}\right), \quad \text{for } j = 0, 1, 2, \dots$$

Where $\hbar$ is the reduced Planck constant. The partition function $Z(T)$ forms a convergent geometric series:

$$Z(T) = \sum_{j=0}^\infty e^{-\frac{\hbar \omega (j + 1/2)}{k_BT}} = e^{-\frac{\hbar \omega}{2k_BT}} \sum_{j=0}^\infty \left(e^{-\frac{\hbar \omega}{k_BT}}\right)^j = \frac{e^{-\frac{\hbar \omega}{2k_BT}}}{1 - e^{-\frac{\hbar \omega}{k_BT}}}$$

Using this explicit function for $Z(T)$, we compute the standard thermodynamic potentials and properties of the oscillator:

**Helmholtz Free Energy:**

$$F(T) = -k_B T \ln Z = \frac{\hbar \omega}{2} + k_B T \ln \left(1 - e^{-\frac{\hbar \omega}{k_BT}}\right)$$

**Entropy:**

$$S(T) = -\left(\frac{\partial F}{\partial T}\right) = -k_B \ln \left(1 - e^{-\frac{\hbar \omega}{k_BT}}\right) + \frac{\hbar \omega}{T} \frac{1}{e^{\frac{\hbar \omega}{k_BT}} - 1}$$

**Internal Energy:**

$$U(T) = F + TS = \hbar \omega \left(\frac{1}{2} + \frac{1}{e^{\frac{\hbar \omega}{k_BT}} - 1}\right)$$

**Heat Capacity:**

$$C_V = \left(\frac{\partial U}{\partial T}\right) = k_B \left(\frac{\hbar \omega}{k_BT}\right)^2 \frac{e^{\frac{\hbar \omega}{k_BT}}}{\left(e^{\frac{\hbar \omega}{k_BT}} - 1\right)^2}$$

Using the hyperbolic identity $\sinh(x) = \frac{e^x - e^{-x}}{2}$, the heat capacity expression can be compactly expressed in terms of its fluctuation structure:

$$C_V = k_B \left(\frac{\hbar \omega}{2k_BT}\right)^2 \frac{1}{\sinh^2\left(\frac{\hbar \omega}{2k_BT}\right)}$$

## 2.7. Thermodynamics of the Grand Canonical Ensemble

Let us consider an open system that is thermally and chemically interacting with a reservoir (a thermal and particle bath). This means the system can freely exchange both energy and particles with its surroundings. For simplicity, we assume the system and reservoir contain only one chemical type of particle.

The energy eigenvalues $E_j(V)$ and particle numbers $n_j$ of the microstates depend on the system's volume $V$. The macroscopic constraints we measure are the average number of particles, $N \triangleq \mathbb{E}[n]$, and the average internal energy, $U \triangleq \mathbb{E}[E]$. Setting our MaxEnt constraint functions to $f_1(x_j) = E_j(V)$ and $f_2(x_j) = n_j$, the core equations yield the **Grand Canonical Ensemble**.

The grand canonical partition function $Z$ is defined as:

$$Z(\lambda_1, \lambda_2, V) = \sum_{j} e^{-\lambda_1 E_j(V) - \lambda_2 n_j}$$

The maximum entropy probability distribution of finding the open system in the $j^{\text{th}}$ microstate is given by:

$$p_j^* = \frac{1}{Z} e^{-\lambda_1 E_j(V) - \lambda_2 n_j}$$

Because the Lagrange multipliers $\lambda_1$ and $\lambda_2$ are constant at global equilibrium, they must map to the invariant intensive properties of the reservoir. We define the absolute temperature $T$ and the chemical potential $\mu$ of the reservoir as follows:

$$T \triangleq \frac{1}{k_B \lambda_1}, \quad \mu \triangleq -k_B T \lambda_2.$$



### 2.7.1 Thermodynamic Properties and Potential

Substituting these definitions into our MaxEnt distribution expressions yields:

$$\begin{align*}
Z(T, \mu, V) &= \sum_{j} e^{\frac{\mu n_j - E_j(V)}{k_BT}} \\
p_j^* &= \frac{1}{Z} e^{\frac{\mu n_j - E_j(V)}{k_BT}} \\
S(U, N, T, V) &= k_B \ln Z + \frac{U}{T} - \frac{\mu N}{T}
\end{align*}$$

Taking partial derivatives of the entropy reveals the thermodynamic definition of temperature and chemical potential:

$$\left(\frac{\partial S}{\partial U}\right)_{N,V} = \frac{1}{T}, \quad \left(\frac{\partial S}{\partial N}\right)_{U,V} = -\frac{\mu}{T}$$

We now define the **Grand Potential** (or Landau Potential) as $\Phi(T, \mu, V) \triangleq -k_B T \ln Z$. Evaluating the partial derivatives of $\Phi$ directly provides our primary thermodynamic parameters:

$$\begin{align*}
S &= -\left(\frac{\partial \Phi}{\partial T}\right)_{\mu, V} \\
N &= -\left(\frac{\partial \Phi}{\partial \mu}\right)_{T, V} \\
\Phi &= U - TS - \mu N
\end{align*}$$

> **Key Takeaway:** The Grand Potential $\Phi(T, \mu, V)$ serves as the characteristic bridging function for open systems. Knowing $\Phi$ uniquely determines the entropy $S$, the expected particle number $N$, and the total expected internal energy $U$.

---

**The First Law for Open Systems:**

By taking the total differential of the maximized entropy $S$ and incorporating the structural parameters, we find:

$$dS = \frac{1}{T}\left( dU - \left\langle \frac{\partial E}{\partial V} \right\rangle dV \right) - \frac{\mu}{T} dN$$

Isolating the internal energy differential $dU$ establishes the fundamental identity of thermodynamics for open systems:

$$dU = T dS + \left(\frac{\partial \Phi}{\partial V}\right)_{T,\mu} dV + \mu dN$$

The first law of thermodynamics states that the internal energy of an open system shifts due to heat transfer, physical work, and the chemical work of adding or removing mass ($dU = \delta Q - P dV + \mu dN$). By comparing this with our MaxEnt derivation, we cleanly recover:

$$\begin{align*}
\delta Q &= T dS \
P &\triangleq -\left(\frac{\partial \Phi}{\partial V}\right)_{T,\mu} = -\left\langle \frac{\partial E}{\partial V} \right\rangle
\end{align*}$$

Using the Euler relation for macroscopic systems, where $\Phi = -PV$, we can express the true **Gibbs Free Energy** ($G$) directly from our terms:

$$G \triangleq U - TS + PV = \mu N$$

> **Conclusion:** Calibrating our statistical multipliers to $T = \frac{1}{k_B \lambda_1}$ and $\mu = -k_B T \lambda_2$ provides an exact recovery of classical open-system thermodynamics. Maximizing Shannon entropy under simultaneous energy and particle constraints generates the rigorous foundational laws governing chemical potential and phase equilibrium.

## 2.8. Thermodynamics of the Microcanonical Ensemble

Consider an isolated system containing a fixed number of particles $N$ confined within a constant volume $V$. Because the system is completely isolated from the rest of the universe, it cannot exchange energy or matter. Therefore, the total internal energy of the system is strictly fixed at a macroscopic value $U$.

Let $\Omega(U, V, N)$ denote the total number of distinct microstates whose energy eigenvalues exactly match $U$. Mathematically, our optimization problem is subject to a hard restriction: the probability $p_j$ must be zero for any state outside this energy shell.

To find the most unbiased distribution among the $\Omega$ accessible states, we maximize the Shannon entropy:

$$H = -\sum_{j=1}^{\Omega} p_j \ln p_j$$

Subject only to the fundamental axiom of total probability:

$$\sum_{j=1}^{\Omega} p_j = 1$$

Because there are no fluctuating macroscopic expectation constraints to differentiate one accessible state from another, maximizing $H$ via a single Lagrange multiplier yields a completely uniform distribution:

$$p_j^* = \frac{1}{\Omega}, \quad \text{for all } j \in \{1, 2, \dots, \Omega\}$$

This result is known as the fundamental postulate of statistical mechanics: **equal a priori probabilities**. If no external information renders a specific microstate special, the most unbiased assumption is that all accessible microstates are equally likely.

### 2.8.1. The Boltzmann Entropy Link

Substituting the uniform distribution $p_j^* = \frac{1}{\Omega}$ back into the definition of thermodynamic entropy ($S = k_B H$) yields:

$$S = -k_B \sum_{j=1}^{\Omega} \frac{1}{\Omega} \ln\left(\frac{1}{\Omega}\right) = -k_B \cdot \Omega \left( \frac{1}{\Omega} (-\ln \Omega) \right)$$

This simplifies directly to the iconic **Boltzmann Entropy Formula**:

$$S(U, V, N) = k_B \ln \Omega$$

Through the Principle of Maximum Entropy, Boltzmann’s statistical definition of entropy for an isolated system is derived cleanly as a special case of Shannon information theory.

### 2.8.2. Deriving Intensive Properties

In the microcanonical framework, entropy $S(U, V, N)$ acts as the core fundamental potential. The macroscopic intensive parameters—**Temperature ($T$)**, **Pressure ($P$)**, and **Chemical Potential ($\mu$)**—are defined internally by analyzing how the total number of accessible microstates scales when we perturb the boundary constraints ($U, V, N$).

Taking the total differential of the fundamental relation $S = S(U, V, N)$ gives:

$$dS = \left(\frac{\partial S}{\partial U}\right)_{V,N} dU + \left(\frac{\partial S}{\partial V}\right)_{U,N} dV + \left(\frac{\partial S}{\partial N}\right)_{U,V} dN$$

Comparing this mathematical expansion to the physical first law rearranged for an isolated system ($dS = \frac{1}{T} dU + \frac{P}{T} dV - \frac{\mu}{T} dN$), we formally define the intensive thermodynamic properties of the system via the structure of its microstate space:

$$\begin{align*}
\frac{1}{T} &\triangleq \left(\frac{\partial S}{\partial U}\right)_{V,N} = k_B \left(\frac{\partial \ln \Omega}{\partial U}\right)_{V,N} \\
\frac{P}{T} &\triangleq \left(\frac{\partial S}{\partial V}\right)_{U,N} = k_B \left(\frac{\partial \ln \Omega}{\partial V}\right)_{U,N} \
-\frac{\mu}{T} &\triangleq \left(\frac{\partial S}{\partial N}\right)_{U,V} = k_B \left(\frac{\partial \ln \Omega}{\partial N}\right)_{U,V}
\end{align*}$$

> **Conclusion:** For an isolated system, you do not calculate a partition function $Z$ to determine thermodynamic properties. Instead, you directly count the number of accessible quantum states $\Omega$. The geometric and physical derivatives of this state-count cleanly expose the system's temperature, pressure, and chemical potential.

# 3. Continuous Principle of Maximum Entropy

## 3.1. Problem Formulation

Let $X$ be a continuous random variable taking values in $\mathbb{R}^d$. To properly transition from the discrete case, we must introduce a baseline reference density $m(x)$, which represents the invariant measure (or prior weight) of the state space.

The mathematically rigorous continuous entropy—often called the **Shannon-Jaynes entropy** or negative **Kullback-Leibler divergence**—is defined as:

$$H_c[p] \triangleq -\int_{\mathbb{R}^d} p(x) \ln \left( \frac{p(x)}{m(x)} \right) dx$$

**The Role of $m(x)$:**

* **In Physics:** $m(x)$ corresponds to the structure of classical phase space. For position $q$ and momentum $p$, Liouville's theorem dictates that $m(q,p) = \frac{1}{h^3}$, ensuring the entropy is invariant under canonical transformations.
* **In Machine Learning:** $m(x)$ is typically chosen as a uniform distribution over a bounded domain, or a reference prior distribution when the domain is infinite.


## 3.2. The Continuous Optimization Framework

Suppose we do not know the exact probability density function $p(x)$, but we are able to measure the expectation values of $m$ independent continuous functions, $f_i(x)$.

Our optimization problem is bounded by the following constraints:

1. **Normalization:** $\int_{\mathbb{R}^d} p(x) \, dx = 1$
2. **Macroscopic Expectations:** $\mathbb{E}[f_i(X)] = \int_{\mathbb{R}^d} p(x) f_i(x) \, dx = \langle f_i(x)\rangle =\bar{f}_i\quad \text{for } i = 1, \dots, m$

To find the most unbiased continuous density function $p^*(x)$, we construct the functional Lagrangian $\mathscr{L}[p]$ using the calculus of variations:

$$\mathscr{L}[p] = -\int_{\mathbb{R}^d} p(x) \ln \left( \frac{p(x)}{m(x)} \right) dx - \gamma \left( \int_{\mathbb{R}^d} p(x) \, dx - 1 \right) - \sum_{i=1}^m \lambda_i \left( \int_{\mathbb{R}^d} p(x) f_i(x) \, dx - \bar{f}_i \right)$$

---


To find the stationary point, we take the functional derivative (Euler-Lagrange derivative) of $\mathscr{L}[p]$ with respect to $p(x)$ and set it to zero:

$$\frac{\delta \mathscr{L}}{\delta p(x)} = -\ln \left( \frac{p(x)}{m(x)} \right) - 1 - \gamma - \sum_{i=1}^m \lambda_i f_i(x) = 0$$

Solving explicitly for the target distribution $p(x)$ yields:

$$p^*(x) = m(x) \exp\left( -(1+\gamma) - \sum_{i=1}^m \lambda_i f_i(x) \right)$$

By enforcing the normalization constraint $\int_{\mathbb{R}^d} p^*(x) \, dx = 1$, we can eliminate the multiplier $\gamma$ by defining the continuous **Partition Function**, $Z$:

$$Z(\lambda_1, \dots, \lambda_m) \triangleq e^{1+\gamma} = \int_{\mathbb{R}^d} m(x) \exp\left( -\sum_{i=1}^m \lambda_i f_i(x) \right) dx$$

Thus, the unique, most unbiased continuous probability density function is given by:

$$p^*(x) = \frac{m(x)}{Z} \exp\left( -\sum_{i=1}^m \lambda_i f_i(x) \right)$$

is the maximum relative entropy density with respect to the reference density $m(x)>0$, provided the constraints are feasible and $Z<\infty$.

## 3.3 The Gaussian Normal as a Max Entropy Distribution

A powerful application of the continuous MaxEnt principle in both statistics and machine learning is proving why the Gaussian distribution is so ubiquitous: **it is the most disordered, least-biased distribution possible for a given mean and variance.**

Let the domain be $X \in (-\infty, \infty)$, and assume a completely uniform reference measure $m(x) = 1$. We enforce two macroscopic constraints:

1. **The Mean:** $\mathbb{E}[X] = \mu$
2. **The Variance:** $\mathbb{E}[(X-\mu)^2] = \sigma^2$

Plugging these constraints ($f_1(x) = x$ and $f_2(x) = (x-\mu)^2$) directly into our continuous MaxEnt solution yields:

$$p^*(x) = \frac{1}{Z} \exp\left( -\lambda_1 x - \lambda_2 (x - \mu)^2 \right)$$
The constraint $\mathbb{E}[X] = \mu$ implies that the linear multiplier must vanish, so $\lambda_1=0$.
By evaluating the standard Gaussian integrals to isolate $\lambda_2$, and $Z$, the multipliers naturally resolve to:

$$\lambda_2 = \frac{1}{2\sigma^2}, \quad Z = \sqrt{2\pi\sigma^2}$$


Substituting these back provides the classic normal distribution density:

$$p^*(x) = \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left( -\frac{(x - \mu)^2}{2\sigma^2} \right)$$

> **Statistical Significance:** If you are modeling data and you only know its mean and variance, assuming *any* distribution other than a Gaussian explicitly introduces arbitrary, unproven assumptions into your machine learning model or physical simulation.

# 4. Application of MaxEnt Inference: Logistic Regrssion

Whenever an engineer uses Logistic Regression or a deep neural network ending in a Softmax layer, they aren't just using a random mathematical trick. They are actually implementing **Jaynes' Principle of Maximum Entropy**. They are creating a mini thermodynamic system where the data categories are microstates, the model weights act as inverse temperatures, and the most stable, unbiased prediction represents the state of thermodynamic equilibrium!

In machine learning classification, we are tasked with predicting a discrete categorical label $Y \in \{1, 2, \dots, C\}$ given an input feature array $\mathbf{X} \in \mathbb{R}^d$. Following the informational philosophy of Jaynes, we want to construct a conditional probability distribution $P(Y =c\mid \mathbf{X}=\mathbf{x})$ that honors the facts extracted from our training data but remains completely uniform and unbiased about everything else.



## 4.1 The Formulation of the Constraints

Suppose we have a training dataset consisting of $N$ observations, $D = \{(\mathbf{x}^{(k)}, y^{(k)})\}_{k=1}^N$. We define a set of joint feature functions $\mathbf{f}(\mathbf{x}, c)\in \mathbb{R}^m$ that capture the interactions between the input features and the output classes, $c\in\{1,2,\cdots, C\}$.
From our training data, we can calculate the empirical joint distribution $$\widetilde{P}(\mathbf{x}, Y) = \frac{1}{N} \sum_{k=1}^N \delta(\mathbf{x} - \mathbf{x}^{(k)})\delta(Y - y^{(k)}).$$
The macroscopic facts available to us are the empirical expectation values of our feature functions:

$$\mathbb{E}_{\widetilde{P}}[\mathbf{f}(\mathbf{x}, c)] = \sum_{\mathbf{x}, Y} \mathbf{f}(\mathbf{x}, c)\widetilde{P}(\mathbf{x}, Y)  = \frac{1}{N} \sum_{k=1}^N \mathbf{f}(\mathbf{x}^{(k)}, y^{(k)}).$$
This is the sample expectation of $\mathbf{f}(\mathbf{x}, c)$.

Our model's distribution $P(\mathbf{x}, c) = \widetilde{P}(\mathbf{x}) P(Y=c \mid \mathbf{X}=\mathbf{x})$ must match these macroscopic constraints exactly. Therefore, for each feature dimension $i$, we require:

$$\mathbb{E}_{P}[f_i(\mathbf{x}, c)] = \mathbb{E}_{\widetilde{P}}[f_i(\mathbf{x}, c)] \implies \sum_{\mathbf{x}} \widetilde{P}(\mathbf{x}) \sum_{c=1}^C P(Y=c \mid  \mathbf{X}=\mathbf{x}) f_i(\mathbf{x}, c) = \frac{1}{N} \sum_{k=1}^N f_i(\mathbf{x}^{(k)}, y^{(k)}).$$

## 4.2. The Conditional MaxEnt Lagrangian

To find the least-biased model, we maximize the **Conditional Shannon Entropy** $H(Y \mid X)$, which measures the average uncertainty remaining in the label $Y$ given that we observe $\mathbf{x}$.

By definition
$$\begin{align*}
H(Y \mid X) &= -\sum_{\mathbf{x}} \sum_{c=1}^C \widetilde{P}(\mathbf{x}, c) \ln P(Y=c \mid  \mathbf{X}=\mathbf{x})
\end{align*}$$

The Entropy of $(Y \mid X=\mathbf{x})$ is:
$$H(Y \mid X=\mathbf{x}) = -\sum_{c=1}^C P(Y=c \mid  \mathbf{X}=\mathbf{x}) \ln P(Y=c \mid  \mathbf{X}=\mathbf{x}).$$
Thus the conditionaly entropy is:
$$\begin{align*}
H(Y \mid X) &= -\sum_{\mathbf{x}} \widetilde{P}(\mathbf{x}) \sum_{c=1}^C P(Y=c \mid  \mathbf{X}=\mathbf{x}) \ln P(Y=c \mid  \mathbf{X}=\mathbf{x}) \\
&= -\sum_{\mathbf{x}} \sum_{c=1}^C \widetilde{P}(\mathbf{x}, c) \ln P(Y=c \mid  \mathbf{X}=\mathbf{x})
\end{align*}$$

Following the discrete MaxEnt framework established previously, we construct the Lagrangian functional $\mathscr{L}$ to maximize $H(Y \mid X)$ subject to our empirical expectation constraints and the fundamental constraint that probabilities must sum to 1 for every individual input $\mathbf{x}$:

$$\begin{align*}
\mathscr{L}(P, {\gamma(\mathbf{x})}, \mathbf{w}^c) = &-\sum_{\mathbf{x}} \widetilde{P}(\mathbf{x}) \sum_{c=1}^C P(Y=c \mid  \mathbf{X}=\mathbf{x}) \ln P(Y=c \mid  \mathbf{X}=\mathbf{x}) \\
&- \sum_{\mathbf{x}} \gamma(\mathbf{x}) \left( \sum_{c=1}^C P(Y=c \mid  \mathbf{X}=\mathbf{x}) - 1 \right) \\
&- \mathbf{w}^c \cdot \left( \sum_{\mathbf{x}} \widetilde{P}(\mathbf{x}) \sum_{c=1}^C P(Y=c \mid  \mathbf{X}=\mathbf{x}) \mathbf{f}(\mathbf{x}, c) - \mathbb{E}_{\widetilde{P}}[\mathbf{f}(\mathbf{x}, c)] \right)
\end{align*}$$

Here, $\mathbf{w}^c = [w^c_1, w^c_2, \dots, w^c_m]^T$ represents the vector of Lagrange multipliers corresponding to our feature expectations, and $\gamma(\mathbf{x})$ represents the point-wise multipliers ensuring local probability normalization.


## 4.3. Solving for the Optimal Distribution

We optimize this functional by taking the partial derivative of $\mathscr{L}$ with respect to the conditional probability of a specific class $c$ given a specific input $\mathbf{x}$, and setting it equal to zero:

$$\frac{\partial \mathscr{L}}{\partial P(Y=c \mid  \mathbf{X}=\mathbf{x})} = -\widetilde{P}(\mathbf{x}) \left( 1 + \ln P(Y=c \mid  \mathbf{X}=\mathbf{x}) \right) - \gamma(\mathbf{x}) - \widetilde{P}(\mathbf{x}) \left( \mathbf{w}^c \cdot \mathbf{f}(\mathbf{x}, c) \right) = 0$$

Dividing through by the empirical data probability $\widetilde{P}(\mathbf{x})$ (on the support of $\widetilde{P}$) and isolating the log probability term gives:

$$\ln P(Y=c \mid  \mathbf{X}=\mathbf{x}) = -\left(1 + \frac{\gamma(\mathbf{x})}{\widetilde{P}(\mathbf{x})}\right) - \mathbf{w}^c \cdot \mathbf{f}(\mathbf{x}, c)$$

Exponentiating both sides reveals the canonical exponential shape of the MaxEnt solution:

$$P^*(Y=c \mid \mathbf{X}=\mathbf{x}) = \exp\left(-\left(1 + \frac{\gamma(\mathbf{x})}{\widetilde{P}(\mathbf{x})}\right)\right) \exp\left( -\mathbf{w}^c \cdot \mathbf{f}(\mathbf{x}, c) \right)$$

By enforcing the structural axiom of total probability
$$\sum_{l=1}^C P^*(Y=l \mid \mathbf{X}=\mathbf{x}) = 1,$$
we eliminate the local multiplier $\gamma(\mathbf{x})$ and define the conditional partition function $Z(\mathbf{x})$:

$$Z(\mathbf{x}) \triangleq \exp\left(1 + \frac{\gamma(\mathbf{x})}{\widetilde{P}(\mathbf{x})}\right) = \sum_{l=1}^C \exp\left( -\mathbf{w}^l \cdot \mathbf{f}(\mathbf{x}, l) \right)$$

This yields the definitive conditional probability distribution:

$$P^*(Y=c \mid \mathbf{X}=\mathbf{x}) = \frac{\exp\left( -\mathbf{w}^c \cdot \mathbf{f}(\mathbf{x}, c) \right)}{\sum_{l=1}^C \exp\left( -\mathbf{w}^l \cdot \mathbf{f}(\mathbf{x}, l) \right)}$$

Since the Lagrange multipliers, $\mathbf{w}$, are arbitray constants without loss of generality we can write:
$$P^*(Y=c \mid \mathbf{X}=\mathbf{x}) = \frac{\exp\left( \mathbf{w}^c \cdot \mathbf{f}(\mathbf{x}, c) \right)}{\sum_{l=1}^C \exp\left( \mathbf{w}^l \cdot \mathbf{f}(\mathbf{x}, l) \right)}$$

## 4.4. Multiclass Logistic Regression (Softmax)

To map this directly to the standard presentation of multi-class classification found in machine learning textbooks, we can choose the specific joint feature functions:

$$\mathbf{f}(\mathbf{x}, c) = \mathbb{I}(Y = c) \mathbf{x}$$

Where $\mathbb{I}$ is the indicator function that equals 1 if the statement is true and 0 otherwise. This decouples the global weight vector $\mathbf{w}$ into class-specific weight vectors $\mathbf{w}_c$ for each category. Substituting this choice back into our general MaxEnt solution yields:


$$P^*(Y=c \mid \mathbf{X}=\mathbf{x}) = \frac{\exp\left( \mathbf{w}_c \cdot \mathbf{x} \right)}{\sum_{l=1}^C \exp\left( \mathbf{w}_l \cdot \mathbf{x} \right)}$$

This is precisely the **Softmax Regression** or **Multiclass Logistic Regression** model.

### 4.4.1 MLE Estimation of $[\mathbf{w}^1,\cdots,\mathbf{w}^C]$

To find the optimal weight vectors $\mathbf{w}^1, \dots, \mathbf{w}^C$ given a training dataset, we maximize the **conditional log-likelihood** of our data.

Suppose your training data consists of $n$ independent observations:
$$\mathscr{D} = \{(\mathbf{x}_1, y_1), (\mathbf{x}_2, y_2), \dots, (\mathbf{x}_n, y_n)\}.$$

The probability of observing the true label $y_i$ for a given input $\mathbf{x}_i$ is:


$$P(Y = y_i \mid \mathbf{X}=\mathbf{x}_i) = \frac{\exp(\mathbf{w}^{y_i} \cdot \mathbf{x}_i)}{\sum_{l=1}^C \exp(\mathbf{w}^l \cdot \mathbf{x}_i)}$$

To find the best weights, we maximize the sum of the log-probabilities across all training examples:

$$\max_{\mathbf{w}^1, \dots, \mathbf{w}^C} \mathscr{L}(\mathbf{w}^1, \dots, \mathbf{w}^C) = \sum_{i=1}^n \ln P(Y = y_i \mid \mathbf{x}_i)$$

Expanding this out gives the classic **multinomial logistic regression objective**:


$$\mathscr{L}(\mathbf{w}^1, \dots, \mathbf{w}^C) = \sum_{i=1}^n \left( \mathbf{w}^{y_i} \cdot \mathbf{x}_i - \ln \sum_{l=1}^C \exp(\mathbf{w}^l \cdot \mathbf{x}_i) \right)$$

---

Unlike linear regression, this equation cannot be solved analytically by setting the derivative to zero and using a closed-form formula. Instead, we use numerical optimization.

To do this, we compute the partial derivative (gradient) of our objective function with respect to a specific class weight vector $\mathbf{w}^c$:

$$\begin{align*}
\nabla_{\mathbf{w}^c} \mathscr{L} &= \sum_{i=1}^n \left( \mathbb{I}(y_i = c) - \frac{\exp\left( \mathbf{w}^c \cdot \mathbf{x}_i \right)}{\sum_{l=1}^C \exp\left( \mathbf{w}^l \cdot \mathbf{x}_i \right)} \right) \mathbf{x}_i \\
&= \sum_{i=1}^n \Big( \mathbb{I}(y_i = c) - P(Y = c \mid \mathbf{x}_i) \Big) \mathbf{x}_i
\end{align*}$$

This gradient has a beautiful physical meaning. For every data point $\mathbf{x}_i$:

* $\mathbb{I}(y_i = c)$ is the **target truth** ($1$ if the true class is $c$, $0$ otherwise).
* $P(Y = c \mid \mathbf{x}_i)$ is the **model's current prediction error** (a probability between $0$ and $1$).
* The term $\big(\text{Truth} - \text{Prediction}\big)$ is the **error signal**. The gradient pushes the weights $\mathbf{w}^c$ to close the gap between your model's predictions and reality.

---


Because the log-likelihood function is strictly **concave** (or convex if you invert it to minimize negative log-likelihood), it has no local minima—only a single, global optimum.

Statisticians and ML practitioners find the weights using iterative numerical optimization algorithms:

**Method A: Gradient Descent (or Stochastic Gradient Descent):**

You initialize all weights $\mathbf{w}^c$ to zero or small random numbers, and repeatedly update them by taking a small step in the direction of the gradient:


$$\mathbf{w}^c \leftarrow \mathbf{w}^c + \eta \cdot \nabla_{\mathbf{w}^c} \mathscr{L}$$


*(where $\eta$ is the learning rate).*

**Method B: Quasi-Newton Methods (L-BFGS):**

In traditional statistical learning packages (like `scikit-learn` in Python), the default solver for multinomial logistic regression / MaxEnt models is often **L-BFGS**. Instead of taking basic gradient steps, it calculates an approximation of the second derivative (the curvature of the loss surface), allowing it to converge to the perfect $\mathbf{w}^1, \dots, \mathbf{w}^C$ values significantly faster than basic gradient descent.



## 4.5 Binary Logistic Regression

When $Y$ takes only two values, $0$ or $1$, the multi-class Softmax collapses directly into standard **Binary Logistic Regression**.

Using our class-specific weight vectors $\mathbf{w}_0$ and $\mathbf{w}_1$, the probability for class $1$ is written as:

$$P^*(Y=1 \mid \mathbf{X}=\mathbf{x}) = \frac{\exp\left( \mathbf{w}_1 \cdot \mathbf{x} \right)}{\exp\left( \mathbf{w}_0 \cdot \mathbf{x} \right) + \exp\left( \mathbf{w}_1 \cdot \mathbf{x} \right)}$$

To eliminate parameter redundancy, we divide the numerator and denominator by $\exp\left( \mathbf{w}_0 \cdot \mathbf{x} \right)$ and define a single unified weight vector $\mathbf{w} = \mathbf{w}_1 - \mathbf{w}_0$:

$$
P^*(Y=1 \mid \mathbf{X}=\mathbf{x}) = \frac{\exp\left( (\mathbf{w}_1 - \mathbf{w}_0) \cdot \mathbf{x} \right)}{1 + \exp\left( (\mathbf{w}_1 - \mathbf{w}_0) \cdot \mathbf{x} \right)} = \frac{\exp\left( \mathbf{w} \cdot \mathbf{x} \right)}{1 + \exp\left( \mathbf{w} \cdot \mathbf{x} \right)}= \frac{1}{1 + \exp\left( -\mathbf{w} \cdot \mathbf{x} \right)}$$

This is precisely the standard **Sigmoid** activation function used in binary logistic regression.

### 4.5.1 The Sigmoid

```python
import numpy as np
import plotly.graph_objects as go

# Define the binary logistic sigmoid function
def sigmoid(x, w):
    return 1 / (1 + np.exp(-w * x))

# Generate a range of x values
x = np.linspace(-10, 10, 500)

# Define different values of w to showcase different behaviors:
# - Small w: gradual slope
# - Large w: steep step-like slope
# - Negative w: flips the orientation of the probability
w_values = [0.2, 0.5, 1.0, 2.5, -1.0]

# Initialize the Plotly figure
fig = go.Figure()

# Add a line trace for each value of w
for w in w_values:
    y = sigmoid(x, w)
    fig.add_trace(go.Scatter(
        x=x,
        y=y,
        mode='lines',
        name=f'w = {w}',
        line=dict(width=2.5)
    ))

# Style the chart layout
fig.update_layout(
    title={
        'text': "Sigmoid Curves for Various Weights <i>w</i>",
        'y':0.9,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'
    },
    xaxis_title="Input Feature (x)",
    yaxis_title="Probability P(Y=1 | x)",
    xaxis=dict(gridcolor='rgba(200, 200, 200, 0.3)', zerolinecolor='gray'),
    yaxis=dict(gridcolor='rgba(200, 200, 200, 0.3)', zerolinecolor='gray', range=[-0.05, 1.05]),
    template="plotly_white",
    legend_title="Weight Value",
    width=800,
    height=500
)

# Render the interactive plot
fig.show()
```

### 4.5.2 Connection to Electron Distribution

The binary logistic regression sigmoid function is mathematically identical to the **Fermi-Dirac distribution** in quantum and statistical physics, which describes the occupancy probability of energy states for electrons (fermions):

$$P(Y=1 \mid \mathbf{X}=x) = \frac{1}{1 + \exp\left(-(wx + b)\right)} \quad \longleftrightarrow \quad f(E) = \frac{1}{1 + \exp\left(\frac{E - \mu}{k_B T}\right)}$$

This deep connection stems from a shared foundational origin: **The Principle of Maximum Entropy**.

* **In Machine Learning:** The sigmoid maximizes Shannon entropy subject to training data constraints for a binary classification outcome ($0$ or $1$).
* **In Physics:** Nature maximizes thermodynamic entropy subject to a energy conservation and the **Pauli Exclusion Principle**, which dictates that a quantum state's electron occupancy can *only* be binary ($0$ or $1$).

Under this physical analogy, the input feature $x$ maps to the energy level $E$, the bias $b$ acts as the chemical potential (Fermi level $\mu$), and the classification weight $w$ behaves identically to the inverse thermal energy scale ($1 / k_B T$). Consequently, a highly certain model with large weights (large $w$) mirrors a physical system near absolute zero ($T \to 0$), freezing the smooth sigmoid curve into a sharp, deterministic step function.

### 4.5.3 Example of a Spam email Filter

Suppose we want to classify a document (e.g., detecting if a short message is *Spam ($Y=1$)* or *Ham ($Y=0$)*) based on a single continuous or indicator feature $x$ (e.g., the frequency of a specific keyword).

Our system has three historical macrostate scenarios defining the background probability distribution $P(x)$ over our feature space $\mathscr{X} = \{x_A, x_B, x_C\}$:

1. **Scenario A (Low Frequency):** $x_A = 0 \implies P(x_A) = 0.50$
2. **Scenario B (Medium Frequency):** $x_B = 1 \implies P(x_B) = 0.35$
3. **Scenario C (High Frequency):** $x_C = 2 \implies P(x_C) = 0.15$

#### Solution Outline

We construct a binary MaxEnt classifier. The conditional distributions parameterized by the baseline weights $w_0$ and target weights $w_1$ are given by:

$$P(Y = 0 \mid \mathbf{X}=x) = \frac{e^{w_0 x}}{e^{w_0 x} + e^{w_1 x}} = \frac{1}{1 + e^{w x}}$$

$$P(Y = 1 \mid \mathbf{X}=x) = \frac{e^{w_1 x}}{e^{w_0 x} + e^{w_1 x}} = \frac{1}{1 + e^{-w x}}$$

Where we apply the change of variables $w = w_1 - w_0$.

---


Assume that from historical data we know the value of $\mathbb{E}\big[x\mathbb{I}(Y=1)\big]$.

We observe
$$\mathbb{E}\big[x\mathbb{I}(Y=1)\big] = \sum_{x \in \mathscr{X}} x \cdot \left[ \frac{1}{1 + e^{-w x}} \right] \cdot P(x).$$

Expanding this summation over our three scenarios $\mathscr{X} = \{0, 1, 2\}$:

$$\mathbb{E}\big[x\mathbb{I}(Y=1)\big] = 0 \cdot \left[\frac{1}{1+e^{0}}\right]P(x_A) + 1 \cdot \left[\frac{1}{1+e^{-w}}\right]P(x_B) + 2 \cdot \left[\frac{1}{1+e^{-2w}}\right]P(x_C)$$

Substituting our scenario probabilities $P(x_B) = 0.35$ and $P(x_C) = 0.15$:

$$\mathbb{E}\big[x\mathbb{I}(Y=1)\big] = 0.35 \left( \frac{1}{1 + e^{-w}} \right) + 0.30 \left( \frac{1}{1 + e^{-2w}} \right)$$

If the empirical data profile tells us that the target baseline expectation $\mathbb{E}\big[x\mathbb{I}(Y=1)\big] = 0.45$, we simply solve the above single-variable equation for $w$ to find the conditional probability.


$$0.45 = 0.35 \left( \frac{1}{1 + e^{-w}} \right) + 0.30 \left( \frac{1}{1 + e^{-2w}} \right)$$


Notice that for $x=2$, the probability term becomes:


$$\frac{1}{1 + e^{-2w}} = \frac{1}{1 + (e^{-w})^2}$$

We can solve this system using a standard numerical root solver to find the exact parameter difference:

$$w \approx \mathbf{0.8664}$$

---

Using $w \approx 0.8664$, the operational conditional probabilities for each feature value are calculated using:


$$P(Y=1 \mid \mathbf{X}=x) = \frac{1}{1 + e^{-0.8664x}} \quad \text{and} \quad P(Y=0 \mid \mathbf{X}=x) = \frac{1}{1 + e^{0.8664x}}$$

```python
import numpy as np
import plotly.graph_objects as go

# Define the learned parameter difference
u_1 = 0.8664

# Define the MaxEnt conditional probability function for Y=1
def p_y1(x, u1):
    return 1 / (1 + np.exp(-u1 * x))

# Generate a smooth range of x values for the sigmoid curve
x_curve = np.linspace(-1, 3, 400)
y_curve = p_y1(x_curve, u_1)

# Scenario coordinates from the problem
scenarios = [
    {"name": "A (Low)", "x": 0, "color": "#ff7f0e"},
    {"name": "B (Medium)", "x": 1, "color": "#2ca02c"},
    {"name": "C (High)", "x": 2, "color": "#d62728"}
]

# Create the interactive figure
fig = go.Figure()

# 1. Add the continuous MaxEnt probability curve
fig.add_trace(go.Scatter(
    x=x_curve,
    y=y_curve,
    mode='lines',
    name=r'$P(Y=1 \mid \mathbf{X}=x)$',
    line=dict(color='#1f77b4', width=3),
    hovertemplate='<b>Feature x</b>: %{x:.2f}<br><b>Probability</b>: %{y:.2%}<extra></extra>'
))

# 2. Add scenario marker points and drop lines
for sc in scenarios:
    prob = p_y1(sc['x'], u_1)

    # Dotted projection lines to axes
    fig.add_trace(go.Scatter(
        x=[sc['x'], sc['x'], None, -1, sc['x']],
        y=[0, prob, None, prob, prob],
        mode='lines',
        line=dict(color=sc['color'], width=1.5, dash='dash'),
        hoverinfo='skip',
        showlegend=False
    ))

    # Standalone scenario point
    fig.add_trace(go.Scatter(
        x=[sc['x']],
        y=[prob],
        mode='markers+text',
        name=f"Scenario {sc['name']}",
        marker=dict(color=sc['color'], size=12, line=dict(color='black', width=1.5)),
        text=[f"<b>Scenario {sc['name']}</b><br>x = {sc['x']}<br>P = {prob*100:.1f}%"],
        textposition="top right",
        hovertemplate='<b>%{text}</b><extra></extra>'
    ))

# 3. Layout adjustments and styling
fig.update_layout(
    title=dict(
        text='<b>MaxEnt 1D Conditional Probability Curve P(Y=1 | X=x)</b>',
        font=dict(size=16)
    ),
    xaxis=dict(
        title='Feature Value (x)',
        range=[-0.5, 2.5],
        gridcolor='rgba(200, 200, 200, 0.3)'
    ),
    yaxis=dict(
        title='Probability',
        range=[0, 1.05],
        tickformat='.0%',
        gridcolor='rgba(200, 200, 200, 0.3)'
    ),
    template='plotly_white',
    width=850,
    height=500,
    showlegend=True,
    hovermode='closest'
)

# Display the interactive chart
fig.show()
```

# 5. Other Applications

Here is a list of other applications of the **Principle of Maximum Entropy (MaxEnt)** that are significantly different from the classical thermodynamic ensembles and standard logistic/softmax regression covered above:

1. Ecology and Species Distribution Modeling (MaxEnt Software)

    In spatial ecology, predicting where a species lives is a massive challenge when you only have "presence-only" data (you know where a bird was spotted, but you don't have a list of all the places it *isn't*).

    * **The Constraints:** The empirical averages of environmental layers (like humidity, annual rainfall, elevation, and temperature) at the known geographic locations where the species was observed.
    * **The MaxEnt Application:** The model generates a spatial probability distribution across the entire map that is as uniform (spread out) as possible, *except* where it is forced to match the environmental constraints of the observed sightings.

2. Natural Language Processing: Part-of-Speech (POS) Tagging

    Before deep learning took over, MaxEnt models were the gold standard for predicting whether a word in a sentence is a noun, verb, or adjective based on surrounding context.

    * **The Constraints:** Syntactic context rules extracted from a training corpus. For example: *"The word immediately following 'the' is a noun 65% of the time,"* or *"If a word ends in '-ing' and follows a form of 'to be', it is a verb 80% of the time."*
    * **The MaxEnt Application:** Instead of making independent, biased assumptions about all other millions of word-tag combinations, MaxEnt finds the most unconstrained distribution over the text paths that perfectly satisfies these grammatical averages.

3. Finance: Asset Pricing and Portfolio Optimization

    In financial markets, predicting the future probability distribution of an asset's return is notoriously difficult because distributions are rarely perfectly Gaussian (they have fat tails and skewness).

    * **The Constraints:** Prior historic moments (mean return, variance) along with forward-looking market constraints, such as the current market prices of derivative options (which represent the market's collective expectation of future volatility).
    * **The MaxEnt Application:** Investors use MaxEnt to infer a "risk-neutral" probability distribution of future stock prices. It provides the safest, least-biased estimate of market risk without assuming a rigid curve like a geometric Brownian motion.

4. Image Reconstruction and Deconvolution (Astronomy & Medical Imaging)

    When capturing images of deep space (via radio telescopes) or internal organs (via MRI), the raw data is often noisy, blurry, or missing spatial frequencies due to hardware limitations.

    * **The Constraints:** The reconstructed pixel matrix, when passed through a mathematical model of the camera lens or telescope dish (a convolution matrix), must match the actual noisy data recorded by the physical sensors within a certain noise tolerance ($\chi^2$ error).
    * **The MaxEnt Application:** Because infinitely many sharp images could technically blur down to match the blurry data, MaxEnt chooses the one image that maximizes pixel entropy (treating pixel intensities like a probability distribution). This effectively yields the "smoothest" or most visually unbiased image that doesn't artificially manufacture sharp edges or artifacts out of thin air.

5. Urban Planning and Traffic Flow (Gravity Models)

    When planning cities or subways, engineers need to estimate how many people will travel from Zone A to Zone B daily.

    * **The Constraints:** The total number of commuters leaving Zone A each morning (total origin capacity), the total number of workers arriving at Zone B (total destination capacity), and the average total budget/time spent traveling across the entire city.
    * **The MaxEnt Application:** MaxEnt distributes the trips between all possible origins and destinations in the most randomized, disordered way possible while adhering to those boundary capacities. This derives the famous "Gravity Model" of human mobility purely from information theory.
